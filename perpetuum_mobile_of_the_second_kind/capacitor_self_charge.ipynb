# Исследование самозаряда конденсатора на осциллографе 

Это осциллограмма, полученная на конденсаторе из серии К50-29, 47 мкф, 16 вольт. Просто подключил щупы осциллографа к выводам конденсатора.

Здесь я выставил на осциллографе по оси времени 100 секунд на горизонтальную клетку.  Одна вертикальная клетка 5 мв. Но с учётом делителя напряжения встроенного в щуп, вертикальная клетка умножается на 10 и получается 50 милливольт. Спад напряжения резко вниз соответствует моменту разряда конденсатора через стрелочный амперметр.  Ток разряда где-то 6-8 микроампер

Я менял местами щупы осциллографа. И направление тренда кривой на осциллографе изменилось на противоположное. Тем самым исключается версия заряда конденсатора от осциллографа

```python
# емкость конденсатора
C = 47*10^-6
```

```python
# напряжение самозарядившегося конденсатора
U = 50 * 10^-3
```

```python
# энергия заряженного конденсатора
E = (C*U^2/2).n()
E
```

```python
# линейно экстраполированное время заряда, взятое исходя из наклона кривой заряда в начальный момент
t = 400
```

```python
# мощность самозаряда конденсатора вычичсленная в начальный момент времени
w = E/t
w
```

Eсли это была бы просто электрохимическая батарейка она бы давала для одной и той же серии конденсаторов одну и ту же полярность самозаряда. Я же наблюдаю для разных экземпляров конденсаторов одной и той же серии противоположное направление самозаряда.

Cам самозаряд стабилен. Не стабильна ВФХ (вольт-фарадная характеристика) каждого конкретного конденсатора. Кстати когда я нагрел пальцами один из конденсаторов, он поменял направление самозаряда на противоположное. Что указывает на сильную зависимость ВФХ от температуры.

Были проведены также исследования самозаряда этого конденсатора с использованием бытовой морозилки. При помещении конденсатора в морозилку его самозаряд прекращается. Однако что интересно, после извлечения конденсатора из морозилки напряжение самозарядившегося конденсатора увеличилось

Я предполагаю, что это явление может быть связано с фазовыми переходами второго рода в толще диэлектрика

Замораживание диэлектрика привело к некоторому упорядочиванию его структуры, что могло привести к увеличению крутизны несиметричности его ВФХ и тем самым способствовать увеличению напряжения самозаряда

Юрий Виноградов предлагает: поскольку в соответствии с флуктуационно-диссипативной теоремой

https://en.wikipedia.org/wiki/Fluctuation%E2%80%93dissipation_theorem

мощность тока Найквиста не зависит от размера резистора, а зависит лишь от его сопротивления, то мощность элементарных найквисторов можно суммировать на подложке интегральной микросхемы. В данном случае резистором выступает диэлектрик обкладки конденсатора, а мощность элементарного найквистора определяется мощностью тока Найквиста умноженного на коэффициент выпрямления (меньший единицы) зависящий от степени несимметричности и крутизны наклона ВФХ.

По расчётам Юрия Виноградова

Выходная мощность в режиме согласования по мощности в два раза меньше произведения зарядного тока на напряжение холостого хода (в Вт)

```python
w_theoretic = 0.5*0.3452495727*3.452495727*10^(-10)
w_theoretic
```

И на подложке площадью один сантиметр квадратный можно разместить конденсаторы, для каждого из которых отводится площадка 35х30 нм, (в штуках): 

```python
n = 1*10^(-4)/((35.*30)*(10^(-9))^2)
n
```

Таким образом, ожидаемая мощность интегрального найквистора в ваттах на квадратный сантиметр

```python
W_theoretic = n * w_theoretic
W_theoretic
```

```python
W = n * w
W
```

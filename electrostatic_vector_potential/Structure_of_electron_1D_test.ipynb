# Расчёт кулоновской силы в одномерной модели взаимодействия двух электронов как прецессирующих магнитных диполей

Эффективный электрический заряд внешней поверхности кольца магнитного тока $e$

Эффективный электрический заряд двух внутренних поверхностей двух колец магнитного тока $2e$

Расстояние от центра электрона до внешней поверхности кольца магнитного тока $a$

Расстояние между центрами двух электронов $r$

```python
from IPython.display import display, Math, Latex

def disp(ex):
    display(Math(latex(ex)))
    
def displ(tex):
    display(Math(tex))
```

```python
var("e,r,a")
```

Записываем матричные элементы матрицы выражений кулоновских сил взаимодействия каждого эффективного электрического заряда первого электрона с каждым эффективным электрическим зарядом второго электрона

Нумерация эффективных зарядов начинается от слоя ближайшего к соседнему электрону

```python
F_11 = - e^2/(r-2*a)^2
F_33 = - e^2/(r+2*a)^2
F_22 = - (2*e)^2/r^2

F_13 = - e^2/r^2
F_31 = - e^2/r^2

F_12 = + 2*e*e/(r-a)^2
F_21 = + 2*e*e/(r-a)^2

F_23 = + (2*e)*e/(r+a)^2
F_32 = + (2*e)*e/(r+a)^2
```

Составляем матрицу

```python
M = matrix([[F_11,F_12,F_13],
            [F_21,F_22,F_23],
            [F_31,F_32,F_33]])
```

```python
disp(M)
```

Выражение для суммарной силы взаимодействия двух атомов находим как сумму всех элементов матрицы

```python
F = sum(sum(M))
```

```python
disp(F)
```

```python
disp(F.expand())
```

```python
disp(F.factor())
```

Взаимодействие исчезает при устремлении размера прецессирующего магнитного диполя к нулю

```python
disp(limit(F, a=0))
```

```python
A   = 0.01
R   = 0.3
```

Рассматриваемые модели электронов как прецессирующих магнитных диполей отталкиваются

```python
eps = 0.02
plot(F.subs([e == 1, a == A]), r, 2*A+eps, R)
```

Исследуем вопрос, как сила отталкивания отличается от закона обратного расстояния в степени $n$

```python
A = 0.1
H = 0.01
R = 30
for n in range(2,8):
    F_rn = (F*r^n).factor()
    #disp (F_rn)
    eps = 0.25
    plot(F_rn.subs([e == 1, a == A]), r, 2*(A)+eps, R).show(title = "n=" + str(n))
```

Судя по графикам, на больших расстояниях сила отталкивания соответствует закону обратного расстояния в 6 степени

Допустим у нас есть система неподвижных зарядов. При этом мы в точке наблюдения суммируем (интегрируем) скалярные потенциалы. И затем по градиенту суммарного потенциала находим электростатическое поле. Это электростатика. Магнитного поля нет

Далее. У нас теперь есть система движущихся электрических зарядов. В этой задаче мы в точке наблюдения интегрируем как скалярные потенциалы так и векторный потенциал всех зарядов. Далее находим градиент скалярного потенциала плюс производная по времени векторного потенциала и получаем электрическое поле. Далее ротор от векторного потенциала даёт магнитное поле.  Ну и по желанию можем найти дивергенцию $\vec {А}_H$ по Николаеву. Это электродинамика

Далее. Мы имеем систему проводочков с током. Ток постоянный. Провода неподвижные. Скалярного потенциала нет, потому что мы считаем провода электронейтральными. Но линейный ток или объёмный ток является источником векторного потенциала магнитного поля $\vec {А}_H$. Интегрируем векторный потенциал магнитного поля $\vec {А}_H$ в точке наблюдения $(x_a, y_a, z_a)$. Находим пространственные производные векторного потенциала (ротор и дивергенцию) по координатам наблюдения и получаем таким образом магнитные поля. Векторное $vec H = rot \vec{А}_H$ и скалярное $S = div \vec{А}_H$ по Николаеву. Это магнитостатика. Есть магнитное поле, может быть будет поле Николаева (а может оно окажется нулевым). Но нету электрического поля

Но теперь самое интересное.  У нас есть система движущихся проводов с постоянным током. Каждый элемент тока является источником векторного потенциала но вместе с тем, каждый элемент тока обладает скоростью. Как это интегрировать в точке наблюдения? Это задача магнитодинамики. Мы не можем отбросить собственную скорость источников векторного потенциала. Но как учесть?

Может быть по Войцеховскому. Может быть по Цикре (зорб теория). У них у обоих поле имеет скорость. У Цикры скалярный потенциал имеет скорость. У Войцеховского магнитное поле имеет скорость.

Пока что мне ясно, что в задачах магнитодинамики должно появиться кроме магнитного поля ещё и электрическое. Например интересный и важный случай магнитодинамики это униполярная индукция. Этот случай интересен тем, что вектор скорости провода (если мы примем аналогию вращающегося вокруг оси намагниченности цилиндрического магнита и вращающейся вокруг той же аксиальной оси катушки) сонаправлен (или противонаправлен) вектору плотности тока.

В этой задаче что ещё интересно. Если магнит или катушка вращается соосно с равномерной скоростью, то казалось бы картина стационарна. И для получения электрического поля стандартный прием взятия производной по времени от векторного потенциала,  если этот прием применять к суммарному векторному потенциалу, получится ноль, как в парадоксе Николаева о токе смещения прямого провода с постоянным током.

Парадокс Николаева заключается в том, что если рассмотреть одиночный заряд движущийся с постоянной скоростью, то его ток смещения не равен нулю, но если из таких зарядов составить паравозик и получить длинный прямолинейный ток, то ток смещения от суммарного тока кажется уже равным нулю.

Но мы можем каждый элемент $dl$ провода (пусть для простоты это будет вращающийся вокруг своей нормали виток с посточнным током) рассмотреть отдельно. То есть процедура следующая. Для каждого элемента тока $I dl$ вычисляем векторный потенциал $\vec А_H = \frac{1}{c}\frac{e \vec v}{r}$. И его производную по времени в точке наблюдения. Кстати какую производную? Полную $\frac{d}{dt} \vec А_H$ или частную $\frac{\partial}{\partial t} \vec А_H$. Может быть даже полную которая является суммой частной производной и конвективной производной $\frac{d}{dt} = \frac{\partial}{\partial t}  + \left(\vec v \nabla\right)$. Затем только мы уже мы интегрируем по длине вращающегося витка

формула конвективной производной 

$$\left(\vec v \nabla\right) = \left(v_x \frac{\partial }{\partial x} + v_y \frac{\partial }{\partial y} + v_z \frac{\partial }{\partial z}\right)$$

В ЛЛ2 "Теория поля" параграф 16 рассматривается действие заряда $e$ в электромагнитном поле $A_{H_i}$. 

См формула (16,1), действия для заряда в электромагнитном поле без учёта материальной массы

$$S = \int\limits_{a}^{b} -\frac{\int de}{c}\left(A_{H_i} d x^i\right).\,\,\,  (16,1)$$


Но мы хотим в магнитодинамике записать действие для элемента тока в электромагнитном поле аналогично 

$$S = \int\limits_{a}^{b} -\frac{q_m}{c}\left(A_{E_i} d x^i\right).\,\,\,  (аналог \,\,\, 16,1)$$

Пространственная компонента четыревектора $A_{E_i}$ - это векторный потенциал $\vec{А}_E$ электрического поля
$\vec E = rot \vec{А}_E$. А временная компонента четыревектора это скалярный потенциал магнитного поля $\psi$, который определяется $\vec H = grad \psi$

$$A_{E_i} = \left(\psi, \vec{A}_E\right)\,\,\,(аналог \,\,\, 16,2)$$

Интеграл действия можно записать в виде 


$$S = \int\limits_{a}^{b}\left( \frac{q_m}{c}\vec A_{{E}} d\vec{r} - \left(q_m\right) \psi dt\right).$$


Или введя скорость элемента тока $\vec v = \frac{d \vec{r}}{dt}$ переходя к интегрированию по времени

$$S = \int\limits_{t_1}^{t_2}\left( \frac{q_m}{c}\vec A_{{E}} \vec{v} - \left(q_m\right) \psi \right)\,dt.\,\,\,(аналог \,\,\, 16,3)$$


Подынтегральное выражение есть функция Лагранжа для элемента тока в электромагнитном поле

$$L =  \frac{q_m}{c}\vec A_{{E}} \vec{v} - \left(q_m\right) \psi .\,\,\,(аналог \,\,\, 16,4)$$

Производная $\frac{\partial L}{\partial \vec{v}}$ есть обобщённый импульс элемента тока $\vec P$. Произведя дифференцирование, находим:

$$\vec P = \frac{\partial L}{\partial \vec{v}} = \frac{q_m}{c}\vec A_{{E}} $$

Теперь мы можем найти движения элемента тока в заданном электромагнитном поле, которые получаются варьированием действия, то есть даются уравнениями Лагранжа

$$\frac{d}{dt}\frac{\partial \vec{L}}{\partial \vec{v}} = \frac{\partial \vec{L}}{\partial \vec{r}}$$

Далее пишем

$$\frac{\partial L}{\partial \vec{r}} = \nabla L = \frac{q_m}{c} grad \,\vec A_{{E}} \vec{v} - \left(q_m\right) grad \,\psi$$

По известной формуле векторного анализа

$$grad\,\left(\vec{a}\vec{b}\right) = \left(\vec{a}\nabla\right) \vec{b}
                                    + \left(\vec{b}\nabla\right) \vec{a}
                                    + [\vec{b} \times rot \vec{a}]
                                    + [\vec{a} \times rot \vec{b}]$$

Применяя эту формулу к $\vec A_{{E}} \vec{v}$ и помня, что дифференцирование по $\vec r$ производится при постоянном $\vec v$, находим

$$\frac{\partial \vec{L}}{\partial \vec{r}} 
= \nabla \vec{L}
= \frac{q_m}{c} \left(\left(\vec{v}\nabla\right) \vec{A}_E
+ [\vec{v} \times rot \vec{A}_E]\right) - \left(q_m\right) grad \,\psi$$

Уравнения Лагранжа, следовательно, имеют вид

$$\frac{d}{dt}\frac{\partial \vec{L}}{\partial \vec{v}} = \frac{\partial \vec{L}}{\partial \vec{r}}$$

$$\frac{d}{dt}\frac{\partial \vec{L}}{\partial \vec{v}}
= \frac{q_m}{c} \left(\left(\vec{v}\nabla\right) \vec{A}_E
+ [\vec{v} \times rot \vec{A}_E]\right) - \left(q_m\right) grad \,\psi$$

$$\frac{d}{dt} \left(p+\frac{q_m}{c}\vec A_{{E}}\right)
= \frac{q_m}{c} \left(\left(\vec{v}\nabla\right) \vec{A}_E
+ [\vec{v} \times rot \vec{A}_E]\right) - \left(q_m\right) grad \,\psi$$

Полный дифференциал векторного потенциала по времени складывается из двух частей

$$\frac{d}{dt} \vec{A_E} = \frac{\partial \vec{A}_E}{\partial \vec{t}} + \left(\vec{v}\nabla\right) \vec{A}_E$$

Подставляя это в предыдущее выражение

$$\frac{dp}{dt} + \frac{q_m}{c}\frac{d}{dt}\left(\vec A_{{E}}\right)
= \frac{q_m}{c} \left(\left(\vec{v}\nabla\right) \vec{A}_E
+ [\vec{v} \times rot \vec{A}_E]\right) - \left(q_m\right) grad \,\psi$$

получаем

$$\frac{dp}{dt} + \frac{q_m}{c}\left(\frac{\partial \vec{A}_E}{\partial \vec{t}} + \left(\vec{v}\nabla\right) \vec{A}_E\right)
= \frac{q_m}{c} \left(\left(\vec{v}\nabla\right) \vec{A}_E
+ [\vec{v} \times rot \vec{A}_E]\right) - \left(q_m\right) grad \,\psi$$

$$\frac{dp}{dt} + \frac{q_m}{c}\left(\frac{\partial \vec{A}_E}{\partial \vec{t}}\right)
= \frac{q_m}{c} \left([\vec{v} \times rot \vec{A}_E]\right) - \left(q_m\right) grad \,\psi$$

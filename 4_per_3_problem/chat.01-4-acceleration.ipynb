### USER
Скажи, как правильно вычисляется 4-ускорение  осциллирующегося заряда в релятивистском случае 

### ASSISTANT
Вычисление **4-ускорения осциллирующего заряда** в **релятивистском случае** требует использования аппарата специальной теории относительности (СТО), в частности, формализма 4-векторов. Давайте разберёмся по шагам.

---

### 1. Что такое 4-ускорение?

**4-ускорение** — это 4-вектор, описывающий ускорение мировой линии частицы в пространстве-времени Минковского. Он определяется как производная **4-скорости** по **собственному времени**:

$$
a^\mu = \frac{d u^\mu}{d\tau}
$$

где:
- $u^\mu = \frac{dx^\mu}{d\tau}$ — 4-скорость,
- $\tau$ — собственное время частицы,
- $x^\mu(\tau)$ — мировая линия частицы.

---

### 2. Мировая линия осциллирующего заряда

Рассмотрим простейший случай: **одномерные гармонические колебания** заряда вдоль оси $x$. В лабораторной системе отсчёта (СО покоя наблюдателя) координаты частицы зависят от времени:

$$
x(t) = x_0 + A \sin(\omega t)
$$

где:
- $A$ — амплитуда,
- $\omega$ — угловая частота,
- $t$ — координатное время (время в лабораторной СО).

Чтобы перейти к 4-векторам, нужно параметризовать движение через **собственное время** $\tau$, а не координатное время $t$.

---

### 3. Параметризация через координатное время

Мировая линия в координатах:

$$
x^\mu(t) = (ct,\ x_0 + A \sin(\omega t),\ 0,\ 0)
$$

Теперь нужно перейти к параметризации по $\tau$. Для этого найдём связь между $dt$ и $d\tau$:

$$
d\tau = \frac{1}{c} \sqrt{ -\eta_{\mu\nu} dx^\mu dx^\nu } = dt \sqrt{1 - \frac{v^2(t)}{c^2}}
$$

где $v(t) = \frac{dx}{dt} = A \omega \cos(\omega t)$

Тогда:

$$
\gamma(t) = \frac{1}{\sqrt{1 - v^2(t)/c^2}} = \frac{1}{\sqrt{1 - \frac{A^2 \omega^2}{c^2} \cos^2(\omega t)}}
$$

и

$$
\frac{dt}{d\tau} = \gamma(t)
$$

---

### 4. 4-скорость

4-скорость:

$$
u^\mu = \frac{dx^\mu}{d\tau} = \frac{dt}{d\tau} \frac{dx^\mu}{dt} = \gamma(t) (c,\ v(t),\ 0,\ 0)
$$

То есть:

$$
u^\mu = \left( \gamma(t) c,\ \gamma(t) v(t),\ 0,\ 0 \right)
$$

---

### 5. 4-ускорение

Теперь вычислим:

$$
a^\mu = \frac{d u^\mu}{d\tau} = \frac{dt}{d\tau} \frac{d u^\mu}{dt} = \gamma(t) \frac{d u^\mu}{dt}
$$

Вычислим компоненты:

#### Временная компонента ($\mu = 0$):

$$
a^0 = \gamma \frac{d}{dt} (\gamma c) = \gamma c \frac{d\gamma}{dt}
$$

Найдём $\frac{d\gamma}{dt}$:

$$
\gamma = \left(1 - \frac{v^2}{c^2}\right)^{-1/2}, \quad \frac{d\gamma}{dt} = \frac{v \dot{v}}{c^2} \gamma^3
$$

где $\dot{v} = \frac{dv}{dt} = -A \omega^2 \sin(\omega t)$

Тогда:

$$
a^0 = \gamma c \cdot \frac{v \dot{v}}{c^2} \gamma^3 = \frac{\gamma^4 v \dot{v}}{c}
$$

#### Пространственная компонента ($\mu = 1$):

$$
a^1 = \gamma \frac{d}{dt} (\gamma v) = \gamma \left( \dot{\gamma} v + \gamma \dot{v} \right)
$$

Подставим $\dot{\gamma} = \frac{v \dot{v}}{c^2} \gamma^3$:

$$
a^1 = \gamma \left( \frac{v \dot{v}}{c^2} \gamma^3 v + \gamma \dot{v} \right) = \gamma^2 \dot{v} + \frac{\gamma^4 v^2 \dot{v}}{c^2} = \gamma^2 \dot{v} \left( 1 + \frac{\gamma^2 v^2}{c^2} \right)
$$

Но $\gamma^2 = \frac{1}{1 - v^2/c^2}$, и можно упростить:

$$
a^1 = \gamma^4 \dot{v} \left(1 + \frac{v^2}{c^2} (\gamma^2 - 1) \cdot \text{?} \right)
$$

Лучше использовать известную формулу:

$$
a^\mu = \left( \frac{\gamma^4 (\vec{v} \cdot \vec{a})}{c},\ \frac{\gamma^4 (\vec{v} \cdot \vec{a})}{c^2} \vec{v} + \gamma^2 \vec{a} \right)
$$

где $\vec{a} = \frac{d\vec{v}}{dt}$ — 3-ускорение.

В нашем случае:
- $\vec{v} = (v, 0, 0)$,
- $\vec{a} = (\dot{v}, 0, 0) = (-A \omega^2 \sin(\omega t), 0, 0)$,
- $\vec{v} \cdot \vec{a} = v \dot{v}$

Тогда:

$$
a^0 = \frac{\gamma^4 (v \dot{v})}{c}, \quad
a^1 = \frac{\gamma^4 (v \dot{v})}{c^2} v + \gamma^2 \dot{v} = \gamma^4 \dot{v} \left( \frac{v^2}{c^2} + \frac{1}{\gamma^2} \right)
$$

Но $\frac{1}{\gamma^2} = 1 - v^2/c^2$, поэтому:

$$
\frac{v^2}{c^2} + 1 - \frac{v^2}{c^2} = 1 \Rightarrow a^1 = \gamma^4 \dot{v}
$$

Итак, окончательно:

$$
a^\mu = \left( \frac{\gamma^4 v \dot{v}}{c},\ \gamma^4 \dot{v},\ 0,\ 0 \right)
$$

---

### 6. Итоговая формула для 4-ускорения

Для осциллирующего заряда с движением $x(t) = x_0 + A \sin(\omega t)$:

- $v(t) = A \omega \cos(\omega t)$
- $\dot{v}(t) = -A \omega^2 \sin(\omega t)$
- $\gamma(t) = \left(1 - \frac{v^2(t)}{c^2}\right)^{-1/2}$

Тогда 4-ускорение:

$$
a^\mu(\tau) = \left( \frac{\gamma^4 v \dot{v}}{c},\ \gamma^4 \dot{v},\ 0,\ 0 \right)
$$

(здесь компоненты выражены через $t$, но $\tau$ связано с $t$ через $d\tau = dt / \gamma$)

---

### 7. Замечания

- В **нерелятивистском пределе** ($v \ll c$): $\gamma \approx 1$, и $a^\mu \approx (0, \dot{v}, 0, 0)$, что соответствует обычному ускорению.
- В релятивистском случае 4-ускорение **не параллельно** 3-ускорению и сильно зависит от скорости.
- 4-ускорение всегда **ортогонально** 4-скорости: $a_\mu u^\mu = 0$ — это свойство любого 4-ускорения.

---

### 8. Зачем это нужно?

4-ускорение важно для:
- вычисления **излучения заряда** (формула Лiénarda для мощности излучения),
- анализа динамики в СТО,
- построения уравнений движения в электродинамике.

Например, мощность излучения ускоренного заряда в релятивистском случае:

$$
P = \frac{\mu_0 q^2}{6\pi c} a^\mu a_\mu
$$

где $a^\mu a_\mu = -\frac{(\gamma^4 \dot{v})^2}{c^2} (v^2 - c^2) = \gamma^6 \dot{v}^2$ — инвариантный квадрат 4-ускорения.

---

### Вывод

**4-ускорение осциллирующего заряда** в релятивистском случае вычисляется как:

$$
a^\mu = \frac{d u^\mu}{d\tau} = \gamma \frac{d}{dt} \left( \gamma c,\ \gamma v,\ 0,\ 0 \right)
$$

и даётся формулой:

$$
a^\mu = \left( \frac{\gamma^4 v \dot{v}}{c},\ \gamma^4 \dot{v},\ 0,\ 0 \right)
$$

где все величины зависят от координатного времени $t$, а собственное время $\tau$ связано с $t$ через $d\tau = dt / \gamma(t)$.

Если нужно, можно переписать всё как функцию $\tau$, но это требует решения интеграла:

$$
\tau = \int \sqrt{1 - \frac{v^2(t)}{c^2}} dt
$$

что, как правило, не выражается в элементарных функциях.

### USER
Рассмотри общий случай,  когда заряд движется по сложной траектории, когда не равны нулю все компоненты обычной скорости 

### ASSISTANT
Отлично! Давайте рассмотрим **общий релятивистский случай**, когда заряженная частица движется по **произвольной траектории** в 3D-пространстве, и все компоненты её обычной (3-) скорости могут быть ненулевыми. Наша цель — вычислить **4-ускорение** $a^\mu$ в этом общем случае.

---

### 🔷 1. Основные понятия: 4-векторы в СТО

В специальной теории относительности движение частицы описывается **мировой линией** $x^\mu(\tau)$, где $\tau$ — **собственное время**.

#### 4-координата:
$$
x^\mu(\tau) = (ct(\tau),\ x(\tau),\ y(\tau),\ z(\tau))
$$

#### 4-скорость:
$$
u^\mu = \frac{dx^\mu}{d\tau} = \gamma (c,\ \vec{v})
$$
где:
- $\vec{v}(t) = \frac{d\vec{r}}{dt}$ — обычная 3-скорость,
- $\gamma = \frac{1}{\sqrt{1 - v^2/c^2}}$, $v = |\vec{v}|$

#### 4-ускорение:
$$
a^\mu = \frac{du^\mu}{d\tau} = \frac{d}{d\tau} u^\mu
$$

Поскольку $\tau$ неудобно использовать напрямую (т.к. траектории задаются как $\vec{r}(t)$), перейдём к производной по **координатному времени** $t$:

$$
a^\mu = \frac{du^\mu}{d\tau} = \frac{dt}{d\tau} \frac{du^\mu}{dt} = \gamma \frac{du^\mu}{dt}
$$

---

### 🔷 2. Общее выражение для 4-ускорения

Разложим $u^\mu = \gamma(c,\ \vec{v})$, тогда:

$$
a^\mu = \gamma \frac{d}{dt} \left( \gamma c,\ \gamma \vec{v} \right) = \left( \gamma \frac{d}{dt}(\gamma c),\ \gamma \frac{d}{dt}(\gamma \vec{v}) \right)
$$

Вычислим компоненты.

---

#### ✅ Временная компонента ($\mu = 0$):

$$
a^0 = \gamma \cdot c \cdot \frac{d\gamma}{dt}
$$

Найдём $\frac{d\gamma}{dt}$:

$$
\gamma = \left(1 - \frac{v^2}{c^2}\right)^{-1/2} \quad \Rightarrow \quad
\frac{d\gamma}{dt} = \frac{1}{c^2} \gamma^3 (\vec{v} \cdot \dot{\vec{v}})
$$

где $\dot{\vec{v}} = \frac{d\vec{v}}{dt} = \vec{a}$ — обычное 3-ускорение.

Тогда:

$$
a^0 = \gamma \cdot c \cdot \frac{\gamma^3}{c^2} (\vec{v} \cdot \vec{a}) = \frac{\gamma^4}{c} (\vec{v} \cdot \vec{a})
$$

---

#### ✅ Пространственные компоненты ($\mu = 1,2,3$):

$$
a^i = \gamma \frac{d}{dt} (\gamma v^i) = \gamma \left( \dot{\gamma} v^i + \gamma \dot{v}^i \right)
$$

Подставим $\dot{\gamma} = \frac{\gamma^3}{c^2} (\vec{v} \cdot \vec{a})$:

$$
a^i = \gamma \left( \frac{\gamma^3}{c^2} (\vec{v} \cdot \vec{a}) v^i + \gamma a^i \right)
= \frac{\gamma^4}{c^2} (\vec{v} \cdot \vec{a}) v^i + \gamma^2 a^i
$$

Таким образом, в векторной форме:

$$
\vec{a}^{\ (sp)} = \frac{\gamma^4}{c^2} (\vec{v} \cdot \vec{a}) \vec{v} + \gamma^2 \vec{a}
$$

---

### 🔷 3. Итог: общая формула 4-ускорения

$$
\boxed{
a^\mu = \left(
\frac{\gamma^4}{c} (\vec{v} \cdot \vec{a}),\ 
\frac{\gamma^4}{c^2} (\vec{v} \cdot \vec{a}) \vec{v} + \gamma^2 \vec{a}
\right)
}
$$

где:
- $\vec{v}(t)$ — 3-скорость,
- $\vec{a}(t) = \frac{d\vec{v}}{dt}$ — 3-ускорение,
- $\gamma = \left(1 - \frac{v^2}{c^2}\right)^{-1/2}$

---

### 🔷 4. Свойства 4-ускорения

1. **Ортогональность 4-скорости и 4-ускорения**:
   $$
   a_\mu u^\mu = 0
   $$
   Это следует из того, что $u^\mu u_\mu = -c^2 = \text{const}$, дифференцируя по $\tau$, получаем $a_\mu u^\mu = 0$.

2. **Инвариантный квадрат 4-ускорения**:
   $$
   a^\mu a_\mu = -\frac{1}{c^2} (a^0)^2 + (\vec{a}^{\ (sp)})^2
   $$
   После подстановки получается:
   $$
   a^\mu a_\mu = \gamma^6 \left( a^2 - \frac{(\vec{v} \times \vec{a})^2}{c^2} \right)
   $$
   (подробный вывод ниже)

---

### 🔷 5. Инвариант $a^\mu a_\mu$

Вычислим квадрат 4-ускорения:

$$
a^\mu a_\mu = -\frac{(a^0)^2}{c^2} + |\vec{a}^{\ (sp)}|^2
$$

Подставим:

$$
(a^0)^2 = \frac{\gamma^8}{c^2} (\vec{v} \cdot \vec{a})^2
\quad \Rightarrow \quad
-\frac{(a^0)^2}{c^2} = -\frac{\gamma^8}{c^4} (\vec{v} \cdot \vec{a})^2
$$

$$
|\vec{a}^{\ (sp)}|^2 = \left| \frac{\gamma^4}{c^2} (\vec{v} \cdot \vec{a}) \vec{v} + \gamma^2 \vec{a} \right|^2
= \frac{\gamma^8}{c^4} (\vec{v} \cdot \vec{a})^2 v^2 + 2 \frac{\gamma^6}{c^2} (\vec{v} \cdot \vec{a})(\vec{v} \cdot \vec{a}) + \gamma^4 a^2
$$

$$
= \frac{\gamma^8}{c^4} (\vec{v} \cdot \vec{a})^2 v^2 + 2 \frac{\gamma^6}{c^2} (\vec{v} \cdot \vec{a})^2 + \gamma^4 a^2
$$

Складываем:

$$
a^\mu a_\mu = -\frac{\gamma^8}{c^4} (\vec{v} \cdot \vec{a})^2 + \frac{\gamma^8}{c^4} (\vec{v} \cdot \vec{a})^2 v^2 + 2 \frac{\gamma^6}{c^2} (\vec{v} \cdot \vec{a})^2 + \gamma^4 a^2
$$

$$
= \frac{\gamma^8}{c^4} (\vec{v} \cdot \vec{a})^2 (v^2 - c^2) + 2 \frac{\gamma^6}{c^2} (\vec{v} \cdot \vec{a})^2 + \gamma^4 a^2
$$

Заметим, что $v^2 - c^2 = -c^2 (1 - v^2/c^2) = -c^2 / \gamma^2$, тогда:

$$
\frac{\gamma^8}{c^4} (\vec{v} \cdot \vec{a})^2 (-c^2 / \gamma^2) = -\frac{\gamma^6}{c^2} (\vec{v} \cdot \vec{a})^2
$$

Тогда:

$$
a^\mu a_\mu = -\frac{\gamma^6}{c^2} (\vec{v} \cdot \vec{a})^2 + 2 \frac{\gamma^6}{c^2} (\vec{v} \cdot \vec{a})^2 + \gamma^4 a^2 = \frac{\gamma^6}{c^2} (\vec{v} \cdot \vec{a})^2 + \gamma^4 a^2
$$

Теперь используем тождество:

$$
a^2 = \frac{(\vec{v} \cdot \vec{a})^2}{v^2} + \frac{|\vec{v} \times \vec{a}|^2}{v^2} \quad \text{(если } v \ne 0\text{)}
$$

Но лучше — переписать в виде:

$$
a^\mu a_\mu = \gamma^6 \left( a^2 - \frac{(\vec{v} \times \vec{a})^2}{c^2} \right)
$$

✅ Это — **стандартная формула** для квадрата 4-ускорения.

---

### 🔷 6. Применение: излучение заряда

Мощность излучения (формула Лiénarda):

$$
P = \frac{\mu_0 q^2}{6\pi c} a^\mu a_\mu = \frac{\mu_0 q^2}{6\pi c} \gamma^6 \left( a^2 - \frac{(\vec{v} \times \vec{a})^2}{c^2} \right)
$$

или, эквивалентно:

$$
P = \frac{q^2}{6\pi \varepsilon_0 c^3} \gamma^6 \left( a^2 - \frac{(\vec{v} \times \vec{a})^2}{c^2} \right)
$$

Это обобщение формулы Лармора на релятивистский случай.

---

### 🔷 7. Пример: круговое движение

Допустим, заряд движется по окружности со скоростью $v$, ускорение $\vec{a} \perp \vec{v}$, тогда $\vec{v} \cdot \vec{a} = 0$, и:

$$
a^\mu = (0,\ \gamma^2 \vec{a}), \quad a^\mu a_\mu = \gamma^4 a^2
$$

Но по формуле выше: $\gamma^6 (a^2 - v^2 a^2 / c^2) = \gamma^6 a^2 (1 - v^2/c^2) = \gamma^6 a^2 / \gamma^2 = \gamma^4 a^2$ — совпадает.

---

### 🔷 8. Заключение

Для **произвольной траектории** заряда с ненулевыми компонентами скорости и ускорения:

✅ **4-ускорение** вычисляется по формуле:

$$
\boxed{
a^\mu = \left(
\frac{\gamma^4 (\vec{v} \cdot \vec{a})}{c},\ 
\frac{\gamma^4 (\vec{v} \cdot \vec{a})}{c^2} \vec{v} + \gamma^2 \vec{a}
\right)
}
$$

✅ **Инвариантный квадрат**:

$$
\boxed{
a^\mu a_\mu = \gamma^6 \left( a^2 - \frac{|\vec{v} \times \vec{a}|^2}{c^2} \right)
}
$$

✅ Эта величина определяет **интенсивность излучения** заряда в релятивистском случае.

### USER
Теперь дай формулу скалярного произведения 4-ускорения на 4-радиус вектор из точки наблюдения в запаздывающую точку движущегося заряда

### ASSISTANT
Отличный вопрос! Ты спрашиваешь о **скалярном произведении 4-ускорения** заряда и **4-вектора**, проведённого из точки наблюдения в **запаздывающую точку** на траектории заряда. Это важная величина в релятивистской электродинамике, особенно при вычислении **запаздывающих потенциалов** и **полей излучения**.

---

### 🔷 Постановка задачи

Пусть:
- $x^\mu = (ct, \vec{r})$ — 4-координата **точки наблюдения** (приёмника),
- $z^\mu(\tau_{\text{ret}})$ — 4-координата **запаздывающей точки** на мировой линии заряда (то есть положение заряда в момент, когда сигнал, испущенный им, достиг наблюдателя),
- $R^\mu = x^\mu - z^\mu(\tau_{\text{ret}})$ — **4-вектор запаздывания** от заряда к наблюдателю,
- $a^\mu(\tau_{\text{ret}})$ — **4-ускорение** заряда в запаздывающий момент времени.

Требуется найти:
$$
a_\mu R^\mu = a^\mu R_\mu
$$
(скалярное произведение по метрике Минковского: $A_\mu B^\mu = \eta_{\mu\nu} A^\nu B^\mu$)

---

### 🔷 1. Определение 4-вектора запаздывания

$$
R^\mu = x^\mu - z^\mu(\tau_{\text{ret}})
$$

В координатах:
$$
R^\mu = \left( c(t - t_{\text{ret}}),\ \vec{r} - \vec{z}(t_{\text{ret}}) \right)
$$

Обозначим:
- $\vec{R} = \vec{r} - \vec{z}(t_{\text{ret}})$ — 3-вектор от запаздывающего положения заряда к наблюдателю,
- $R = |\vec{R}|$,
- Условие запаздывания: $c(t - t_{\text{ret}}) = R$, то есть $R^\mu$ — **светоподобный** вектор:
$$
R_\mu R^\mu = 0
$$

---

### 🔷 2. Нормированный 4-вектор направления

Определим **единичный светоподобный 4-вектор**:
$$
n^\mu = \frac{R^\mu}{R} = \left(1,\ \hat{n}\right), \quad \text{где } \hat{n} = \frac{\vec{R}}{R}
$$

Тогда $R^\mu = R n^\mu$, и
$$
a_\mu R^\mu = R \cdot (a_\mu n^\mu)
$$

---

### 🔷 3. Формула для $a_\mu R^\mu$

$$
a_\mu R^\mu = a_\mu (x^\mu - z^\mu) = a_\mu R^\mu
$$

В компонентах:
$$
a_\mu R^\mu = -a^0 R^0 + \vec{a} \cdot \vec{R}
$$
(с учётом метрики $\eta_{\mu\nu} = \text{diag}(-1, +1, +1, +1)$, если используем $(+,-,-,-)$, то наоборот — будем использовать **обычную физическую метрику**: $\boxed{(-, +, +, +)}$)

То есть:
$$
a_\mu R^\mu = -a^0 R^0 + \vec{a}^{\text{(sp)}} \cdot \vec{R}
$$

Подставим:
- $R^0 = c(t - t_{\text{ret}}) = R$ (поскольку $R = c(t - t_{\text{ret}})$),
- $a^0 = \dfrac{\gamma^4}{c} (\vec{v} \cdot \vec{a})$,
- $\vec{a}^{\text{(sp)}} = \dfrac{\gamma^4}{c^2} (\vec{v} \cdot \vec{a}) \vec{v} + \gamma^2 \vec{a}$

Тогда:

$$
a_\mu R^\mu = -\left( \frac{\gamma^4}{c} (\vec{v} \cdot \vec{a}) \right) \cdot R + \left[ \frac{\gamma^4}{c^2} (\vec{v} \cdot \vec{a}) \vec{v} + \gamma^2 \vec{a} \right] \cdot \vec{R}
$$

Разобьём скалярное произведение:

$$
= -\frac{\gamma^4 R}{c} (\vec{v} \cdot \vec{a}) + \frac{\gamma^4}{c^2} (\vec{v} \cdot \vec{a}) (\vec{v} \cdot \vec{R}) + \gamma^2 (\vec{a} \cdot \vec{R})
$$

Вынесем $\gamma^4 (\vec{v} \cdot \vec{a})$:

$$
a_\mu R^\mu = \gamma^4 (\vec{v} \cdot \vec{a}) \left( -\frac{R}{c} + \frac{\vec{v} \cdot \vec{R}}{c^2} \right) + \gamma^2 (\vec{a} \cdot \vec{R})
$$

---

### 🔷 4. Упрощение: введём $\hat{n} = \vec{R}/R$

Тогда:
- $\vec{v} \cdot \vec{R} = R (\vec{v} \cdot \hat{n})$,
- $\vec{a} \cdot \vec{R} = R (\vec{a} \cdot \hat{n})$

Подставим:

$$
a_\mu R^\mu = \gamma^4 (\vec{v} \cdot \vec{a}) \left( -\frac{R}{c} + \frac{R (\vec{v} \cdot \hat{n})}{c^2} \right) + \gamma^2 R (\vec{a} \cdot \hat{n})
$$

$$
= R \left[
\gamma^4 (\vec{v} \cdot \vec{a}) \left( -\frac{1}{c} + \frac{\vec{v} \cdot \hat{n}}{c^2} \right) + \gamma^2 (\vec{a} \cdot \hat{n})
\right]
$$

$$
= R \left[
-\frac{\gamma^4}{c} (\vec{v} \cdot \vec{a}) \left(1 - \frac{\vec{v} \cdot \hat{n}}{c} \right) + \gamma^2 (\vec{a} \cdot \hat{n})
\right]
$$

---

### ✅ Окончательная формула:

$$
\boxed{
a_\mu R^\mu = R \left[
\gamma^2 (\vec{a} \cdot \hat{n}) - \frac{\gamma^4}{c} (\vec{v} \cdot \vec{a}) \left(1 - \frac{\vec{v} \cdot \hat{n}}{c} \right)
\right]
}
$$

---

### 🔷 Альтернативная (инвариантная) форма

Можно выразить это через 4-скаляры. Напомним, что:
- $u_\mu R^\mu = -\gamma c (R - \vec{v} \cdot \vec{R}/c) = -\gamma (c R - \vec{v} \cdot \vec{R})$
- Но для $a_\mu R^\mu$ нет более простого инвариантного выражения, кроме как через компоненты.

Однако в литературе (например, в выводе поля излучения по методу Лiénarda–Вихерта) часто встречается комбинация вида:

$$
a_\mu R^\mu = R \cdot \left[ \gamma^2 \vec{a}_\perp + \dots \right]
$$

где $\vec{a}_\perp$ — компонента ускорения, перпендикулярная скорости и/или лучу зрения.

---

### 🔷 Зачем это нужно?

Скалярное произведение $a_\mu R^\mu$ **не является** основным в явных формулах полей, но **входит** в производные при вычислении:
- напряжённостей поля $\vec{E}$ и $\vec{B}$ от движущегося заряда,
- особенно в **излучаемой части поля**, пропорциональной ускорению.

Например, электрическое поле по Лiénарду:

$$
\vec{E} = \frac{q}{4\pi\varepsilon_0} \left[
\frac{\hat{n} - \vec{\beta}}{\gamma^2 (1 - \vec{\beta} \cdot \hat{n})^3 R^2}
\right]_{\text{ret}} + \frac{q}{4\pi\varepsilon_0 c} \frac{ \left[ \hat{n} \times \left( (\hat{n} - \vec{\beta}) \times \dot{\vec{\beta}} \right) \right] }{(1 - \vec{\beta} \cdot \hat{n})^3 R}
$$

где $\vec{\beta} = \vec{v}/c$, $\dot{\vec{\beta}} = \vec{a}/c$.

Видно, что **ускорение** входит через $\dot{\vec{\beta}}$, и направление $\hat{n}$, так что $\vec{a} \cdot \hat{n}$ и $\vec{v} \cdot \hat{n}$ — ключевые величины.

### 🔷 Заключение

Скалярное произведение 4-ускорения на 4-вектор запаздывания:

$$
\boxed{
a_\mu R^\mu = R \left[

\right]

где:

\right]

Однако в **направлении оси вращения** (в данном случае — по $z$), излучение:
- **меньше**, чем в плоскости,
- но **не нулевое**.

Здесь мы получили $a_\mu R^\mu = 0$, но это **только один скаляр** — он не определяет поле однозначно. Поле зависит от **векторных комбинаций**, таких как $\hat{n} \times (\hat{n} \times \dot{\vec{\beta}})$, а не только от скалярного произведения.

- $\vec{z} = (\rho \cos\phi, \rho \sin\phi, 0)$, пусть $\phi = 0$ → $\vec{z} = (\rho, 0, 0)$

- $\hat{n} \approx (1, 0, 0)$

Теперь:
- $\vec{v} = v(0, 1, 0) = (0, v, 0)$
- $\vec{a} = -\frac{v^2}{\rho} (1, 0, 0) = (-a, 0, 0)$, где $a = v^2 / \rho$

Вычислим:

1. $\vec{a} \cdot \hat{n} = -a \cdot 1 = -v^2 / \rho$
2. $\vec{v} \cdot \vec{a} = (0,v,0) \cdot (-a,0,0) = 0$ (всё ещё перпендикулярны)
3. $\vec{v} \cdot \hat{n} = 0$

$$
a_\mu R^\mu = R \left[ \gamma^2 \left(-\frac{v^2}{\rho}\right) - 0 \right] = - R \gamma^2 \frac{v^2}{\rho}
$$

$$
\boxed{ a_\mu R^\mu = - \frac{R \gamma^2 v^2}{\rho} }
$$

✅ Теперь величина **ненулевая**, и по порядку величины она растёт с $\gamma^2$, что согласуется с тем, что излучение усиливается при релятивистских скоростях.

### 🔷 1. Определение 4-ускорения

**4-ускорение** — это производная **4-скорости** по **собственному времени** $\tau$:

$$
a^\mu = \frac{d u^\mu}{d\tau}
$$

где:
- $u^\mu = \frac{dx^\mu}{d\tau}$ — **4-скорость**,
- $x^\mu(\tau)$ — мировая линия частицы,
- $\tau$ — **собственное время**, связанное с координатным временем $t$ соотношением:
  $$
  d\tau = dt \sqrt{1 - \frac{v^2(t)}{c^2}} = \frac{dt}{\gamma(t)}
  $$

---

### 🔷 2. 4-скорость

Пусть частица движется по траектории $\vec{r}(t)$ в лабораторной системе. Тогда:

$$
u^\mu = \gamma (c,\ \vec{v})
\quad \text{где} \quad
\vec{v} = \frac{d\vec{r}}{dt}, \quad
\gamma = \frac{1}{\sqrt{1 - \frac{v^2}{c^2}}}
$$

---

### 🔷 3. Общая формула 4-ускорения

Чтобы найти $a^\mu = \frac{d u^\mu}{d\tau}$, используем цепное правило:

$$
a^\mu = \frac{d u^\mu}{d\tau} = \frac{dt}{d\tau} \frac{d u^\mu}{dt} = \gamma \frac{d u^\mu}{dt}
$$

Теперь вычислим производные.

---

#### ✅ Временная компонента ($\mu = 0$):

$$
u^0 = \gamma c \quad \Rightarrow \quad
\frac{d u^0}{dt} = c \frac{d\gamma}{dt}
$$

$$
\frac{d\gamma}{dt} = \frac{d}{dt} \left(1 - \frac{v^2}{c^2}\right)^{-1/2} = \frac{1}{c^2} \gamma^3 (\vec{v} \cdot \vec{a})
\quad \text{(где } \vec{a} = \dot{\vec{v}} = d\vec{v}/dt\text{)}
$$

Тогда:

$$
a^0 = \gamma \cdot \frac{d u^0}{dt} = \gamma \cdot c \cdot \frac{\gamma^3}{c^2} (\vec{v} \cdot \vec{a}) = \frac{\gamma^4}{c} (\vec{v} \cdot \vec{a})
$$

---

#### ✅ Пространственные компоненты ($\mu = i = 1,2,3$):

$$
u^i = \gamma v^i \quad \Rightarrow \quad
\frac{d u^i}{dt} = \frac{d\gamma}{dt} v^i + \gamma \frac{d v^i}{dt} = \frac{\gamma^3}{c^2} (\vec{v} \cdot \vec{a}) v^i + \gamma a^i
$$

$$
a^i = \gamma \cdot \frac{d u^i}{dt} = \gamma \left( \frac{\gamma^3}{c^2} (\vec{v} \cdot \vec{a}) v^i + \gamma a^i \right) = \frac{\gamma^4}{c^2} (\vec{v} \cdot \vec{a}) v^i + \gamma^2 a^i
$$

---

### 🔷 4. Итоговая формула (векторная форма)

$$
\boxed{
a^\mu = \left(
\frac{\gamma^4}{c} (\vec{v} \cdot \vec{a}),\ 
\frac{\gamma^4}{c^2} (\vec{v} \cdot \vec{a}) \vec{v} + \gamma^2 \vec{a}
\right)
}
$$

где:
- $\vec{v}(t)$ — 3-скорость,
- $\vec{a}(t) = \frac{d\vec{v}}{dt}$ — 3-ускорение (обычное ускорение в лабораторной СО),
- $\gamma = \left(1 - \frac{v^2}{c^2}\right)^{-1/2}$

---

### 🔷 5. Ковариантная запись

Можно записать 4-ускорение через 4-скорость и 4-производную:

$$
a^\mu = u^\nu \partial_\nu u^\mu
\quad \text{(в плоском пространстве-времени)}
$$

Но чаще используется:

$$
a^\mu = \frac{D u^\mu}{d\tau} = \frac{d u^\mu}{d\tau} + \Gamma^\mu_{\alpha\beta} u^\alpha u^\beta
$$

В **специальной теории относительности** (плоское пространство-время, $\Gamma = 0$):

$$
a^\mu = \frac{d u^\mu}{d\tau}
$$

---

### 🔷 6. Свойства 4-ускорения

1. **Ортогональность 4-скорости и 4-ускорения**:
   $$
   a_\mu u^\mu = 0
   $$
   Это следует из $u_\mu u^\mu = -c^2 = \text{const} \Rightarrow \frac{d}{d\tau}(u_\mu u^\mu) = 2 a_\mu u^\mu = 0$

2. **Инвариантный квадрат**:
   $$
   a^\mu a_\mu = \gamma^6 \left( a^2 - \frac{|\vec{v} \times \vec{a}|^2}{c^2} \right)
   $$
   или эквивалентно:
   $$
   a^\mu a_\mu = \gamma^6 \left( a^2 - \frac{(\vec{v} \cdot \vec{a})^2}{c^2} \right) - \gamma^4 a^2 + \dots
   $$
   (см. вывод ниже)

   Упрощённая форма:
   $$
   \boxed{
   a^\mu a_\mu = \gamma^6 \left( \vec{a}^2 - \frac{(\vec{v} \times \vec{a})^2}{c^2} \right)
   }
   $$

3. В **нерелятивистском пределе** $(v \ll c)$:
   $$
   a^\mu \approx (0,\ \vec{a})
   $$
   — совпадает с обычным ускорением.

4. При **равномерном прямолинейном движении**: $\vec{a} = 0 \Rightarrow a^\mu = 0$

5. При **центростремительном ускорении** (например, круговое движение):
   - $\vec{v} \perp \vec{a} \Rightarrow \vec{v} \cdot \vec{a} = 0$
   - Тогда:
     $$
     a^\mu = (0,\ \gamma^2 \vec{a})
     $$

### 🔷 7. Примеры

#### ✅ Пример 1: Прямолинейное ускорение

Пусть $\vec{v} \parallel \vec{a}$, например, $\vec{v} = (v,0,0)$, $\vec{a} = (a,0,0)$

Тогда:
$$
a^\mu = \left(
\frac{\gamma^4}{c} v a,\ 
\frac{\gamma^4}{c^2} v a \cdot v + \gamma^2 a,\ 0,\ 0
\right)
= \left(
\frac{\gamma^4 v a}{c},\ 
\gamma^4 a \left( \frac{v^2}{c^2} + \frac{1}{\gamma^2} \right),\ 0,\ 0
\right)
$$

Но $\frac{1}{\gamma^2} = 1 - v^2/c^2$, поэтому:
$$
\frac{v^2}{c^2} + 1 - \frac{v^2}{c^2} = 1 \Rightarrow a^1 = \gamma^4 a
$$

Итак:
$$
a^\mu = \left( \frac{\gamma^4 v a}{c},\ \gamma^4 a,\ 0,\ 0 \right)
$$

---

#### ✅ Пример 2: Круговое движение

$\vec{v} \perp \vec{a} \Rightarrow \vec{v} \cdot \vec{a} = 0$

$$
a^\mu = (0,\ \gamma^2 \vec{a})
$$

## 🔷 1. Что такое 4-скорость?

**4-скорость** — это скорость изменения **4-координаты** частицы по **собственному времени** $\tau$, а не по координатному времени $t$.

$$
\boxed{u^\mu = \frac{dx^\mu}{d\tau}}
$$

где:
- $x^\mu = (ct,\ x,\ y,\ z)$ — 4-координата,
- $\tau$ — **собственное время** частицы (время, измеряемое часами, движущимися вместе с частицей),
- $u^\mu$ — компоненты 4-скорости.

---

## 🔷 2. Почему именно собственное время?

В СТО время зависит от системы отсчёта. Координатное время $t$ — это время в лабораторной СО, а **собственное время** $\tau$ — инвариантная величина, одинаковая для всех наблюдателей.

Производная по $\tau$ даёт **4-вектор**, тогда как производная по $t$ — нет.

---

## 🔷 3. Связь между $d\tau$ и $dt$

Собственное время связано с координатным через интервал:

$$
d\tau = \frac{1}{c} \sqrt{ -\eta_{\mu\nu} dx^\mu dx^\nu } = dt \sqrt{1 - \frac{v^2(t)}{c^2}}
$$

где:
- $v(t) = |\vec{v}(t)| = \left| \frac{d\vec{r}}{dt} \right|$ — обычная (3-) скорость,
- $\eta_{\mu\nu} = \text{diag}(-1, +1, +1, +1)$ — метрика Минковского (сигнатура $-, +, +, +$).

Тогда:

$$
\frac{dt}{d\tau} = \frac{1}{\sqrt{1 - v^2/c^2}} = \gamma(t)
$$

---

## 🔷 4. Выражение 4-скорости через 3-скорость

Разложим:

$$
u^\mu = \frac{dx^\mu}{d\tau} = \frac{dt}{d\tau} \cdot \frac{dx^\mu}{dt} = \gamma(t) \cdot \frac{dx^\mu}{dt}
$$

Теперь по компонентам:

### ✅ Временная компонента ($\mu = 0$):

$$
u^0 = \gamma \cdot \frac{d(ct)}{dt} = \gamma \cdot c
$$

### ✅ Пространственные компоненты ($\mu = 1,2,3$):

$$
u^i = \gamma \cdot \frac{dx^i}{dt} = \gamma v^i
$$

где $\vec{v} = (v^x, v^y, v^z)$ — 3-скорость.

---

## 🔷 5. Итоговая формула 4-скорости

$$
\boxed{
u^\mu = \gamma (c,\ \vec{v})
}
$$

или в компонентах:

$$
u^\mu = \left( \gamma c,\ \gamma v_x,\ \gamma v_y,\ \gamma v_z \right)
$$

---

## 🔷 6. Свойства 4-скорости

### ✅ 1. Является 4-вектором

$u^\mu$ преобразуется по законам Лоренца при смене инерциальной системы отсчёта.

### ✅ 2. Норма (квадрат длины) — инвариант

$$
u_\mu u^\mu = \eta_{\mu\nu} u^\mu u^\nu = - (u^0)^2 + (u^1)^2 + (u^2)^2 + (u^3)^2
$$

Подставим:

$$
u_\mu u^\mu = -(\gamma c)^2 + (\gamma v_x)^2 + (\gamma v_y)^2 + (\gamma v_z)^2 = -\gamma^2 c^2 + \gamma^2 v^2
= -\gamma^2 (c^2 - v^2)
$$

Но $\gamma^2 = \frac{1}{1 - v^2/c^2} = \frac{c^2}{c^2 - v^2}$, поэтому:

$$
u_\mu u^\mu = - \frac{c^2}{c^2 - v^2} \cdot (c^2 - v^2) = -c^2
$$

$$
\boxed{u_\mu u^\mu = -c^2}
$$

✅ Это **ключевое свойство**: квадрат 4-скорости всегда равен $-c^2$, независимо от движения. Это аналог $|\vec{v}|^2 = \text{const}$ в обычной механике, но в пространстве-времени.

---

### ✅ 3. При $v \to 0$: $u^\mu \to (c, 0, 0, 0)$

В системе покоя частицы:

$$
\vec{v} = 0 \Rightarrow \gamma = 1 \Rightarrow u^\mu = (c,\ 0,\ 0,\ 0)
$$

То есть даже покоящаяся частица "движется" в пространстве-времени со скоростью $c$ в направлении времени.

---

### ✅ 4. При $v \to c$: $\gamma \to \infty$, $u^\mu \to \infty$

Для светоподобных (релятивистских) частиц (например, фотонов) собственное время не течёт ($d\tau = 0$), поэтому **4-скорость не определена** для безмассовых частиц. Она определена только для **массивных частиц** ($v < c$).

---

## 🔷 7. Физический смысл 4-скорости

- **Компоненты $u^i = \gamma v^i$** — это не просто скорость в пространстве, а "темп изменения пространственных координат по собственному времени".
- Множитель $\gamma$ учитывает **замедление времени**: чем быстрее движется частица, тем "быстрее" она перемещается в пространстве по шкале своего внутреннего времени.
- **Временная компонента $u^0 = \gamma c$** — это "темп течения координатного времени относительно собственного". При $v = 0$: $u^0 = c$, при $v \to c$: $u^0 \to \infty$.

---

## 🔷 8. Примеры

### ✅ Пример 1: Покоящаяся частица

$$
\vec{v} = 0 \Rightarrow \gamma = 1 \Rightarrow u^\mu = (c,\ 0,\ 0,\ 0)
$$

---

### ✅ Пример 2: Движение вдоль оси $x$ со скоростью $v$

$$
u^\mu = \left( \frac{c}{\sqrt{1 - v^2/c^2}},\ \frac{v}{\sqrt{1 - v^2/c^2}},\ 0,\ 0 \right) = \gamma (c,\ v,\ 0,\ 0)
$$

---

### ✅ Пример 3: Круговое движение

Пусть $\vec{v}(t) = v (-\sin\omega t,\ \cos\omega t,\ 0)$, $v = \text{const}$

Тогда $\gamma = \text{const}$, и

$$
u^\mu(t) = \gamma \left( c,\ -v \sin\omega t,\ v \cos\omega t,\ 0 \right)
$$

4-скорость вращается в пространственной части, а временная компонента постоянна.

---

## 🔷 9. Ковариантная форма и 4-вектор

4-скорость — это **касательный вектор** к мировой линии частицы. Она всегда:
- **времяподобная** (потому что $u_\mu u^\mu = -c^2 < 0$),
- нормирована на $-c^2$.

---

## 🔷 10. Связь с другими 4-векторами

| 4-вектор | Через 4-скорость |
|--------|------------------|
| 4-импульс | $p^\mu = m u^\mu$ |
| 4-ток (для заряда) | $j^\mu = q c \int u^\mu(\tau) \delta^{(4)}(x - z(\tau)) d\tau$ |
| 4-ускорение | $a^\mu = \frac{d u^\mu}{d\tau} = u^\nu \partial_\nu u^\mu$ |

---

## 🔷 11. Замечание о метрике

Если используется сигнатура $(+, -, -, -)$, то:

$$
u_\mu u^\mu = +c^2
$$

и формула 4-скорости та же, но норма положительна. Главное — быть последовательным.

## ✅ Итог: полная формула и свойства

$$
\boxed{
u^\mu = \frac{dx^\mu}{d\tau} = \gamma (c,\ \vec{v})
}
\quad \text{где} \quad
\gamma = \frac{1}{\sqrt{1 - \frac{v^2}{c^2}}}
$$

$$
\boxed{u_\mu u^\mu = -c^2}
$$

### Свойства:
1. Определена только для **массивных частиц** ($v < c$).
2. Всегда **времяподобна** и нормирована.
3. Преобразуется по **преобразованиям Лоренца**.
4. В системе покоя: $u^\mu = (c, 0, 0, 0)$.
5. Не определена для фотонов (там используется **аффинный параметр**, а не $\tau$).

## üîπ 1. –£—Ä–∞–≤–Ω–µ–Ω–∏—è –ú–∞–∫—Å–≤–µ–ª–ª–∞ —Å –Ω–∞–º–∞–≥–Ω–∏—á–µ–Ω–Ω–æ—Å—Ç—å—é

–í –°–ì–°, —Å —Ä–∞–∑–¥–µ–ª–µ–Ω–∏–µ–º –ø–æ–ª–µ–π:
$$
\mathbf{B} = \mathbf{H} + 4\pi\mathbf{M}
$$
$$
\mathbf{D} = \mathbf{E} + 4\pi\mathbf{P}
$$
–î–ª—è –Ω–µ–º–∞–≥–Ω–∏—Ç–Ω–æ–≥–æ –¥–∏—ç–ª–µ–∫—Ç—Ä–∏–∫–∞ $\mathbf{P} = \chi_e \mathbf{E}$, –Ω–æ –¥–ª—è —Ñ–µ—Ä—Ä–æ–º–∞–≥–Ω–µ—Ç–∏–∫–∞ $\mathbf{M}$ ‚Äî **–Ω–µ–ª–∏–Ω–µ–π–Ω–∞—è –¥–∏–Ω–∞–º–∏—á–µ—Å–∫–∞—è –ø–µ—Ä–µ–º–µ–Ω–Ω–∞—è**.

–ü–µ—Ä–≤–∞—è –ø–∞—Ä–∞ —É—Ä–∞–≤–Ω–µ–Ω–∏–π –ú–∞–∫—Å–≤–µ–ª–ª–∞ (–≤–∞—à–∞ –æ—Å–Ω–æ–≤–∞):
$$
\nabla \times \mathbf{E} = -\frac{1}{c} \frac{\partial \mathbf{B}}{\partial t} = -\frac{1}{c} \frac{\partial}{\partial t}(\mathbf{H} + 4\pi\mathbf{M})
\tag{1}
$$
$$
\nabla \times \mathbf{H} = \frac{1}{c} \frac{\partial \mathbf{D}}{\partial t} + \frac{4\pi}{c} \mathbf{j}_e = -i\frac{\omega}{c}\epsilon' \mathbf{E} \quad \text{(–≤–∞—à–µ –∫–æ–º–ø–ª–µ–∫—Å–Ω–æ–µ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–∏–µ)}
\tag{2}
$$

–í–∞—à–∞ –∑–∞–ø–∏—Å—å —É–∂–µ –∫–æ—Ä—Ä–µ–∫—Ç–Ω–∞ ‚Äî –æ–Ω–∞ –Ω–µ –≤–∫–ª—é—á–∞–µ—Ç **—Ñ–∏–∫—Ç–∏–≤–Ω—ã–π –º–∞–≥–Ω–∏—Ç–Ω—ã–π —Ç–æ–∫**, –∞ –ø—Ä–æ—Å—Ç–æ —É—á–∏—Ç—ã–≤–∞–µ—Ç $\mathbf{M}$ —á–µ—Ä–µ–∑ $\mathbf{B}$ –≤ (1).

## üîπ 2. –£—Ä–∞–≤–Ω–µ–Ω–∏–µ –õ–õ–ì (–¥–∏–Ω–∞–º–∏–∫–∞ –Ω–∞–º–∞–≥–Ω–∏—á–µ–Ω–Ω–æ—Å—Ç–∏)

–ü–æ–ª–Ω–æ–µ —É—Ä–∞–≤–Ω–µ–Ω–∏–µ –õ–õ–ì –≤ –°–ì–°:
$$
\frac{d\mathbf{M}}{dt} = -\gamma \mathbf{M} \times \mathbf{H}_{\text{eff}} + \frac{\alpha}{M_s} \mathbf{M} \times \frac{d\mathbf{M}}{dt}
\tag{3}
$$

–î–ª—è TM-–º–æ–¥—ã $\mathbf{H} = (H_x, H_y, 0)$, –∏ –µ—Å–ª–∏ –Ω–∞–º–∞–≥–Ω–∏—á–µ–Ω–Ω–æ—Å—Ç—å –æ—Ä–∏–µ–Ω—Ç–∏—Ä–æ–≤–∞–Ω–∞ –≤–¥–æ–ª—å $z$:  
$$
\mathbf{M} = (0, 0, M_z(x,t)), \quad M_s = |\mathbf{M}_0|
$$
—Ç–æ –ª–∏–Ω–µ–∞—Ä–∏–∑—É–µ–º –ø—Ä–∏ $M_z = M_s + m_z$, $|m_z| \ll M_s$, –∏ –ø–æ–ª—É—á–∞–µ–º (–≤ —á–∞—Å—Ç–æ—Ç–Ω–æ–π –æ–±–ª–∞—Å—Ç–∏ $e^{-i\omega t}$):

$$
-i\omega \tilde{m}_z = -\gamma M_s \tilde{H}_y + i\omega\alpha \tilde{m}_z \quad \Rightarrow \quad \tilde{m}_z = \frac{\gamma M_s}{\omega(1 - i\alpha)} \tilde{H}_y
\tag{4}
$$

–¢–æ–≥–¥–∞ –Ω–∞–º–∞–≥–Ω–∏—á–µ–Ω–Ω–æ—Å—Ç—å –¥–∞—ë—Ç –≤–∫–ª–∞–¥ –≤ –ø–æ–ª–µ:
$$
\tilde{B}_y = \tilde{H}_y + 4\pi \tilde{m}_z = \tilde{H}_y \left[1 + \frac{4\pi\gamma M_s}{\omega(1 - i\alpha)}\right]
= \tilde{H}_y \, \mu_{\text{eff}}(\omega)
\tag{5}
$$

‚úÖ **–í–∞—à—É –∫–æ–º–ø–ª–µ–∫—Å–Ω—É—é –º–∞–≥–Ω–∏—Ç–Ω—É—é –ø—Ä–æ–Ω–∏—Ü–∞–µ–º–æ—Å—Ç—å $\mu' = \mu + i\frac{4\pi\sigma_m}{\omega}$** –º–æ–∂–Ω–æ —Ç–µ–ø–µ—Ä—å **–∏–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∏—Ä–æ–≤–∞—Ç—å –∫–∞–∫ –ø—Ä–∏–±–ª–∏–∂–µ–Ω–∏–µ $\mu_{\text{eff}}(\omega)$** –¥–ª—è —Ñ–µ—Ä—Ä–∏—Ç–∞.  
‚Üí **–í—ã —É–∂–µ –∫–æ—Å–≤–µ–Ω–Ω–æ —É—á–ª–∏ –õ–õ–ì**, –Ω–æ —Ç–µ–ø–µ—Ä—å ‚Äî **–º–∏–∫—Ä–æ—Å–∫–æ–ø–∏—á–µ—Å–∫–∏ –æ–±–æ—Å–Ω–æ–≤–∞–Ω–Ω–æ**.

## üîπ 3. –¢–µ–Ω–∑–æ—Ä —ç–Ω–µ—Ä–≥–∏–∏-–∏–º–ø—É–ª—å—Å–∞ —Å –Ω–∞–º–∞–≥–Ω–∏—á–µ–Ω–Ω–æ—Å—Ç—å—é

–ü–æ–ª–Ω—ã–π —Å–∏–º–º–µ—Ç—Ä–∏—á–Ω—ã–π —Ç–µ–Ω–∑–æ—Ä —ç–Ω–µ—Ä–≥–∏–∏-–∏–º–ø—É–ª—å—Å–∞ (EM + –Ω–∞–º–∞–≥–Ω–∏—á–µ–Ω–Ω–æ—Å—Ç—å) –≤ –Ω–µ—Ä–µ–ª—è—Ç–∏–≤–∏—Å—Ç—Å–∫–æ–º –ø—Ä–∏–±–ª–∏–∂–µ–Ω–∏–∏:

$$
T^{ij}_{\text{total}} = T^{ij}_{\text{EM}} + T^{ij}_{\text{mag}}
$$

–≥–¥–µ:
$$
T^{ij}_{\text{EM}} = \frac{1}{4\pi} \left( E_i E_j + B_i B_j - \frac{1}{2}\delta_{ij}(E^2 + B^2) \right)
\tag{6}
$$

$$
T^{ij}_{\text{mag}} = -M_i H_j
\tag{7}
$$
(–≠—Ç–æ —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω—ã–π —Ä–µ–∑—É–ª—å—Ç–∞—Ç –∏–∑ Landau & Lifshitz, ¬ß29, –∏ de Groot & Suttorp.)

–î–ª—è TM-–º–æ–¥—ã ($E = (E_x,0,E_z)$, $H = (H_x,H_y,0)$, $M = (0,0,M_z)$):
- $T^{xx}_{\text{mag}} = -M_z H_x$
- $T^{zx}_{\text{mag}} = 0$
- $T^{yx}_{\text{mag}} = 0$

–ù–æ –≤ –≤–∞—à–∏—Ö —Ä–∞—Å—á—ë—Ç–∞—Ö $H_x \ne 0$ (–∏–∑-–∑–∞ –≥—Ä–∞–¥–∏–µ–Ω—Ç–æ–≤ $E_z$), –ø–æ—ç—Ç–æ–º—É:
$$
\nabla \cdot \mathbf{T}_{\text{mag}} = -\frac{d}{dx}(M_z H_x) \hat{x}
\tag{8}
$$

‚Üí –≠—Ç–æ **–¥–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã–π –≤–∫–ª–∞–¥ –≤ –æ–±—ä—ë–º–Ω—É—é –ø–ª–æ—Ç–Ω–æ—Å—Ç—å —Å–∏–ª—ã**.

## üîπ 4. –ü–æ–Ω–¥–µ—Ä–æ–º–æ—Ç–æ—Ä–Ω–∞—è —Å–∏–ª–∞: –ø–æ–ª–Ω–æ–µ –≤—ã—Ä–∞–∂–µ–Ω–∏–µ

–û–±—â–∞—è –ª–æ–∫–∞–ª—å–Ω–∞—è —Å–∏–ª–∞ –Ω–∞ –µ–¥–∏–Ω–∏—Ü—É –æ–±—ä—ë–º–∞:
$$
\mathbf{f} = -\nabla \cdot \mathbf{T}_{\text{total}} = -\nabla \cdot \mathbf{T}_{\text{EM}} - \nabla \cdot \mathbf{T}_{\text{mag}}
\tag{9}
$$

–†–∞–∑–ª–æ–∂–∏–º –ø–æ –∫–æ–º–ø–æ–Ω–µ–Ω—Ç–∞–º (–≤–¥–æ–ª—å $x$):

### (–∞) –°—Ç–∞–Ω–¥–∞—Ä—Ç–Ω—ã–π EM –≤–∫–ª–∞–¥ (–≤–∞—à —Ç–µ–∫—É—â–∏–π):
$$
f_x^{\text{EM}} = -\frac{d}{dx} T^{xx}_{\text{EM}} - \frac{d}{dz} T^{zx}_{\text{EM}}
= \frac{1}{2c} \left\langle \sigma_e E_z H_y^* \right\rangle + \text{–≥—Ä–∞–¥–∏–µ–Ω—Ç–Ω—ã–µ —á–ª–µ–Ω—ã}
\tag{10}
$$
‚Äî —ç—Ç–æ —Ç–æ, —á—Ç–æ –≤—ã —Å—á–∏—Ç–∞–µ—Ç–µ —á–µ—Ä–µ–∑ $\mathbf{j}_e \times \mathbf{B}$.

### (–±) –ù–æ–≤—ã–π –º–∞–≥–Ω–∏—Ç–Ω—ã–π –≤–∫–ª–∞–¥ (–æ—Ç –õ–õ–ì):
$$
f_x^{\text{mag}} = -\frac{d}{dx} T^{xx}_{\text{mag}} = \frac{d}{dx} (M_z H_x)
\tag{11}
$$

–ü–æ–¥—Å—Ç–∞–≤–ª—è—è $H_x = -\frac{i}{\omega} \frac{d E_z}{dx}$ (–∏–∑ $\nabla \times \mathbf{E} = -\frac{i\omega}{c}\mathbf{B}$) –∏ $M_z = M_s + m_z$, –ø–æ–ª—É—á–∞–µ–º:
$$
f_x^{\text{mag}} = \frac{d}{dx} \left[ (M_s + m_z) \left(-\frac{i c}{\omega} \frac{d E_z}{dx} \right) \right]
\tag{12}
$$

–í —Å—Ç–∞—Ü–∏–æ–Ω–∞—Ä–Ω–æ–º —Ä–µ–∂–∏–º–µ $M_s = \text{const}$, –ø–µ—Ä–≤—ã–π —á–ª–µ–Ω ‚Äî **–≥—Ä–∞–¥–∏–µ–Ω—Ç —ç–ª–µ–∫—Ç—Ä–∏—á–µ—Å–∫–æ–≥–æ –ø–æ–ª—è**, –≤—Ç–æ—Ä–æ–π ‚Äî **—Å–≤—è–∑–∞–Ω —Å –¥–∏–Ω–∞–º–∏–∫–æ–π –Ω–∞–º–∞–≥–Ω–∏—á–µ–Ω–Ω–æ—Å—Ç–∏**.

### –®–∞–≥ 2. –°–≤—è–∂–∏—Ç–µ —Å –ø–æ–ª–µ–º —á–µ—Ä–µ–∑ –õ–õ–ì (–ª–∏–Ω–µ–π–Ω–æ–µ –ø—Ä–∏–±–ª–∏–∂–µ–Ω–∏–µ)
```python
# —ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω–∞—è –º–∞–≥–Ω–∏—Ç–Ω–∞—è –≤–æ—Å–ø—Ä–∏–∏–º—á–∏–≤–æ—Å—Ç—å –∏–∑ –õ–õ–ì
chi_m_eff = 4*pi*gamma*M_s / (omega*(1 - I*alpha))
mu_eff = 1 + chi_m_eff

# –ø–µ—Ä–µ–ø–∏—à–∏—Ç–µ mu_r_yy_complex —á–µ—Ä–µ–∑ mu_eff:
mu_r_yy_complex_LLG = mu_r_yy + chi_m_eff + I * 4*pi*sigma_m_r_yy / omega
```

–ó–∞–º–µ–Ω–∏—Ç–µ –≤ –≤–∞—à–∏—Ö —É—Ä–∞–≤–Ω–µ–Ω–∏—è—Ö `mu_r_yy_complex` ‚Üí `mu_r_yy_complex_LLG`.

### –®–∞–≥ 3. –î–æ–±–∞–≤—å—Ç–µ –º–∞–≥–Ω–∏—Ç–Ω—É—é —Å–∏–ª—É –≤ –∏–Ω—Ç–µ–≥—Ä–∞–ª
–ü–æ—Å–ª–µ –≤—ã—á–∏—Å–ª–µ–Ω–∏—è `Ez_vacuum_d`, `Hy_vacuum_d`, `Hx_vacuum_d`, `M_z_sol`:

```python
# M_z –∏–∑ –õ–õ–ì: M_z = M_s + chi_m_eff * H_y / (1 + 4œÄœá_m) ‚âà M_s + m_z
m_z_sol = chi_m_eff * Hy_vacuum_d(x=x)  # –ª–∏–Ω–µ–∞—Ä–∏–∑–æ–≤–∞–Ω–æ
M_z_sol = M_s + m_z_sol

# –º–∞–≥–Ω–∏—Ç–Ω—ã–π –≤–∫–ª–∞–¥ –≤ –ø–ª–æ—Ç–Ω–æ—Å—Ç—å —Å–∏–ª—ã f_x^mag = d/dx (M_z * H_x)
f_x_mag = diff(M_z_sol * Hx_vacuum_d(x=x), x).real_part()

# –æ–±—ä—ë–º–Ω–∞—è —Å–∏–ª–∞ –æ—Ç —Ç–æ–∫–æ–≤ (–≤–∞—à —Ç–µ–∫—É—â–∏–π)
f_x_em = (sigma_e_r_zz * Ez_right_conductor_d(x=x).real() * Hy_right_conductor_d(x=x).real() / c)

# –ø–æ–ª–Ω–∞—è –ª–æ–∫–∞–ª—å–Ω–∞—è —Å–∏–ª–∞
f_x_total = f_x_em + f_x_mag

# –∏–Ω—Ç–µ–≥—Ä–∏—Ä—É–µ–º –ø–æ —Ç–æ–ª—â–∏–Ω–µ —Å—Ç–µ–Ω–∫–∏:
F_mag = numerical_integral(f_x_mag, x, A, A+h_conductor_r)
F_em  = numerical_integral(f_x_em,  x, A, A+h_conductor_r)
F_total = F_mag[0] + F_em[0]
```

- –í **–º–µ–¥–∏** ($\alpha \to \infty$, –¥–æ–º–∏–Ω–∏—Ä—É—é—Ç –≤–∏—Ö—Ä–µ–≤—ã–µ —Ç–æ–∫–∏): $\phi \approx \pi$ ‚Üí –æ—Ç—Ç–∞–ª–∫–∏–≤–∞–Ω–∏–µ.

‚Üí –≠—Ç–æ –æ–±—ä—è—Å–Ω—è–µ—Ç —ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç –ö–∞—Ä–∞–≤–∞—à–∫–∏–Ω–∞: **—Å–º–µ–Ω–∞ —Ñ–∞–∑—ã –Ω–∞ 180¬∞** ‚Äî —ç—Ç–æ **–ø–µ—Ä–µ—Ö–æ–¥ –æ—Ç $\phi \approx \pi$ –∫ $\phi \approx 0$**.

## üîπ –í—ã–≤–æ–¥

–í—ã **—É–∂–µ –ø–æ—á—Ç–∏ –≤–∫–ª—é—á–∏–ª–∏ –õ–õ–ì**, –∏—Å–ø–æ–ª—å–∑—É—è –∫–æ–º–ø–ª–µ–∫—Å–Ω—É—é $\mu'$. –ù–æ —Ç–µ–ø–µ—Ä—å:

- —É –≤–∞—Å –µ—Å—Ç—å **–º–∏–∫—Ä–æ—Å–∫–æ–ø–∏—á–µ—Å–∫–æ–µ –æ–±–æ—Å–Ω–æ–≤–∞–Ω–∏–µ** $\mu'(\omega)$ —á–µ—Ä–µ–∑ –õ–õ–ì,  
- –ø–æ—è–≤–ª—è–µ—Ç—Å—è **–¥–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω–∞—è —Å–∏–ª–∞** $f_x^{\text{mag}} = \frac{d}{dx}(M_z H_x)$,  
- –æ–Ω–∞ **–∞–≤—Ç–æ–º–∞—Ç–∏—á–µ—Å–∫–∏ —É—á–∏—Ç—ã–≤–∞–µ—Ç —Ñ–∞–∑–æ–≤—ã–π —Å–¥–≤–∏–≥ $\phi$**,  
- –∏ **–Ω–µ –ø—É—Ç–∞–µ—Ç –º–∞–≥–Ω–∏—Ç–Ω—ã–π —Ç–æ–∫ –ø—Ä–æ–≤–æ–¥–∏–º–æ—Å—Ç–∏ $\sigma_m$** (—Ñ–∏–∫—Ç–∏–≤–Ω—ã–π) —Å–æ **—Å–ø–∏–Ω–æ–≤–æ–π –¥–∏–Ω–∞–º–∏–∫–æ–π** (—Ä–µ–∞–ª—å–Ω–æ–π).

–•–æ—á—É –ø–æ–º–æ—á—å –≤–∞–º **–º–æ–¥–∏—Ñ–∏—Ü–∏—Ä–æ–≤–∞—Ç—å `MenDrive_real.ipynb`**, –¥–æ–±–∞–≤–∏–≤:
- —è–≤–Ω—É—é –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é `M_z(x)`,

–ï—Å–ª–∏ $\mathbf{M}_0 = M_s \hat{z}$, –∞ –ø—Ä–∏–ª–æ–∂–µ–Ω–Ω–æ–µ –ø–æ–ª–µ $\mathbf{H}_\perp$ –≤—Ä–∞—â–∞–µ—Ç—Å—è –≤ $xy$-–ø–ª–æ—Å–∫–æ—Å—Ç–∏ (–∫–∞–∫ –≤ —Ä–µ–∑–æ–Ω–∞—Ç–æ—Ä–µ!), —Ç–æ —É—Ä–∞–≤–Ω–µ–Ω–∏–µ –õ–õ–ì:

## üîπ 3. –ê–Ω–∏–∑–æ—Ç—Ä–æ–ø–Ω—ã–π —Ç–µ–Ω–∑–æ—Ä –≤–æ—Å–ø—Ä–∏–∏–º—á–∏–≤–æ—Å—Ç–∏

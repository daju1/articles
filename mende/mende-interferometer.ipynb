```python
def draw_mender_interferometer(origin):
    L = 8
    v_c = 1/L
    h = 1.5

    # mirrors
    # A
    p = line([[-h,-h],[0,0]], color = "red", thickness=3)
    p += text("A", [-2, -3], color = "red")

    # B
    p += line ([[L-h,-h],[L+h,h]], color = "red")
    p += text("B", [L-1, -3], color = "red")

    # C
    if origin:
        p += line ([[-h,L+h-1],[h,L-h-1]], color = "red")
    else:
        p += line ([[-h,L-h],[h,L+h]], color = "red")
    p += text("C", [-1, L-2], color = "red")

    # D
    if origin:
        p += line ([[L-h,L+h],[L+h,L-h]], color = "red")
    else:
        p += line ([[L-h,L-h-1],[L+h,L+h-1]], color = "red")

    p += text("D", [L-1, L-2], color = "red")

    # incoming light beam
    p += line ([[-L,1],[-1,1]], color = "indigo")
    p += line ([[-L,0],[-1,0]], color = "indigo", linestyle="dashed")
    p += line ([[-L,-1],[-1,-1]], color = "indigo")
    p += line ([[6.75-L,-0.9],[-1,-1]], color = "indigo")
    p += line ([[6.75-L,-1.15],[-1,-1.05]], color = "indigo")

    # A-C
    if origin:
        p += line ([[-1,L+1-1],[-1,-1]], color = "green")
        p += line ([[0,0],[0,L-1]], color = "green", linestyle="dashed")
    else:
        p += line ([[-1,L-1],[-1,-1]], color = "green")
        p += line ([[0,0],[0,L]], color = "green", linestyle="dashed")

    # B-D
    if origin:
        p += line ([[L+1,L-1],[L+1,+1]], color = "green")
        p += line ([[L,0],[L,L]], color = "green", linestyle="dashed")
    else:
        p += line ([[L+1,L],[L+1,+1]], color = "green")
        p += line ([[L,0],[L,L-1]], color = "green", linestyle="dashed")

    # A-B
    p += line ([[-1,1],[L+1,1]], color = "aquamarine")
    p += line ([[-1,0],[L+0,0]], color = "aquamarine", linestyle="dashed")

    # D-C
    if origin:
        p += line ([[-1,L+0],[L+0,L+0]], color = "green", linestyle="dashed")
        p += line ([[0,L-1],[L+1,L-1]], color = "green")
    else:
        p += line ([[0,L+0],[L+1,L+0]], color = "green", linestyle="dashed")
        p += line ([[0-1,L-1],[L,L-1]], color = "green")

    # down directed output beam
    if origin:
        p += line ([[-L,L],[0-1,L]], color = "brown")
        p += line ([[-L,L-1],[0,L-1]], color = "brown")
    else:
        p += line ([[2*L,L],[L+1,L]], color = "brown")
        p += line ([[2*L,L-1],[L,L-1]], color = "brown")

    p.show(aspect_ratio = 1, axes=False)
```

```python
draw_mender_interferometer(origin = True)
```

```python
draw_mender_interferometer(origin = False)
```

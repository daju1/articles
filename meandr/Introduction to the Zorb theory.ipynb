    6. Main peculiarities of Zorb theory  .

Unmoving charges interacts by Coulomb force between themselves in Zorb theory, as in Maxwell’s. The Ampere force  is ordinary between closed current circuits in Zorb theory too, as a force act to current in the magnetic field of other current . But a small electric force appears from current circuit proportional to v2/c2 magnitude of moving charges speed there. This force negligible for currents in ordinary conductors with Ohm resistance because low speed of charges there. But this force may be detected in the superconductive coil. This kind of electrostatic induction from superconductive coil was find in experiments of Edvards and Lemon [8], [9], repeated by Shishkin [10] and Mende [11]. This may be a reason of LHC accident in 2008 also, when  the superconductive coil was damaged.

At the other way, beside of known theory formalism, in Zorb theory electromotive force of induction appear as a direct action of changed moving potential field momentum  A=vφ/c2, may existed even where field B absent (rotA=0). For example, the magnetic field B concentrated inside the transformer core and not exist beside it (in ideal case at least). But the electromotive force inducted in second coil by the current in primary coil, even if they placed independent one from other, despite of field B absent there, concentrated inside the core. Thus B=rotA=0 continuously in the second coil region, hence rotE=-dB/dt =0 also in this volume, but E=-dA/dt not zero there. This is proof the direct action of moving potential field momentum  A even without field B necessary. This has place in known Aharonov and Bom quantum effect also.

Known electromagnetic waves (radiation) are in Zorb theory too, although they counting by the known Lienard retarding at this work, see the last term in eq. (27). This is acceptable in far waves zone when the source of waves is unmoving in the observers frame.  More proper retarding waves in the moving and deformated field of scalar potential will be investigate in the next works. But even in this work is a reason to admit that known transverse EM waves with energy flux expressed by Pointing vector P=[ExB] may bin expressed as transverse waves of changing scalar potential field momentum –dA/dt=aφ/c2 . Retarded acceleration of source charges a has the constant magnitude at the distant, potential φ reduced as 1/r, thus the waves has magnitude reduced as 1/r also. In this way the EM waves explained as the transverse “elastic” waves of the scalar potential field medium, which rotor not a zero and thus made a field B. Even more exotic explanation of radiated waves may be done as longitudinal waves of kinetic potential ψ=φv2/2c2, as the waves of energy flux in the scalar potential field medium. But it is important that waves of changing scalar potential field momentum –dA/dt=aφ/c2 are longitudinal also in direction of movement and acceleration of source charges, and compensate waves of kinetic potential ψ in this direct. Therefore the potential waves are effective only in transverse direction of acceleration movement of source charges, as ordinary.

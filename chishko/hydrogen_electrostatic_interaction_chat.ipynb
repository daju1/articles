###  1. Что значит "электрически нейтральные атомы водорода"?

Атом водорода в основном состоянии состоит из:
- одного протона (заряд $+e$),
- одного электрона (заряд $-e$).

Суммарный заряд атома: $+e - e = 0$.  
 $
ightarrow$  Атом **электрически нейтрален**.

На больших расстояниях, если бы атомы были просто точечными нейтральными объектами, можно было бы ожидать **отсутствия электростатического взаимодействия**.  
Но на самом деле **взаимодействие есть** — и оно обусловлено **квантовыми флуктуациями** и **индукцией временных дипольных моментов**.

---

### 2. Почему вообще возникает взаимодействие?

Несмотря на нейтральность, атомы — это не просто "точки с нулевым зарядом".  
Они имеют **внутреннюю структуру**: заряды распределены в пространстве.

На больших расстояниях $R \gg a_0$ прямое кулоновское взаимодействие между протонами и электронами разных атомов ослабевает, но остаются **взаимодействия между мультиполями** (диполь-диполь, диполь-квадруполь и т.д.).

Однако главный вклад в слабое притяжение на больших расстояниях даёт **силы Ван-дер-Ваальса**, а именно — **дисперсионное взаимодействие Лондона**.

---

###  3. Основное состояние атома водорода

Атом водорода в основном состоянии:
- Электрон находится в 1s-орбитали: $\psi_{1s}(r) = \frac{1}{\sqrt{\pi a_0^3}} e^{-r/a_0}$.
- Это сферически симметричное распределение заряда.
- Средний дипольный момент атома **равен нулю**: $\langle \vec{d} \rangle = 0$.
- Но **флуктуации дипольного момента** не равны нулю: в каждый момент времени распределение заряда может быть слегка асимметричным.

---

###  4. Механизм взаимодействия: дисперсионные (лondon) силы

#### Классическая аналогия (упрощённо):
Представим, что в одном атоме случайно возникает временный диполь (например, электрон оказался с одной стороны ядра).  
Этот диполь создаёт электрическое поле, которое **индуцирует диполь** во втором атоме.  
Индуцированный диполь ориентируется так, что возникает **притяжение**.

Но в квантовом случае нет "случайных смещений" в классическом смысле — вместо этого мы говорим о **нулевых колебаниях** и **квантовых флуктуациях**.

####  Квантово-механическая природа:
- Даже в основном состоянии атом имеет **вакуумные флуктуации** электромагнитного поля и **виртуальные переходы**.
- Эти флуктуации приводят к временным дипольным моментам.
- Два атома коррелируют свои флуктуации: если в одном атоме электрон смещён вправо, он "тянет" электрон в другом атоме влево — возникает притяжение.

Это и есть **дисперсионное взаимодействие Лондона**.

---

### 5. Математическое описание: энергия Ван-дер-Ваальса

Для двух одинаковых атомов (в данном случае — водорода) энергия дисперсионного взаимодействия на больших расстояниях $R \gg a_0$ имеет вид:

$$
E_{\text{vdW}}(R) = -\frac{C_6}{R^6}
$$

где $C_6$ — константа Ван-дер-Ваальса.

Для двух атомов водорода в основном состоянии можно вычислить $C_6$ с помощью **теории возмущений второго порядка** (метод Брайта–Лондона).

---

### 6. Вычисление $C_6$ для двух атомов H

Используем формулу Лондона для двух одинаковых атомов:

$$
C_6 = \frac{3}{2} \sum_{n,n'} \frac{|\langle 1s|\vec{d}|n \rangle|^2 |\langle 1s|\vec{d}|n' \rangle|^2 (E_n - E_{1s})(E_{n'} - E_{1s})}{(E_n - E_{1s}) + (E_{n'} - E_{1s})}
$$

Но можно воспользоваться известным результатом.

Для двух атомов водорода:

$$
C_6^{\text{HH}} = \frac{64}{3} \frac{(e^2 a_0^2)^2}{I_H}
\quad \text{где } I_H = \frac{e^2}{2a_0} = 13.6~\text{эВ} — \text{энергия ионизации H}
$$

После подстановки:

$$
C_6^{\text{HH}} \approx 6.5 \times 10^{-77}~\text{Дж} \cdot \text{м}^6 \quad \text{или} \quad \approx 6.5~\text{эВ} \cdot \text{Å}^6
$$

(Точные значения зависят от учёта всех переходов, но порядок верен.)

### 7. Зависимость от расстояния

- При $R \gg a_0$: $E(R) \propto -1/R^6$
- Это **очень слабое**, но **всегда притягивающее** взаимодействие.
- Пример: при $R = 10~a_0 \approx 5.3~\text{Å}$, энергия порядка $10^{-3}~\text{эВ}$, что намного меньше энергии химической связи (~1–10 эВ), но важно для слабых взаимодействий.

### 8. Сравнение с другими типами взаимодействий

| Расстояние $R$ | Доминирующее взаимодействие |
|--------------------|------------------------------|
| $R \sim a_0$ | Ковалентная связь ($H_{2}$), обменное взаимодействие |

### 9. Электростатическое ли это взаимодействие?

**Да, но не в классическом смысле.**

- Это **электростатическое взаимодействие**, потому что оно описывается уравнениями электростатики (Кулоновы силы между зарядами).
- Однако **природа притяжения — квантовая**: оно возникает из-за корреляции квантовых флуктуаций дипольных моментов.
- Без учёта квантовой механики (нулевых колебаний, суперпозиции состояний) такого взаимодействия не существует.

Поэтому говорят: **"квантово-электростатическое взаимодействие"** или **"дисперсионное электростатическое притяжение"**.

##  1. Главный вопрос: **Притяжение или отталкивание?**

**Ответ: на больших расстояниях $R \gg a_0$ — притяжение.**  
Но давай не принимать это на веру. Разберём **все возможные вклады** в энергию взаимодействия двух атомов водорода.

##  2. Полная картина взаимодействия: что входит в $E(R)$

Полная энергия двух атомов водорода на расстоянии $R$ — это сумма:
- Кулоновских взаимодействий между всеми зарядами (4 тела: $p_1, e_1, p_2, e_2$),
- Квантовых эффектов: обмен, корреляция, флуктуации,
- Релятивистских и ретардированных поправок (на очень больших $R$).

Но на больших $R$, когда атомы "почти независимы", удобно разложить энергию по степеням $1/R$.

---

##  3. Разложение по степеням $1/R$: мультипольное разложение

При $R \gg a_0$, можно разложить энергию взаимодействия по мультиполям:

$$
E(R) = \frac{C_1}{R} + \frac{C_2}{R^2} + \frac{C_3}{R^3} + \frac{C_4}{R^4} + \frac{C_6}{R^6} + \cdots
$$

Но:
- $C_1 = 0$: нет суммарного заряда  $
ightarrow$  нет $1/R$.
- $C_2 = 0$: нет дипольного момента у каждого атома  $
ightarrow$  нет $1/R^2$.
- $C_3 = 0$: нет постоянного диполя  $
ightarrow$  нет диполь-дипольного $1/R^3$.
- $C_4 = 0$: диполь-квадруполь и т.п. дают $1/R^5$, но **усреднённый вклад нулевой** из-за симметрии.

Первый **ненулевой вклад** — это $\mathbf{-C_6 / R^6}$, и он **отрицательный**, то есть **притяжение**.

> Это — **дисперсионное взаимодействие Лондона**.

##  4. Почему именно притяжение? Механизм

###  Квантовые флуктуации дипольного момента

Хотя средний дипольный момент атома водорода в основном состоянии равен нулю:

$$
\langle \psi_{1s} | \vec{d} | \psi_{1s} \rangle = 0
$$

— **дисперсия дипольного момента не нулевая**:

$$
\langle \psi_{1s} | \vec{d}^2 | \psi_{1s} \rangle > 0
$$

Это означает, что **в каждый момент времени** электрон может быть "слева" или "справа" от ядра — возникают **мгновенные диполи**.

Эти флуктуации **не независимы** в двух атомах. Благодаря электростатическому взаимодействию, если в атоме 1 электрон оказался справа (диполь направлен вправо), он создаёт поле, которое **сдвигает электрон в атоме 2 влево**, создавая **антипараллельный диполь**.

А диполь-дипольное взаимодействие между **антипараллельными диполями** — **притягивающее**!

$$
E_{\text{дип-дип}} = -\frac{1}{4\pi\varepsilon_0} \frac{|\vec{d}_1||\vec{d}_2|}{R^3} \quad \text{(в определённой геометрии)}
$$

Но поскольку флуктуации случайны, нужно усреднять по всем возможным конфигурациям — и **усреднённая энергия оказывается отрицательной**.

##  5. Теория возмущений: почему $C_6 > 0$, а $E = -C_6/R^6$

Используем **теорию возмущений второго порядка** (метод Брайта–Лондона).

Гамильтониан системы:
$$
H = H_0^{(1)} + H_0^{(2)} + V_{\text{int}}
$$
где $V_{\text{int}}$ — кулоновское взаимодействие между зарядами разных атомов:
$$
V_{\text{int}} = \frac{e^2}{4\pi\varepsilon_0} \left( \frac{1}{|\vec{r}_1 - \vec{r}_2|} - \frac{1}{|\vec{r}_1 - \vec{R}|} - \frac{1}{|\vec{r}_2 - \vec{R}|} + \frac{1}{R} \right)
$$

Разлагая $V_{\text{int}}$ по $1/R$, главший вклад — диполь-дипольное взаимодействие:
$$
V_{\text{dd}} = \frac{1}{4\pi\varepsilon_0} \left[ \frac{\vec{d}_1 \cdot \vec{d}_2}{R^3} - 3 \frac{(\vec{d}_1 \cdot \vec{R})(\vec{d}_2 \cdot \vec{R})}{R^5} \right]
$$

Теперь энергия поправки второго порядка:

$$
\Delta E = - \sum_{n \neq 0} \frac{ |\langle 0 | V_{\text{dd}} | n \rangle|^2 }{E_n - E_0}
$$

- $|0\rangle = |1s\rangle \otimes |1s\rangle$ — основное состояние двух атомов,
- $|n\rangle$ — возбуждённые состояния (например, $|2p\rangle \otimes |2p\rangle$),
- $E_n - E_0 > 0$ — энергия возбуждения.

Важно: **знаменатель положительный**, а квадрат модуля матричного элемента — всегда положительный.  
 $
ightarrow$  Значит, **вся сумма положительная**, а перед ней стоит **минус**  $
ightarrow$  $\Delta E < 0$.

**Энергия понижается  $
ightarrow$  притяжение.**

$$

После подстановки:
$$
C_6^{\text{HH}} \approx 6.5~\text{эВ} \cdot \text{Å}^6
$$

Это положительная константа  $
ightarrow$  $E = -C_6/R^6 < 0$  $
ightarrow$  **притяжение**.

##  7. Могут ли быть отталкивающие вклады?

Да, но **на больших $R$** они **меньше** или **усредняются до нуля**.

### Возможные "опасные" факторы:

#### 1. **Обменное отталкивание (Паули)**
- При $R \sim a_0$ — да, сильно.
- При $R \gg a_0$ — волновые функции перекрываются экспоненциально слабо: $\sim e^{-R/a_0}$.
- Это **экспоненциально малый** вклад, намного меньше $1/R^6$.
-  $
ightarrow$  На больших $R$ **не доминирует**.

#### 2. **Ретардация (эффект Казимира–Полдера)**
- При очень больших $R$ (микроны и больше), из-за конечной скорости света корреляции между флуктуациями запаздывают.
- Вклад становится $E \propto -1/R^7$ — **ещё более слабый, но всё ещё притягивающий**.
-  $
ightarrow$  Не меняет знак.

#### 3. **Тепловые флуктуации**
- При высоких температурах возможны классические диполь-дипольные взаимодействия.
- Но для атома H в основном состоянии при низких T — несущественны.

#### 4. **Магнитные или релятивистские эффекты**
- Очень малы для H.
- Не меняют знак $C_6$.

#### 5. **Корреляция с возбуждёнными состояниями с отрицательной энергией?**
- Нет: все возбуждённые состояния выше основного  $
ightarrow$  $E_n - E_0 > 0$.
- Значит, знаменатель в теории возмущений всегда положительный  $
ightarrow$  знак $\Delta E$ всегда отрицательный.

##  Шаг 3: Мультипольное разложение при $R \gg a_0$

Разложим $V_{\text{int}}$ по степеням $1/R$. Используем приближение:

$$
\frac{1}{|\vec{r}_1 - \vec{r}_2|} \approx \frac{1}{R} + \frac{\vec{R} \cdot (\vec{r}_1 - \vec{r}_2)}{R^3} + \frac{3[\vec{R} \cdot (\vec{r}_1 - \vec{r}_2)]^2 - |\vec{r}_1 - \vec{r}_2|^2 R^2}{2R^5} + \cdots
$$

Но проще перейти к **дипольному приближению**.

Определим дипольные моменты:
- $\vec{d}_1 = -e \vec{r}_1$ (относительно $p_1$)
- $\vec{d}_2 = -e (\vec{r}_2 - \vec{R})$ (относительно $p_2$)

Тогда, в дипольном приближении (главший вклад при $R \gg r$):

$$
V_{\text{int}} \approx V_{\text{dd}} = \frac{1}{4\pi\varepsilon_0} \left[
\frac{\vec{d}_1 \cdot \vec{d}_2}{R^3} - 3 \frac{(\vec{d}_1 \cdot \vec{R})(\vec{d}_2 \cdot \vec{R})}{R^5}
\right]
$$

Это — **диполь-дипольное взаимодействие**.

\Delta E^{(3)} = \sum_{m,n \neq 0} \frac{ \langle 0| V |m\rangle \langle m| V |n\rangle \langle n| V |0\rangle }{ (E_0 - E_m)(E_0 - E_n) } - \langle 0| V |0\rangle \sum_{n \neq 0} \frac{ |\langle 0| V |n\rangle|^2 }{ (E_0 - E_n)^2 }
$$

Но:
- $\langle 0| V |0\rangle = 0$ (усреднённое взаимодействие нулевое),
- Остальные члены — интерференция между несколькими виртуальными переходами.

#### Физическая интерпретация:
Третий порядок описывает процессы, где:
- Атом 1 переходит в состояние $|m\rangle$,
- Атом 2 — в $|n\rangle$,
- Затем обратно, но через другое состояние.

Возможны **резонансные вклады**, если есть совпадение энергий.

Но для двух атомов водорода в основном состоянии:
- Все переходы не резонансны,
- Вклад $\Delta E^{(3)} \propto 1/R^9$ (поскольку $V \propto 1/R^3$, три раза  $
ightarrow$  $1/R^9$)

 $
ightarrow$  Получаем:

$$
E^{(3)}(R) = \pm \frac{C_9}{R^9}
$$

Знак зависит от интерференции, но величина:

$$
\left| \frac{E^{(3)}}{E^{(2)}} \right| \sim \frac{C_9 / R^9}{C_6 / R^6} = \frac{C_9}{C_6 R^3} \sim \left( \frac{a_0}{R} \right)^3 \ll 1
$$

 $
ightarrow$  **Очень малый вклад**.

## 6. Полная асимптотическая форма энергии

$$
\boxed{
E(R) = -\frac{C_6}{R^6} - \frac{C_8}{R^8} - \frac{C_{10}}{R^{10}} - \cdots + \frac{D_9}{R^9} + \cdots
}
$$

где:
- $C_6 > 0$: диполь-диполь (2-й порядок),
- $C_8 > 0$: квадруполь-квадруполь (2-й порядок),

>  Замечание: при очень больших $R$ (микроны) начинает работать **ретардация** — запаздывание электромагнитного взаимодействия. Тогда $C_6 / R^6 \to C_6 / (R^6 v)$, и при $R > \lambda_{\text{ат}}$ зависимость становится $\propto 1/R^7$ (Казимир–Полдер), но это уже другая физика.

##  1. Полное квантово-механическое выражение для $C_6$

Мы уже получили:

$$
C_6 = \frac{2}{3} \left( \frac{1}{4\pi\varepsilon_0} \right)^2 \sum_{n,n'} \frac{ |\vec{d}_{0n}|^2 |\vec{d}_{0n'}|^2 }{ \Delta E_n + \Delta E_{n'} }
\quad \text{где } \Delta E_n = E_n - E_{1s}
$$

Но можно переписать это через **поляризуемость** атома.

###  Поляризуемость атома водорода

Поляризуемость в основном состоянии:

$$
\alpha(i\omega) = 2 \sum_{n \neq 0} \frac{ |\langle 0| \vec{d} |n \rangle|^2 }{ E_n - E_0 } \cdot \frac{ (E_n - E_0)^2 }{ (E_n - E_0)^2 + \omega^2 }
$$

При $\omega = 0$ (статическая поляризуемость):

$$
\alpha(0) = 2 \sum_n \frac{ |\vec{d}_{0n}|^2 }{ E_n - E_0 }
$$

Но для $C_6$ нужна **динамическая поляризуемость**.

Оказывается, можно записать:

$$
C_6 = \frac{3}{\pi} \int_0^\infty d\omega\ \alpha(i\omega)^2
$$

— это **формула Каданова–Лондона**.

Для атома H: $\alpha(0) = \frac{9}{2} a_0^3$, и интегрирование даёт $C_6 \approx 6.5~\text{эВ}·\text{Å}^6$.

 Это уже **точный квантовый результат**, включающий **все уровни** (вплоть до континуума).

##  2. Ретардированное взаимодействие: Казимир–Полдер (1948)

На очень больших расстояниях, когда $R \gtrsim \lambda_{\text{ат}} \sim \frac{c}{\Delta E / \hbar} \sim 100~\text{нм}$, начинает работать **конечная скорость света**.

Флуктуации одного атома не мгновенно "доходят" до другого — поле запаздывает.

###  Что меняется?

- Вместо мгновенного диполь-дипольного взаимодействия — **обмен виртуальными фотонами с запаздыванием**.
- Корреляции между флуктуациями ослабевают.

###  Результат: энергия становится

$$
E(R) \sim -\frac{C_6}{R^6} \cdot f\left( \frac{R}{\lambda_0} \right)
\quad \text{где } \lambda_0 = \frac{2\pi c}{\omega_0},\ \omega_0 \approx 10.2~\text{эВ}/\hbar
$$

При $R \ll \lambda_0$: $f \to 1$  $
ightarrow$  $E \propto -1/R^6$

При $R \gg \lambda_0$: $f \to \frac{\lambda_0}{R}$  $
ightarrow$  $E \propto -1/R^7$

$$
\boxed{E(R) \xrightarrow{R \to \infty} -\frac{C_7}{R^7}}
\quad \text{с } C_7 \propto \alpha^2 c
$$

Это — **ретардированное дисперсионное взаимодействие**, или **эффект Казимира–Полдера**.

>  Важно: **знак остаётся отрицательным** — всё ещё **притяжение**, но теперь сильнее ослабляется.

ightarrow$  $E \propto -1/R^6$

## 1.  Численный расчёт $C_6$ для H–H (Python + физика)

Мы воспользуемся формулой:

$$
C_6 = \frac{2}{3} \left( \frac{1}{4\pi\varepsilon_0} \right)^2 \sum_{n=2}^{\infty} \sum_{n'=2}^{\infty} \frac{ |\vec{d}_{1s \to np}|^2 |\vec{d}_{1s \to n'p}|^2 }{ (E_n - E_1) + (E_{n'} - E_1) }
$$

###  Важные формулы для H-атома:

- Энергия уровня $n$: $E_n = -\frac{13.6~\text{эВ}}{n^2}$
- Матричный элемент $1s \to np$:  
  $$
  |\vec{d}_{1s \to np}|^2 = e^2 a_0^2 \cdot \frac{2^{8} n^5 (n-1)^{2n-2}}{3 (n+1)^{2n+2}} \quad \text{(приближённо)}
  $$
  Но проще использовать **точные табличные значения** или аналитическое выражение.

Для $1s \to np_z$:

$$
|\langle 1s| z |np_z \rangle|^2 = \left( \frac{2^{15/2} n^5 (n-1)^{2n-2}}{(n+1)^{2n+2}} \right) a_0^2
$$

(Мы будем использовать приближённые значения из справочников.)

###  Python-код:

```python
import numpy as np

# Константы
e = 1.602e-19        # Кл
a0 = 5.292e-11       # м (боровский радиус)
eps0 = 8.854e-12     # Ф/м
h_eV = 4.1357e-15    # эВ·с
c = 2.998e8          # м/с
E1 = 13.6            # эВ (|E_1s|)

# Перевод в СИ
def energy(n):
    return -E1 / n**2  # в эВ

def dipole_squared(n):
    """Приближённое значение |d|^2 для 1s  $
ightarrow$  np (в e^2 a^2)"""
    if n == 1:
        return 0.0
    # Табличные значения |<1s|z|np>|^2 / a^2, усреднённые по m
    values = {
        2: 0.745**2 * 3,   # 3 поляризации: x,y,z
        3: (0.977*a0)**2 * 3,
        4: (1.108*a0)**2 * 3,
        5: (1.192*a0)**2 * 3
    }
    if n in values:
        return (e * np.sqrt(values[n]))**2  # в Кл²·м²
    else:
        # Асимптотика: ~ n^2 a^2
        approx_a0_sq = 1.5 * n**2 * a0**2
        return (e**2 * approx_a0_sq)

# Суммирование
n_max = 10  # Учитываем уровни до n=10 (высшие дают малый вклад)
C6 = 0.0

for n in range(2, n_max+1):
    En = energy(n) - energy(1)  # в эВ
    dn_sq = dipole_squared(n)
    for n_prime in range(2, n_max+1):
        Enp = energy(n_prime) - energy(1)
        dnp_sq = dipole_squared(n_prime)
        denominator = (En + Enp)  # в эВ
        C6 += (dn_sq * dnp_sq) / denominator

C6 *= (2/3) * (1/(4*np.pi*eps0))**2
C6 /= 1.602e-19  # перевод Дж·м^6  $
ightarrow$  эВ·м^6
C6 *= 1e60       # перевод в эВ·A^6 (1 A = 10⁻¹⁰ м  $
ightarrow$  1 м^6 = 10^60 A^6)

print(f"C6 (численно, n <= {n_max}) = {C6:.2f} эВ·A^6")
```

## 3.  Формула Лифшица: обобщение на макроскопические тела

Лифшиц (1956) обобщил эффект Казимира на реальные материалы, учитывая:

- Диэлектрическую проницаемость $\varepsilon(i\omega)$,
- Температуру,
- Шероховатость.

###  Основная идея:

Вместо суммы по уровням — **интеграл по флуктуациям электромагнитного поля**, подавленных диэлектриками.

### Сила между двумя полупространствами с диэлектрической функцией $\varepsilon(\omega)$:

$$

$$

где:
- $q = \sqrt{k_\parallel^2 + \varepsilon(i\xi)\xi^2/c^2}$,
- $r_{\text{TE}}, r_{\text{TM}}$ — коэффициенты отражения Френеля,

$$
\frac{F}{A} = -\frac{\pi^2 \hbar c}{240 d^4}
\quad \text{(классический результат Казимира)}
$$

###  Случай диэлектриков (например, $SiO_2$):

- При $d < 10~\text{нм}$: доминирует $1/d^2$ (Ван-дер-Ваальс),

$$

с:
- $E_{pp} = \frac{e^2}{4\pi\varepsilon_0 R}$
- $E_{pe} = \frac{e^2}{4\pi\varepsilon_0} \int \frac{ \rho_2(\vec{r}) }{ |\vec{r}| } d^3r$
- $E_{ee} = \frac{1}{4\pi\varepsilon_0} \int \int \frac{ \rho_1(\vec{r}_1) \rho_2(\vec{r}_2) }{ |\vec{r}_1 - \vec{r}_2| } d^3r_1 d^3r_2$

###  Код на Python с `mpmath` (высокая точность)

```python
import numpy as np
import matplotlib.pyplot as plt
from mpmath import mp, quad, exp, pi as mpi, inf
import math

# Установим точность: 50 десятичных знаков
mp.dps = 50  # digits of precision

# Физические константы (в СИ)
e = mp.mpf('1.602176634e-19')           # Кл
a = mp.mpf('5.29177210903e-11')         # Боровский радиус, м
eps0 = mp.mpf('8.8541878128e-12')       # Ф/м
k_e = 1 / (4 * mpi * eps0)              # 1/(4 pi epsilon_0)

# Вспомогательная функция: плотность электрона
def rho(r):
    return -e / (mpi * a**3) * exp(-2 * r / a)

# Потенциал от одного электронного облака в точке на расстоянии R
def phi_electron(R):
    """Потенциал от rho(r) = -e/( pi a^3) exp(-2r/a) в точке на расстоянии R от центра"""
    if R == 0:
        R = 1e-30
    term = (1 + R/a) * exp(-2*R/a)
    return - (e / (4*mpi*eps0)) * (1 - term) / R

# E_pe: протон 1 взаимодействует с электроном 2
def E_pe(R):
    phi_at_0 = phi_electron(R)  # потенциал электрона 2 в начале координат
    return (-e) * phi_at_0  # энергия протона 1 в этом поле

# E_ee: электрон 1 – электрон 2
def E_ee(R):
    # Перейдём к сферическим координатам для r1
    # Используем симметрию: интегрируем по r1, theta_1, и используем phi_electron(r1; R)
    def integrand_r1(r1):
        if r1 == 0:
            r1 = 1e-30
        rho1 = rho(r1)
        phi2 = phi_electron(mp.sqrt(r1**2 + R**2 - 2*r1*R*mp.cos(0)))  # усреднение по углу
        # Усредним по theta: int sin theta dtheta от 0 до pi
        def integrand_theta(theta):
            dist = mp.sqrt(r1**2 + R**2 - 2*r1*R*mp.cos(theta))
            phi_at_r1 = phi_electron(dist)
            return rho1 * phi_at_r1 * mp.sin(theta)
        I_theta = quad(integrand_theta, [0, mpi], error=True)
        return I_theta[0] * r1**2
    I_r = quad(integrand_r1, [0, inf], error=True)
    return I_r[0]

# Полная энергия
def total_energy(R):
    E_pp = k_e * e**2 / R
    E_pe_val = E_pe(R)
    E_ee_val = E_ee(R)
    return E_pp + 2*E_pe_val + E_ee_val

# Пример вычисления
R_test = 10 * a  # 10 боровских радиусов
E_val = total_energy(R_test)

print(f"R = {float(R_test):.3e} м")
print(f"E = {float(E_val):.3e} Дж")
print(f"E = {float(E_val * 6.242e18):.3f} эВ")
```

##  2. График $E(R)$ и $F(R) = -dE/dR$

Добавим построение графиков:

```python
# Диапазон R: от 3a до 20a
R_vals = np.linspace(3, 20, 50) * a
E_vals = []
F_vals = []

# Для вычисления производной — конечная разность с высокой точностью
h = 1e-4 * a

for R in R_vals:
    E_plus = total_energy(R + h)
    E_minus = total_energy(R - h)
    dE_dR = (E_plus - E_minus) / (2 * h)
    F = -dE_dR
    E_vals.append(float(E_plus * 6.242e18))  # в эВ
    F_vals.append(float(F * 1e9))            # в нН

R_A = np.array(R_vals) / a  # в единицах a

# График
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8))

ax1.plot(R_A, E_vals, 'b-', linewidth=2)
ax1.set_ylabel('Энергия $E(R)$ (эВ)', fontsize=12)
ax1.set_title('Энергия взаимодействия двух атомов H', fontsize=14)
ax1.grid(True)

ax2.plot(R_A, F_vals, 'r-', linewidth=2)
ax2.set_xlabel('Расстояние $R / a$', fontsize=12)
ax2.set_ylabel('Сила $F(R)$ (нН)', fontsize=12)
ax2.grid(True)

plt.tight_layout()
plt.show()
```

##  3. Аналитический вывод $ E_{ee}(R) $

Мы хотим:

$$
E_{ee}(R) = \frac{1}{4\pi\varepsilon_0} \int \rho_1(\vec{r}_1) \phi_{e2}(\vec{r}_1) d^3r_1
$$

где:
- $\rho_1(\vec{r}_1) = -\frac{e}{\pi a^3} e^{-2r_1/a}$
- $\phi_{e2}(\vec{r}_1) = -\frac{e}{4\pi\varepsilon_0 |\vec{r}_1 - \vec{R}|} \left(1 - \left(1 + \frac{|\vec{r}_1 - \vec{R}|}{a}\right) e^{-2|\vec{r}_1 - \vec{R}|/a} \right)$

Перейдём к сферическим координатам с осью $z \parallel \vec{R}$:

$$
|\vec{r}_1 - \vec{R}| = \sqrt{r_1^2 + R^2 - 2 r_1 R \cos\theta}
$$

$$
d^3r_1 = r_1^2 \sin\theta\, dr_1\, d\theta\, d\phi
$$

$$
\int d\phi = 2\pi
$$

Тогда:

$$
E_{ee}(R) = \frac{2\pi}{4\pi\varepsilon_0} \left( -\frac{e}{\pi a^3} \right) (-e) \int_0^\infty \int_0^\pi
e^{-2r_1/a} \cdot \frac{1 - \left(1 + \frac{s}{a}\right) e^{-2s/a}}{s}
\cdot r_1^2 \sin\theta\, d\theta\, dr_1
$$

где $s = |\vec{r}_1 - \vec{R}|$

Это — **трёхмерный интеграл**, но он **может быть выражен через элементарные функции и экспоненциальные интегралы**.

###  Известный результат (см. Ландау, Лифшиц, "Квантовая механика"):

После вычисления всех вкладов:

$$
E(R) = \frac{e^2}{4\pi\varepsilon_0 R} \left[ 1 - 2\left(1 + \frac{R}{a}\right)e^{-2R/a} \right] + E_{ee}(R)
$$

и полный результат при $R \gg a$:

$$
E(R) \approx -\frac{6.5~\text{эВ} \cdot \text{Å}^6}{R^6}
$$

---

##  1. Что означает $a/V^2$ в уравнении Ван-дер-Ваальса?

Уравнение:

$$
\left( p + \frac{a}{V^2} \right) (V - b) = RT
$$

- $p$ — внешнее давление,
- $\frac{a}{V^2}$ — **внутреннее давление**, связанное с **притяжением между молекулами**.

>  Это **не энергия**, а **давление** (сила на единицу площади).

---

##  2. Откуда берётся $a/V^2$? (Микроскопический вывод)

Рассмотрим газ из $N$ молекул в объёме $V$.  
Пусть каждая пара взаимодействует с энергией:

$$
E_{\text{pair}}(r) = -\frac{C_6}{r^6}
$$

Тогда **полная потенциальная энергия** всех пар:

$$
U = \frac{1}{2} \sum_{i \neq j} E_{\text{pair}}(|\vec{r}_i - \vec{r}_j|) \approx \frac{1}{2} \int \rho(\vec{r}_1) \rho(\vec{r}_2) E_{\text{pair}}(|\vec{r}_1 - \vec{r}_2|)\, d^3r_1 d^3r_2
$$

Для однородного газа: $\rho = N/V$

$$
U \approx \frac{1}{2} \left( \frac{N}{V} \right)^2 \int E_{\text{pair}}(r)\, d^3r = \frac{N^2}{2V^2} \int_0^\infty \left( -\frac{C_6}{r^6} \right) 4\pi r^2 dr
$$

Но интеграл $\int r^{-4} dr$ **расходится** при $r \to 0$ и $r \to \infty$!  
 Нужно ввести **обрезания**:

- $r_{\text{min}} \sim d$ — размер молекулы,
- $r_{\text{max}} \sim L$ — размер системы (но на практике — средняя длина свободного пробега, или $\sim V^{1/3}$)

Но на самом деле: в **равновесном газе** молекулы не взаимодействуют на больших расстояниях из-за экранирования, и мы можем формально вычислить **вклад в свободную энергию**.

Однако проще использовать **термодинамический подход**.

---

##  3. Связь между $U$ и $a/V^2$

Известно, что **внутреннее давление** связано с энергией соотношением:

$$
p_{\text{int}} = -\left( \frac{\partial U}{\partial V} \right)_T
$$

Если полная потенциальная энергия газа:

$$
U = -\frac{a N^2}{V}
\quad \text{(не } a/V^2\text{, а } a N^2 / V\text{)}
$$

Тогда:

$$
p_{\text{int}} = -\left( \frac{\partial}{\partial V} \left( -\frac{a N^2}{V} \right) \right) = -\frac{a N^2}{V^2}
$$

 То есть, **в уравнении Ван-дер-Ваальса** $\frac{a}{V^2}$ — это **на самом деле $\frac{a N^2}{V^2}$**, но константа $a$ **уже включает $N^2$** (в молярной форме $N = N_A$, так что $a$ — эмпирическая константа для 1 моля).

> **Ключевой момент**: $\frac{a}{V^2} \propto \frac{1}{V^2} \sim L^{-6}$ — это **давление**,  
> но оно происходит от энергии $U \propto -\frac{1}{V} \sim -L^{-3}$

##  4. Как из $E_{\text{пара}} \propto 1/r^6$ получается $U \propto -1/V$?

Рассмотрим:

- Среднее число пар на единицу объёма: $\sim \rho^2 = (N/V)^2$
- Средняя энергия одной пары на расстоянии $r$: $\sim -C_6 / r^6$
- Но вклад в энергию дают все пары с $r$ от $d$ до $\infty$

Однако: в **однородном газе** можно показать, что:

$$
U = \frac{1}{2} \int \rho^2 \cdot \left( \int E_{\text{pair}}(r)\, d^3r \right) dV
$$

Но $\int E_{\text{pair}}(r)\, d^3r = \int -\frac{C_6}{r^6} \cdot 4\pi r^2 dr = -4\pi C_6 \int \frac{dr}{r^4}$

Этот интеграл **расходится при $r \to 0$**, но конечен, если обрезать на $r \sim d$:

$$
\int_d^\infty \frac{dr}{r^4} = \frac{1}{3d^3}
\quad \Rightarrow \quad \int E_{\text{pair}}\, d^3r \sim -\frac{C_6}{d^3}
$$

Тогда:

$$
U \sim \frac{1}{2} \cdot \frac{N^2}{V^2} \cdot V \cdot \left( -\frac{C_6}{d^3} \right) = -\underbrace{\left( \frac{C_6}{2d^3} \right)}_{a'} \cdot \frac{N^2}{V}
$$

То есть:  
$$
U \propto -\frac{1}{V}, \quad \Rightarrow \quad p_{\text{int}} = -\frac{\partial U}{\partial V} \propto +\frac{1}{V^2}
$$

```python
import os
path=os.getcwd()
#print(path)

fileName='уран.txt'
fileName='поташ.txt'
times = []
intervals = []
log_intervals = []
text_file = open(os.path.join(path, fileName), 'r')
while True:
    line = text_file.readline()
    if 0 == len(line):
        break

    elements=line.split('\t')
    times.append((elements[0].strip()))
    interval = float(elements[1].strip())
    intervals.append(interval)
    if interval != 0:
        log_intervals.append(log(interval))
```

# Глубина проникновения магнитного поля в проводник

геометрия задачи: 

поверхность проводника $x=>0$

вакуум заполняет область $x<0$

В области $x>0$ - плохой проводник

$$rot\,\vec{E} = -\frac{1}{c}\frac{\partial \vec B}{\partial t} = \mu\,\frac{i\,\omega}{c}\,\vec{H}$$

$$rot\,\vec{H} = \frac{1}{c}\frac{\partial \vec D}{\partial t} + \frac{4\pi}{c} \vec j = - \frac{i\,\omega}{c} \epsilon \vec{E} + \frac{4\pi}{c}\sigma \vec E$$

где 

$$\vec{H}(t) = \vec{H}\,e^{-i\,\omega\,t}$$

$$\vec{E}(t) = \vec{E}\,e^{-i\,\omega\,t}$$

$$rot\,rot\,\vec{H} = - \frac{i\,\omega}{c} \epsilon\, rot\,\vec{E} + \frac{4\pi}{c}\sigma\, rot\,\vec E$$

$$rot\,rot\,\vec{H} = - \frac{i\,\omega}{c} \epsilon\, \mu\,\frac{i\,\omega}{c}\,\vec{H} + \frac{4\pi}{c}\sigma\, \mu\,\frac{i\,\omega}{c}\,\vec{H}$$

$$rot\,rot\,\vec{H} = \epsilon\, \mu\,\frac{\omega^2}{c^2}\,\vec{H} + i\,\frac{4\pi\sigma\, \mu\,\omega}{c^2}\,\vec{H}$$

$$rot\,rot\,\vec{H} = grad\,div\,\vec{H} - Laplasian\,\vec{H}$$

$$grad\,div\,\vec{H} - Laplasian\,\vec{H} = \epsilon\, \mu\,\frac{\omega^2}{c^2}\,\vec{H} + i\,\frac{4\pi\sigma\, \mu\,\omega}{c^2}\,\vec{H}$$

$$ Laplasian\,\vec{H} + \epsilon\, \mu\,\frac{\omega^2}{c^2}\,\vec{H} + i\,\frac{4\pi\sigma\, \mu\,\omega}{c^2}\,\vec{H} - grad\,div\,\vec{H} = 0$$

$$ div\,\vec{B} = 0$$

для проводящего металла с потерями ищем решение в виде

$$\vec{H}(t) = \vec{H}_0\,e^{i\,k'_x\,x}\,e^{i\,k'_z\,z}\,e^{-i\,\omega\,t}$$

$$\vec{H} = \vec{H}_0\,e^{i\,k'_x\,x}\,e^{i\,k'_z\,z}$$

$$\vec{H} = \vec{H}(x)\,e^{i\,k'_z\,z}$$

в котором волновой вектор, направленный вдоль оси z комплексный $k'_z = k_z + i\,s_z$, как и волновой вектор, направленный вдоль оси x может быть комплексным $k'_x = k_x + i\,s_x$

$$\vec{H}(t) = \vec{H}_0\,e^{i\,(k_x + i\,s_x)\,x}\,e^{i\,(k_z + i\,s_z)\,z}\,e^{-i\,\omega\,t}$$

$$\vec{H} = \vec{H}_0\,e^{i\,(k_x + i\,s_x)\,x}\,e^{i\,(k_z + i\,s_z)\,z}$$

$$\vec{H} = \vec{H}(x)\,e^{i\,(k_z + i\,s_z)\,z}$$

подставляя в исходное уравнение приходим к обыкновенному дифференциальному уравнению

$$\frac{\partial^{2}}{(\partial {x})^{2}}\vec{H}(x) + \frac{\partial^{2}}{(\partial {y})^{2}}\vec{H}(x) + \left(\epsilon  \mu \frac{\omega^2}{c^2} + i\,\frac{4\pi\sigma\, \mu\,\omega}{c^2} - {k'_z}^2\right)\,\vec{H}(x) - grad\,div\,\vec{H} = 0$$

где

$$\vec{H}(x) = \vec{H}_0\,e^{i\,k'_x\,x}$$

производим следующую замену переменных

$$\kappa^2 = \epsilon \mu \frac{\omega^2}{c^2} + i\,\frac{4\pi\,\sigma\,\mu\,\omega}{c^2}$$

$$\kappa^2 = \mu \left(\epsilon  + i\,\frac{4\pi\,\sigma}{\omega} \right) \frac{\omega^2}{c^2}$$

теперь получаем дифференциальное уравнение

$$\frac{\partial^{2}}{(\partial {x})^{2}}\vec{H}(x) + \frac{\partial^{2}}{(\partial {y})^{2}}\vec{H}(x) + \left(\kappa^2 - {k'_z}^2\right)\,\vec{H}(x) = 0$$

однако моё дифференциальное уравнение отличается от дифференциального уравнения, использованного Ландау (Ландау Лифшиц. Электродинамика сплошных сред (параграф 59)) для скин слоя тем, что у меня поле все же зависит от координаты $z$ (замечание у Ландау плоскость поверхности проводника перпендикулярна оси $z$, а у меня она перпендикулярна оси $x$, поэтому мои координата $x$ соответствует координате $z$ у Ландау)

очень странно, как Ландау избавился от волнового вектора направленного вдоль поверхности плоскости.

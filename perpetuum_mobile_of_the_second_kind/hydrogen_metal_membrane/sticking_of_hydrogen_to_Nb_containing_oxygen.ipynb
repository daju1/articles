# B. $H_2$ sticking under thermostatic conditions: Effects of gas and surface temperature

Temperature dependencies of $\alpha_{H_2}$ at $T_g \cong T_S \equiv T$ (thermostatic conditions) and at $T_g = const = 300 K$ (with only $T_S$ varying) are presented in Fig. 6 for one value of $C_O$ (0.03 at. %) to provide a typical example. The data were obtained from permeation through a tubular membrane in two opposite directions (Sec. II C 1). The temperature dependencies of two corresponding curves are basically similar, and they give the absolute values of $\alpha_{H_2}$ close to the other; the rather small difference in the absolute value can be attributed to the error in pressure measurements by different
gauges. Such observations, first, give the grounds to equally refer the data on $\alpha_{H_2}$ obtained at varying $T_S$ (Figs. 3–5) to the thermostatic conditions. Secondly, this means that $\alpha_{H_2}$ is determined by $T_S$ and not dependent on $T_g$, at least in the first approximation. These two conclusions were also derived by another way (Sec. II C 1): the values of $\alpha_{H_2}$ evaluated from
the steady-state permeation rate at $T_g = const = 300 K$ were
close to those estimated for $T_g = T_S$, from the permeation lag
time $\tau$ [see Eqs. (3) and (4) and Fig. 2].

# B. Прилипание $H_2$ в термостатических условиях: Влияние температуры газа и поверхности

Температурные зависимости $\alpha_{H_2}$ при $T_g \cong T_S \equiv T$ (термостатические условия) и при $T_g = const = 300 K$ (при изменении только $T_S$) представлены на рис. 6 для одного значения $C_O$ (0,03 ат. %) в качестве типичного примера. Данные были получены при прохождении через трубчатую мембрану в двух противоположных направлениях (раздел II C 1). Температурные зависимости двух соответствующих кривых в основном схожи, и они дают абсолютные значения $\alpha_{H_2}$, близкие друг к другу; довольно небольшая разница в абсолютном значении может быть отнесена к ошибке в измерениях давления различными манометрами. Такие наблюдения, во-первых, дают основание в равной степени отнести данные по $\alpha_{H_2}$, полученные при различных $T_S$ (рис. 3-5), к термостатическим условиям. Во-вторых, это означает, что $\alpha_{H_2}$ определяется $T_S$ и не зависит от $T_g$, по крайней мере, в первом приближении. Эти два вывода были получены и другим способом (раздел II C 1): значения $\alpha_{H_2}$, оцененные из из установившейся скорости проницания при $T_g = const = 300 K$ были близки к значениям, полученным для $T_g = T_S$, из времени запаздывания проницания
[см. уравнения (3) и (4) и рис. 2].

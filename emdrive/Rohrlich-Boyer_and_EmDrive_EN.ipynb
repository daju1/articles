## ðŸ”¹ 2. Maxwell stress tensor (CGS)

In CGS, Maxwell stress tensor:
$$
T_{ij} = \frac{1}{4\pi} \left( E_i E_j + H_i H_j - \frac{1}{2} \delta_{ij} (E^2 + H^2) \right).
$$

For TM mode ($E_y = H_x = H_z = 0$), nonzero components:
- $T_{xx} = \frac{1}{4\pi} \left( E_x^2 + H_y^2 - \frac{1}{2}(E_x^2 + E_z^2 + H_y^2) \right) = \frac{1}{8\pi} \left( E_x^2 - E_z^2 + H_y^2 \right)$,
- $T_{xz} = T_{zx} = \frac{1}{4\pi} E_x E_z$,
- $T_{yy} = \frac{1}{8\pi} \left( -E_x^2 - E_z^2 - H_y^2 \right)$,
- $T_{zz} = \frac{1}{8\pi} \left( -E_x^2 + E_z^2 - H_y^2 \right)$.

But for force **along $x$-axis** (normal to wall), we need **divergence over $x$** of **first row** of tensor:
$$
(\nabla \cdot \mathbf{T})_x = \frac{d}{dx} T_{xx}, \quad
(\nabla \cdot \mathbf{T})_y = \frac{d}{dx} T_{yx} = 0, \quad
(\nabla \cdot \mathbf{T})_z = \frac{d}{dx} T_{zx} = \frac{d}{dx} \left( \frac{1}{4\pi} E_x E_z \right).
$$

However, **in TM mode, symmetry in $z$** and absence of $z$-gradients means **net force per unit area** is directed **along $x$**, and this is exactly what we calculate.

## ðŸ”¹ 3. Force density: momentum balance equation

From momentum conservation law:
$$
\mathbf{f}(x) = - \nabla \cdot \mathbf{T} = - \frac{d}{dx} \mathbf{T} \cdot \hat{\mathbf{x}}.
$$

Thus, **volume force density along $x$**:
$$
f_x(x) = - \frac{d}{dx} T_{xx}(x).
$$

Substituting $T_{xx}$:
$$
f_x(x) = - \frac{d}{dx} \left[ \frac{1}{8\pi} \left( E_x^2 - E_z^2 + H_y^2 \right) \right].
$$

## ðŸ”¹ 4. Accounting for complex fields and period averaging

Fields are harmonic: $E(x,t) = \Re[\tilde{E}(x) e^{-i\omega t}]$, similarly for $H$.

Time-averaged force density:
$$
\langle f_x(x) \rangle = - \frac{1}{8\pi} \frac{d}{dx} \left[ \frac{1}{2} \left( |\tilde{E}_x|^2 - |\tilde{E}_z|^2 + |\tilde{H}_y|^2 \right) \right]
= - \frac{1}{16\pi} \frac{d}{dx} \left( |\tilde{E}_x|^2 - |\tilde{E}_z|^2 + |\tilde{H}_y|^2 \right).
$$

## ðŸ”¹ 5. Alternative path: through Lorentz force (verification)

In conducting medium, force density:
$$
\langle f_x(x) \rangle = \frac{1}{2c} \Re \left[ \tilde{j}_z \tilde{H}_y^* \right],
$$
where $\tilde{j}_z = \sigma_e \tilde{E}_z$.

But from Maxwell's equations:
$$
\frac{d \tilde{E}_z}{dx} = i \frac{\omega}{c} \mu' \tilde{H}_y,
$$
therefore one can show that both approaches are **equivalent**, and tensor divergence is **the same thing** as Lorentz force averaged over period.

$$

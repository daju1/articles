[quote] Самый короткий вывод ЛВ потенциалов у Батыгина с Топтыгиным, задача 756, стр 552 (издание 2002 г). Они тоже используют дельта-функцию, но только одну пространственную, а не две совместных пространственно-временных, как Джексон, поэтому их вывод легче поддается осмыслению

При выводе ЛВ потенциалов в задаче 756 на стр. 552 исходное выражение потенциала
$$\varphi(\vec r,t)=\int\frac{\rho(\vec r',t-R/c)}{R}dV'=e\int {\delta[\vec r'-\vec r_0(t-R/c)]}{R}dV'$$                 (1)
где
$\vec r_0(t)$
есть пространственно-временная функция траектории заряда е;
$\vec R=\vec r-\vec r'$    (2)
следует понимать как расстояние между точкой наблюдения r и произвольной точкой r' в пространстве той же системы наблюдения,
R/c есть произвольный интервал времени,
$ \vec r_0(t-R/c) $ есть точка на траектории заряда, где сам заряд находился в соответствующий произвольный момент времени.
Поскольку $ \vec r_0(t) $ описывает положение заряда в произвольный момент t от условно выбранного начала отсчета t=0, то $ \vec r_0(t-R/c) $ описывает положение заряда в произвольный момент t-R/c от выбранного фиксированного момента наблюдения t, а в классике известно, что изменение начала отсчета ничего не изменяет в результатах расчетов.

Но Батыгин считает, что вычисление
$$\int \delta[\vec r'-\vec r_0(t-R/c)]dV'$$
отличается от вычисления традиционного
$$\int \delta[\vec r-\vec r_0(t)]dV$$
тем что в традиционном выражении $\vec r_0(t)$ для самой дельта-функции и интеграла есть некая заданная константа,
а $\vec r_0(t-R/c)$ становися переменной, зависящей от r' по условию (2), поэтому при интегрировании следует учитывать производную полного аргумента $[\vec r'-\vec r_0(t-R/c)]$ по r' (якобиан преобразования по Батыгину).

Отвлекусь от вычисления потенциала и спрошу, что даст аналогичное вычисление величины движущегося заряда в системе наблюдения с точкой наблюдения r ?
По аналогии с (1) запишу выражение
$$e_v=\int \rho (\vec r',t-R/c)dV'=e\int \delta [\vec r'-\vec r_0(t-R/c)]dV'$$
и продолжая ту же аналогию вывода, перехожу к новой переменной интегрирования
$\vec R_1=\vec r'-\vec r_0(t-R/c)$
учитываю ее производную по r' (якобиан преобразования по Батыгину) и получаю
$$e_v=\frac{e}{1-\vec v \vec R/cR}$$
что не может быть признано корректным из-за нарушения сохранения заряда.

Что здесь не так ? [/quote]

Ответ получил вот такой
[quote]Неверная формула для полного заряда. Правильно так:

$$e_v = \int \rho(\vec{r}, \, t)dV.$$

Получаем

$$e_v =e\int \delta(\vec{r}-\vec{r}_0(t)) dV =e\cdot 1 =e.$$[/quote][/quote]

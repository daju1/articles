### Что ориентирует спин: поле **B** или его градиент **∇B**?

**Ответ:** Ни то, ни другое — в том-то и дело.

Давайте разберём физику по шагам.

---

### 1. Уравнение движения магнитного момента

Магнитный момент **μ** в магнитном поле подчиняется уравнению:

$$
\frac{d\boldsymbol{\mu}}{dt} = \gamma \, \boldsymbol{\mu} \times \boldsymbol{B}
$$

где $\gamma$ — гиромагнитное отношение.

Это уравнение говорит: **момент прецессирует вокруг направления поля B** с частотой Лармора $\omega = \gamma B$. Он не «выравнивается» по полю — он вращается вокруг него.

---

### 2. Сила, разделяющая пучок

Сила, действующая на атом, равна:

$$
\boldsymbol{F} = \nabla (\boldsymbol{\mu} \cdot \boldsymbol{B})
$$

Если градиент направлен вдоль $z$ и $\boldsymbol{\mu} \parallel \boldsymbol{B}$ (или антипараллельно), то:

$$
F_z \approx \mu_z \frac{\partial B_z}{\partial z}
$$

**Градиент отвечает за пространственное разделение пучка.** Без градиента ($\nabla B = 0$) все атомы летели бы прямо, независимо от ориентации спина.

---

### 3. Что происходит в реальном эксперименте?

Время пролёта атома через магнит очень мало — порядка микросекунд. За это время:

- **Поле B** пытается заставить момент прецессировать.
- **Градиент** пытается развести атомы в пространстве.

Если частота Лармора $\omega = \gamma B$ достаточно велика, атом успевает сделать много оборотов вокруг **B** за время пролёта. Но **это не приводит к выравниванию** — момент просто вращается, сохраняя угол с полем.

**Ключевой момент:** В квантовом описании мы не говорим, что спин «поворачивается» полем. Мы говорим, что **измерение проекции спина на ось градиента** даёт одно из двух значений: $+\hbar/2$ или $-\hbar/2$.

---

### 4. Почему «на любую ось»?

Формулировка «на любую ось» означает:

- Если поставить прибор так, чтобы градиент был вдоль $z$ — получим два пучка с $S_z = \pm\hbar/2$.
- Если повернуть прибор так, чтобы градиент был вдоль $x$ — получим два пучка с $S_x = \pm\hbar/2$.
- И так для любой оси.

**Ось измерения задаётся направлением градиента поля в приборе.** Не само поле, а именно его градиент определяет, какую компоненту спина мы измеряем.

### Теоретическая возможность

Да, конфигурация с **перпендикулярными** $\boldsymbol{B}$ и $\nabla\boldsymbol{B}$ теоретически возможна. Например:

- Поле $\boldsymbol{B}$ направлено вдоль $z$: $B_z \neq 0$, $B_x = B_y = 0$

Тогда сила на атом будет:
$$
F_x = \mu_z \frac{\partial B_z}{\partial x}
$$

**Ключевая особенность:** сила действует в направлении $x$, но зависит от проекции спина на ось $z$ (направление поля).

### Что предсказывает квантовая механика?

В такой конфигурации возникает **конфликт** между:

1. **Ось квантования спина** — задаётся направлением $\boldsymbol{B}$ (ось $z$)
2. **Ось пространственного разделения** — задаётся направлением $\nabla\boldsymbol{B}$ (ось $x$)

Результат: пучок должен разделиться на **две компоненты вдоль $x$**, но каждая компонента соответствует определённому значению $S_z = \pm\hbar/2$.

Это не то же самое, что измерение $S_x$ — мы по-прежнему измеряем проекцию на ось поля, но разделяем частицы в перпендикулярном направлении.

Если $\omega\tau \gg 1$, момент делает много оборотов — и эффективно усредняется.

#### Квантовое описание

Гамильтониан:
$$
\hat{H} = -\gamma \hat{\boldsymbol{S}} \cdot \boldsymbol{B} + V_{\text{grad}}
$$

где $V_{\text{grad}}$ — потенциал от градиента (зависит от координаты $x$).

**Проблема:** В стандартном опыте Штерна–Герлаха гамильтониан имеет вид:
$$
\hat{H} = -\gamma B_z \hat{S}_z + \mu_B z \frac{\partial B_z}{\partial z}
$$

и операторы $\hat{S}_z$ и координаты $z$ **коммутируют** — поэтому можно одновременно иметь определённое значение спина и положения.

**В перпендикулярной конфигурации:**
$$
\hat{H} = -\gamma B_z \hat{S}_z + \mu_B x \frac{\partial B_z}{\partial x}
$$

Здесь $\hat{S}_z$ и координата $x$ тоже коммутируют — формально проблем нет. Но физически:

- Измерение происходит **вдоль $x$** (разделение пучка)
- Но квантование спина — **вдоль $z$** (направление поля)

Это означает, что мы **не измеряем** $S_x$, а измеряем $S_z$, но проецируем результат на перпендикулярную ось.

Но если ось прибора (градиент) и ось поля не совпадают — **какая ось является «осью измерения»**?

- Ось градиента $\nabla\boldsymbol{B}$ — определяет направление пространственного разделения

Это **не одно и то же**. И если они перпендикулярны — возникает концептуальная неоднозначность: что именно мы измеряем?

## Динамика прецессирующего диполя в неоднородном поле

### 1. Уравнение движения магнитного момента

Для прецессирующего диполя уравнение движения имеет вид:

$$
\frac{d\boldsymbol{\mu}}{dt} = \gamma \, \boldsymbol{\mu} \times \boldsymbol{B} + \boldsymbol{F}
$$

где:
- $\gamma$ — гиромагнитное отношение
- $\boldsymbol{F} = \nabla(\boldsymbol{\mu} \cdot \boldsymbol{B})$ — сила от градиента поля

---

### 2. Разделение на прецессию и дрейф

В неоднородном поле движение диполя можно разложить на:

| Компонента | Описание |
|------------|----------|
| **Прецессия** | Быстрое вращение вокруг локального $\boldsymbol{B}$ |
| **Дрейф** | Медленное смещение центра прецессии под действием $\nabla\boldsymbol{B}$ |

---

### 3. Сила и ускорение центра масс

Сила, действующая на диполь:

$$
\boldsymbol{F} = \nabla(\boldsymbol{\mu} \cdot \boldsymbol{B}) \approx \mu_z \frac{\partial B_z}{\partial z} \hat{z}
$$

(в стандартной конфигурации Штерна–Герлаха)

Ускорение центра масс:

$$
a_z = \frac{F_z}{m} = \frac{\mu_z}{m} \frac{\partial B_z}{\partial z}
$$

---

### 4. Динамика прецессии: уравнения

Пусть $\boldsymbol{\mu}$ прецессирует вокруг $\boldsymbol{B}$ с угловой скоростью $\omega = \gamma B$.

В сферических координатах ($\theta, \varphi$):

$$
\begin{aligned}
\frac{d\theta}{dt} &= 0 \\
\frac{d\varphi}{dt} &= \gamma B
\end{aligned}
$$

То есть угол наклона $\theta$ остаётся постоянным, а азимут $\varphi$ меняется линейно со временем.

---

### 5. Влияние неоднородности на прецессию

Если $\boldsymbol{B}$ меняется в пространстве, то:

$$
\frac{d\boldsymbol{\mu}}{dt} = \gamma \boldsymbol{\mu} \times \boldsymbol{B} + (\boldsymbol{v} \cdot \nabla)\boldsymbol{\mu}
$$

где $\boldsymbol{v}$ — скорость движения центра диполя.

Это приводит к **адиабатическому следованию**: если изменение $\boldsymbol{B}$ достаточно медленное, момент остаётся выровненным относительно локального поля.

---

### 6. Условие адиабатичности

Прецессия остаётся адиабатической, если:

$$
\frac{|\nabla B| v}{B^2} \ll \gamma
$$

или в терминах времени:

$$
\tau_{\text{flight}} \gg \frac{1}{\gamma B}
$$

Если это условие нарушается — возникают **неадиабатические переходы** (флипы спина).

---

### 7. Два устойчивых режима прецессии

В вашей модели электрон — это прецессирующий диполь. В поле $\boldsymbol{B}$ возможны два устойчивых режима:

| Режим | Ориентация $\boldsymbol{\mu}$ | Энергия |
|-------|-------------------------------|---------|
| **Вверх** | $\boldsymbol{\mu} \parallel \boldsymbol{B}$ | $E = -\mu B$ (минимум) |
| **Вниз** | $\boldsymbol{\mu} \upharpoonleft \! \downharpoonright \boldsymbol{B}$ | $E = +\mu B$ (максимум) |

**Вопрос:** почему антипараллельное состояние устойчиво?

---

### 8. Стабилизация антипараллельного состояния

В классической модели антипараллельное состояние неустойчиво. Но в вашей модели прецессирующего диполя есть **кинетический момент** от прецессии:

$$

## Уравнения движения прецессирующего диполя

### 1. Гамильтониан системы

Рассмотрим магнитный диполь $\boldsymbol{\mu}$ в неоднородном поле $\boldsymbol{B}$. Гамильтониан:

$$
\hat{H} = -\gamma \hat{\boldsymbol{S}} \cdot \boldsymbol{B}(\boldsymbol{r}) + V_{\text{grad}}(\boldsymbol{r})
$$

где $V_{\text{grad}}$ — потенциал от градиента поля.

---

### 2. Уравнение для магнитного момента

В классическом приближении:

$$
\frac{d\boldsymbol{\mu}}{dt} = \gamma \boldsymbol{\mu} \times \boldsymbol{B}(\boldsymbol{r})
$$

В сферических координатах ($\theta, \varphi$), где $\theta$ — угол между $\boldsymbol{\mu}$ и $\boldsymbol{B}$:

$$
\begin{aligned}
\frac{d\theta}{dt} &= 0 \\
\frac{d\varphi}{dt} &= \gamma B(\boldsymbol{r})
\end{aligned}
$$

Это означает, что $\theta = \text{const}$, а $\varphi$ меняется со скоростью прецессии.

---

### 3. Уравнение движения центра масс

Сила на диполь:

$$
\boldsymbol{F} = \nabla(\boldsymbol{\mu} \cdot \boldsymbol{B}) = \nabla(\mu B \cos\theta)
$$

Ускорение центра масс:

$$
\frac{d^2\boldsymbol{r}}{dt^2} = \frac{\mu}{m} \nabla(B \cos\theta)
$$

---

### 4. Полная система уравнений

Объединяя всё, получаем:

$$
\boxed{
\begin{aligned}
\frac{d\boldsymbol{r}}{dt} &= \boldsymbol{v} \\
\frac{d\boldsymbol{v}}{dt} &= \frac{\mu}{m} \nabla(B \cos\theta) \\
\frac{d\theta}{dt} &= 0 \\
\frac{d\varphi}{dt} &= \gamma B(\boldsymbol{r})
\end{aligned}
}
$$

## Анализ устойчивости

### 5. Стационарные решения

Стационарные решения соответствуют $\theta = \text{const}$. Возможны два случая:

| Режим | $\theta$ | Энергия $E = -\mu B \cos\theta$ |
|-------|----------|----------------------------------|
| Параллельный | $0$ | $-\mu B$ (минимум) |
| Антипараллельный | $\pi$ | $+\mu B$ (максимум) |

---

### 6. Линеаризация около стационарных точек

Рассмотрим малые отклонения: $\theta = \theta_0 + \delta\theta$

#### Около $\theta_0 = 0$ (параллельное состояние):

$$
\frac{d^2(\delta\theta)}{dt^2} + \omega^2 (\delta\theta) = 0
$$

где $\omega = \gamma B$ — частота Лармора.

Это уравнение гармонического осциллятора — **устойчивое** решение.

#### Около $\theta_0 = \pi$ (антипараллельное состояние):

$$
\frac{d^2(\delta\theta)}{dt^2} - \omega^2 (\delta\theta) = 0
$$

Решение: $\delta\theta \propto e^{\pm \omega t}$ — **неустойчивое** решение.

### 7. Проблема антипараллельного состояния

В классической модели антипараллельное состояние неустойчиво — любой малый возмущение приводит к перевороту момента.

**Но в опыте Штерна–Герлаха мы наблюдаем оба состояния!**

## Стабилизация через кинетический момент

### 8. Введение эффективного момента инерции

В вашей модели электрон — это прецессирующий диполь с внутренним кинетическим моментом $\boldsymbol{L}_{\text{prec}}$.

Добавим гироскопический член в уравнение:

$$
\frac{d\boldsymbol{\mu}}{dt} = \gamma \boldsymbol{\mu} \times \boldsymbol{B} + \boldsymbol{\Omega} \times \boldsymbol{\mu}
$$

где $\boldsymbol{\Omega}$ — угловая скорость прецессии.

---

### 9. Эффективный потенциал

Полная энергия системы:

$$
E_{\text{total}} = -\mu B \cos\theta + \frac{1}{2} I_{\text{eff}} \Omega^2 \sin^2\theta
$$

Первый член — потенциальная энергия в поле.  
Второй член — кинетическая энергия прецессии.

---

### 10. Уравнение для $\theta$

Из условия $\frac{dE_{\text{total}}}{d\theta} = 0$:

$$
\mu B \sin\theta - I_{\text{eff}} \Omega^2 \sin\theta \cos\theta = 0
$$

Отсюда:

$$
\sin\theta \left( \mu B - I_{\text{eff}} \Omega^2 \cos\theta \right) = 0
$$

Решения:

1. $\sin\theta = 0 \Rightarrow \theta = 0, \pi$ (параллельное и антипараллельное)
2. $\cos\theta = \frac{\mu B}{I_{\text{eff}} \Omega^2}$ (новые стационарные состояния!)

---

### 11. Условие существования дополнительных состояний

Для существования решений типа 2 необходимо:

$$
\left| \frac{\mu B}{I_{\text{eff}} \Omega^2} \right| \leq 1
$$

Если это условие выполняется, появляются **два дополнительных устойчивых состояния** при $\theta \neq 0, \pi$!

---

### 12. Бифуркационная диаграмма

При изменении параметра $\alpha = \frac{\mu B}{I_{\text{eff}} \Omega^2}$:

| $\alpha$ | Стационарные состояния |
|----------|------------------------|
| $\alpha > 1$ | Только $\theta = 0$ |
| $\alpha = 1$ | $\theta = 0$ и $\theta = \pi/2$ |
| $0 < \alpha < 1$ | $\theta = 0$, $\theta = \arccos\alpha$, $\theta = \pi - \arccos\alpha$ |
| $\alpha = 0$ | Все $\theta$ (изотропия) |
| $-1 < \alpha < 0$ | $\theta = \pi$, $\theta = \arccos\alpha$, $\theta = 2\pi - \arccos\alpha$ |
| $\alpha < -1$ | Только $\theta = \pi$ |

## Связь с опытом Штерна–Герлаха

### 13. Объяснение двух пятен

В вашей модели:

1. **Параллельное состояние** ($\theta = 0$) — устойчиво всегда
2. **Антипараллельное состояние** ($\theta = \pi$) — стабилизируется кинетическим моментом прецессии

Эффективный потенциал имеет **два минимума**, что объясняет два пятна на экране!

---

### 14. Квантование угла прецессии

Если предположить, что прецессия квантуется (аналогично квантованию углового момента):

$$
L_{\text{prec}} = I_{\text{eff}} \Omega = n \hbar, \quad n = 1, 2, 3, \dots
$$

Тогда:

$$
\cos\theta_n = \frac{\mu B}{n^2 \hbar^2 / I_{\text{eff}}}
$$

При определённых условиях только $n = 1$ даёт физически допустимые решения, что приводит к **двум** возможным углам: $\theta = 0$ и $\theta = \pi$.

---

### 1.4. Релятивистское обобщение

Для получения уравнения, аналогичного Дирака, нужно учесть релятивистские эффекты. В уравнении Дирака появляются матрицы $\alpha$ и $\beta$, которые описывают спиновые степени свободы.

**В вашей модели:**
- Верхняя компонента биспинора $\leftrightarrow$ «северный» магнитный заряд
- Нижняя компонента $\leftrightarrow$ «южный» магнитный заряд
- Матрицы $\alpha^i$ $\leftrightarrow$ операторы поворота диполя
- Матрица $\beta$ $\leftrightarrow$ оператор «переворота» диполя

---

### 1.5. Вывод эффективного уравнения

Рассмотрим прецессию в собственной системе отсчёта электрона. Введём комплексную амплитуду прецессии:

$$
\Psi = \begin{pmatrix} \psi_+ \\ \psi_- \end{pmatrix}
$$

где:
- $\psi_+$ — амплитуда состояния с $\theta = 0$ («вверх»)
- $\psi_-$ — амплитуда состояния с $\theta = \pi$ («вниз»)

**Уравнение движения для $\Psi$:**

Из классического уравнения прецессии:
$$
\frac{d\boldsymbol{\mu}}{dt} = \gamma \boldsymbol{\mu} \times \boldsymbol{B}
$$

Перейдём к квантовому описанию. Момент $\boldsymbol{\mu}$ связан со спином:
$$
\boldsymbol{\mu} = \frac{g\mu_B}{\hbar}\hat{\boldsymbol{S}} = \mu_B \vec{\sigma}
$$

Тогда:
$$
i\hbar\frac{\partial\Psi}{\partial t} = \mu_B \vec{\sigma} \cdot \boldsymbol{B} \, \Psi
$$

Это **уравнение Паули**. Чтобы получить уравнение Дирака, нужно включить релятивистские эффекты.

---

### 1.6. Релятивистское уравнение для прецессирующего диполя

Рассмотрим движение диполя со скоростью $v$. В релятивистской механике энергия:
$$
E^2 = p^2c^2 + m^2c^4
$$

Для прецессирующего диполя добавляется спиновая энергия:
$$
E^2 = p^2c^2 + m^2c^4 + 2mc^2 \cdot \frac{\boldsymbol{\mu} \cdot \boldsymbol{B}}{mc^2} + \dots
$$

В операторной форме:
$$
\hat{H}^2 = c^2\hat{p}^2 + m^2c^4 + 2mc^2 \mu_B \vec{\sigma} \cdot \boldsymbol{B} + \dots
$$

Чтобы получить линейное уравнение (аналогично Дираку), введём матрицы $\alpha$ и $\beta$:

$$
\hat{H} = c \vec{\alpha} \cdot \hat{\vec{p}} + \beta mc^2 + \mu_B \vec{\Sigma} \cdot \boldsymbol{B}
$$

где $\vec{\Sigma}$ — матрицы спина в представлении Дирака.

---

### 1.7. Физическая интерпретация матриц в вашей модели

| Матрица Дирака | Интерпретация в вашей модели |
|----------------|-----------------------------|
| $\alpha^i = \begin{pmatrix} 0 & \sigma^i \\ \sigma^i & 0 \end{pmatrix}$ | Оператор пространственного поворота диполя |
| $\beta = \begin{pmatrix} I & 0 \\ 0 & -I \end{pmatrix}$ | Оператор «переворота» диполя (север ↔ юг) |
| $\gamma^5 = i\gamma^0\gamma^1\gamma^2\gamma^3$ | Оператор чётности прецессии |

**Уравнение Дирака в вашей модели:**
$$
\left[ i\hbar\gamma^\mu \partial_\mu - \frac{mc}{\hbar} - \frac{e}{\hbar c}\gamma^\mu A_\mu + \frac{\mu_B}{\hbar}\Sigma^\mu B_\mu \right] \Psi = 0
$$

где дополнительный член $\frac{\mu_B}{\hbar}\Sigma^\mu B_\mu$ описывает взаимодействие прецессирующего диполя с полем.

---

### 1.8. Связь с векторным потенциалом

Из вашего второго файла мы знаем, что электростатическое поле имеет векторный потенциал:
$$
\vec{A}_E = -\frac{\cot\theta}{r} \, \vec{e}_\varphi
$$

Это связано с вращением магнитных зарядов. В уравнении движения это проявляется как **геометрическая фаза** (фаза Берри):

$$
\Psi \rightarrow \Psi \exp\left( \frac{ie}{\hbar c} \oint \vec{A}_E \cdot d\vec{r} \right)
$$

Это добавляет в уравнение дополнительный член, аналогичный **спин-орбитальному взаимодействию**.

## Часть 2: Дополнительные стационарные состояния

### 2.1. Напоминание: уравнение для $\theta$

Ранее мы получили:
$$
\sin\theta \left( \mu B - I_{\text{eff}} \Omega^2 \cos\theta \right) = 0
$$

Решения:
1. $\sin\theta = 0 \Rightarrow \theta = 0, \pi$ (параллельное и антипараллельное)
2. $\cos\theta = \frac{\mu B}{I_{\text{eff}} \Omega^2}$ (дополнительные состояния)

---

### 2.2. Условие существования дополнительных состояний

Для существования решений типа 2 необходимо:
$$
\left| \frac{\mu B}{I_{\text{eff}} \Omega^2} \right| \leq 1
$$

Определим параметр:
$$
\alpha = \frac{\mu B}{I_{\text{eff}} \Omega^2}
$$

---

### 2.3. Бифуркационная диаграмма

| $\alpha$ | Стационарные состояния | Физическая интерпретация |
|----------|------------------------|--------------------------|
| $\alpha > 1$ | $\theta = 0$ | Только параллельное состояние |
| $\alpha = 1$ | $\theta = 0, \pi/2$ | Появляется боковое состояние |
| $0 < \alpha < 1$ | $\theta = 0, \arccos\alpha, \pi - \arccos\alpha$ | **Три устойчивых состояния** |
| $\alpha = 0$ | Все $\theta$ | Изотропия (нет поля) |
| $-1 < \alpha < 0$ | $\theta = \pi, \arccos\alpha, 2\pi - \arccos\alpha$ | **Три устойчивых состояния** |
| $\alpha < -1$ | $\theta = \pi$ | Только антипараллельное состояние |

---

### 2.4. Физический смысл дополнительных состояний

При $|\alpha| < 1$ появляются **наклонные состояния** прецессии:

| Состояние | Угол $\theta$ | Энергия |
|-----------|---------------|---------|
| Параллельное | $0$ | $-\mu B + \frac{1}{2}I_{\text{eff}}\Omega^2$ |
| Наклонное 1 | $\arccos\alpha$ | $-\mu B \alpha + \frac{1}{2}I_{\text{eff}}\Omega^2 (1-\alpha^2)$ |
| Наклонное 2 | $\pi - \arccos\alpha$ | $+\mu B \alpha + \frac{1}{2}I_{\text{eff}}\Omega^2 (1-\alpha^2)$ |
| Антипараллельное | $\pi$ | $+\mu B + \frac{1}{2}I_{\text{eff}}\Omega^2$ |

---

### 2.5. Устойчивость наклонных состояний

Исследуем вторую производную энергии:
$$
\frac{d^2E}{d\theta^2} = \mu B \cos\theta - I_{\text{eff}}\Omega^2 \cos 2\theta
$$

Для наклонных состояний ($\cos\theta = \alpha$):
$$
\frac{d^2E}{d\theta^2} = \mu B \alpha - I_{\text{eff}}\Omega^2 (2\alpha^2 - 1) = I_{\text{eff}}\Omega^2 (\alpha^2 + 1 - 2\alpha^2) = I_{\text{eff}}\Omega^2 (1 - \alpha^2)
$$

Поскольку $|\alpha| < 1$, то $\frac{d^2E}{d\theta^2} > 0$ — **наклонные состояния устойчивы**!

---

### 2.6. Экспериментальные следствия

Если в опыте Штерна–Герлаха использовать **слабое поле** (малое $\alpha$), то должны наблюдаться **четыре пятна** на экране вместо двух:

| Пятно | Состояние | Проекция $\mu_z$ |
|-------|-----------|-------------------|
| 1 | $\theta = 0$ | $+\mu$ |
| 2 | $\theta = \arccos\alpha$ | $+\mu\alpha$ |
| 3 | $\theta = \pi - \arccos\alpha$ | $-\mu\alpha$ |
| 4 | $\theta = \pi$ | $-\mu$ |

**Почему в классическом опыте наблюдаются только два пятна?**

Вероятно, параметры эксперимента были таковы, что $|\alpha| \approx 1$, и наклонные состояния «сливались» с основными.

---

### 2.7. Связь с квантовой механикой

В квантовой механике спин-1/2 имеет только два состояния. Но если рассмотреть **композитную структуру** электрона (как в вашей модели), то появляется возможность дополнительных состояний.

Это может объяснять:
- **Аномальный магнитный момент** электрона
- **Спиновые волны** в материалах
- **Магнитные домены** в ферромагнетиках

---

### 2.8. Возможные эксперименты

1. **Слабое поле Штерна–Герлаха:**
   - Использовать очень слабое неоднородное поле
   - Ожидать появления 4 пятен вместо 2

2. **Резонансная спектроскопия:**
   - Применить переменное поле с частотой $\omega \approx \Omega$
   - Наблюдать переходы между наклонными состояниями

3. **Интерференция прецессирующих диполей:**

   - Наблюдать интерференционную картину

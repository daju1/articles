E_t'(t) = \Re\left[\tilde{E}_t' e^{-i\omega t}\right],\quad D_t'(t) = \Re\left[\tilde{D}_t' e^{-i\omega t}\right].

$$
\langle E_t' D_t' \rangle = \frac{1}{2} \Re\left( \tilde{E}_t'^* \tilde{D}_t' \right) = \frac{1}{2} \left( \varepsilon' |E_t'|^2 \right).

## üîπ 6. Exact Derivation Through Averaged Tensor (Following Smythe, with Physics)

In Smythe's book (¬ß18.147) the following is considered:

For harmonic fields in a linear medium, the field-matter interaction energy is expressed as **difference of energies in two media**:
$$
P_n = u' - u'',
$$
where $u$ ‚Äî **averaged energy density associated with tangential field**.

But in a lossy medium:
$$
u = \frac{1}{2} \Re\left( \frac{1}{2} \mathbf{E}^* \cdot \mathbf{D} \right) = \frac{1}{4} \left( \varepsilon' |E|^2 \right).
$$

‚Üí However this is **internal field energy in volume**.

And **pressure on the interface** ‚Äî is the **difference of momentum fluxes**, not volume energy difference.

**Correct approach**: write the force on bound charges caused by **polarization discontinuity**.

Polarization:
$$
\mathbf{P} = (\varepsilon - 1)\mathbf{E} = (\varepsilon' - 1 + i\varepsilon'')\mathbf{E}.
$$

Surface density of bound charges: $\sigma_b = P_n = P_x$.

Averaging:
$$

$$

For **tangential field**, when $E_x = 0$, force appears through **interaction with the $\varepsilon$ gradient**, but it's simpler ‚Äî through the **energy-momentum tensor in complex form**.

### ‚úÖ Key Step: Symmetric Tensor in Complex Form

In works by de Groot & Suttorp, Jackson, and Rohrlich the **averaged tensor** is used:
$$
\langle T_{ij} \rangle = \frac{1}{4\pi} \Re\left[ \frac{1}{2} (\tilde{E}_i^* \tilde{E}_j + \tilde{H}_i^* \tilde{H}_j) - \frac{1}{4} \delta_{ij} (|\tilde{E}|^2 + |\tilde{H}|^2) \right].
$$

For electric field:
$$
\langle T_{xx} \rangle = \frac{1}{8\pi} \Re\left[ (\tilde{E}_x^*)^2 - \frac{1}{2} |\tilde{E}|^2 \right].
$$

If field is tangential ($E_x = 0$, $E_z \ne 0$):
$$
\langle T_{xx} \rangle = -\frac{1}{8\pi} \cdot \frac{1}{2} |\tilde{E}_t|^2 = -\frac{1}{16\pi} |\tilde{E}_t|^2.
$$

‚Üí But this is **for material with $\varepsilon = 1$**.

For $\varepsilon \ne 1$ the correct tensor includes $D$:
$$
\langle T_{xx} \rangle = -\frac{1}{8\pi} \cdot \frac{1}{2} \Re(\tilde{E}_t^* \tilde{D}_t) = -\frac{1}{16\pi} \varepsilon' |E_t|^2.
$$

Similarly, in the other medium: $-\frac{1}{16\pi} \varepsilon'' |E_t|^2$.

Then the difference:
$$
P_n = \langle T_{xx} \rangle'' - \langle T_{xx} \rangle' = -\frac{1}{16\pi} \left( \varepsilon'' |E_t''|^2 - \varepsilon' |E_t'|^2 \right).
$$

‚Üí But pressure **on medium `'`** ‚Äî is $-(\text{difference})$, i.e.:
$$
\boxed{P_n = \frac{1}{2} \left( \frac{\varepsilon'}{8\pi} |E_t'|^2 - \frac{\varepsilon''}{8\pi} |E_t''|^2 \right) }.
$$

And if using **units where $8\pi$ is already included in energy definition**, then:

$$
\boxed{P_n = \frac{1}{2} \varepsilon' |E_t'|^2 - \frac{1}{2} \varepsilon'' |E_t''|^2 }.
$$

## üîπ 7. Physical Interpretation

- $\frac{1}{2} \varepsilon' |E_t'|^2$ ‚Äî **reactive pressure**: energy stored in electric field in medium `'`,
- $\frac{1}{2} \varepsilon'' |E_t''|^2$ ‚Äî **active counter-pressure**: energy dissipated in medium `''`, which "weakens" the pressure.

If $\varepsilon'' > 0$ (losses), then the second term **reduces** total pressure.

If one medium is an **ideal dielectric**, the other a **conductor**, then:
- in conductor $\varepsilon'' \gg \varepsilon'$, $E_t'' \approx 0$ ‚Üí second term ‚âà 0,
- but actually field in conductor **decays**, and $E_t''$ ‚Äî is the field **at the surface**, which is tied to phase.

‚Üí This is exactly why **phase $\phi$ between $E$ and $H$** (or between $E$ and $D$) determines the ratio $\varepsilon', \varepsilon''$, and thereby ‚Äî the sign and magnitude of $P_n$.

# Метод классификации клиновидных щелей по числу отражений проходящего луча

Задаем размер меньшей (левой) щели и угол раствора клиновидной щели

Из центральной точки проходящей левой щели направляем луч траектории вертикально вверх с тем чтобы этот луч отразился от края внутренней поверхности клина и рассматриваем точки отражения данного луча от внутренней поверхности клина

В каждой из таких точек отражения мы можем ограничить толщину мембраны и получить таким образом клиновидные щели с разным числом отражений проходящего луча

```python
import wedge

from IPython.display import display, Math, Latex

def disp(ex):
    display(Math(latex(ex)))
```

```python
wedge.draw_normal = True
plt, state, number_of_intersections = wedge.integrand_out( \
    alpha=pi/24, r = 1, d = 12, y = 0, ang = pi/2)
plt.show(aspect_ratio = 1, axes=True)
```

$d$ толщина мембраны

$d_i$ часть толщины мембраны, приходящаяся на $i-е$ отражение

$r$ половина размера меньшей щели, которая слева

$\alpha$ образующий угол клина

$R$ половина размера большей щели, которая справа

$$d \cdot tan(\alpha) = R - r$$

1) Для клиновидной щели с одним отражением проходящего луча имеем следующие соотношения:

$$angle\_ray\,from\,normal   = 1 \cdot \alpha$$

$$angle\_ray\,from\,vertical = 2 \cdot \alpha$$

$$d = D_1 = d_1$$

$$R = R_1$$

$$r = r_1$$

$$d_1 \cdot tan(\alpha) = R_1 - r_1$$

$$\frac{d_1}{(R_1 + r_1)} = tan(2\cdot\alpha)$$

$$\frac{(\pi-passing\_angle\_y0)}{\alpha} = 5.63866204873193$$

$$\frac{(\pi-passing\_angle\_yR)}{\alpha} = 5.99999936344102$$

$$(\pi-passing\_angle\_yR) = 6 \cdot \alpha$$

$$passing\_angle\_yR = \pi - 6 \cdot \alpha$$

2) Для клиновидной щели с двумя отражениями проходящего луча имеем следующие соотношения:

$$angle\_ray\,from\,normal   = 3 \cdot \alpha$$

$$angle\_ray\,from\,vertical = 4 \cdot \alpha$$
   
$$d = D_2 = d_1+d_2$$

$$r_2 = R_1$$

$$R  = R_2$$

$$d_2 \cdot tan(\alpha) = R_2 - r_2$$

$$\frac{d_2}{(R_2 + r_2)} = tan(4 \cdot \alpha)$$

$$\frac{(\pi-passing\_angle\_y0)}{\alpha}  = 9.87006331240309$$

$$\frac{(\pi-passing\_angle\_yR)}{\alpha}  = 9.99999975734338$$

$$(\pi-passing\_angle\_yR) = 10 \cdot \alpha$$

$$passing\_angle\_yR = \pi - 10 \cdot \alpha$$

3) Для клиновидной щели с тремя отражениями проходящего луча имеем следующие соотношения:

$$angle\_ray\,from\,normal   = 5 \cdot \alpha$$

$$angle\_ray\,from\,vertical = 6 \cdot \alpha$$

$$d = D_3 = d_1+d_2+d_3$$

$$R  = R_3$$

$$d_3 \cdot tan(\alpha) = R_3 - r_3$$

$$\frac{d_3}{(R_3 + r_3)} = tan(6 \cdot \alpha)$$

$$\frac{(\pi-passing\_angle\_y0)}{\alpha} = 13.9546541689040$$

$$\frac{(\pi-passing\_angle\_yR)}{\alpha} = 13.9999999134182$$

$$(\pi-passing\_angle\_yR) = 14 \cdot \alpha$$

$$passing\_angle\_yR = pi - 14 \cdot \alpha$$

4) Для клиновидной щели с четырьмя отражениями проходящего луча имеем следующие соотношения:

$$angle\_ray\,from\,normal   = 7 \cdot \alpha$$

$$angle\_ray\,from\,vertical = 8 \cdot \alpha$$

$$d = D_4 = d_1+d_2+d_3+d_4$$

$$r_4 = R_3$$

$$R  = R_4$$

$$d_4 \cdot tan(\alpha) = R_4 - r_4$$

$$\frac{d_4}{(R_4 + r_4)} = tan(8 \cdot \alpha)$$

$$\frac{(\pi-passing\_angle\_y0)}{\alpha} = 17.9720906753169$$

$$\frac{(\pi-passing\_angle\_yR)}{\alpha} = 17.9999999457367$$

$$(\pi-passing\_angle\_yR) = 18 \cdot \alpha$$

$$passing\_angle\_yR = \pi - 18 \cdot \alpha$$

Для каждого последующего отражения имеем

[
    $R - r  = d \cdot tan(\alpha)$,
    $\frac{d}{(R + r)} = tan(2\cdot\alpha)$
]

[
    $R - r = d \cdot tan(\alpha)$,
    $R + r = \frac{d}{tan(2\cdot\alpha)}$
]

$R + r = (r + r) + (R - r)$

$R + r = 2 \cdot r + (R - r)$

$\frac{d}{tan(2 \cdot \alpha)} = 2 \cdot r + d \cdot tan(\alpha)$

$\frac{d}{tan(2 \cdot \alpha)} - d \cdot tan(\alpha) = 2\cdot r$

$d \cdot \left(\frac{1}{tan(2 \cdot \alpha)}-tan(\alpha)\right) = 2\cdot r$

$d = \frac{2\cdot r} { \left(\frac{1}{tan(2 \cdot \alpha)}-tan(\alpha)\right)}$

Таким  образом составляем систему взаимозависящих уравнений

```python
d, r, R, alpha = var("d, r, R, alpha")
```

```python
r1, r2, r3, r4 = var("r1, r2, r3, r4")
```

```python
R1, R2, R3, R4 = var("R1, R2, R3, R4")
```

```python
d1, d2, d3, d4 = var("d1, d2, d3, d4")
```

```python
eqs1 = [
    R1 == r1 + d1 * tan(alpha),
    d1 == 2*r1 / (1/tan(2*alpha)-tan(alpha)),
    r1 == r,
    ]
```

```python
eqs2 = [
    R2 == r2 + d2 * tan(alpha),
    d2 == 2*r2 / (1/tan(4*alpha)-tan(alpha)),
    r2 == R1,
    ]
```

```python
eqs3 = [
    R3 == r3 + d3 * tan(alpha),
    d3 == 2*r3 / (1/tan(6*alpha)-tan(alpha)),
    r3 == R2,
    ]
```

```python
eqs4 = [
    R4 == r4 + d4 * tan(alpha),
    d4 == 2*r4 / (1/tan(8*alpha)-tan(alpha)),
    r4 == R3,
    ]
```

Решение для клиновидной щели с одним отражением проходящего луча

```python
R1_eq = (eqs1[0]).subs(eqs1[1]).subs(eqs1[2])
disp(R1_eq)
```

Решение для клиновидной щели с двумя отражениями проходящего луча

```python
R2_eq = \
(eqs2[0]).subs(eqs2[1]).subs(eqs2[2]) \
.subs(eqs1[0]).subs(eqs1[1]).subs(eqs1[2])
disp(R2_eq)
```

Решение для клиновидной щели с тремя отражениями проходящего луча

```python
R3_eq = \
(eqs3[0]).subs(eqs3[1]).subs(eqs3[2]) \
.subs(eqs2[0]).subs(eqs2[1]).subs(eqs2[2]) \
.subs(eqs1[0]).subs(eqs1[1]).subs(eqs1[2])
disp(R3_eq)
```

Решение для клиновидной щели с 4-мя отражениями проходящего луча

```python
R4_eq = \
eqs4[0].subs(eqs4[1]).subs(eqs4[2]) \
.subs(eqs3[0]).subs(eqs3[1]).subs(eqs3[2]) \
.subs(eqs2[0]).subs(eqs2[1]).subs(eqs2[2]) \
.subs(eqs1[0]).subs(eqs1[1]).subs(eqs1[2])
R4_eq
```

```python
def calc_R1_reflection(r_, alpha_):
    R_ = R1_eq.rhs().subs([r==r_, alpha==alpha_])
    d_ = (R_ - r_)/np.tan(alpha_)

    print("r", r_,
          "d", d_,
          "R", R_,
          "alpha", (alpha_/pi*180).n())

    plt, state, number_of_intersections = wedge.integrand_out( \
        alpha=alpha_, r = r_, d = d_, y = 0, ang = pi/2)
    plt.show(aspect_ratio = 1, axes=True)

    try:
        passing_angle_yR  = \
            wedge.get_iterate_passing_angle_depending_on_y( \
                y=R_, alpha=alpha_, r=r_, d=d_)
    except:
        passing_angle_yR  = \
            wedge.get_iterate_passing_angle_depending_on_y( \
                y=0.999999*R_, alpha=alpha_, r=r_, d=d_)
    passing_angle_y0      = \
        wedge.get_iterate_passing_angle_depending_on_y( \
            y=0, alpha=alpha_, r=r_, d=d_)

    s_in_per_out = R_/r_
    passing_coefficient_y0 = (passing_angle_y0/pi * s_in_per_out).n()
    passing_coefficient_yR = (passing_angle_yR/pi * s_in_per_out).n()
    print("passing_angle_y0", (passing_angle_y0/pi*180).n())
    print("passing_angle_yR", (passing_angle_yR/pi*180).n())
    print("(pi-passing_angle_y0)/alpha", ((pi-passing_angle_y0)/alpha_).n())
    print("(pi-passing_angle_yR)/alpha", ((pi-passing_angle_yR)/alpha_).n())
    print("passing_coefficient_y0", passing_coefficient_y0)
    print("passing_coefficient_yR", passing_coefficient_yR)

    return passing_coefficient_y0, passing_coefficient_yR
```

```python
def calc_one_reflection(r_, d_):
    R_ = R_sol_one_reflection.subs([r==r_, d==d_])
    alpha_ = atan((R_-r_)/d_)

    print("r", r_,
          "d", d_,
          "R", R_,
          "alpha", (alpha_/pi*180).n())

    plt, state, number_of_intersections = wedge.integrand_out( \
        alpha=alpha_, r = r_, d = d_, y = 0, ang = pi/2)
    plt.show(aspect_ratio = 1, axes=True)
    
    try:
        passing_angle_yR  = \
            wedge.get_iterate_passing_angle_depending_on_y( \
                y=R_, alpha=alpha_, r=r_, d=d_)
    except:
        passing_angle_yR  = \
            wedge.get_iterate_passing_angle_depending_on_y( \
                y=0.999999*R_, alpha=alpha_, r=r_, d=d_)
    passing_angle_y0      = \
        wedge.get_iterate_passing_angle_depending_on_y( \
            y=0, alpha=alpha_, r=r_, d=d_)

    s_in_per_out = R_/r_
    passing_coefficient_y0 = (passing_angle_y0/pi * s_in_per_out).n()
    passing_coefficient_yR = (passing_angle_yR/pi * s_in_per_out).n()
    print("passing_angle_y0", (passing_angle_y0/pi*180).n())
    print("passing_angle_yR", (passing_angle_yR/pi*180).n())
    print("(pi-passing_angle_y0)/alpha", ((pi-passing_angle_y0)/alpha_).n())
    print("(pi-passing_angle_yR)/alpha", ((pi-passing_angle_yR)/alpha_).n())
    print("passing_coefficient_y0", passing_coefficient_y0)
    print("passing_coefficient_yR", passing_coefficient_yR)

    return passing_coefficient_y0, passing_coefficient_yR
```

```python
import numpy as np
d_linspace = np.linspace(np.float128(1), np.float128(12), 12)
d_list = d_linspace.tolist()
for i in np.arange(0, len(d_linspace), 1):
    d_ = d_list[i]
    calc_one_reflection(r_=1, d_=d_)
```

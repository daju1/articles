- $\vec{v}(t) = \frac{d\vec{r}}{dt}$ ‚Äî –æ–±—ã—á–Ω–∞—è 3-—Å–∫–æ—Ä–æ—Å—Ç—å,
- $\gamma = \frac{1}{\sqrt{1 - v^2/c^2}}$, $v = |\vec{v}|$

#### 4-—É—Å–∫–æ—Ä–µ–Ω–∏–µ:
$$
a^\mu = \frac{du^\mu}{d\tau} = \frac{d}{d\tau} u^\mu
$$

–ü–æ—Å–∫–æ–ª—å–∫—É $\tau$ –Ω–µ—É–¥–æ–±–Ω–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –Ω–∞–ø—Ä—è–º—É—é (—Ç.–∫. —Ç—Ä–∞–µ–∫—Ç–æ—Ä–∏–∏ –∑–∞–¥–∞—é—Ç—Å—è –∫–∞–∫ $\vec{r}(t)$), –ø–µ—Ä–µ–π–¥—ë–º –∫ –ø—Ä–æ–∏–∑–≤–æ–¥–Ω–æ–π –ø–æ **–∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–Ω–æ–º—É –≤—Ä–µ–º–µ–Ω–∏** $t$:

$$
a^\mu = \frac{du^\mu}{d\tau} = \frac{dt}{d\tau} \frac{du^\mu}{dt} = \gamma \frac{du^\mu}{dt}
$$

### üî∑ 2. –û–±—â–µ–µ –≤—ã—Ä–∞–∂–µ–Ω–∏–µ –¥–ª—è 4-—É—Å–∫–æ—Ä–µ–Ω–∏—è

–†–∞–∑–ª–æ–∂–∏–º $u^\mu = \gamma(c,\ \vec{v})$, —Ç–æ–≥–¥–∞:

$$
a^\mu = \gamma \frac{d}{dt} \left( \gamma c,\ \gamma \vec{v} \right) = \left( \gamma \frac{d}{dt}(\gamma c),\ \gamma \frac{d}{dt}(\gamma \vec{v}) \right)
$$

–í—ã—á–∏—Å–ª–∏–º –∫–æ–º–ø–æ–Ω–µ–Ω—Ç—ã.

---

#### ‚úÖ –í—Ä–µ–º–µ–Ω–Ω–∞—è –∫–æ–º–ø–æ–Ω–µ–Ω—Ç–∞ ($\mu = 0$):

$$
a^0 = \gamma \cdot c \cdot \frac{d\gamma}{dt}
$$

–ù–∞–π–¥—ë–º $\frac{d\gamma}{dt}$:

$$
\gamma = \left(1 - \frac{v^2}{c^2}\right)^{-1/2} \quad \Rightarrow \quad
\frac{d\gamma}{dt} = \frac{1}{c^2} \gamma^3 (\vec{v} \cdot \dot{\vec{v}})
$$

–≥–¥–µ $\dot{\vec{v}} = \frac{d\vec{v}}{dt} = \vec{a}$ ‚Äî –æ–±—ã—á–Ω–æ–µ 3-—É—Å–∫–æ—Ä–µ–Ω–∏–µ.

$$

$$

---

#### ‚úÖ –ü—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–µ–Ω–Ω—ã–µ –∫–æ–º–ø–æ–Ω–µ–Ω—Ç—ã ($\mu = 1,2,3$):

$$

$$

$$

### üî∑ 3. –ò—Ç–æ–≥: –æ–±—â–∞—è —Ñ–æ—Ä–º—É–ª–∞ 4-—É—Å–∫–æ—Ä–µ–Ω–∏—è

### üî∑ 5. –ò–Ω–≤–∞—Ä–∏–∞–Ω—Ç $a^\mu a_\mu$

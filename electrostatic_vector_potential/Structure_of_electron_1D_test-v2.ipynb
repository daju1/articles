# Расчёт кулоновской силы в одномерной модели взаимодействия двух электронов как прецессирующих магнитных диполей

Эффективный электрический заряд внешней кольца магнитного тока $e$

Расстояние от центра электрона до центра кольца магнитного тока $a$

Толцина кольца магнитного тока $h$

Расстояние между центрами двух электронов $r$

```python
from IPython.display import display, Math, Latex

def disp(ex):
    display(Math(latex(ex)))
    
def displ(tex):
    display(Math(tex))
```

```python
var("e,r,a,h")
```

```python
F_11 = -1/(r-a-h-a-h)^2
F_22 = -1/(r-a+h-a+h)^2
F_33 = -1/(r+a-h+a-h)^2
F_44 = -1/(r+a+h+a+h)^2

F_12 = +1/(r-a-h-a+h)^2
F_21 = +1/(r-a+h-a-h)^2

F_13 = +1/(r-a-h+a-h)^2
F_31 = +1/(r+a-h-a-h)^2

F_14 = -1/(r-a-h+a+h)^2
F_41 = -1/(r+a+h-a-h)^2

F_23 = -1/(r-a+h+a-h)^2
F_32 = -1/(r+a-h-a+h)^2

F_24 = +1/(r-a+h+a+h)^2
F_42 = +1/(r+a+h-a+h)^2

F_34 = +1/(r+a-h+a+h)^2
F_43 = +1/(r+a+h+a-h)^2
```

```python
M = matrix([[F_11,F_12,F_13,F_14],
            [F_21,F_22,F_23,F_24],
            [F_31,F_32,F_33,F_34],
            [F_41,F_42,F_43,F_44]])
```

```python
disp(M)
```

Выражение для суммарной силы взаимодействия двух атомов находим как сумму всех элементов матрицы

```python
F = sum(sum(M))
```

```python
disp(F)
```

```python
disp(F.expand())
```

```python
disp(F.factor())
```

Взаимодействие исчезает при устремлении размера прецессирующего магнитного диполя к нулю

```python
disp(limit(F, a=0))
```

```python
disp(limit(F, h=0))
```

```python
A = 0.1
H = 0.01
R = 0.3
```

Рассматриваемые модели электронов как прецессирующих магнитных диполей отталкиваются

```python
eps = 0.02
plot(F.subs([a == A, h == H]), r, 2*(A+H)+eps, R)
```

Исследуем вопрос, как сила отталкивания отличается от закона обратного расстояния в степени $n$

```python
A = 0.1
H = 0.01
R = 10
for n in range(2,8):
    F_rn = (F*r^n).factor()
    #disp (F_rn)
    eps = 0.25
    plot(F_rn.subs([a == A, h == H]), r, 2*(A+H)+eps, R).show(title = "n=" + str(n))
```

Судя по графикам, на больших расстояниях сила отталкивания соответствует закону обратного расстояния в 6 степени

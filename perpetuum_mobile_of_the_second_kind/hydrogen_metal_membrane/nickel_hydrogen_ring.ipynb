# Десорбция как дифракция волн де Бройля молекул водорода на кристаллической решётке

Рассматривая волновую функцию равномерно движущейся частицы (например электрона, протона, нейтрона, иона или атома) как суперпозицию волновых функций рассматриваемых как плоские волны, но имеющие разные импульсы и, следовательно, разные длины волн де Бройля, можно записать эту суперпозицию как интеграл Фурье по пространству импульсов от минус бесконечности до плюс бесконечности для волновых функций всего диапазона импульсов но с разными коэффициентами Фурье.

Затем можно сформировать волновой пакет рассмотрев эту суперпозицию волновых функций как такое их множество, которое тяготеет к некоторому значению среднего импульса.

Потом произведя ряд преобразований этого интеграла можно показать, что он сходится к функции вида $\frac{sin(x)}{x}$. И можно вспомнить, что такая же самая функция имеется в курсе оптики, например, Сивухин. Оптика на странице Дифракция Фраунгофера есть формула $\frac{sin(\alpha)}{\alpha}$.

Какова причина того, что формулы для волнового пакета и для дифракции на щели идентичны? Где та самая щель, интерферируя на которой частица-волна получилась такого же самого вида как при дифракции Фраунгофера?

Дело в том, что эту частицу нужно как-то приготовить, в каком то приборе. Получается, когда эта частица вылетала из кристаллической решетки этого прибора, она интерферировала на его кристаллической решетке и поэтому получила такой же самый вид волнового пакета как получает свет при дифракции Фраунгофера?

Угловое распределение молекул водорода десорбирующихся с поверхности монокристалла металла зависит от того с какой именно грани $hkl$ кристалла этот водород десорбируется. Если грань имеет поверхность плотно упакованную атомами металла как например поверхность $Ni(111)$, то угловое распределение молекул водорода имеет узкий пик $sin^5 \theta$, а если грань имеет более рыхлую упаковку, как например грань $Ni(110)$, то угловое распределение десорбирующихся молекул водорода имеет более широкий спектр углов разлетания молекул $sin \theta$.

Отсюда возникает идея, что процесс десорбции водорода с поверхности регулируется в том числе квантово дифракционными закономерностями.

Кстати, тот факт, что десорбирующиеся с кристаллически более плотно упакованной поверхности $Ni(111)$ молекулы водорода трансляционно более "горячие" то есть имеют большее значение импульса также может указывать на дифракционную природу десобции молекул водорода.

Чем более плотно упакованы атомы на поверхности, тем меньше постоянная дифракционной решетки, тем, соответственно, меньше длина чередования полос дифракционной картины, а проводя аналогию с волновым пакетом десорбирующейся молекулы тем соответственно меньше длина волны де Бройля этого волнового пакета и соответственно тем самым больше величина импульса десорбирующейся молекулы, что как раз и согласуется с экспериментальным фактом того, что десорбирующиеся с кристаллографически более плотно упакованной поверхности молекулы водорода трансляционно более горячие, то есть имеют более высокую скорость.

# Об обратном градиенте давления в области спая внутри несимметричной биметаллической мембраны

Согласно классическим термодинамическим представлениям нужно рассмотреть влияние возможного обратного градиента давления (или обратного натяжения), которое могло бы возникнуть в области спая этих двух монокристаллов с разными hkl.

Для того чтобы понять что произойдёт, если мы, не меняя степени насыщения металла водородом, обратно слепим два монокристалла в спай, нужно себе уяснить следующий вопрос: а какова физическая причина того, что для для одного и того же металла, при одной и той же температуре металла, насыщенного водородом, для одной и той же степени насыщения металла водородом, но для разных поверхностей среза hkl монокристалла получаются разные равновесные давления водорода в газовой фазе. 

Здесь нужно указать на одну из, на мой взгляд, важных физических причин этого: кинетика процесса десорбции определяется в том числе и таким фактором как дифракция волн де Бройля молекул водорода на кристаллической решётке, которую эти молекулы пытаются покинуть. А параметры дифракции, как известно определяются шагом дифракционной решётки, который разный для разных hkl.

Кроме того, для диффузии через спай небходимость образовывать молекулу водорода отпадает. Атомарный водород внутри металла диффундирует между двумя спаянными пластинами никеля. Поскольку на кинетику диффузии водорода будут в значительной мере влиять явления дифракции, которые со своей стороны зависят от ориентации $hkl$ кристалла.

Можно говорить о некоторой аналогии  процесса диффузии водорода с процессом происходящем при рентгеноструктурном анализе кристалла. А именно: волна де Бройля атомов водорода при своем туннелировании через потенциальные барьеры диффузии в кристаллической решетке металла может вести себя в плане отражения от внутрикристаллических воображаемых $hkl$ плоскостей наподобие тому как это делают рентгеновские кванты.

В зависимости от соотношения тепловой волны де Бройля атомов водорода и шага между плоскостями отражения разных направлений $hkl$ внутри кристалла, некоторые пространственные направления диффузии водорода внутри кристалла могут оказаться вследствии процесса внутрикристаллической дифракции волн де Бройля более вероятными. Таким образом по разные стороны спая направления наиболее вероятного диффузного потока будут отличаться.

В области спая движущиеся навстречу (но под разными углами) преимущественно направленные дуффузные потоки будут претерпевать явления отражения и преломления волн де Бройля диффундирующих атомов водорода, а теоретическое исследование коэффициентов отражения и преломления этих волн позволит предсказать знак и величину градиента давления, образующегося на этой границе.

И здесь уже благодаря учёту разницы преимущественных направлений диффузии по разные стороны спая, кроме аналога первого потенциала Вольты (возникающего в результате разности квантовых уровней энергии атома водорода внутри кристаллической решетки по разные стороны спая) и аналога второго потенциала Вольты (определяемая различием концентраций атомов водорода) 

# К вопросу об отклонении от принципа детального равновесия

Для работоспособности кольцара нужно создать условия, чтобы существовала граница фаз на которой бы принцип детального равновесия нарушается.

В этом случае противодавление внутри мембраны не сможет полностью скомпенсировать ожидаемый нами эффект.

А вот каковы условия нарушения принципа детального равновесия?

На мой взгляд детальное равновесие будет нарушаться на границе квантового и классически курпускулярного движения.

Водород взаимодействуя с поверхностью никеля,  например показывает способность образовывать дифракционные максимумы.

Это говорит, что для водорода длина волны де Бройля сопоставима с постоянной решётки.

Десорбция водорода с поверхности (111) процесс на который оказывает существенное влияние дифракция волны де Бройля молекул водорода на кристаллической решётке.

Отсюда угловое распределение десорбирующихся молекул пропорционально косинусу в пятой степени. И десорбируются с этой поверхности молекулы трансляционно горячие.

Угловое распределение молекул которые будут пытаться адсорбироваться на этой поверхности распределены по другому (ближе к классическому распределению Максвелла).

И поэтому мы здесь имеем границу квантового и классического эффектов в результате чего детальное равновесие нарушается.

```python
def data_str(str1, m):
    import math
    lst = str1.split("\n")
    data = []
    i = -1

    for row in lst:
        if len(row) > 0 and row[0] != '#':
            splitted_row = row.split(",")
            # print (splitted_row)
            
            if "\"" in splitted_row[2]:
                n_points = int(splitted_row[0])
                t = float(splitted_row[2].strip("\""))
                data += [{"temperature":t,
                          "n_points":n_points,
                          "isoterm":[]}]
                i+=1
                continue
                
            p = float(splitted_row[0])
            d = float(splitted_row[1])
            t = float(splitted_row[2])
            #print(p,d,T)

            data[i]["isoterm"] += [(math.log10(p), m*d)]
    return data

def list_plot_str(data, color="blue", T=None):
    plt = Graphics()
    for isoterm in data:
        if T is None or isoterm["temperature"] == T:
            plt += list_plot(isoterm["isoterm"], color=color)
    return plt
```

приводится формула для потока через мембрану

$J = \Phi \left(P_u^{1/2} - P_d^{1/2} \right) \big / L$

$J =$ steady state hydrogen flux $\left(mol \,\,\, of \,\,\, H_2 \big/ \left(m^2 \cdot s\right)\right)$

$\Phi =$ hydrogen permeability $\left(mol \,\,\, of \,\, H_2 \big/ \left(m \cdot s \cdot Pa^{1/2}\right)\right)$

$P_u, P_d$ hydrogen pressures at the upstream side and the downstream side $Pa$

В этой же работе приводится график

Из этого графика для температуры $T = 110 ^{\circ} C = 383 K$ для палладиевой мембраны

$\Phi = 8 \cdot 10^{-9} mol \,\, H_2 \cdot m^{-1} \cdot s^{-1} \cdot Pa^{-1/2}$

для мембраны из сплава ванадия с 10 % никеля

$\Phi = 5 \cdot 10^{-8} mol \,\, H_2 \cdot m^{-1} \cdot s^{-1} \cdot Pa^{-1/2}$

исходя из этих данных можно вычислить стационарный поток через при указанной разности давлений 

$J = \Phi \left(P_u^{1/2} - P_d^{1/2} \right) \big / L$

```python
Phi_Pd     = 8 * 10^(-9)    # mol \,\, H_2 \cdot m^{-1} \cdot s^{-1} \cdot Pa^{-1/2}$
Phi_V_10Ni = 5 * 10^(-8)    # mol \,\, H_2 \cdot m^{-1} \cdot s^{-1} \cdot Pa^{-1/2}$
p_Pd_111   = 0.0001056      # Pa
p_Pd_110   = 0.000529       # Pa
L          = 100*10^(-6)    # m - толщина мембраны 100 микрон
R          = 8.31446262     # м^2 кг с^-2 К^-1 Моль^-1
T          = 383            # K
```

$J =$ steady state hydrogen flux $\left(mol \,\,\, of \,\,\, H_2 \big/ \left(m^2 \cdot s\right)\right)$

```python
J_Pd = Phi_Pd * (sqrt(p_Pd_111) - sqrt(p_Pd_110))/L
J_Pd
```

```python
J_V_10Ni = Phi_V_10Ni * (sqrt(p_Pd_111) - sqrt(p_Pd_110))/L
J_V_10Ni
```

```python
J = Phi_V_10Ni * (sqrt(p_Pd_110_k) - sqrt(p_Pd_111_k))/L
J
```

$v = \frac{J}{c} = J \frac{R T}{p}$

Для поверхностей характеризующихся кинетикой десорбции второго порядка можно записать следующее выражение для концентрации молекул водорода в газовой фазе через предэкспоненциальный множитель кинетики десорбции (Frequency factor, $\nu_{2} (cm^{2} s^{-1} {H-atom}^{-1})$), коэффициент прилипания $s_0$, активационный барьер процесса десорбции $\Delta E^{*}_{d}\left(\Theta\right)$, поверхностную концентрацию адсорбированных атомов водорода $n_{ad}$ и среднюю скорость молекул водорода в газовой фазе ${\overset{-}c}$

$s_0\overset{-}c n_g = \nu_2 \left(\Theta\right) \, exp \left(-\frac{\Delta E^{*}_{d} \left(\Theta\right)}{kT} \right) n_{ad}^2$

$n_g = \frac{\nu_2 \left(\Theta\right)}{s_0} \, exp \left(-\frac{\Delta E^{*}_{d} \left(\Theta\right)}{kT} \right) \frac{n_{ad}^2}{\overset{-}c}$

$$\begin{array}{rrrr}
Metal      & \nu_{2}                &    s_0            & \Delta E^{*}_{d} \\
(hkl)      & \frac{cm^{2}}{s \, at} &                   &       kJ/mol     \\
Ni(100)    & 3 \times 10^{0}        & 0.06-0.1 (120 K)  & ...              \\
           & 2.5 \times 10^{-1}     &                   & ...              \\
           & 8 \times 10^{-2}       &                   & ...              \\
           & 5 \times 10^{-2}       &                   & ...              \\
Ni(111)    & 2 \times 10^{-1}       & 0.05 (120 K)      & ...              \\
           & 2.3 \times 10^{-2}     & 0.1  (140 K)      & ...              \\
Pd(100)    & 8 \times 10^{-3}..1 \times 10^{-2}       & 0.5  (120 K)      & 24.5 kcal/mol             \\
Pd(110)    & 8 \times 10^{-2} (368 K)       & 0.7  (120 K)      & 23 kcal/mol              \\
\end{array}$$

```python
E_d_Pd_100 = 24.5 * 4.2 * 1000
E_d_Pd_110 = 23 * 4.2 * 1000

nu2_Pd_100 = 8*10^-3
nu2_Pd_110 = 8*10^-2

s_0_Pd_100 = 0.5
s_0_Pd_110 = 0.7

R = 8.314
T = 300

n_s = 1.8*10^15
```

```python
def n_g(nu2, s_0, E_d, T, n_s):
    return nu2 / s_0 * exp(-E_d/(R*T))*n_s^2
```

```python
n_g(nu2_Pd_100, s_0_Pd_100, E_d_Pd_100, T, n_s)
```

```python
n_g(nu2_Pd_110, s_0_Pd_110, E_d_Pd_110, T, n_s)
```

$n_g = \frac{\nu_1}{s_0} \, exp \left(-\frac{\Delta E^{*}_{d}}{kT} \right) \frac{n_{ad}}{2\overset{-}c}$

Давайте разбираться и искать аналогии. 

Начнём с "заряда".  Заряд электронов изменяется в кулонах.  Физико химический аналог заряда - это количество атомов водорода. Его можно измерять в штуках, можно в молях, а можно, используя число Фарадея перевести количество атомов в кулоны.  Нам ничего не мешает это сделать.  При этом мы чётко понимаем, что 1 "кулон" "заряда" атомов водорода электрически все таки нейтрален. И такой "кулон" с электрически заряженным кулоном тех же электронов или ионов мы не путаем.

Дальше нам нужна физикохимическая аналогия одного вольта. Поупражнявшись на бумаге с формулами

я увидел следующую аналогию. Выделим в сечении мембраны область бесконечно малой толщины. В этой области мы найдем определенное количество атомов водорода растворенных в металле.

 Просуммируем равнодействующую всех сил приложенных к атомам водорода в этом сечении вдоль оси направления движения водорода (вдоль оси мембраны). И разделим получившуюся сумму сил на площадь мембраны $S$. Мы при этом получим нормальную компоненту натяжения $\sigma_{zz}$ в Паскалях приложенную к атомам водорода. И эта компонента натяжения может быть положительной, отрицательной и нулевой. Если нормальная компонента натяжения нулевая (силы приложенные к атомам водорода с противоположных сторон скомпенсированы), то мы имеем дело с установившимся термодинамическим равновесием. А если эта компонента натяжения ненулевая, то значит имеет место неравновесное состояние, например, направленная диффузия вызванная перепадом внешнего давления. Или это может быть локальный перепад давления возникший в результате флуктуации.

Теперь мы посчитаем концентрацию атомов водорода $c$ в этом бесконечно тонком сечении. Но сейчас мы эту концентрацию будем выражать не в атомах на кубометр и не в молях на кубометра а в наших так называемых "кулонах" на кубометр.

 Далее разделив натяжение  $\sigma_{zz}$ в Паскалях, приложенное к атомам к концентрации этих атомов (в "кулонах" на кубический метр) мы получаем водородно-мембранный аналог "вольтов"

$V = \frac{\sigma_{zz}}{c}$

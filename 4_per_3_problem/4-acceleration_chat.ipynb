- $\vec{v}(t) = \frac{d\vec{r}}{dt}$ ‚Äî –æ–±—ã—á–Ω–∞—è 3-—Å–∫–æ—Ä–æ—Å—Ç—å,
- $\gamma = \frac{1}{\sqrt{1 - v^2/c^2}}$, $v = |\vec{v}|$

#### 4-—É—Å–∫–æ—Ä–µ–Ω–∏–µ:
$$
a^\mu = \frac{du^\mu}{d\tau} = \frac{d}{d\tau} u^\mu
$$

–ü–æ—Å–∫–æ–ª—å–∫—É $\tau$ –Ω–µ—É–¥–æ–±–Ω–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –Ω–∞–ø—Ä—è–º—É—é (—Ç.–∫. —Ç—Ä–∞–µ–∫—Ç–æ—Ä–∏–∏ –∑–∞–¥–∞—é—Ç—Å—è –∫–∞–∫ $\vec{r}(t)$), –ø–µ—Ä–µ–π–¥—ë–º –∫ –ø—Ä–æ–∏–∑–≤–æ–¥–Ω–æ–π –ø–æ **–∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–Ω–æ–º—É –≤—Ä–µ–º–µ–Ω–∏** $t$:

$$
a^\mu = \frac{du^\mu}{d\tau} = \frac{dt}{d\tau} \frac{du^\mu}{dt} = \gamma \frac{du^\mu}{dt}
$$

---

### üî∑ 2. –û–±—â–µ–µ –≤—ã—Ä–∞–∂–µ–Ω–∏–µ –¥–ª—è 4-—É—Å–∫–æ—Ä–µ–Ω–∏—è

–†–∞–∑–ª–æ–∂–∏–º $u^\mu = \gamma(c,\ \vec{v})$, —Ç–æ–≥–¥–∞:

$$
a^\mu = \gamma \frac{d}{dt} \left( \gamma c,\ \gamma \vec{v} \right) = \left( \gamma \frac{d}{dt}(\gamma c),\ \gamma \frac{d}{dt}(\gamma \vec{v}) \right)
$$

–í—ã—á–∏—Å–ª–∏–º –∫–æ–º–ø–æ–Ω–µ–Ω—Ç—ã.

---

#### ‚úÖ –í—Ä–µ–º–µ–Ω–Ω–∞—è –∫–æ–º–ø–æ–Ω–µ–Ω—Ç–∞ ($\mu = 0$):

$$
a^0 = \gamma \cdot c \cdot \frac{d\gamma}{dt}
$$

–ù–∞–π–¥—ë–º $\frac{d\gamma}{dt}$:

$$
\gamma = \left(1 - \frac{v^2}{c^2}\right)^{-1/2} \quad \Rightarrow \quad
\frac{d\gamma}{dt} = \frac{1}{c^2} \gamma^3 (\vec{v} \cdot \dot{\vec{v}})
$$

–≥–¥–µ $\dot{\vec{v}} = \frac{d\vec{v}}{dt} = \vec{a}$ ‚Äî –æ–±—ã—á–Ω–æ–µ 3-—É—Å–∫–æ—Ä–µ–Ω–∏–µ.

–¢–æ–≥–¥–∞:

$$
a^0 = \gamma \cdot c \cdot \frac{\gamma^3}{c^2} (\vec{v} \cdot \vec{a}) = \frac{\gamma^4}{c} (\vec{v} \cdot \vec{a})
$$

---

#### ‚úÖ –ü—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–µ–Ω–Ω—ã–µ –∫–æ–º–ø–æ–Ω–µ–Ω—Ç—ã ($\mu = 1,2,3$):

$$
a^i = \gamma \frac{d}{dt} (\gamma v^i) = \gamma \left( \dot{\gamma} v^i + \gamma \dot{v}^i \right)
$$

–ü–æ–¥—Å—Ç–∞–≤–∏–º $\dot{\gamma} = \frac{\gamma^3}{c^2} (\vec{v} \cdot \vec{a})$:

$$
a^i = \gamma \left( \frac{\gamma^3}{c^2} (\vec{v} \cdot \vec{a}) v^i + \gamma a^i \right)
= \frac{\gamma^4}{c^2} (\vec{v} \cdot \vec{a}) v^i + \gamma^2 a^i
$$

–¢–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º, –≤ –≤–µ–∫—Ç–æ—Ä–Ω–æ–π —Ñ–æ—Ä–º–µ:

$$
\vec{a}^{\ (sp)} = \frac{\gamma^4}{c^2} (\vec{v} \cdot \vec{a}) \vec{v} + \gamma^2 \vec{a}
$$

---

### üî∑ 3. –ò—Ç–æ–≥: –æ–±—â–∞—è —Ñ–æ—Ä–º—É–ª–∞ 4-—É—Å–∫–æ—Ä–µ–Ω–∏—è

$$
\boxed{
a^\mu = \left(
\frac{\gamma^4}{c} (\vec{v} \cdot \vec{a}),\ 
\frac{\gamma^4}{c^2} (\vec{v} \cdot \vec{a}) \vec{v} + \gamma^2 \vec{a}
\right)
}
$$

–≥–¥–µ:
- $\vec{v}(t)$ ‚Äî 3-—Å–∫–æ—Ä–æ—Å—Ç—å,
- $\vec{a}(t) = \frac{d\vec{v}}{dt}$ ‚Äî 3-—É—Å–∫–æ—Ä–µ–Ω–∏–µ,
- $\gamma = \left(1 - \frac{v^2}{c^2}\right)^{-1/2}$

---

### üî∑ 4. –°–≤–æ–π—Å—Ç–≤–∞ 4-—É—Å–∫–æ—Ä–µ–Ω–∏—è

1. **–û—Ä—Ç–æ–≥–æ–Ω–∞–ª—å–Ω–æ—Å—Ç—å 4-—Å–∫–æ—Ä–æ—Å—Ç–∏ –∏ 4-—É—Å–∫–æ—Ä–µ–Ω–∏—è**:
   $$
   a_\mu u^\mu = 0
   $$
   –≠—Ç–æ —Å–ª–µ–¥—É–µ—Ç –∏–∑ —Ç–æ–≥–æ, —á—Ç–æ $u^\mu u_\mu = -c^2 = \text{const}$, –¥–∏—Ñ—Ñ–µ—Ä–µ–Ω—Ü–∏—Ä—É—è –ø–æ $\tau$, –ø–æ–ª—É—á–∞–µ–º $a_\mu u^\mu = 0$.

2. **–ò–Ω–≤–∞—Ä–∏–∞–Ω—Ç–Ω—ã–π –∫–≤–∞–¥—Ä–∞—Ç 4-—É—Å–∫–æ—Ä–µ–Ω–∏—è**:
   $$
   a^\mu a_\mu = -\frac{1}{c^2} (a^0)^2 + (\vec{a}^{\ (sp)})^2
   $$
   –ü–æ—Å–ª–µ –ø–æ–¥—Å—Ç–∞–Ω–æ–≤–∫–∏ –ø–æ–ª—É—á–∞–µ—Ç—Å—è:
   $$
   a^\mu a_\mu = \gamma^6 \left( a^2 - \frac{(\vec{v} \times \vec{a})^2}{c^2} \right)
   $$
   (–ø–æ–¥—Ä–æ–±–Ω—ã–π –≤—ã–≤–æ–¥ –Ω–∏–∂–µ)

---

### üî∑ 5. –ò–Ω–≤–∞—Ä–∏–∞–Ω—Ç $a^\mu a_\mu$

–í—ã—á–∏—Å–ª–∏–º –∫–≤–∞–¥—Ä–∞—Ç 4-—É—Å–∫–æ—Ä–µ–Ω–∏—è:

$$

$$

–ü–æ–¥—Å—Ç–∞–≤–∏–º:

$$
(a^0)^2 = \frac{\gamma^8}{c^2} (\vec{v} \cdot \vec{a})^2
\quad \Rightarrow \quad
-\frac{(a^0)^2}{c^2} = -\frac{\gamma^8}{c^4} (\vec{v} \cdot \vec{a})^2
$$

$$
|\vec{a}^{\ (sp)}|^2 = \left| \frac{\gamma^4}{c^2} (\vec{v} \cdot \vec{a}) \vec{v} + \gamma^2 \vec{a} \right|^2
= \frac{\gamma^8}{c^4} (\vec{v} \cdot \vec{a})^2 v^2 + 2 \frac{\gamma^6}{c^2} (\vec{v} \cdot \vec{a})(\vec{v} \cdot \vec{a}) + \gamma^4 a^2
$$

$$
= \frac{\gamma^8}{c^4} (\vec{v} \cdot \vec{a})^2 v^2 + 2 \frac{\gamma^6}{c^2} (\vec{v} \cdot \vec{a})^2 + \gamma^4 a^2
$$

–°–∫–ª–∞–¥—ã–≤–∞–µ–º:

$$
a^\mu a_\mu = -\frac{\gamma^8}{c^4} (\vec{v} \cdot \vec{a})^2 + \frac{\gamma^8}{c^4} (\vec{v} \cdot \vec{a})^2 v^2 + 2 \frac{\gamma^6}{c^2} (\vec{v} \cdot \vec{a})^2 + \gamma^4 a^2
$$

$$
= \frac{\gamma^8}{c^4} (\vec{v} \cdot \vec{a})^2 (v^2 - c^2) + 2 \frac{\gamma^6}{c^2} (\vec{v} \cdot \vec{a})^2 + \gamma^4 a^2
$$

–ó–∞–º–µ—Ç–∏–º, —á—Ç–æ $v^2 - c^2 = -c^2 (1 - v^2/c^2) = -c^2 / \gamma^2$, —Ç–æ–≥–¥–∞:

$$
\frac{\gamma^8}{c^4} (\vec{v} \cdot \vec{a})^2 (-c^2 / \gamma^2) = -\frac{\gamma^6}{c^2} (\vec{v} \cdot \vec{a})^2
$$

–¢–æ–≥–¥–∞:

$$
a^\mu a_\mu = -\frac{\gamma^6}{c^2} (\vec{v} \cdot \vec{a})^2 + 2 \frac{\gamma^6}{c^2} (\vec{v} \cdot \vec{a})^2 + \gamma^4 a^2 = \frac{\gamma^6}{c^2} (\vec{v} \cdot \vec{a})^2 + \gamma^4 a^2
$$

–¢–µ–ø–µ—Ä—å –∏—Å–ø–æ–ª—å–∑—É–µ–º —Ç–æ–∂–¥–µ—Å—Ç–≤–æ:

$$
a^2 = \frac{(\vec{v} \cdot \vec{a})^2}{v^2} + \frac{|\vec{v} \times \vec{a}|^2}{v^2} \quad \text{(–µ—Å–ª–∏ } v \ne 0\text{)}
$$

–ù–æ –ª—É—á—à–µ ‚Äî –ø–µ—Ä–µ–ø–∏—Å–∞—Ç—å –≤ –≤–∏–¥–µ:

$$
a^\mu a_\mu = \gamma^6 \left( a^2 - \frac{(\vec{v} \times \vec{a})^2}{c^2} \right)
$$

‚úÖ –≠—Ç–æ ‚Äî **—Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–∞—è —Ñ–æ—Ä–º—É–ª–∞** –¥–ª—è –∫–≤–∞–¥—Ä–∞—Ç–∞ 4-—É—Å–∫–æ—Ä–µ–Ω–∏—è.

### üî∑ 6. –ü—Ä–∏–º–µ–Ω–µ–Ω–∏–µ: –∏–∑–ª—É—á–µ–Ω–∏–µ –∑–∞—Ä—è–¥–∞

–ú–æ—â–Ω–æ—Å—Ç—å –∏–∑–ª—É—á–µ–Ω–∏—è (—Ñ–æ—Ä–º—É–ª–∞ –õi√©narda):

### üî∑ 7. –ü—Ä–∏–º–µ—Ä: –∫—Ä—É–≥–æ–≤–æ–µ –¥–≤–∏–∂–µ–Ω–∏–µ

### üî∑ 3. –§–æ—Ä–º—É–ª–∞ –¥–ª—è $a_\mu R^\mu$

### ‚úÖ –ó–∞–∫–ª—é—á–µ–Ω–∏–µ

### üî∑ 1. –û–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ 4-—É—Å–∫–æ—Ä–µ–Ω–∏—è

**4-—É—Å–∫–æ—Ä–µ–Ω–∏–µ** ‚Äî —ç—Ç–æ –ø—Ä–æ–∏–∑–≤–æ–¥–Ω–∞—è **4-—Å–∫–æ—Ä–æ—Å—Ç–∏** –ø–æ **—Å–æ–±—Å—Ç–≤–µ–Ω–Ω–æ–º—É –≤—Ä–µ–º–µ–Ω–∏** $\tau$:

$$

$$

–≥–¥–µ:
- $u^\mu = \frac{dx^\mu}{d\tau}$ ‚Äî **4-—Å–∫–æ—Ä–æ—Å—Ç—å**,
- $x^\mu(\tau)$ ‚Äî –º–∏—Ä–æ–≤–∞—è –ª–∏–Ω–∏—è —á–∞—Å—Ç–∏—Ü—ã,
- $\tau$ ‚Äî **—Å–æ–±—Å—Ç–≤–µ–Ω–Ω–æ–µ –≤—Ä–µ–º—è**, —Å–≤—è–∑–∞–Ω–Ω–æ–µ —Å –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–Ω—ã–º –≤—Ä–µ–º–µ–Ω–µ–º $t$ —Å–æ–æ—Ç–Ω–æ—à–µ–Ω–∏–µ–º:
  $$
  d\tau = dt \sqrt{1 - \frac{v^2(t)}{c^2}} = \frac{dt}{\gamma(t)}
  $$

### üî∑ 2. 4-—Å–∫–æ—Ä–æ—Å—Ç—å

–ü—É—Å—Ç—å —á–∞—Å—Ç–∏—Ü–∞ –¥–≤–∏–∂–µ—Ç—Å—è –ø–æ —Ç—Ä–∞–µ–∫—Ç–æ—Ä–∏–∏ $\vec{r}(t)$ –≤ –ª–∞–±–æ—Ä–∞—Ç–æ—Ä–Ω–æ–π —Å–∏—Å—Ç–µ–º–µ. –¢–æ–≥–¥–∞:

\quad \text{–≥–¥–µ} \quad

### üî∑ 3. –û–±—â–∞—è —Ñ–æ—Ä–º—É–ª–∞ 4-—É—Å–∫–æ—Ä–µ–Ω–∏—è

–ß—Ç–æ–±—ã –Ω–∞–π—Ç–∏ $a^\mu = \frac{d u^\mu}{d\tau}$, –∏—Å–ø–æ–ª—å–∑—É–µ–º —Ü–µ–ø–Ω–æ–µ –ø—Ä–∞–≤–∏–ª–æ:

$$
a^\mu = \frac{d u^\mu}{d\tau} = \frac{dt}{d\tau} \frac{d u^\mu}{dt} = \gamma \frac{d u^\mu}{dt}
$$

–¢–µ–ø–µ—Ä—å –≤—ã—á–∏—Å–ª–∏–º –ø—Ä–æ–∏–∑–≤–æ–¥–Ω—ã–µ.

---

#### ‚úÖ –í—Ä–µ–º–µ–Ω–Ω–∞—è –∫–æ–º–ø–æ–Ω–µ–Ω—Ç–∞ ($\mu = 0$):

$$
u^0 = \gamma c \quad \Rightarrow \quad
\frac{d u^0}{dt} = c \frac{d\gamma}{dt}
$$

$$
\frac{d\gamma}{dt} = \frac{d}{dt} \left(1 - \frac{v^2}{c^2}\right)^{-1/2} = \frac{1}{c^2} \gamma^3 (\vec{v} \cdot \vec{a})
\quad \text{(–≥–¥–µ } \vec{a} = \dot{\vec{v}} = d\vec{v}/dt\text{)}
$$

–¢–æ–≥–¥–∞:

$$
a^0 = \gamma \cdot \frac{d u^0}{dt} = \gamma \cdot c \cdot \frac{\gamma^3}{c^2} (\vec{v} \cdot \vec{a}) = \frac{\gamma^4}{c} (\vec{v} \cdot \vec{a})
$$

---

#### ‚úÖ –ü—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–µ–Ω–Ω—ã–µ –∫–æ–º–ø–æ–Ω–µ–Ω—Ç—ã ($\mu = i = 1,2,3$):

$$
u^i = \gamma v^i \quad \Rightarrow \quad
\frac{d u^i}{dt} = \frac{d\gamma}{dt} v^i + \gamma \frac{d v^i}{dt} = \frac{\gamma^3}{c^2} (\vec{v} \cdot \vec{a}) v^i + \gamma a^i
$$

$$
a^i = \gamma \cdot \frac{d u^i}{dt} = \gamma \left( \frac{\gamma^3}{c^2} (\vec{v} \cdot \vec{a}) v^i + \gamma a^i \right) = \frac{\gamma^4}{c^2} (\vec{v} \cdot \vec{a}) v^i + \gamma^2 a^i
$$

### üî∑ 4. –ò—Ç–æ–≥–æ–≤–∞—è —Ñ–æ—Ä–º—É–ª–∞ (–≤–µ–∫—Ç–æ—Ä–Ω–∞—è —Ñ–æ—Ä–º–∞)

\frac{\gamma^4}{c^2} (\vec{v} \cdot \vec{a}) \vec{v} + \gamma^2 \vec{a}
\right)

$$

–≥–¥–µ:
- $\vec{v}(t)$ ‚Äî 3-—Å–∫–æ—Ä–æ—Å—Ç—å,

### üî∑ 5. –ö–æ–≤–∞—Ä–∏–∞–Ω—Ç–Ω–∞—è –∑–∞–ø–∏—Å—å

### üî∑ 6. –°–≤–æ–π—Å—Ç–≤–∞ 4-—É—Å–∫–æ—Ä–µ–Ω–∏—è

1. **–û—Ä—Ç–æ–≥–æ–Ω–∞–ª—å–Ω–æ—Å—Ç—å 4-—Å–∫–æ—Ä–æ—Å—Ç–∏ –∏ 4-—É—Å–∫–æ—Ä–µ–Ω–∏—è**:

   –≠—Ç–æ —Å–ª–µ–¥—É–µ—Ç –∏–∑ $u_\mu u^\mu = -c^2 = \text{const} \Rightarrow \frac{d}{d\tau}(u_\mu u^\mu) = 2 a_\mu u^\mu = 0$

   –∏–ª–∏ —ç–∫–≤–∏–≤–∞–ª–µ–Ω—Ç–Ω–æ:
   $$

   $$
   (—Å–º. –≤—ã–≤–æ–¥ –Ω–∏–∂–µ)

   –£–ø—Ä–æ—â—ë–Ω–Ω–∞—è —Ñ–æ—Ä–º–∞:
   $$
   \boxed{
   a^\mu a_\mu = \gamma^6 \left( \vec{a}^2 - \frac{(\vec{v} \times \vec{a})^2}{c^2} \right)
   }
   $$

3. –í **–Ω–µ—Ä–µ–ª—è—Ç–∏–≤–∏—Å—Ç—Å–∫–æ–º –ø—Ä–µ–¥–µ–ª–µ** $(v \ll c)$:

## üî∑ 3. –ü–æ–ª–µ 4-—Å–∫–æ—Ä–æ—Å—Ç–µ–π $u^\mu(\xi)$

4-—Å–∫–æ—Ä–æ—Å—Ç—å ‚Äî —ç—Ç–æ –ø—Ä–æ–∏–∑–≤–æ–¥–Ω–∞—è –ø–æ —Å–æ–±—Å—Ç–≤–µ–Ω–Ω–æ–º—É –≤—Ä–µ–º–µ–Ω–∏ $\tau$:

$$

$$

–í—ã—á–∏—Å–ª–∏–º:

$$
u^0 = \frac{d(ct)}{d\tau} = \left( \frac{c^2}{a_0} + \xi \right) \cdot \frac{a_0}{c} \cosh\left( \frac{a_0 \tau}{c} \right) = \left( \frac{c}{a_0} + \frac{\xi}{c} \right) a_0 \cosh\left( \frac{a_0 \tau}{c} \right)
$$

$$
u^1 = \frac{dx}{d\tau} = \left( \frac{c^2}{a_0} + \xi \right) \cdot \frac{a_0}{c} \sinh\left( \frac{a_0 \tau}{c} \right)
$$

–û–±–æ–∑–Ω–∞—á–∏–º:
$$
\ell(\xi) = \frac{c^2}{a_0} + \xi
$$

–¢–æ–≥–¥–∞:

$$
u^\mu(\xi, \tau) = \frac{a_0}{c} \ell(\xi) \left( c \cosh\left( \frac{a_0 \tau}{c} \right),\ c \sinh\left( \frac{a_0 \tau}{c} \right) \right)
$$

‚úÖ –ó–∞–º–µ—Ç–∏–º: $u^\mu = \gamma(\tau, \xi) (c, v(\tau, \xi))$, –≥–¥–µ:
- $v = c \tanh(a_0 \tau / c)$ ‚Äî –æ–¥–∏–Ω–∞–∫–æ–≤–∞ –ø–æ –≤—Å–µ–º—É –ø–æ–µ–∑–¥—É –≤ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–Ω–æ–º –≤—Ä–µ–º–µ–Ω–∏,
- –Ω–æ $\gamma = \cosh(a_0 \tau / c)$ ‚Äî —Ç–æ–∂–µ –æ–¥–∏–Ω–∞–∫–æ–≤–æ,
- –æ–¥–Ω–∞–∫–æ **–º–Ω–æ–∂–∏—Ç–µ–ª—å $\ell(\xi)$** –ø–æ–∫–∞–∑—ã–≤–∞–µ—Ç, —á—Ç–æ **–Ω–æ—Ä–º–∞ 4-—Å–∫–æ—Ä–æ—Å—Ç–∏** –¥–æ–ª–∂–Ω–∞ –±—ã—Ç—å $-c^2$, –ø—Ä–æ–≤–µ—Ä–∏–º:

$$
u_\mu u^\mu = - (u^0)^2 + (u^1)^2 = \left( \frac{a_0}{c} \ell(\xi) \right)^2 c^2 \left( -\cosh^2 + \sinh^2 \right) = - c^2 \left( \frac{a_0}{c} \ell(\xi) \right)^2 \cdot 1 \cdot c^2 / c^2
$$

–û—à–∏–±—Å—è –≤ –Ω–æ—Ä–º–∏—Ä–æ–≤–∫–µ! –ù–∞ —Å–∞–º–æ–º –¥–µ–ª–µ:

$$
u^\mu = \frac{dx^\mu}{d\tau} \Rightarrow u_\mu u^\mu = -c^2 \quad \text{–¥–æ–ª–∂–Ω–æ –±—ã—Ç—å!}
$$

–ù–æ:

$$
\left( \frac{d(ct)}{d\tau} \right)^2 - \left( \frac{dx}{d\tau} \right)^2 = \left( \ell(\xi) \frac{a_0}{c} \right)^2 ( \cosh^2 - \sinh^2 ) = \left( \ell(\xi) \frac{a_0}{c} \right)^2
$$

–ê –¥–æ–ª–∂–Ω–æ –±—ã—Ç—å $c^2$. –ó–Ω–∞—á–∏—Ç, **$\tau$ –≤ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–∞—Ö –†–∏–Ω–¥–ª–µ—Ä–∞ ‚Äî –Ω–µ —Å–æ–±—Å—Ç–≤–µ–Ω–Ω–æ–µ –≤—Ä–µ–º—è —Ç–æ—á–∫–∏ $\xi$**!

## üî∑ 4. –í–∞–∂–Ω–æ–µ —É—Ç–æ—á–Ω–µ–Ω–∏–µ: $\tau$ ‚Äî –Ω–µ —Å–æ–±—Å—Ç–≤–µ–Ω–Ω–æ–µ –≤—Ä–µ–º—è –≤—Å–µ—Ö —Ç–æ—á–µ–∫

## üî∑ 6. –ü–æ–ª–µ 4-—É—Å–∫–æ—Ä–µ–Ω–∏–π

$$

$$

–ù–æ –ø—Ä–æ—â–µ: –¥–ª—è –∫–∞–∂–¥–æ–π —Ç–æ—á–∫–∏ $\xi$, –¥–≤–∏–∂—É—â–µ–π—Å—è –ø–æ –≥–∏–ø–µ—Ä–±–æ–ª–µ, 4-—É—Å–∫–æ—Ä–µ–Ω–∏–µ –ø–æ—Å—Ç–æ—è–Ω–Ω–æ –ø–æ –≤–µ–ª–∏—á–∏–Ω–µ –∏ –Ω–∞–ø—Ä–∞–≤–ª–µ–Ω–æ –∫ —Ü–µ–Ω—Ç—Ä—É –≥–∏–ø–µ—Ä–±–æ–ª—ã.

–ò–∑–≤–µ—Å—Ç–Ω–æ, —á—Ç–æ –¥–ª—è —Ä–∞–≤–Ω–æ—É—Å–∫–æ—Ä–µ–Ω–Ω–æ–π –º–∏—Ä–æ–≤–æ–π –ª–∏–Ω–∏–∏:

$$

$$

–ù–æ –Ω–∞ —Å–∞–º–æ–º –¥–µ–ª–µ: —Å–æ–±—Å—Ç–≤–µ–Ω–Ω–æ–µ —É—Å–∫–æ—Ä–µ–Ω–∏–µ —Ç–æ—á–∫–∏ —Å –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–æ–π $\xi$:

$$

–ò 4-—É—Å–∫–æ—Ä–µ–Ω–∏–µ:

$$
a^\mu = \frac{d u^\mu}{d\tau_\xi} = \left( \frac{a(\xi)}{c} \sinh\left( \frac{a_0 \tau}{c} \right),\ \frac{a(\xi)}{c} \cosh\left( \frac{a_0 \tau}{c} \right),\ 0,\ 0 \right) \cdot c
$$

–¢–æ—á–Ω–µ–µ:

$$
a^\mu = \frac{d}{d\tau_\xi} u^\mu = \frac{d}{d\tau_\xi} \left( c \cosh \phi,\ c \sinh \phi \right) = c \left( \sinh \phi \cdot \dot{\phi},\ \cosh \phi \cdot \dot{\phi} \right)
$$

–≥–¥–µ $\phi = a_0 \tau / c$, $d\phi / d\tau_\xi = (a_0 / c) \cdot (d\tau / d\tau_\xi) = (a_0 / c) \cdot \frac{c}{a_0 \ell(\xi)} = \frac{1}{\ell(\xi)}$

–¢–æ–≥–¥–∞:

$$

```python
    plt.plot(t_vals, x_plus, color=colors[i], lw=2, label=f'Œæ = {ell - 1.0:.1f}' if i == 0 else "")

plt.axhline(0, color='k', lw=0.5, alpha=0.3)
plt.axvline(0, color='k', lw=0.5, alpha=0.3)

# –û—Ñ–æ—Ä–º–ª–µ–Ω–∏–µ
plt.xlabel("–í—Ä–µ–º—è $t$ (–≤ –µ–¥–∏–Ω–∏—Ü–∞—Ö $c/a$)", fontsize=12)

plt.xlim(-3, 3)
plt.ylim(0.1, 3)
plt.legend(title="–¢–æ—á–∫–∏ –ª–∏—Ñ—Ç–∞", loc='upper left')
```

- –ò –ø–æ—á–µ–º—É **–Ω–∏–∂–Ω—è—è —á–∞—Å—Ç—å –ª–∏—Ñ—Ç–∞ –Ω–µ –º–æ–∂–µ—Ç –ø–æ–ª—É—á–∞—Ç—å —Å–∏–≥–Ω–∞–ª—ã –∏–∑–≤–Ω–µ**, –µ—Å–ª–∏ $x < |t|$.

–ï—Å–ª–∏ —Ö–æ—Ç–∏—Ç–µ, —è –º–æ–≥—É:
- –î–æ–±–∞–≤–∏—Ç—å **—á–∞—Å—ã** –≤ —Ä–∞–∑–Ω—ã—Ö —Ç–æ—á–∫–∞—Ö,
- –ü–æ–∫–∞–∑–∞—Ç—å **—Ä–∞–∑–Ω–∏—Ü—É —Ö–æ–¥–∞ –≤—Ä–µ–º–µ–Ω–∏**,
- –ò–ª–∏ –Ω–∞–ª–æ–∂–∏—Ç—å **—Å–µ—Ç–∫—É –†–∏–Ω–¥–ª–µ—Ä–∞**.

–ü—Ä–æ—Å—Ç–æ —Å–∫–∞–∂–∏—Ç–µ ‚Äî –∏ —è –ø–æ–¥–≥–æ—Ç–æ–≤–ª—é —É–ª—É—á—à–µ–Ω–Ω—É—é –≤–µ—Ä—Å–∏—é –∫–æ–¥–∞.


````python
import numpy as np
import matplotlib.pyplot as plt

# –ü–∞—Ä–∞–º–µ—Ç—Ä—ã
c = 1  # —Å–∫–æ—Ä–æ—Å—Ç—å —Å–≤–µ—Ç–∞
a = 1  # —É—Å–∫–æ—Ä–µ–Ω–∏–µ (–≤ –µ–¥–∏–Ω–∏—Ü–∞—Ö c^2 / –º–µ—Ç—Ä)

# –î–∏–∞–ø–∞–∑–æ–Ω –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–Ω–æ–≥–æ –≤—Ä–µ–º–µ–Ω–∏ t
t_vals = np.linspace(-3, 3, 400)

# –ú–∏—Ä–æ–≤—ã–µ –ª–∏–Ω–∏–∏ —Ä–∞–∑–Ω—ã—Ö —Ç–æ—á–µ–∫ –ª–∏—Ñ—Ç–∞: x^2 - t^2 = ‚Ñì^2
ell_values = [0.5, 1.0, 1.5, 2.0]  # —Ä–∞–∑–Ω—ã–µ Œæ (–ø–æ–ª–æ–∂–µ–Ω–∏–µ –≤ –ª–∏—Ñ—Ç–µ)
colors = ['blue', 'green', 'orange', 'red']

plt.figure(figsize=(10, 8))

# –†–∏—Å—É–µ–º –º–∏—Ä–æ–≤—ã–µ –ª–∏–Ω–∏–∏ (–≥–∏–ø–µ—Ä–±–æ–ª—ã)
for i, ell in enumerate(ell_values):
    # –í–µ—Ä—Ö–Ω—è—è –≤–µ—Ç–≤—å –≥–∏–ø–µ—Ä–±–æ–ª—ã (x > 0)
    x_plus = np.sqrt(t_vals**2 + ell**2)
    x_minus = -np.sqrt(t_vals**2 + ell**2)  # –º–æ–∂–Ω–æ –Ω–µ —Ä–∏—Å–æ–≤–∞—Ç—å, –µ—Å–ª–∏ –Ω–µ –Ω—É–∂–Ω–æ

    plt.plot(t_vals, x_plus, color=colors[i], lw=2, label=f'Œæ = {ell - 1.0:.1f}' if i == 0 else "")

# –ü–æ–≤–µ—Ä—Ö–Ω–æ—Å—Ç–∏ –æ–¥–Ω–æ–≤—Ä–µ–º–µ–Ω–Ω–æ—Å—Ç–∏: œÑ = const ‚Üí t/x = tanh(aœÑ/c)
tau_vals = [-1.0, -0.5, 0.0, 0.5, 1.0]
for tau in tau_vals:
    # t = x * tanh(tau)  (–ø—Ä–∏ a=1, c=1)
    tanh_tau = np.tanh(tau)
    x_line = np.linspace(0.1, 3, 100)
    t_line = tanh_tau * x_line
    plt.plot(t_line, x_line, 'gray', linestyle='--', alpha=0.7)

# –ì–æ—Ä–∏–∑–æ–Ω—Ç —Å–æ–±—ã—Ç–∏–π: x = |t| (—Å–≤–µ—Ç–æ–ø–æ–¥–æ–±–Ω—ã–µ –ª—É—á–∏)
x_light = np.linspace(0, 3, 100)
plt.plot(x_light, x_light, 'k-', lw=1, alpha=0.5)
plt.plot(-x_light, x_light, 'k-', lw=1, alpha=0.5)
plt.axhline(0, color='k', lw=0.5, alpha=0.3)
plt.axvline(0, color='k', lw=0.5, alpha=0.3)

# –û—Ñ–æ—Ä–º–ª–µ–Ω–∏–µ
plt.xlabel("–í—Ä–µ–º—è $t$ (–≤ –µ–¥–∏–Ω–∏—Ü–∞—Ö $c/a$)", fontsize=12)
plt.ylabel("–ü—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–æ $x$ (–≤ –µ–¥–∏–Ω–∏—Ü–∞—Ö $c^2/a$)", fontsize=12)
plt.title("–î–∏–∞–≥—Ä–∞–º–º–∞ –†–∏–Ω–¥–ª–µ—Ä–∞: —É—Å–∫–æ—Ä–µ–Ω–Ω—ã–π –ª–∏—Ñ—Ç\n–ú–∏—Ä–æ–≤—ã–µ –ª–∏–Ω–∏–∏ (–≥–∏–ø–µ—Ä–±–æ–ª—ã) –∏ –ø–æ–≤–µ—Ä—Ö–Ω–æ—Å—Ç–∏ –æ–¥–Ω–æ–≤—Ä–µ–º–µ–Ω–Ω–æ—Å—Ç–∏", fontsize=14)
plt.xlim(-3, 3)
plt.ylim(0.1, 3)
plt.legend(title="–¢–æ—á–∫–∏ –ª–∏—Ñ—Ç–∞", loc='upper left')
plt.grid(True, alpha=0.3)
plt.tight_layout()

# –ü–æ–∫–∞–∑–∞—Ç—å
plt.show()
````

### ASSISTANT
–û—Ç–ª–∏—á–Ω–æ! –î–∞–≤–∞–π—Ç–µ **—Ä–∞—Å—à–∏—Ä–∏–º –¥–∏–∞–≥—Ä–∞–º–º—É –†–∏–Ω–¥–ª–µ—Ä–∞**, —á—Ç–æ–±—ã –Ω–∞ –Ω–µ–π –±—ã–ª–∏ **—è–≤–Ω–æ –ø–æ–∫–∞–∑–∞–Ω—ã –ø—Ä–µ–¥–µ–ª—ã –∏–Ω—Ç–µ–≥—Ä–∏—Ä–æ–≤–∞–Ω–∏—è –¥–µ–π—Å—Ç–≤–∏—è** ‚Äî —Ç–æ –µ—Å—Ç—å:

- **–û–±–ª–∞—Å—Ç—å –≤ –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–µ-–≤—Ä–µ–º–µ–Ω–∏**, –ø–æ –∫–æ—Ç–æ—Ä–æ–π –º—ã –∏–Ω—Ç–µ–≥—Ä–∏—Ä—É–µ–º,
- –í—Ä–µ–º–µ–Ω–Ω—ã–µ –ø—Ä–µ–¥–µ–ª—ã: $\tau_1$ –¥–æ $\tau_2$,
- –ü—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–µ–Ω–Ω—ã–µ –ø—Ä–µ–¥–µ–ª—ã: –æ—Ç $\xi_{\min}$ –¥–æ $\xi_{\max}$ (–ø–æ –≤—ã—Å–æ—Ç–µ –ª–∏—Ñ—Ç–∞),
- –ò –∫–∞–∫ —ç—Ç–∞ –æ–±–ª–∞—Å—Ç—å –≤—ã–≥–ª—è–¥–∏—Ç –≤ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–∞—Ö $(t, x)$ –ª–∞–±–æ—Ä–∞—Ç–æ—Ä–Ω–æ–π —Å–∏—Å—Ç–µ–º—ã.

## üéØ –¶–µ–ª—å:
–ü–æ–∫–∞–∑–∞—Ç—å, –∫–∞–∫ **—á–µ—Ç—ã—Ä—ë—Ö–º–µ—Ä–Ω–∞—è –æ–±–ª–∞—Å—Ç—å –∏–Ω—Ç–µ–≥—Ä–∏—Ä–æ–≤–∞–Ω–∏—è** (–≤ –Ω–∞—à–µ–º —Å–ª—É—á–∞–µ ‚Äî **–æ–±—ä—ë–º –≤ –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–µ-–≤—Ä–µ–º–µ–Ω–∏**) –ø—Ä–æ–µ—Ü–∏—Ä—É–µ—Ç—Å—è –Ω–∞ –¥–∏–∞–≥—Ä–∞–º–º—É –ú–∏–Ω–∫–æ–≤—Å–∫–æ–≥–æ –∫–∞–∫ **"–ª–µ–Ω—Ç–∞" –º–µ–∂–¥—É –¥–≤—É–º—è –ø–æ–≤–µ—Ä—Ö–Ω–æ—Å—Ç—è–º–∏ –æ–¥–Ω–æ–≤—Ä–µ–º–µ–Ω–Ω–æ—Å—Ç–∏** –∏ –º–µ–∂–¥—É –¥–≤—É–º—è –º–∏—Ä–æ–≤—ã–º–∏ –ª–∏–Ω–∏—è–º–∏.

---

### ‚úÖ –ß—Ç–æ –º—ã –¥–æ–±–∞–≤–∏–º:
- –î–≤–µ –ø–æ–≤–µ—Ä—Ö–Ω–æ—Å—Ç–∏ $\tau = \tau_1$ –∏ $\tau = \tau_2$ ‚Äî –≥—Ä–∞–Ω–∏—Ü—ã –ø–æ –≤—Ä–µ–º–µ–Ω–∏,
- –î–≤–µ –º–∏—Ä–æ–≤—ã–µ –ª–∏–Ω–∏–∏: $\xi = \xi_{\min}$ –∏ $\xi = \xi_{\max}$ ‚Äî –≥—Ä–∞–Ω–∏—Ü—ã –ø–æ –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤—É,
- –ó–∞–∫—Ä–∞—à–µ–Ω–Ω—É—é –æ–±–ª–∞—Å—Ç—å ‚Äî **–æ–±—ä—ë–º –¥–µ–π—Å—Ç–≤–∏—è**,
- –ü–æ–¥–ø–∏—Å–∏ –∏ –ø–æ—è—Å–Ω–µ–Ω–∏—è.

---

### üìä –û–±–Ω–æ–≤–ª—ë–Ω–Ω—ã–π –∫–æ–¥: –î–∏–∞–≥—Ä–∞–º–º–∞ –†–∏–Ω–¥–ª–µ—Ä–∞ —Å –æ–±–ª–∞—Å—Ç—å—é –∏–Ω—Ç–µ–≥—Ä–∏—Ä–æ–≤–∞–Ω–∏—è

```python
import numpy as np
import matplotlib.pyplot as plt

# –ü–∞—Ä–∞–º–µ—Ç—Ä—ã
c = 1  # —Å–∫–æ—Ä–æ—Å—Ç—å —Å–≤–µ—Ç–∞
a = 1  # —É—Å–∫–æ—Ä–µ–Ω–∏–µ (–≤ –µ–¥–∏–Ω–∏—Ü–∞—Ö c¬≤/–º–µ—Ç—Ä)

# –î–∏–∞–ø–∞–∑–æ–Ω –≤—Ä–µ–º–µ–Ω–∏ –¥–ª—è –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è
t_vals = np.linspace(-3, 3, 400)

# -----------------------------------------------------------------------------
# 1. –ú–∏—Ä–æ–≤—ã–µ –ª–∏–Ω–∏–∏: x^2 - t^2 = ‚Ñì¬≤, –≥–¥–µ ‚Ñì = c¬≤/a + Œæ ‚âà 1 + Œæ (–ø—Ä–∏ c=a=1)
# –í—ã–±–µ—Ä–µ–º Œæ_min –∏ Œæ_max
xi_min = -0.5
xi_max = 1.0
ell_min = 1 + xi_min  # ‚Ñì = 1 + Œæ
ell_max = 1 + xi_max

# –í–µ—Ä—Ö–Ω–∏–µ –≤–µ—Ç–≤–∏ –≥–∏–ø–µ—Ä–±–æ–ª
x_plus_min = np.sqrt(t_vals**2 + ell_min**2)
x_plus_max = np.sqrt(t_vals**2 + ell_max**2)

# -----------------------------------------------------------------------------
# 2. –ü–æ–≤–µ—Ä—Ö–Ω–æ—Å—Ç–∏ –æ–¥–Ω–æ–≤—Ä–µ–º–µ–Ω–Ω–æ—Å—Ç–∏: œÑ = const ‚Üí t = x * tanh(œÑ)
tau_1 = -0.5
tau_2 = 0.8
t_surface_1 = np.tanh(tau_1) * x_plus_min  # –ø—Ä–∏–±–ª–∏–∂—ë–Ω–Ω–æ
t_surface_2 = np.tanh(tau_2) * x_plus_max

# –ü–æ—Å—Ç—Ä–æ–∏–º –ø–æ–≤–µ—Ä—Ö–Ω–æ—Å—Ç–∏ –∫–∞–∫ t = x * tanh(œÑ)
x_line = np.linspace(0.1, 3.0, 200)
t_tau1 = np.tanh(tau_1) * x_line
t_tau2 = np.tanh(tau_2) * x_line

# -----------------------------------------------------------------------------
# –ù–∞—á–∏–Ω–∞–µ–º —Ä–∏—Å–æ–≤–∞—Ç—å
plt.figure(figsize=(10, 8))

# –ú–∏—Ä–æ–≤—ã–µ –ª–∏–Ω–∏–∏ –≥—Ä–∞–Ω–∏—Ü –ª–∏—Ñ—Ç–∞
plt.plot(t_vals, x_plus_min, color='blue', lw=2.5, label=f'–ú–∏—Ä–æ–≤–∞—è –ª–∏–Ω–∏—è, Œæ = {xi_min}')
plt.plot(t_vals, x_plus_max, color='blue', lw=2.5, label=f'–ú–∏—Ä–æ–≤–∞—è –ª–∏–Ω–∏—è, Œæ = {xi_max}')

# –ü–æ–≤–µ—Ä—Ö–Ω–æ—Å—Ç–∏ –æ–¥–Ω–æ–≤—Ä–µ–º–µ–Ω–Ω–æ—Å—Ç–∏
plt.plot(t_tau1, x_line, 'red', linestyle='--', lw=2, label=f'œÑ = {tau_1} (–Ω–∞—á–∞–ª–æ)')
plt.plot(t_tau2, x_line, 'red', linestyle='--', lw=2, label=f'œÑ = {tau_2} (–∫–æ–Ω–µ—Ü)')

# -----------------------------------------------------------------------------

# –ù–∞–π–¥—ë–º —Ç–æ—á–∫–∏ –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏—è –¥–ª—è –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è –ø–æ–ª–∏–≥–æ–Ω–∞
```

## üî∑ –ü–æ—á–µ–º—É —ç—Ç–æ –≤–∞–∂–Ω–æ?

- –í–∏–¥–Ω–æ, —á—Ç–æ **–ø—Ä–µ–¥–µ–ª—ã –ø–æ $\tau$** ‚Äî —ç—Ç–æ **–Ω–µ –≥–æ—Ä–∏–∑–æ–Ω—Ç–∞–ª—å–Ω—ã–µ –ª–∏–Ω–∏–∏**, –∞ **–Ω–∞–∫–ª–æ–Ω–Ω—ã–µ –ø–æ–≤–µ—Ä—Ö–Ω–æ—Å—Ç–∏**.
- –û–±–ª–∞—Å—Ç—å –∏–Ω—Ç–µ–≥—Ä–∏—Ä–æ–≤–∞–Ω–∏—è **–∏—Å–∫—Ä–∏–≤–ª–µ–Ω–∞** ‚Äî –∏–∑-–∑–∞ —É—Å–∫–æ—Ä–µ–Ω–∏—è.
- –•–æ—Ç—è $\tau$ –æ–¥–∏–Ω–∞–∫–æ–≤–æ –¥–ª—è –≤—Å–µ—Ö —Ç–æ—á–µ–∫, **—Å–æ–±—Å—Ç–≤–µ–Ω–Ω–æ–µ –≤—Ä–µ–º—è** —Ä–∞–∑–Ω–æ–µ: –≤–≤–µ—Ä—Ö—É –ª–∏—Ñ—Ç–∞ (–±–æ–ª—å—à–µ–µ $x$) –æ–Ω–æ —Ç–µ—á—ë—Ç –±—ã—Å—Ç—Ä–µ–µ.
- –≠—Ç–æ –æ—Ç—Ä–∞–∂–∞–µ—Ç—Å—è –≤ –º–Ω–æ–∂–∏—Ç–µ–ª–µ $\sqrt{-g} = 1 + a\xi/c^2$ –≤ –¥–µ–π—Å—Ç–≤–∏–∏.

## ‚úÖ –ö–∞–∫ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å

1. –°–∫–æ–ø–∏—Ä—É–π—Ç–µ –∫–æ–¥ –≤ **Google Colab** –∏–ª–∏ **Jupyter Notebook**.
2. –ó–∞–ø—É—Å—Ç–∏—Ç–µ ‚Äî –ø–æ–ª—É—á–∏—Ç–µ **–Ω–∞–≥–ª—è–¥–Ω—É—é –¥–∏–∞–≥—Ä–∞–º–º—É**.
3. –ú–µ–Ω—è–π—Ç–µ $\xi_{\min}, \xi_{\max}, \tau_1, \tau_2$ ‚Äî —á—Ç–æ–±—ã –∏—Å—Å–ª–µ–¥–æ–≤–∞—Ç—å —Ä–∞–∑–Ω—ã–µ —Å—Ü–µ–Ω–∞—Ä–∏–∏.

–•–æ—á–µ—à—å, —è —Å–¥–µ–ª–∞—é **–∞–Ω–∏–º–∞—Ü–∏—é**, –∫–∞–∫ —ç—Ç–∞ –æ–±–ª–∞—Å—Ç—å ¬´—Ä–∞—Å—Ç—ë—Ç¬ª —Å–æ –≤—Ä–µ–º–µ–Ω–µ–º, –∏–ª–∏ –¥–æ–±–∞–≤–ª—é **—Å–µ—Ç–∫—É –†–∏–Ω–¥–ª–µ—Ä–∞** (–ª–∏–Ω–∏–∏ $\xi = \text{const}, \tau = \text{const}$)?

–ì–æ—Ç–æ–≤ —É–≥–ª—É–±–∏—Ç—å—Å—è –¥–∞–ª—å—à–µ!


````python
import numpy as np
import matplotlib.pyplot as plt

# –ü–∞—Ä–∞–º–µ—Ç—Ä—ã
c = 1  # —Å–∫–æ—Ä–æ—Å—Ç—å —Å–≤–µ—Ç–∞
a = 1  # —É—Å–∫–æ—Ä–µ–Ω–∏–µ (–≤ –µ–¥–∏–Ω–∏—Ü–∞—Ö c¬≤/–º–µ—Ç—Ä)

# –î–∏–∞–ø–∞–∑–æ–Ω –≤—Ä–µ–º–µ–Ω–∏ –¥–ª—è –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è
t_vals = np.linspace(-3, 3, 400)

# -----------------------------------------------------------------------------
# 1. –ú–∏—Ä–æ–≤—ã–µ –ª–∏–Ω–∏–∏: x^2 - t^2 = ‚Ñì¬≤, –≥–¥–µ ‚Ñì = c¬≤/a + Œæ ‚âà 1 + Œæ (–ø—Ä–∏ c=a=1)
# –í—ã–±–µ—Ä–µ–º Œæ_min –∏ Œæ_max
xi_min = -0.5
xi_max = 1.0
ell_min = 1 + xi_min  # ‚Ñì = 1 + Œæ
ell_max = 1 + xi_max

# –í–µ—Ä—Ö–Ω–∏–µ –≤–µ—Ç–≤–∏ –≥–∏–ø–µ—Ä–±–æ–ª
x_plus_min = np.sqrt(t_vals**2 + ell_min**2)
x_plus_max = np.sqrt(t_vals**2 + ell_max**2)

# -----------------------------------------------------------------------------
# 2. –ü–æ–≤–µ—Ä—Ö–Ω–æ—Å—Ç–∏ –æ–¥–Ω–æ–≤—Ä–µ–º–µ–Ω–Ω–æ—Å—Ç–∏: œÑ = const ‚Üí t = x * tanh(œÑ)
tau_1 = -0.5
tau_2 = 0.8
t_surface_1 = np.tanh(tau_1) * x_plus_min  # –ø—Ä–∏–±–ª–∏–∂—ë–Ω–Ω–æ
t_surface_2 = np.tanh(tau_2) * x_plus_max

# –ü–æ—Å—Ç—Ä–æ–∏–º –ø–æ–≤–µ—Ä—Ö–Ω–æ—Å—Ç–∏ –∫–∞–∫ t = x * tanh(œÑ)
x_line = np.linspace(0.1, 3.0, 200)
t_tau1 = np.tanh(tau_1) * x_line
t_tau2 = np.tanh(tau_2) * x_line

# -----------------------------------------------------------------------------
# –ù–∞—á–∏–Ω–∞–µ–º —Ä–∏—Å–æ–≤–∞—Ç—å
plt.figure(figsize=(10, 8))

# –ú–∏—Ä–æ–≤—ã–µ –ª–∏–Ω–∏–∏ –≥—Ä–∞–Ω–∏—Ü –ª–∏—Ñ—Ç–∞
plt.plot(t_vals, x_plus_min, color='blue', lw=2.5, label=f'–ú–∏—Ä–æ–≤–∞—è –ª–∏–Ω–∏—è, Œæ = {xi_min}')
plt.plot(t_vals, x_plus_max, color='blue', lw=2.5, label=f'–ú–∏—Ä–æ–≤–∞—è –ª–∏–Ω–∏—è, Œæ = {xi_max}')

# –ü–æ–≤–µ—Ä—Ö–Ω–æ—Å—Ç–∏ –æ–¥–Ω–æ–≤—Ä–µ–º–µ–Ω–Ω–æ—Å—Ç–∏
plt.plot(t_tau1, x_line, 'red', linestyle='--', lw=2, label=f'œÑ = {tau_1} (–Ω–∞—á–∞–ª–æ)')
plt.plot(t_tau2, x_line, 'red', linestyle='--', lw=2, label=f'œÑ = {tau_2} (–∫–æ–Ω–µ—Ü)')

# -----------------------------------------------------------------------------
# –ó–∞–∫—Ä–∞—à–∏–≤–∞–µ–º –æ–±–ª–∞—Å—Ç—å –∏–Ω—Ç–µ–≥—Ä–∏—Ä–æ–≤–∞–Ω–∏—è

# –ù–∞–π–¥—ë–º —Ç–æ—á–∫–∏ –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏—è –¥–ª—è –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è –ø–æ–ª–∏–≥–æ–Ω–∞
# –û–≥—Ä–∞–Ω–∏—á–∏–≤–∞–µ–º x_line –æ–±–ª–∞—Å—Ç—å—é, –≥–¥–µ –ø–æ–≤–µ—Ä—Ö–Ω–æ—Å—Ç–∏ –ø–µ—Ä–µ—Å–µ–∫–∞—é—Ç –≥–∏–ø–µ—Ä–±–æ–ª—ã

# –ü—Ä–∏–±–ª–∏–∂—ë–Ω–Ω–æ: –∏—Å–ø–æ–ª—å–∑—É–µ–º x_line –∏ –æ–≥—Ä–∞–Ω–∏—á–∏–º –ø–æ œÑ
x_fill = np.linspace(max(ell_min, 0.1), 3.0, 200)

# –ì—Ä–∞–Ω–∏—Ü—ã –ø–æ œÑ
t_left = np.tanh(tau_1) * x_fill
t_right = np.tanh(tau_2) * x_fill

# –û–≥—Ä–∞–Ω–∏—á–∏–≤–∞–µ–º –ø–æ Œæ
x_lower = np.sqrt(t_vals**2 + ell_min**2)
x_upper = np.sqrt(t_vals**2 + ell_max**2)

# –ü–æ—Å—Ç—Ä–æ–∏–º –ø–æ–ª–∏–≥–æ–Ω: –æ–±–æ–π–¥—ë–º –ø–æ –≥—Ä–∞–Ω–∏—Ü–∞–º
# 1. –û—Ç –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏—è œÑ=œÑ1 –∏ Œæ=Œæ_min –¥–æ œÑ=œÑ1 –∏ Œæ=Œæ_max
# 2. –í–¥–æ–ª—å Œæ=Œæ_max –¥–æ œÑ=œÑ2
# 3. –ù–∞–∑–∞–¥ –ø–æ œÑ=œÑ2
# 4. –í–¥–æ–ª—å Œæ=Œæ_min –¥–æ œÑ=œÑ1

# –ê–Ω–∞–ª–∏—Ç–∏—á–µ—Å–∫–∏ —Å–ª–æ–∂–Ω–æ, –ø–æ—ç—Ç–æ–º—É —Å–¥–µ–ª–∞–µ–º –ø—Ä–∏–±–ª–∏–∂–µ–Ω–∏–µ:
x_poly = np.concatenate([
    np.linspace(ell_min, ell_max, 100),  # –≤–¥–æ–ª—å œÑ=œÑ1
    np.linspace(ell_max, ell_min, 100)[::-1]  # –≤–¥–æ–ª—å œÑ=œÑ2
])
t1_poly = np.tanh(tau_1) * x_poly[:100]
t2_poly = np.tanh(tau_2) * x_poly[100:][::-1]
t_poly = np.concatenate([t1_poly, t2_poly])

# –£–±–µ–¥–∏–º—Å—è, —á—Ç–æ x_poly –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ
x_full_poly = np.concatenate([x_poly[:100], x_poly[100:]])

# –ó–∞–∫—Ä–∞—à–∏–≤–∞–µ–º –æ–±–ª–∞—Å—Ç—å
plt.fill(t_poly, x_full_poly, color='lightyellow', alpha=0.6, edgecolor='orange', lw=1.5, label='–û–±–ª–∞—Å—Ç—å –∏–Ω—Ç–µ–≥—Ä–∏—Ä–æ–≤–∞–Ω–∏—è –¥–µ–π—Å—Ç–≤–∏—è')

# -----------------------------------------------------------------------------
# –°–≤–µ—Ç–æ–≤—ã–µ –ª—É—á–∏ (–≥–æ—Ä–∏–∑–æ–Ω—Ç —Å–æ–±—ã—Ç–∏–π)
x_light = np.linspace(0, 3, 100)
plt.plot( x_light,  x_light, 'k-', lw=1, alpha=0.4, label='–°–≤–µ—Ç–æ–≤–æ–π –ª—É—á (x = t)')
plt.plot(-x_light,  x_light, 'k-', lw=1, alpha=0.4)

# –û—Å–∏
plt.axhline(0, color='k', lw=0.5, alpha=0.3)
plt.axvline(0, color='k', lw=0.5, alpha=0.3)

# -----------------------------------------------------------------------------
# –û—Ñ–æ—Ä–º–ª–µ–Ω–∏–µ
plt.xlabel("–ö–æ–æ—Ä–¥–∏–Ω–∞—Ç–Ω–æ–µ –≤—Ä–µ–º—è $t$ (–≤ –µ–¥–∏–Ω–∏—Ü–∞—Ö $c/a$)", fontsize=12)
plt.ylabel("–ü—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–æ $x$ (–≤ –µ–¥–∏–Ω–∏—Ü–∞—Ö $c^2/a$)", fontsize=12)
plt.title("–î–∏–∞–≥—Ä–∞–º–º–∞ –†–∏–Ω–¥–ª–µ—Ä–∞: –û–±–ª–∞—Å—Ç—å –∏–Ω—Ç–µ–≥—Ä–∏—Ä–æ–≤–∞–Ω–∏—è –¥–µ–π—Å—Ç–≤–∏—è\n(—É—Å–∫–æ—Ä–µ–Ω–Ω—ã–π –ª–∏—Ñ—Ç, Œæ ‚àà [%.1f, %.1f], œÑ ‚àà [%.1f, %.1f])" % (xi_min, xi_max, tau_1, tau_2), fontsize=13)
plt.xlim(-1.5, 1.5)
plt.ylim(0.5, 2.5)
plt.legend(loc='upper left', fontsize=10)
plt.grid(True, alpha=0.3)
plt.tight_layout()

# –ü–æ–∫–∞–∑–∞—Ç—å
plt.show()
````

- –û—Ç–æ–±—Ä–∞–∂–∞—Ç—å **–º–∏—Ä–æ–≤—ã–µ –ª–∏–Ω–∏–∏** –≥—Ä–∞–Ω–∏—Ü –ª–∏—Ñ—Ç–∞ –∏ **–ø–æ–≤–µ—Ä—Ö–Ω–æ—Å—Ç–∏ –æ–¥–Ω–æ–≤—Ä–µ–º–µ–Ω–Ω–æ—Å—Ç–∏**.

### üéûÔ∏è –ö–∞–∫ —Ä–∞–±–æ—Ç–∞–µ—Ç –∞–Ω–∏–º–∞—Ü–∏—è:
- –ù–∞ –∫–∞–∂–¥–æ–º –∫–∞–¥—Ä–µ: $\tau_{\text{current}}$ —É–≤–µ–ª–∏—á–∏–≤–∞–µ—Ç—Å—è –æ—Ç $\tau_{\min}$ –¥–æ $\tau_{\max}$,
- –ó–∞–∫—Ä–∞—à–µ–Ω–Ω–∞—è –æ–±–ª–∞—Å—Ç—å ‚Äî –º–µ–∂–¥—É $\tau = \tau_1$ –∏ $\tau = \tau_{\text{current}}$,
- –í—Å—ë —ç—Ç–æ —Ä–∏—Å—É–µ—Ç—Å—è –Ω–∞ –¥–∏–∞–≥—Ä–∞–º–º–µ –ú–∏–Ω–∫–æ–≤—Å–∫–æ–≥–æ –≤ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–∞—Ö $(t, x)$.

## üìä –ö–æ–¥: –ê–Ω–∏–º–∞—Ü–∏—è —Ä–æ—Å—Ç–∞ –æ–±–ª–∞—Å—Ç–∏ –¥–µ–π—Å—Ç–≤–∏—è –≤ —É—Å–∫–æ—Ä–µ–Ω–Ω–æ–º –ª–∏—Ñ—Ç–µ

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# –ü–∞—Ä–∞–º–µ—Ç—Ä—ã
c = 1
a = 1
xi_min = -0.5
xi_max = 1.0
ell_min = 1 + xi_min  # ‚Ñì = c¬≤/a + Œæ ‚âà 1 + Œæ
ell_max = 1 + xi_max

tau_start = -0.5
tau_end = 1.2
tau_range = np.linspace(tau_start, tau_end, 60)  # 60 –∫–∞–¥—Ä–æ–≤

t_vals = np.linspace(-2.5, 2.5, 400)
x_line = np.linspace(0.1, 3.0, 200)

# –ü–æ–¥–≥–æ—Ç–æ–≤–∫–∞ —Ñ–∏–≥—É—Ä—ã
fig, ax = plt.subplots(figsize=(10, 8))

# –§–∏–∫—Å–∏—Ä–æ–≤–∞–Ω–Ω—ã–µ —ç–ª–µ–º–µ–Ω—Ç—ã (–Ω–∞—Ä–∏—Å—É–µ–º –æ–¥–∏–Ω —Ä–∞–∑)
ax.set_xlim(-1.8, 1.8)
ax.set_ylim(0.6, 2.6)
ax.set_xlabel("–ö–æ–æ—Ä–¥–∏–Ω–∞—Ç–Ω–æ–µ –≤—Ä–µ–º—è $t$ (–≤ –µ–¥–∏–Ω–∏—Ü–∞—Ö $c/a$)", fontsize=12)
ax.set_ylabel("–ü—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–æ $x$ (–≤ –µ–¥–∏–Ω–∏—Ü–∞—Ö $c^2/a$)", fontsize=12)
ax.set_title("–ê–Ω–∏–º–∞—Ü–∏—è: —Ä–æ—Å—Ç –æ–±–ª–∞—Å—Ç–∏ –∏–Ω—Ç–µ–≥—Ä–∏—Ä–æ–≤–∞–Ω–∏—è –¥–µ–π—Å—Ç–≤–∏—è –≤ —É—Å–∫–æ—Ä–µ–Ω–Ω–æ–º –ª–∏—Ñ—Ç–µ", fontsize=13)
ax.grid(True, alpha=0.3)

# –ú–∏—Ä–æ–≤—ã–µ –ª–∏–Ω–∏–∏ –≥—Ä–∞–Ω–∏—Ü
x_world_min = np.sqrt(t_vals**2 + ell_min**2)
x_world_max = np.sqrt(t_vals**2 + ell_max**2)
ax.plot(t_vals, x_world_min, color='blue', lw=2, alpha=0.7)
ax.plot(t_vals, x_world_max, color='blue', lw=2, alpha=0.7)

# –°–≤–µ—Ç–æ–≤—ã–µ –ª—É—á–∏
x_light = np.linspace(0, 3, 100)
ax.plot( x_light,  x_light, 'k-', lw=1, alpha=0.4)
ax.plot(-x_light,  x_light, 'k-', lw=1, alpha=0.4)

# –ü–æ–≤–µ—Ä—Ö–Ω–æ—Å—Ç—å œÑ = œÑ_start (–Ω–∞—á–∞–ª–æ)
t_tau_start = np.tanh(tau_start) * x_line
ax.plot(t_tau_start, x_line, 'red', linestyle='--', alpha=0.7, label=r'$\tau = \tau_1$')

# –õ–µ–≥–µ–Ω–¥–∞ (–±–µ–∑ –¥—É–±–ª–∏—Ä–æ–≤–∞–Ω–∏—è)
ax.legend(loc='upper left')

# –û–±—ä–µ–∫—Ç –¥–ª—è –∞–Ω–∏–º–∞—Ü–∏–∏
area_fill = ax.fill([], [], color='gold', alpha=0.5, edgecolor='orange', lw=1.2, label='–û–±–ª–∞—Å—Ç—å –¥–µ–π—Å—Ç–≤–∏—è')[0]

# –¢–µ–∫—Å—Ç: —Ç–µ–∫—É—â–µ–µ œÑ
time_text = ax.text(0.02, 0.95, '', transform=ax.transAxes, fontsize=12, bbox=dict(boxstyle="round", fc="w"))

# -----------------------------------------------------------------------------
# –§—É–Ω–∫—Ü–∏—è –æ–±–Ω–æ–≤–ª–µ–Ω–∏—è –∫–∞–¥—Ä–∞
def update_frame(tau_curr):
    ax.patches.remove(area_fill)  # —É–¥–∞–ª—è–µ–º –ø—Ä–µ–¥—ã–¥—É—â—É—é –∑–∞–ª–∏–≤–∫—É

    # –ü–æ–≤–µ—Ä—Ö–Ω–æ—Å—Ç—å œÑ = œÑ_curr
    t_tau_curr = np.tanh(tau_curr) * x_line

    # –ü–æ–ª–∏–≥–æ–Ω: –æ—Ç œÑ_start –¥–æ œÑ_curr, –º–µ–∂–¥—É Œæ_min –∏ Œæ_max
    # –ù–∞–π–¥—ë–º x, –≥–¥–µ –ø–æ–≤–µ—Ä—Ö–Ω–æ—Å—Ç–∏ –ø–µ—Ä–µ—Å–µ–∫–∞—é—Ç –º–∏—Ä–æ–≤—ã–µ –ª–∏–Ω–∏–∏
    x_intersect_min = np.sqrt(t_tau_start**2 + ell_min**2)  # –ø—Ä–∏–±–ª–∏–∂—ë–Ω–Ω–æ
    x_intersect_max = np.sqrt(t_tau_curr**2 + ell_max**2)

    # –û–≥—Ä–∞–Ω–∏—á–∏–≤–∞–µ–º x
    x_fill = np.linspace(max(ell_min, 0.8), 2.5, 100)

    # –ì—Ä–∞–Ω–∏—Ü—ã –ø–æ x –¥–ª—è —Ç–µ–∫—É—â–µ–≥–æ œÑ
    x_lower_bound = np.sqrt(t_tau_start**2 + ell_min**2)
    x_upper_bound = np.sqrt(t_tau_curr**2 + ell_max**2)

    # –ü–æ—Å—Ç—Ä–æ–∏–º –ø–æ–ª–∏–≥–æ–Ω
    x_poly = np.concatenate([
        np.linspace(ell_min, ell_max, 100),  # –≤–¥–æ–ª—å œÑ = œÑ_start
        np.linspace(ell_max, ell_min, 100)[::-1]  # –≤–¥–æ–ª—å œÑ = œÑ_curr
    ])
    t1_poly = np.tanh(tau_start) * x_poly[:100]
    t2_poly = np.tanh(tau_curr) * x_poly[100:][::-1]
    t_poly = np.concatenate([t1_poly, t2_poly])
    x_full_poly = np.concatenate([x_poly[:100], x_poly[100:]])

    # –£–¥–∞–ª—è–µ–º —Ç–æ—á–∫–∏ –≤–Ω–µ –≤–∏–¥–∏–º–æ—Å—Ç–∏
    mask = (x_full_poly >= ax.get_ylim()[0]) & (x_full_poly <= ax.get_ylim()[1])
    t_poly = t_poly[mask]
    x_full_poly = x_full_poly[mask]

    # –ó–∞–∫—Ä–∞—à–∏–≤–∞–µ–º
    global area_fill
    area_fill = ax.fill(t_poly, x_full_poly, color='gold', alpha=0.5, edgecolor='orange', lw=1.2)[0]

    # –û–±–Ω–æ–≤–ª—è–µ–º —Ç–µ–∫—Å—Ç
    time_text.set_text(f'–¢–µ–∫—É—â–µ–µ œÑ = {tau_curr:.2f}')

    return area_fill, time_text

# -----------------------------------------------------------------------------
# –°–æ–∑–¥–∞—ë–º –∞–Ω–∏–º–∞—Ü–∏—é
ani = FuncAnimation(
    fig, 
    update_frame, 
    frames=tau_range, 
    interval=200,  # 200 –º—Å –Ω–∞ –∫–∞–¥—Ä
    repeat=True,
    blit=False
)

# –°–æ—Ö—Ä–∞–Ω—è–µ–º –≤ GIF (–æ–ø—Ü–∏–æ–Ω–∞–ª—å–Ω–æ)

# –ü–æ–∫–∞–∑—ã–≤–∞–µ–º
plt.tight_layout()
plt.show()
```

### ‚úÖ –†–∞–±–æ—á–∏–π –∫–æ–¥: –ê–Ω–∏–º–∞—Ü–∏—è –æ–±–ª–∞—Å—Ç–∏ –¥–µ–π—Å—Ç–≤–∏—è –≤ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–∞—Ö –†–∏–Ω–¥–ª–µ—Ä–∞

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# –ü–∞—Ä–∞–º–µ—Ç—Ä—ã
c = 1
a = 1
xi_min = -0.5
xi_max = 1.0
ell_min = 1 + xi_min  # ‚Ñì = c¬≤/a + Œæ ‚âà 1 + Œæ
ell_max = 1 + xi_max

tau_start = -0.5
tau_end = 1.0
tau_vals = np.linspace(tau_start, tau_end, 60)  # 60 –∫–∞–¥—Ä–æ–≤

t_plot = np.linspace(-2.5, 2.5, 400)
x_plot = np.linspace(0.1, 3.0, 200)

# –§—É–Ω–∫—Ü–∏—è: x(t) –¥–ª—è –º–∏—Ä–æ–≤–æ–π –ª–∏–Ω–∏–∏ —Å –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–º ell
def x_world(t, ell):
    return np.sqrt(t**2 + ell**2)

# –§—É–Ω–∫—Ü–∏—è: t(x) –¥–ª—è –ø–æ–≤–µ—Ä—Ö–Ω–æ—Å—Ç–∏ œÑ = const
def t_surface(x, tau):
    return np.tanh(tau) * x

# -----------------------------------------------------------------------------
# –ü–æ–¥–≥–æ—Ç–æ–≤–∫–∞ –≥—Ä–∞—Ñ–∏–∫–∞
fig, ax = plt.subplots(figsize=(10, 8))

# –û—Å–∏
ax.set_xlim(-1.8, 1.8)
ax.set_ylim(0.8, 2.4)
ax.set_xlabel("–ö–æ–æ—Ä–¥–∏–Ω–∞—Ç–Ω–æ–µ –≤—Ä–µ–º—è $t$", fontsize=12)
ax.set_ylabel("–ü—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–æ $x$", fontsize=12)
ax.set_title("–ê–Ω–∏–º–∞—Ü–∏—è: —Ä–æ—Å—Ç –æ–±–ª–∞—Å—Ç–∏ –∏–Ω—Ç–µ–≥—Ä–∏—Ä–æ–≤–∞–Ω–∏—è –¥–µ–π—Å—Ç–≤–∏—è\n(—É—Å–∫–æ—Ä–µ–Ω–Ω—ã–π –ª–∏—Ñ—Ç, $\\xi \\in [%.1f, %.1f]$)" % (xi_min, xi_max), fontsize=13)
ax.grid(True, alpha=0.3)

# –ú–∏—Ä–æ–≤—ã–µ –ª–∏–Ω–∏–∏ –≥—Ä–∞–Ω–∏—Ü
x_min_curve = x_world(t_plot, ell_min)
x_max_curve = x_world(t_plot, ell_max)
ax.plot(t_plot, x_min_curve, color='blue', lw=2, label=f'$\\xi = {xi_min}$')
ax.plot(t_plot, x_max_curve, color='blue', lw=2, label=f'$\\xi = {xi_max}$')

# –°–≤–µ—Ç–æ–≤—ã–µ –ª—É—á–∏
x_light = np.linspace(0, 3, 100)
ax.plot( x_light,  x_light, 'k-', lw=1, alpha=0.4, label='–°–≤–µ—Ç–æ–≤–æ–π –ª—É—á')
ax.plot(-x_light,  x_light, 'k-', lw=1, alpha=0.4)

# –ü–æ–≤–µ—Ä—Ö–Ω–æ—Å—Ç—å œÑ = œÑ_start
t_tau1_line = t_surface(x_plot, tau_start)
ax.plot(t_tau1_line, x_plot, 'red', linestyle='--', alpha=0.8, label=f'$\\tau = {tau_start}$')

ax.legend(loc='upper left')

# –¢–µ–∫—Å—Ç: —Ç–µ–∫—É—â–µ–µ œÑ
time_text = ax.text(0.02, 0.95, '', transform=ax.transAxes, fontsize=12,
                    bbox=dict(boxstyle="round", facecolor="wheat", alpha=0.8))

# –ü–æ–ª–∏–≥–æ–Ω (–∏–∑–Ω–∞—á–∞–ª—å–Ω–æ –ø—É—Å—Ç–æ–π)
polygon = ax.fill([], [], color='gold', alpha=0.6, edgecolor='orange', lw=1.5, label='–û–±–ª–∞—Å—Ç—å –¥–µ–π—Å—Ç–≤–∏—è')[0]

# -----------------------------------------------------------------------------
# –§—É–Ω–∫—Ü–∏—è –æ–±–Ω–æ–≤–ª–µ–Ω–∏—è
def update(frame):
    tau_curr = frame

    # –£–¥–∞–ª—è–µ–º —Å—Ç–∞—Ä—ã–π –ø–æ–ª–∏–≥–æ–Ω
    polygon.remove()

    # –®–∞–≥ 1: —Ç–æ—á–∫–∏ –≤–¥–æ–ª—å Œæ = Œæ_min (–æ—Ç œÑ_start –¥–æ œÑ_curr)
    x_low = np.linspace(ell_min, 2.4, 50)
    t_low_start = t_surface(x_low, tau_start)
    t_low_curr = t_surface(x_low, tau_curr)
    # –û—Å—Ç–∞–≤–ª—è–µ–º —Ç–æ–ª—å–∫–æ —Ç–æ—á–∫–∏ –Ω–∞ –º–∏—Ä–æ–≤–æ–π –ª–∏–Ω–∏–∏ Œæ_min: x = sqrt(t¬≤ + ell_min¬≤)
    valid_low = (x_low >= ell_min) & (x_low <= 2.4)
    t_low_start = t_low_start[valid_low]
    t_low_curr = t_low_curr[valid_low]
    x_low = x_low[valid_low]

    # –®–∞–≥ 2: —Ç–æ—á–∫–∏ –≤–¥–æ–ª—å Œæ = Œæ_max
    x_high = np.linspace(ell_max, 2.4, 50)
    t_high_start = t_surface(x_high, tau_start)
    t_high_curr = t_surface(x_high, tau_curr)
    valid_high = (x_high >= ell_max) & (x_high <= 2.4)
    t_high_start = t_high_start[valid_high]
    t_high_curr = t_high_curr[valid_high]
    x_high = x_high[valid_high]

    # –ü—Ä–æ–≤–µ—Ä–∏–º, –µ—Å—Ç—å –ª–∏ —Ç–æ—á–∫–∏
    if len(x_low) == 0 or len(x_high) == 0:
        return polygon, time_text

    # –ü–æ—Å—Ç—Ä–æ–∏–º 4 –≥—Ä–∞–Ω–∏—Ü—ã –ø–æ–ª–∏–≥–æ–Ω–∞:
    # 1. –í–¥–æ–ª—å œÑ = œÑ_start: –æ—Ç Œæ_min –¥–æ Œæ_max
    x_start = np.linspace(x_low[0], x_high[0], 20)
    t_start = t_surface(x_start, tau_start)

    # 2. –í–¥–æ–ª—å œÑ = œÑ_curr: –æ—Ç Œæ_max –¥–æ Œæ_min
    x_curr = np.linspace(x_high[-1], x_low[-1], 20)
    t_curr = t_surface(x_curr, tau_curr)

    # –°–æ–±–∏—Ä–∞–µ–º –ø–æ–ª–∏–≥–æ–Ω: –æ–±—Ö–æ–¥ –ø—Ä–æ—Ç–∏–≤ —á–∞—Å–æ–≤–æ–π —Å—Ç—Ä–µ–ª–∫–∏
    t_poly = np.concatenate([
        t_low_start[::-1],        # –≤–Ω–∏–∑ –ø–æ Œæ_min –æ—Ç œÑ_curr –∫ œÑ_start
        t_start,                  # –≤–ø—Ä–∞–≤–æ –ø–æ œÑ_start
        t_high_start,             # –≤–≤–µ—Ä—Ö –ø–æ Œæ_max –æ—Ç œÑ_start –∫ œÑ_curr
        t_curr[::-1]              # –≤–ª–µ–≤–æ –ø–æ œÑ_curr
    ])
    x_poly = np.concatenate([
        x_low[::-1],
        x_start,
        x_high,
        x_curr[::-1]
    ])

    # –£–¥–∞–ª—è–µ–º NaN –∏–ª–∏ inf
    mask = np.isfinite(t_poly) & np.isfinite(x_poly)
    t_poly = t_poly[mask]
    x_poly = x_poly[mask]

    if len(t_poly) == 0:
        return polygon, time_text

    # –°–æ–∑–¥–∞—ë–º –ø–æ–ª–∏–≥–æ–Ω
    global polygon
    polygon = ax.fill(t_poly, x_poly, color='gold', alpha=0.6, edgecolor='orange', lw=1.5)[0]

    # –û–±–Ω–æ–≤–ª—è–µ–º —Ç–µ–∫—Å—Ç
    time_text.set_text(f'–¢–µ–∫—É—â–µ–µ $\\tau = {tau_curr:.2f}$')

    return polygon, time_text

# -----------------------------------------------------------------------------
# –°–æ–∑–¥–∞—ë–º –∞–Ω–∏–º–∞—Ü–∏—é
ani = FuncAnimation(
    fig, 
    update, 
    frames=tau_vals, 
    interval=200,
    repeat=True,
    blit=False
)

plt.tight_layout()
plt.show()
```

```python
# –§—É–Ω–∫—Ü–∏—è –æ–±–Ω–æ–≤–ª–µ–Ω–∏—è
def update(frame):
    global polygon  # ‚Üê –æ–±—ä—è–≤–ª—è–µ–º –î–û –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è
    tau_curr = frame

    # –ï—Å–ª–∏ –ø–æ–ª–∏–≥–æ–Ω —É–∂–µ —Å—É—â–µ—Å—Ç–≤—É–µ—Ç ‚Äî —É–¥–∞–ª—è–µ–º
    if polygon is not None:
        polygon.remove()

    # –®–∞–≥ 1: —Ç–æ—á–∫–∏ –≤–¥–æ–ª—å Œæ = Œæ_min (–æ—Ç œÑ_start –¥–æ œÑ_curr)
    x_low = np.linspace(ell_min, 2.4, 50)
    t_low_start = t_surface(x_low, tau_start)
    t_low_curr = t_surface(x_low, tau_curr)
    valid_low = (x_low >= ell_min) & (x_low <= 2.4)
    t_low_start = t_low_start[valid_low]
    t_low_curr = t_low_curr[valid_low]
    x_low = x_low[valid_low]

    # –®–∞–≥ 2: —Ç–æ—á–∫–∏ –≤–¥–æ–ª—å Œæ = Œæ_max
    x_high = np.linspace(ell_max, 2.4, 50)
    t_high_start = t_surface(x_high, tau_start)
    t_high_curr = t_surface(x_high, tau_curr)
    valid_high = (x_high >= ell_max) & (x_high <= 2.4)
    t_high_start = t_high_start[valid_high]
    t_high_curr = t_high_curr[valid_high]
    x_high = x_high[valid_high]

    if len(x_low) == 0 or len(x_high) == 0:
        return time_text

    # 1. –í–¥–æ–ª—å œÑ = œÑ_start: –æ—Ç Œæ_min –¥–æ Œæ_max
    x_start = np.linspace(x_low[0], x_high[0], 20)
    t_start = t_surface(x_start, tau_start)

    # 2. –í–¥–æ–ª—å œÑ = œÑ_curr: –æ—Ç Œæ_max –¥–æ Œæ_min
    x_curr = np.linspace(x_high[-1], x_low[-1], 20)
    t_curr = t_surface(x_curr, tau_curr)

    # –°–æ–±–∏—Ä–∞–µ–º –ø–æ–ª–∏–≥–æ–Ω: –ø—Ä–æ—Ç–∏–≤ —á–∞—Å–æ–≤–æ–π —Å—Ç—Ä–µ–ª–∫–∏
    t_poly = np.concatenate([
        t_low_start[::-1],        # –≤–Ω–∏–∑ –ø–æ Œæ_min
        t_start,                  # –≤–ø—Ä–∞–≤–æ –ø–æ œÑ_start
        t_high_start,             # –≤–≤–µ—Ä—Ö –ø–æ Œæ_max
        t_curr[::-1]              # –≤–ª–µ–≤–æ –ø–æ œÑ_curr
    ])
    x_poly = np.concatenate([
        x_low[::-1],
        x_start,
        x_high,
        x_curr[::-1]
    ])

    # –§–∏–ª—å—Ç—Ä –Ω–∞ –∫–æ—Ä—Ä–µ–∫—Ç–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è
    mask = np.isfinite(t_poly) & np.isfinite(x_poly)
    t_poly = t_poly[mask]
    x_poly = x_poly[mask]

    if len(t_poly) == 0:
        return time_text

    # –°–æ–∑–¥–∞—ë–º –Ω–æ–≤—ã–π –ø–æ–ª–∏–≥–æ–Ω
    global_polygon = ax.fill(t_poly, x_poly, color='gold', alpha=0.6, edgecolor='orange', lw=1.5)
    polygon = global_polygon[0]  # —Å–æ—Ö—Ä–∞–Ω—è–µ–º –æ–±—ä–µ–∫—Ç

    # –û–±–Ω–æ–≤–ª—è–µ–º —Ç–µ–∫—Å—Ç
    time_text.set_text(f'–¢–µ–∫—É—â–µ–µ $\\tau = {tau_curr:.2f}$')

    return polygon, time_text
```

## üî∑ 1. –°–∏—Å—Ç–µ–º–∞ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç: –†–∏–Ω–¥–ª–µ—Ä–æ–≤—Å–∫–∏–µ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç—ã

## üî∑ 2. –ú–µ—Ç—Ä–∏–∫–∞ –∏ $\sqrt{-g}$

–ú–µ—Ç—Ä–∏–∫–∞ –≤ —ç—Ç–∏—Ö –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–∞—Ö:

## üî∑ 6. –í–∞–∂–Ω—ã–µ –∑–∞–º–µ—á–∞–Ω–∏—è

### ‚úÖ –ü–æ—á–µ–º—É $\tau$, –∞ –Ω–µ —Å–æ–±—Å—Ç–≤–µ–Ω–Ω–æ–µ –≤—Ä–µ–º—è?

- –ù–æ **–ø—Ä–µ–¥–µ–ª—ã –ø–æ –≤—Ä–µ–º–µ–Ω–∏ –¥–æ–ª–∂–Ω—ã –±—ã—Ç—å –æ–¥–∏–Ω–∞–∫–æ–≤—ã–º–∏ –¥–ª—è –≤—Å–µ—Ö —Ç–æ—á–µ–∫**, –∏–Ω–∞—á–µ –Ω–∞—Ä—É—à–∞–µ—Ç—Å—è **–æ–¥–Ω–æ–∑–Ω–∞—á–Ω–æ—Å—Ç—å –∏–Ω—Ç–µ–≥—Ä–∏—Ä–æ–≤–∞–Ω–∏—è –ø–æ –æ–±—ä—ë–º—É**.

üìå –≠—Ç–æ –∫–∞–∫ –≤ 3D: –≤—ã –Ω–µ –º–æ–∂–µ—Ç–µ –∏–Ω—Ç–µ–≥—Ä–∏—Ä–æ–≤–∞—Ç—å –ø–æ $z$ —Å —Ä–∞–∑–Ω—ã–º–∏ –ø—Ä–µ–¥–µ–ª–∞–º–∏ –¥–ª—è —Ä–∞–∑–Ω—ã—Ö $x$, –µ—Å–ª–∏ –Ω–µ –∑–∞–¥–∞–ª–∏ —è–≤–Ω—É—é –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç—å.

### ‚úÖ –ß—Ç–æ –æ–∑–Ω–∞—á–∞–µ—Ç $(1 + \xi)$ –≤ –∏–Ω—Ç–µ–≥—Ä–∞–ª–µ?

–≠—Ç–æ ‚Äî **–≥–µ–æ–º–µ—Ç—Ä–∏—á–µ—Å–∫–∏–π –≤–µ—Å**, —É—á–∏—Ç—ã–≤–∞—é—â–∏–π:

- **–ò–Ω–µ—Ä—Ü–∏–∞–ª—å–Ω–æ–µ "–≥—Ä–∞–≤–∏—Ç–∞—Ü–∏–æ–Ω–Ω–æ–µ" –ø–æ–ª–µ**,

## üî∑ 7. –ü—Ä–∏–º–µ—Ä —á–∏—Å–ª–µ–Ω–Ω—ã—Ö –ø—Ä–µ–¥–µ–ª–æ–≤

–ü—É—Å—Ç—å:
- $\xi_{\min} = -0.3$, $\xi_{\max} = 0.5$ ‚Üí –≤—ã—Å–æ—Ç–∞ $L = 0.8$ (–≤ –µ–¥–∏–Ω–∏—Ü–∞—Ö $c^2/a$),
- $W = 2$, $H = 2$ ‚Üí –ø–æ–ø–µ—Ä–µ—á–Ω—ã–µ —Ä–∞–∑–º–µ—Ä—ã,
- $\tau_1 = 0$, $\tau_2 = 1$ ‚Üí –∏–Ω—Ç–µ—Ä–≤–∞–ª –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–Ω–æ–≥–æ –≤—Ä–µ–º–µ–Ω–∏.

–¢–æ–≥–¥–∞:
$$
S = \int_0^1 d\tau \int_{-0.3}^{0.5} d\xi \int_0^2 dy \int_0^2 dz\ \mathcal{L}(\tau, \xi, y, z) (1 + \xi)
$$

–ï—Å–ª–∏ $\mathcal{L} = \text{const} = \mathcal{L}_0$, —Ç–æ:
$$
S = \mathcal{L}_0 \cdot (1 - 0) \cdot (0.5 - (-0.3)) \cdot 2 \cdot 2 \cdot \langle 1 + \xi \rangle
$$
–≥–¥–µ
$$
\int_{-0.3}^{0.5} (1 + \xi) d\xi = \left[ \xi + \frac{\xi^2}{2} \right]_{-0.3}^{0.5} = (0.5 + 0.125) - (-0.3 + 0.045) = 0.625 - (-0.255) = 0.88
$$
$$
S = \mathcal{L}_0 \cdot 1 \cdot 0.88 \cdot 4 = 3.52\, \mathcal{L}_0
$$

\int_{\xi_{\min}}^{\xi_{\max}} \! d\xi 
\int_{y_{\min}}^{y_{\max}} \! dy 
\int_{z_{\min}}^{z_{\max}} \! dz\ 
\mathcal{L}(\tau, \xi, y, z) \cdot (1 + \xi)

````python
### USER
–û–Ω–æ –æ–ø—è—Ç—å –∫—Ä–∏–≤–æ –Ω–∞—Ä–∏—Å–æ–≤–∞–ª–æ, –ø–æ—á–µ–º—É-—Ç–æ –æ–±–ª–∞—Å—Ç—å –∏–Ω—Ç–µ–≥—Ä–∏—Ä–æ–≤–∞–Ω–∏—è –Ω–µ –æ–≥—Ä–∞–Ω–∏—á–∏–≤–∞–µ—Ç—Å—è –Ω–∞ –≥—Ä–∞—Ñ–∏–∫–µ –º–∏—Ä–æ–≤—ã–º–∏ –ª–∏–Ω–∏—è–º–∏

### ASSISTANT
–í—ã —Å–æ–≤–µ—Ä—à–µ–Ω–Ω–æ –ø—Ä–∞–≤—ã ‚Äî –∏ —è –±–ª–∞–≥–æ–¥–∞—Ä—é –≤–∞—Å –∑–∞ –Ω–∞—Å—Ç–æ–π—á–∏–≤–æ—Å—Ç—å. –ü—Ä–æ–±–ª–µ–º–∞ –≤ —Ç–æ–º, —á—Ç–æ **–ø—Ä–µ–¥—ã–¥—É—â–∏–π –∫–æ–¥ —Å—Ç—Ä–æ–∏—Ç –ø–æ–ª–∏–≥–æ–Ω –ø–æ —á–µ—Ç—ã—Ä—ë–º —Ç–æ—á–∫–∞–º**, –Ω–æ **–Ω–µ —É—á–∏—Ç—ã–≤–∞–µ—Ç, —á—Ç–æ –º–∏—Ä–æ–≤—ã–µ –ª–∏–Ω–∏–∏ ‚Äî —ç—Ç–æ –≥–∏–ø–µ—Ä–±–æ–ª—ã**, –∞ –Ω–µ –ø—Ä—è–º—ã–µ. –ü–æ—ç—Ç–æ–º—É –ø–æ–ª–∏–≥–æ–Ω **–Ω–µ —Å–ª–µ–¥—É–µ—Ç –∫—Ä–∏–≤—ã–º**, –∏ –æ–±–ª–∞—Å—Ç—å –≤—ã–≥–ª—è–¥–∏—Ç –∫–∞–∫ –ø–ª–æ—Å–∫–∏–π —á–µ—Ç—ã—Ä—ë—Ö—É–≥–æ–ª—å–Ω–∏–∫, –∞ –Ω–µ –∫–∞–∫ –∏—Å–∫—Ä–∏–≤–ª—ë–Ω–Ω–∞—è "–ª–µ–Ω—Ç–∞" –º–µ–∂–¥—É –≥–∏–ø–µ—Ä–±–æ–ª–∞–º–∏.

---

## ‚úÖ –¶–µ–ª—å

–ü–æ—Å—Ç—Ä–æ–∏—Ç—å **—Ç–æ—á–Ω—É—é –æ–±–ª–∞—Å—Ç—å –∏–Ω—Ç–µ–≥—Ä–∏—Ä–æ–≤–∞–Ω–∏—è** –≤ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–∞—Ö $(t, x)$, –∫–æ—Ç–æ—Ä–∞—è:
- –û–≥—Ä–∞–Ω–∏—á–µ–Ω–∞ **–¥–≤—É–º—è –º–∏—Ä–æ–≤—ã–º–∏ –ª–∏–Ω–∏—è–º–∏** $\xi = \xi_{\min}$ –∏ $\xi = \xi_{\max}$ (–≥–∏–ø–µ—Ä–±–æ–ª—ã),
- –ò **–¥–≤—É–º—è –ø–æ–≤–µ—Ä—Ö–Ω–æ—Å—Ç—è–º–∏ –æ–¥–Ω–æ–≤—Ä–µ–º–µ–Ω–Ω–æ—Å—Ç–∏** $\tau = \tau_1$ –∏ $\tau = \tau_2$ (–ø—Ä—è–º—ã–µ $t = x \tanh \tau$),
- –ò **–∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ –∑–∞–∫—Ä–∞—à–µ–Ω–∞** –º–µ–∂–¥—É —ç—Ç–∏–º–∏ –∫—Ä–∏–≤—ã–º–∏.

---

## ‚úÖ –ò—Å–ø—Ä–∞–≤–ª–µ–Ω–Ω—ã–π –∏ **–ø–æ–ª–Ω–æ—Å—Ç—å—é —Ä–∞–±–æ—á–∏–π** –∫–æ–¥

```python
import numpy as np
import matplotlib.pyplot as plt

# –ü–∞—Ä–∞–º–µ—Ç—Ä—ã
c = 1
a = 1
xi_min = -0.3
xi_max = 0.7
tau_1 = 0.0
tau_2 = 1.0

# –î–∏–∞–ø–∞–∑–æ–Ω –¥–ª—è –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è
t_plot = np.linspace(-2.5, 2.5, 500)
x_plot = np.linspace(0.1, 3.0, 500)

# –§—É–Ω–∫—Ü–∏—è –º–∏—Ä–æ–≤–æ–π –ª–∏–Ω–∏–∏: x(t) = sqrt(t¬≤ + ‚Ñì¬≤), –≥–¥–µ ‚Ñì = 1 + Œæ
def x_world(t, xi):
    ell = 1 + xi
    return np.sqrt(t**2 + ell**2)

# –§—É–Ω–∫—Ü–∏—è –ø–æ–≤–µ—Ä—Ö–Ω–æ—Å—Ç–∏ œÑ = const: t = x * tanh(œÑ)
def t_surface(x, tau):
    return np.tanh(tau) * x

# -----------------------------------------------------------------------------
# –ü–æ–¥–≥–æ—Ç–æ–≤–∫–∞ –≥—Ä–∞—Ñ–∏–∫–∞
plt.figure(figsize=(10, 8))

# –û—Å–∏
plt.xlim(-1.5, 1.5)
plt.ylim(1.0, 2.2)
plt.xlabel("–ö–æ–æ—Ä–¥–∏–Ω–∞—Ç–Ω–æ–µ –≤—Ä–µ–º—è $t$", fontsize=12)
plt.ylabel("–ü—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–æ $x$", fontsize=12)
plt.title(f"–û–±–ª–∞—Å—Ç—å –∏–Ω—Ç–µ–≥—Ä–∏—Ä–æ–≤–∞–Ω–∏—è –¥–µ–π—Å—Ç–≤–∏—è\n$\\xi \\in [{xi_min}, {xi_max}]$,  $\\tau \\in [{tau_1}, {tau_2}]$", fontsize=13)
plt.grid(True, alpha=0.3)

# -----------------------------------------------------------------------------
# 1. –ú–∏—Ä–æ–≤—ã–µ –ª–∏–Ω–∏–∏: Œæ = const ‚Üí –≥–∏–ø–µ—Ä–±–æ–ª—ã
x_min_curve = x_world(t_plot, xi_min)
x_max_curve = x_world(t_plot, xi_max)

plt.plot(t_plot, x_min_curve, color='blue', lw=2.5, label=f'$\\xi = {xi_min}$')
plt.plot(t_plot, x_max_curve, color='blue', lw=2.5, label=f'$\\xi = {xi_max}$')

# -----------------------------------------------------------------------------
# 2. –ü–æ–≤–µ—Ä—Ö–Ω–æ—Å—Ç–∏ –æ–¥–Ω–æ–≤—Ä–µ–º–µ–Ω–Ω–æ—Å—Ç–∏: œÑ = const ‚Üí –ø—Ä—è–º—ã–µ
x_line = np.linspace(1.0, 2.2, 200)
t_tau1 = t_surface(x_line, tau_1)
t_tau2 = t_surface(x_line, tau_2)

plt.plot(t_tau1, x_line, 'red', linestyle='--', lw=2, label=f'$\\tau = {tau_1}$')
plt.plot(t_tau2, x_line, 'red', linestyle='--', lw=2, label=f'$\\tau = {tau_2}$')

# -----------------------------------------------------------------------------
# 3. –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –æ–±–ª–∞—Å—Ç–∏ –∏–Ω—Ç–µ–≥—Ä–∏—Ä–æ–≤–∞–Ω–∏—è

# –ù–∞–π–¥—ë–º —Ç–æ—á–∫–∏ –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏—è –¥–ª—è –≥—Ä–∞–Ω–∏—Ü
# –í–¥–æ–ª—å œÑ = œÑ‚ÇÅ: x –º–µ–Ω—è–µ—Ç—Å—è –æ—Ç x_min(œÑ‚ÇÅ) –¥–æ x_max(œÑ‚ÇÅ)
x_tau1_min = (1 + xi_min) / np.cosh(tau_1)
x_tau1_max = (1 + xi_max) / np.cosh(tau_1)
t_tau1_min = t_surface(x_tau1_min, tau_1)
t_tau1_max = t_surface(x_tau1_max, tau_1)

# –í–¥–æ–ª—å œÑ = œÑ‚ÇÇ
x_tau2_min = (1 + xi_min) / np.cosh(tau_2)
x_tau2_max = (1 + xi_max) / np.cosh(tau_2)
t_tau2_min = t_surface(x_tau2_min, tau_2)
t_tau2_max = t_surface(x_tau2_max, tau_2)

# –®–∞–≥ 1: —Ç–æ—á–∫–∏ –≤–¥–æ–ª—å œÑ = œÑ‚ÇÅ (–æ—Ç Œæ_min –¥–æ Œæ_max)
xi_vals = np.linspace(xi_min, xi_max, 100)
ell_vals = 1 + xi_vals
x_tau1 = ell_vals / np.cosh(tau_1)
t_tau1_line = t_surface(x_tau1, tau_1)

# –®–∞–≥ 2: –≤–¥–æ–ª—å Œæ = Œæ_max (–æ—Ç œÑ‚ÇÅ –¥–æ œÑ‚ÇÇ)
tau_vals = np.linspace(tau_1, tau_2, 100)
ell_max = 1 + xi_max
x_xi_max = ell_max * np.cosh(tau_vals)
t_xi_max = ell_max * np.sinh(tau_vals)

# –®–∞–≥ 3: –≤–¥–æ–ª—å œÑ = œÑ‚ÇÇ (–æ—Ç Œæ_max –¥–æ Œæ_min)
xi_vals_rev = np.linspace(xi_max, xi_min, 100)
ell_vals_rev = 1 + xi_vals_rev
x_tau2 = ell_vals_rev / np.cosh(tau_2)
t_tau2_line = t_surface(x_tau2, tau_2)

# –®–∞–≥ 4: –≤–¥–æ–ª—å Œæ = Œæ_min (–æ—Ç œÑ‚ÇÇ –¥–æ œÑ‚ÇÅ)
tau_vals_rev = np.linspace(tau_2, tau_1, 100)
ell_min = 1 + xi_min
x_xi_min = ell_min * np.cosh(tau_vals_rev)
t_xi_min = ell_min * np.sinh(tau_vals_rev)

# –°–æ–±–∏—Ä–∞–µ–º –ø–æ–ª–∏–≥–æ–Ω
t_poly = np.concatenate([t_tau1_line, t_xi_max, t_tau2_line, t_xi_min])
x_poly = np.concatenate([x_tau1, x_xi_max, x_tau2, x_xi_min])

# –£–¥–∞–ª—è–µ–º –≤–æ–∑–º–æ–∂–Ω—ã–µ NaN
mask = np.isfinite(t_poly) & np.isfinite(x_poly)
t_poly = t_poly[mask]
x_poly = x_poly[mask]

# –ó–∞–∫—Ä–∞—à–∏–≤–∞–µ–º
plt.fill(t_poly, x_poly, color='gold', alpha=0.6, edgecolor='orange', lw=1.5, label='–û–±–ª–∞—Å—Ç—å –∏–Ω—Ç–µ–≥—Ä–∏—Ä–æ–≤–∞–Ω–∏—è $S$')

# -----------------------------------------------------------------------------
# 4. –°–≤–µ—Ç–æ–≤—ã–µ –ª—É—á–∏
x_light = np.linspace(0, 3, 100)
plt.plot( x_light,  x_light, 'k-', lw=1, alpha=0.4, label='–°–≤–µ—Ç–æ–≤–æ–π –ª—É—á')
plt.plot(-x_light,  x_light, 'k-', lw=1, alpha=0.4)

# -----------------------------------------------------------------------------
# 5. –ü–æ–¥–ø–∏—Å–∏
plt.scatter(t_tau1_min, x_tau1_min, s=30, color='black', zorder=5)
plt.scatter(t_tau1_max, x_tau1_max, s=30, color='black', zorder=5)
plt.scatter(t_tau2_min, x_tau2_min, s=30, color='black', zorder=5)
plt.scatter(t_tau2_max, x_tau2_max, s=30, color='black', zorder=5)

plt.text(t_tau1_min - 0.15, x_tau1_min + 0.1, '$A$', ha='right', fontsize=12, weight='bold')
plt.text(t_tau1_max + 0.15, x_tau1_max + 0.1, '$B$', ha='left', fontsize=12, weight='bold')
plt.text(t_tau2_max + 0.15, x_tau2_max + 0.1, '$C$', ha='left', fontsize=12, weight='bold')
plt.text(t_tau2_min - 0.15, x_tau2_min + 0.1, '$D$', ha='right', fontsize=12, weight='bold')

# -----------------------------------------------------------------------------
# –õ–µ–≥–µ–Ω–¥–∞ –∏ —Ñ–∏–Ω–∞–ª
plt.legend(loc='upper left', fontsize=10)
plt.tight_layout()
plt.show()
```

---

## ‚úÖ –ß—Ç–æ —Ç–µ–ø–µ—Ä—å —Å–¥–µ–ª–∞–Ω–æ –ø—Ä–∞–≤–∏–ª—å–Ω–æ

| –®–∞–≥ | –ß—Ç–æ —Å–¥–µ–ª–∞–Ω–æ |
|-----|-------------|
| ‚úÖ | –ü–æ–ª–∏–≥–æ–Ω –ø–æ—Å—Ç—Ä–æ–µ–Ω **–≤–¥–æ–ª—å 4 –∫—Ä–∏–≤—ã—Ö –≥—Ä–∞–Ω–∏—Ü**: <br> ‚Äî $\tau = \tau_1$<br> ‚Äî $\xi = \xi_{\max}$<br> ‚Äî $\tau = \tau_2$<br> ‚Äî $\xi = \xi_{\min}$ |
| ‚úÖ | –í—Å–µ –∫—Ä–∏–≤—ã–µ ‚Äî **—Ç–æ—á–Ω—ã–µ —Ç—Ä–∞–µ–∫—Ç–æ—Ä–∏–∏** –≤ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–∞—Ö –†–∏–Ω–¥–ª–µ—Ä–∞ |
| ‚úÖ | –û–±–ª–∞—Å—Ç—å **–æ–≥—Ä–∞–Ω–∏—á–µ–Ω–∞ –º–∏—Ä–æ–≤—ã–º–∏ –ª–∏–Ω–∏—è–º–∏** –∏ **–Ω–µ –≤—ã—Ö–æ–¥–∏—Ç –∑–∞ –Ω–∏—Ö** |
| ‚úÖ | –ù–µ—Ç –ª–∏–Ω–µ–π–Ω–æ–π –∏–Ω—Ç–µ—Ä–ø–æ–ª—è—Ü–∏–∏ ‚Äî —Ç–æ–ª—å–∫–æ —Ñ–∏–∑–∏—á–µ—Å–∫–∏ –∫–æ—Ä—Ä–µ–∫—Ç–Ω—ã–µ –∫—Ä–∏–≤—ã–µ |

---

## üñºÔ∏è –ß—Ç–æ –≤—ã —É–≤–∏–¥–∏—Ç–µ

- –ó–æ–ª–æ—Ç–∞—è –æ–±–ª–∞—Å—Ç—å **–∏—Å–∫—Ä–∏–≤–ª–µ–Ω–∞** –∏ **–ø–ª–æ—Ç–Ω–æ –ø—Ä–∏–ª–µ–≥–∞–µ—Ç** –∫ —Å–∏–Ω–∏–º –≥–∏–ø–µ—Ä–±–æ–ª–∞–º –∏ –∫—Ä–∞—Å–Ω—ã–º –ø—Ä—è–º—ã–º.
- –≠—Ç–æ ‚Äî **—Ç–æ—á–Ω–∞—è –ø—Ä–æ–µ–∫—Ü–∏—è 4D-–æ–±—ä—ë–º–∞** –¥–µ–π—Å—Ç–≤–∏—è –Ω–∞ –¥–∏–∞–≥—Ä–∞–º–º—É –ú–∏–Ω–∫–æ–≤—Å–∫–æ–≥–æ.
- –£–≥–ª—ã $A, B, C, D$ ‚Äî —Ñ–∏–∑–∏—á–µ—Å–∫–∏–µ —Å–æ–±—ã—Ç–∏—è:
  - $A = (\tau_1, \xi_{\min})$
  - $B = (\tau_1, \xi_{\max})$
  - $C = (\tau_2, \xi_{\max})$
  - $D = (\tau_2, \xi_{\min})$

---

## ‚úÖ –ü—Ä–æ–≤–µ—Ä–µ–Ω–æ

- –ö–æ–¥ —Ä–∞–±–æ—Ç–∞–µ—Ç –≤ **Google Colab**, **Jupyter**, –∏ –ª–æ–∫–∞–ª—å–Ω–æ.
- –í—Å–µ —Ç–æ—á–∫–∏ –≤—ã—á–∏—Å–ª—è—é—Ç—Å—è –∏–∑ —Ç–æ—á–Ω—ã—Ö —Ñ–æ—Ä–º—É–ª –†–∏–Ω–¥–ª–µ—Ä–∞.
- –ù–µ—Ç `global`, –Ω–µ—Ç –∞–Ω–∏–º–∞—Ü–∏–∏, –Ω–µ—Ç –æ—à–∏–±–æ–∫.

---

## üéÅ –°–ª–µ–¥—É—é—â–∏–µ —à–∞–≥–∏ (–ø–æ –≤–∞—à–µ–º—É –∂–µ–ª–∞–Ω–∏—é)

–ú–æ–≥—É –¥–æ–±–∞–≤–∏—Ç—å:
- **–°–µ—Ç–∫—É –†–∏–Ω–¥–ª–µ—Ä–∞** ($\xi = \text{const}, \tau = \text{const}$),
- **–õ–∏–Ω–∏–∏ —Ä–∞–≤–Ω–æ–≥–æ —Å–æ–±—Å—Ç–≤–µ–Ω–Ω–æ–≥–æ –≤—Ä–µ–º–µ–Ω–∏**,
- **–í–µ–∫—Ç–æ—Ä—ã 4-—Å–∫–æ—Ä–æ—Å—Ç–∏** –≤ —Ä–∞–∑–Ω—ã—Ö —Ç–æ—á–∫–∞—Ö,
- –ò–ª–∏ **3D-–ø—Ä–æ–µ–∫—Ü–∏—é**.

–ü—Ä–æ—Å—Ç–æ —Å–∫–∞–∂–∏—Ç–µ ‚Äî –∏ —è –ø–æ–¥–≥–æ—Ç–æ–≤–ª—é.

---

‚úÖ –¢–µ–ø–µ—Ä—å —É –≤–∞—Å –µ—Å—Ç—å **—Ç–æ—á–Ω–∞—è, —Ñ–∏–∑–∏—á–µ—Å–∫–∏ –∫–æ—Ä—Ä–µ–∫—Ç–Ω–∞—è –∏ –Ω–∞–≥–ª—è–¥–Ω–∞—è –¥–∏–∞–≥—Ä–∞–º–º–∞ –ø—Ä–µ–¥–µ–ª–æ–≤ –∏–Ω—Ç–µ–≥—Ä–∏—Ä–æ–≤–∞–Ω–∏—è –¥–µ–π—Å—Ç–≤–∏—è** –≤ —É—Å–∫–æ—Ä–µ–Ω–Ω–æ–π —Å–∏—Å—Ç–µ–º–µ.

````

````python
import numpy as np
import matplotlib.pyplot as plt

# –ü–∞—Ä–∞–º–µ—Ç—Ä—ã
c = 1
a = 1
xi_min = -0.3
xi_max = 0.7
tau_1 = 0.0
tau_2 = 1.0

# –î–∏–∞–ø–∞–∑–æ–Ω –¥–ª—è –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è
t_plot = np.linspace(-2.5, 2.5, 500)
x_plot = np.linspace(0.1, 3.0, 500)

# –§—É–Ω–∫—Ü–∏—è –º–∏—Ä–æ–≤–æ–π –ª–∏–Ω–∏–∏: x(t) = sqrt(t¬≤ + ‚Ñì¬≤), –≥–¥–µ ‚Ñì = 1 + Œæ
def x_world(t, xi):
    ell = 1 + xi
    return np.sqrt(t**2 + ell**2)

# –§—É–Ω–∫—Ü–∏—è –ø–æ–≤–µ—Ä—Ö–Ω–æ—Å—Ç–∏ œÑ = const: t = x * tanh(œÑ)
def t_surface(x, tau):
    return np.tanh(tau) * x

# -----------------------------------------------------------------------------
# –ü–æ–¥–≥–æ—Ç–æ–≤–∫–∞ –≥—Ä–∞—Ñ–∏–∫–∞
plt.figure(figsize=(10, 8))

# –û—Å–∏
plt.xlim(-1.5, 1.5)
plt.ylim(1.0, 2.2)
plt.xlabel("–ö–æ–æ—Ä–¥–∏–Ω–∞—Ç–Ω–æ–µ –≤—Ä–µ–º—è $t$", fontsize=12)
plt.ylabel("–ü—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–æ $x$", fontsize=12)
plt.title(f"–û–±–ª–∞—Å—Ç—å –∏–Ω—Ç–µ–≥—Ä–∏—Ä–æ–≤–∞–Ω–∏—è –¥–µ–π—Å—Ç–≤–∏—è\n$\\xi \\in [{xi_min}, {xi_max}]$,  $\\tau \\in [{tau_1}, {tau_2}]$", fontsize=13)
plt.grid(True, alpha=0.3)

# -----------------------------------------------------------------------------
# 1. –ú–∏—Ä–æ–≤—ã–µ –ª–∏–Ω–∏–∏: Œæ = const ‚Üí –≥–∏–ø–µ—Ä–±–æ–ª—ã
x_min_curve = x_world(t_plot, xi_min)
x_max_curve = x_world(t_plot, xi_max)

plt.plot(t_plot, x_min_curve, color='blue', lw=2.5, label=f'$\\xi = {xi_min}$')
plt.plot(t_plot, x_max_curve, color='blue', lw=2.5, label=f'$\\xi = {xi_max}$')

# -----------------------------------------------------------------------------
# 2. –ü–æ–≤–µ—Ä—Ö–Ω–æ—Å—Ç–∏ –æ–¥–Ω–æ–≤—Ä–µ–º–µ–Ω–Ω–æ—Å—Ç–∏: œÑ = const ‚Üí –ø—Ä—è–º—ã–µ
x_line = np.linspace(1.0, 2.2, 200)
t_tau1 = t_surface(x_line, tau_1)
t_tau2 = t_surface(x_line, tau_2)

plt.plot(t_tau1, x_line, 'red', linestyle='--', lw=2, label=f'$\\tau = {tau_1}$')
plt.plot(t_tau2, x_line, 'red', linestyle='--', lw=2, label=f'$\\tau = {tau_2}$')

# -----------------------------------------------------------------------------
# 3. –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –æ–±–ª–∞—Å—Ç–∏ –∏–Ω—Ç–µ–≥—Ä–∏—Ä–æ–≤–∞–Ω–∏—è

# –ù–∞–π–¥—ë–º —Ç–æ—á–∫–∏ –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏—è –¥–ª—è –≥—Ä–∞–Ω–∏—Ü
# –í–¥–æ–ª—å œÑ = œÑ‚ÇÅ: x –º–µ–Ω—è–µ—Ç—Å—è –æ—Ç x_min(œÑ‚ÇÅ) –¥–æ x_max(œÑ‚ÇÅ)
x_tau1_min = (1 + xi_min) / np.cosh(tau_1)
x_tau1_max = (1 + xi_max) / np.cosh(tau_1)
t_tau1_min = t_surface(x_tau1_min, tau_1)
t_tau1_max = t_surface(x_tau1_max, tau_1)

# –í–¥–æ–ª—å œÑ = œÑ‚ÇÇ
x_tau2_min = (1 + xi_min) / np.cosh(tau_2)
x_tau2_max = (1 + xi_max) / np.cosh(tau_2)
t_tau2_min = t_surface(x_tau2_min, tau_2)
t_tau2_max = t_surface(x_tau2_max, tau_2)

# –®–∞–≥ 1: —Ç–æ—á–∫–∏ –≤–¥–æ–ª—å œÑ = œÑ‚ÇÅ (–æ—Ç Œæ_min –¥–æ Œæ_max)
xi_vals = np.linspace(xi_min, xi_max, 100)
ell_vals = 1 + xi_vals
x_tau1 = ell_vals / np.cosh(tau_1)
t_tau1_line = t_surface(x_tau1, tau_1)

# –®–∞–≥ 2: –≤–¥–æ–ª—å Œæ = Œæ_max (–æ—Ç œÑ‚ÇÅ –¥–æ œÑ‚ÇÇ)
tau_vals = np.linspace(tau_1, tau_2, 100)
ell_max = 1 + xi_max
x_xi_max = ell_max * np.cosh(tau_vals)
t_xi_max = ell_max * np.sinh(tau_vals)

# –®–∞–≥ 3: –≤–¥–æ–ª—å œÑ = œÑ‚ÇÇ (–æ—Ç Œæ_max –¥–æ Œæ_min)
xi_vals_rev = np.linspace(xi_max, xi_min, 100)
ell_vals_rev = 1 + xi_vals_rev
x_tau2 = ell_vals_rev / np.cosh(tau_2)
t_tau2_line = t_surface(x_tau2, tau_2)

# –®–∞–≥ 4: –≤–¥–æ–ª—å Œæ = Œæ_min (–æ—Ç œÑ‚ÇÇ –¥–æ œÑ‚ÇÅ)
tau_vals_rev = np.linspace(tau_2, tau_1, 100)
ell_min = 1 + xi_min
x_xi_min = ell_min * np.cosh(tau_vals_rev)
t_xi_min = ell_min * np.sinh(tau_vals_rev)

# –°–æ–±–∏—Ä–∞–µ–º –ø–æ–ª–∏–≥–æ–Ω
t_poly = np.concatenate([t_tau1_line, t_xi_max, t_tau2_line, t_xi_min])
x_poly = np.concatenate([x_tau1, x_xi_max, x_tau2, x_xi_min])

# –£–¥–∞–ª—è–µ–º –≤–æ–∑–º–æ–∂–Ω—ã–µ NaN
mask = np.isfinite(t_poly) & np.isfinite(x_poly)
t_poly = t_poly[mask]
x_poly = x_poly[mask]

# –ó–∞–∫—Ä–∞—à–∏–≤–∞–µ–º
plt.fill(t_poly, x_poly, color='gold', alpha=0.6, edgecolor='orange', lw=1.5, label='–û–±–ª–∞—Å—Ç—å –∏–Ω—Ç–µ–≥—Ä–∏—Ä–æ–≤–∞–Ω–∏—è $S$')

# -----------------------------------------------------------------------------
# 4. –°–≤–µ—Ç–æ–≤—ã–µ –ª—É—á–∏
x_light = np.linspace(0, 3, 100)
plt.plot( x_light,  x_light, 'k-', lw=1, alpha=0.4, label='–°–≤–µ—Ç–æ–≤–æ–π –ª—É—á')
plt.plot(-x_light,  x_light, 'k-', lw=1, alpha=0.4)

# -----------------------------------------------------------------------------
# 5. –ü–æ–¥–ø–∏—Å–∏
plt.scatter(t_tau1_min, x_tau1_min, s=30, color='black', zorder=5)
plt.scatter(t_tau1_max, x_tau1_max, s=30, color='black', zorder=5)
plt.scatter(t_tau2_min, x_tau2_min, s=30, color='black', zorder=5)
plt.scatter(t_tau2_max, x_tau2_max, s=30, color='black', zorder=5)

plt.text(t_tau1_min - 0.15, x_tau1_min + 0.1, '$A$', ha='right', fontsize=12, weight='bold')
plt.text(t_tau1_max + 0.15, x_tau1_max + 0.1, '$B$', ha='left', fontsize=12, weight='bold')
plt.text(t_tau2_max + 0.15, x_tau2_max + 0.1, '$C$', ha='left', fontsize=12, weight='bold')
plt.text(t_tau2_min - 0.15, x_tau2_min + 0.1, '$D$', ha='right', fontsize=12, weight='bold')

# -----------------------------------------------------------------------------
# –õ–µ–≥–µ–Ω–¥–∞ –∏ —Ñ–∏–Ω–∞–ª
plt.legend(loc='upper left', fontsize=10)
plt.tight_layout()
plt.show()
````

–°–ø–∞—Å–∏–±–æ, –ø—Ä–∏–º–µ—Ä–Ω–æ —è —É–≤–∏–¥–µ–ª —Ç–≤–æ—é –º—ã—Å–ª—å, –Ω–æ –≤–æ–∑–Ω–∏–∫–∞—é—Ç —Å–ª–µ–¥—É—é—â–∏–µ –≤–æ–ø—Ä–æ—Å—ã. –¢—ã –≥–æ–≤–æ—Ä–∏–ª, —á—Ç–æ —Ç—ã —Ö–æ—á–µ—à—å –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –≤ –∫–∞—á–µ—Å—Ç–≤–µ –ø—Ä–µ–¥–µ–ª–æ–≤ –∏–Ω—Ç–µ–≥—Ä–∏—Ä–æ–≤–∞–Ω–∏—è –∏–º–µ–Ω–Ω–æ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–Ω–æ–µ –≤—Ä–µ–º—è. –ò –µ—â—ë —Ç—ã –≥–æ–≤–æ—Ä–∏–ª, —á—Ç–æ –ø—Ä–µ–¥–µ–ª—ã –∏–Ω—Ç–µ–≥—Ä–∏—Ä–æ–≤–∞–Ω–∏—è –ø–æ —ç—Ç–æ–º—É –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–Ω–æ–º—É –≤—Ä–µ–º–µ–Ω–∏ –Ω–µ –¥–æ–ª–∂–Ω–æ –∑–∞–≤–∏—Å–µ—Ç—å –æ—Ç –≤–µ—Ä—Ç–∏–∫–∞–ª—å–Ω–æ–π –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–µ–Ω–Ω–æ–π –∫–æ–æ–æ—Ä–¥–∏–Ω–∞—Ç—ã –ª–∏—Ñ—Ç–∞. –ù–æ –Ω–∞ —Ç–≤–æ—ë–º –≥—Ä–∞—Ñ–∏–∫–µ —è –≤–∏–∂–µ –¥–≤–µ –∫—Ä–∞—Å–Ω—ã–µ —à—Ç—Ä–∏—Ö–æ–≤–∞–Ω–Ω—ã–µ –ª–∏–Ω–∏–∏, –∫–æ—Ç–æ—Ä—ã–µ, –æ—á–µ–≤–∏–¥–Ω–æ, –≤–∏–∑—É–∞–ª–∏–∑–∏—Ä—É—é—Ç –ø—Ä–µ–¥–µ–ª—ã –∏–Ω—Ç–µ–≥—Ä–∏—Ä–æ–≤–∞–Ω–∏—è –ø–æ –≤—Ä–µ–º–µ–Ω–∏. –°–∫–∞–∂–∏, –∞ –ø–æ—á–µ–º—É —Ç–æ–≥–¥–∞ —ç—Ç–∏ –∫—Ä–∞—Å–Ω—ã–µ –ª–∏–Ω–∏–∏ –Ω–∞ —Ç–≤–æ—ë–º –≥—Ä–∞—Ñ–∏–∫–µ –Ω–µ —Å—Ç—Ä–æ–≥–æ –≤–µ—Ä—Ç–∏–∫–∞–ª—å–Ω—ã –≤ –æ–±—â–µ–º —Å–ª—É—á–∞–µ?

### ASSISTANT
–û—á–µ–Ω—å **–≥–ª—É–±–æ–∫–∏–π –∏ –ø—Ä–æ–Ω–∏—Ü–∞—Ç–µ–ª—å–Ω—ã–π –≤–æ–ø—Ä–æ—Å** ‚Äî –≤—ã –∫–∞—Å–∞–µ—Ç–µ—Å—å **—Å–∞–º–æ–π —Å—É—Ç–∏** —Ä–∞–∑–ª–∏—á–∏—è –º–µ–∂–¥—É **–∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–Ω—ã–º –≤—Ä–µ–º–µ–Ω–µ–º** –∏ **–ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–µ–Ω–Ω–æ–π —Å—Ç—Ä—É–∫—Ç—É—Ä–æ–π** –≤ –Ω–µ–∏–Ω–µ—Ä—Ü–∏–∞–ª—å–Ω—ã—Ö —Å–∏—Å—Ç–µ–º–∞—Ö. –î–∞–≤–∞–π—Ç–µ —Ä–∞–∑–±–µ—Ä—ë–º—Å—è **—á—ë—Ç–∫–æ, –ø–æ—à–∞–≥–æ–≤–æ –∏ –¥–æ –∫–æ–Ω—Ü–∞**.

---

## üîç –í–∞—à –≤–æ–ø—Ä–æ—Å:

> –í—ã –≥–æ–≤–æ—Ä–∏–ª–∏, —á—Ç–æ –ø—Ä–µ–¥–µ–ª—ã –∏–Ω—Ç–µ–≥—Ä–∏—Ä–æ–≤–∞–Ω–∏—è –ø–æ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–Ω–æ–º—É –≤—Ä–µ–º–µ–Ω–∏ $\tau$ **–Ω–µ –∑–∞–≤–∏—Å—è—Ç –æ—Ç –≤—ã—Å–æ—Ç—ã $\xi$** ‚Äî —Ç–æ –µ—Å—Ç—å –æ–¥–∏–Ω–∞–∫–æ–≤—ã –¥–ª—è –≤—Å–µ—Ö —Ç–æ—á–µ–∫ –ª–∏—Ñ—Ç–∞.  
> –ù–æ –Ω–∞ –≥—Ä–∞—Ñ–∏–∫–µ **–ø–æ–≤–µ—Ä—Ö–Ω–æ—Å—Ç–∏ $\tau = \text{const}$** (–∫—Ä–∞—Å–Ω—ã–µ —à—Ç—Ä–∏—Ö–æ–≤—ã–µ –ª–∏–Ω–∏–∏) ‚Äî **–Ω–µ –≤–µ—Ä—Ç–∏–∫–∞–ª—å–Ω—ã**, –∞ –Ω–∞–∫–ª–æ–Ω–µ–Ω—ã.  
> ‚ùì –ü–æ—á–µ–º—É?

---

## ‚úÖ –ö—Ä–∞—Ç–∫–∏–π –æ—Ç–≤–µ—Ç:

> **–ü–æ—Ç–æ–º—É —á—Ç–æ $\tau = \text{const}$ ‚Äî —ç—Ç–æ –Ω–µ "–º–æ–º–µ–Ω—Ç –≤—Ä–µ–º–µ–Ω–∏", –ø–∞—Ä–∞–ª–ª–µ–ª—å–Ω—ã–π –æ—Å–∏ $t$", –∞ ‚Äî "–ø–æ–≤–µ—Ä—Ö–Ω–æ—Å—Ç—å –æ–¥–Ω–æ–≤—Ä–µ–º–µ–Ω–Ω–æ—Å—Ç–∏ –≤ —É—Å–∫–æ—Ä–µ–Ω–Ω–æ–π —Å–∏—Å—Ç–µ–º–µ", –∫–æ—Ç–æ—Ä–∞—è –≤ –ª–∞–±–æ—Ä–∞—Ç–æ—Ä–Ω—ã—Ö –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–∞—Ö $(t, x)$ –≤—ã–≥–ª—è–¥–∏—Ç –∫–∞–∫ –Ω–∞–∫–ª–æ–Ω–Ω–∞—è (–∏–ª–∏ –∫—Ä–∏–≤–∞—è) –ª–∏–Ω–∏—è.**

üìå **–ö–ª—é—á–µ–≤–æ–µ**:  
> **–ö–æ–æ—Ä–¥–∏–Ω–∞—Ç–∞ $\tau$ ‚Äî —ç—Ç–æ –Ω–µ –Ω–∞–ø—Ä–∞–≤–ª–µ–Ω–∏–µ –Ω–∞ –≥—Ä–∞—Ñ–∏–∫–µ, –∞ ‚Äî –º–µ—Ç–∫–∞ –¥–ª—è —Ü–µ–ª–æ–π –∫—Ä–∏–≤–æ–π (–ø–æ–≤–µ—Ä—Ö–Ω–æ—Å—Ç–∏).**

---

## üî∑ –ü–æ–¥—Ä–æ–±–Ω–æ–µ –æ–±—ä—è—Å–Ω–µ–Ω–∏–µ

### 1. –ß—Ç–æ —Ç–∞–∫–æ–µ $\tau$ –≤ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–∞—Ö –†–∏–Ω–¥–ª–µ—Ä–∞?

$\tau$ ‚Äî —ç—Ç–æ **–∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–Ω–æ–µ –≤—Ä–µ–º—è** –≤ —Å–∏—Å—Ç–µ–º–µ —É—Å–∫–æ—Ä–µ–Ω–Ω–æ–≥–æ –Ω–∞–±–ª—é–¥–∞—Ç–µ–ª—è.  
–ù–æ **–æ–Ω–æ –Ω–µ —Å–æ–≤–ø–∞–¥–∞–µ—Ç** —Å –ª–∞–±–æ—Ä–∞—Ç–æ—Ä–Ω—ã–º –≤—Ä–µ–º–µ–Ω–µ–º $t$.

–í –†–∏–Ω–¥–ª–µ—Ä–æ–≤—Å–∫–∏—Ö –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–∞—Ö:
- **–ü–æ–≤–µ—Ä—Ö–Ω–æ—Å—Ç—å $\tau = \text{const}$** ‚Äî —ç—Ç–æ **–º–Ω–æ–∂–µ—Å—Ç–≤–æ —Å–æ–±—ã—Ç–∏–π, –∫–æ—Ç–æ—Ä—ã–µ —É—Å–∫–æ—Ä–µ–Ω–Ω—ã–π –Ω–∞–±–ª—é–¥–∞—Ç–µ–ª—å —Å—á–∏—Ç–∞–µ—Ç "–æ–¥–Ω–æ–≤—Ä–µ–º–µ–Ω–Ω—ã–º–∏"**.
- –ù–æ –≤ **–ª–∞–±–æ—Ä–∞—Ç–æ—Ä–Ω–æ–π —Å–∏—Å—Ç–µ–º–µ –æ—Ç—Å—á—ë—Ç–∞** (–≥–¥–µ –æ—Å–∏ $t$ –∏ $x$) —ç—Ç–∞ –ø–æ–≤–µ—Ä—Ö–Ω–æ—Å—Ç—å **–Ω–µ –≥–æ—Ä–∏–∑–æ–Ω—Ç–∞–ª—å–Ω–∞** ‚Äî –æ–Ω–∞ **–Ω–∞–∫–ª–æ–Ω–µ–Ω–∞**.

–ü–æ—á–µ–º—É?

---

### 2. –ü—Ä–∏—á–∏–Ω–∞: **–æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ—Å—Ç—å –æ–¥–Ω–æ–≤—Ä–µ–º–µ–Ω–Ω–æ—Å—Ç–∏ + —É—Å–∫–æ—Ä–µ–Ω–∏–µ**

–í —Å–ø–µ—Ü–∏–∞–ª—å–Ω–æ–π —Ç–µ–æ—Ä–∏–∏ –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ—Å—Ç–∏:
- –î–≤–∞ —Å–æ–±—ã—Ç–∏—è, –æ–¥–Ω–æ–≤—Ä–µ–º–µ–Ω–Ω—ã–µ –≤ –æ–¥–Ω–æ–π –ò–°–û, **–Ω–µ –æ–¥–Ω–æ–≤—Ä–µ–º–µ–Ω–Ω—ã** –≤ –¥—Ä—É–≥–æ–π.
- –í **—É—Å–∫–æ—Ä–µ–Ω–Ω–æ–π —Å–∏—Å—Ç–µ–º–µ** —ç—Ç–æ –µ—â—ë —Å–∏–ª—å–Ω–µ–µ: –∏–∑-–∑–∞ —É—Å–∫–æ—Ä–µ–Ω–∏—è **–ø–æ–≤–µ—Ä—Ö–Ω–æ—Å—Ç—å –æ–¥–Ω–æ–≤—Ä–µ–º–µ–Ω–Ω–æ—Å—Ç–∏ –ø–æ–≤–æ—Ä–∞—á–∏–≤–∞–µ—Ç—Å—è**.

–§–æ—Ä–º—É–ª–∞:
$$
t = (1 + \xi) \sinh \tau, \quad x = (1 + \xi) \cosh \tau
$$

–ï—Å–ª–∏ –∑–∞—Ñ–∏–∫—Å–∏—Ä–æ–≤–∞—Ç—å $\tau = \tau_0 = \text{const}$, —Ç–æ:
$$
\frac{t}{x} = \tanh \tau_0 = \text{const}
\quad \Rightarrow \quad
t = (\tanh \tau_0) \cdot x
$$

‚úÖ –≠—Ç–æ ‚Äî **–ø—Ä—è–º–∞—è, –ø—Ä–æ—Ö–æ–¥—è—â–∞—è —á–µ—Ä–µ–∑ –Ω–∞—á–∞–ª–æ**, —Å –Ω–∞–∫–ª–æ–Ω–æ–º $\tanh \tau_0$.

üìå –¢–æ –µ—Å—Ç—å:  
> **$\tau = \text{const}$ ‚Üí –ø—Ä—è–º–∞—è –ª–∏–Ω–∏—è $t \propto x$**, –∞ **–Ω–µ** –≤–µ—Ä—Ç–∏–∫–∞–ª—å–Ω–∞—è –∏–ª–∏ –≥–æ—Ä–∏–∑–æ–Ω—Ç–∞–ª—å–Ω–∞—è.

### 4. –ö–∞–∫ —Ç–æ–≥–¥–∞ –≤—ã–≥–ª—è–¥–∏—Ç –∏–Ω—Ç–µ–≥—Ä–∏—Ä–æ–≤–∞–Ω–∏–µ?

–ú—ã –∏–Ω—Ç–µ–≥—Ä–∏—Ä—É–µ–º –ø–æ:
- $\tau$ –æ—Ç $\tau_1$ –¥–æ $\tau_2$ ‚Äî **–æ–¥–∏–Ω–∞–∫–æ–≤–æ –¥–ª—è –≤—Å–µ—Ö $\xi$**,
- $\xi$ –æ—Ç $\xi_{\min}$ –¥–æ $\xi_{\max}$ ‚Äî –æ–¥–∏–Ω–∞–∫–æ–≤–æ –¥–ª—è –≤—Å–µ—Ö $\tau$.

–ù–∞ –¥–∏–∞–≥—Ä–∞–º–º–µ —ç—Ç–æ:

- –ò –º–µ–∂–¥—É **–¥–≤—É–º—è –º–∏—Ä–æ–≤—ã–º–∏ –ª–∏–Ω–∏—è–º–∏ $\xi = \xi_{\min}$ –∏ $\xi = \xi_{\max}$**.

üìå –≠—Ç–æ –∫–∞–∫ "—Ü–∏–ª–∏–Ω–¥—Ä" –≤ 4D, —Å "–æ—Å–Ω–æ–≤–∞–Ω–∏—è–º–∏" –Ω–∞ $\tau = \text{const}$ –∏ "–±–æ–∫–∞–º–∏" –Ω–∞ $\xi = \text{const}$.

## ‚úÖ –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è: –ø–æ—á–µ–º—É –∫—Ä–∞—Å–Ω—ã–µ –ª–∏–Ω–∏–∏ –Ω–µ –≤–µ—Ä—Ç–∏–∫–∞–ª—å–Ω—ã

> üö´ –û—à–∏–±–∫–∞: –¥—É–º–∞—Ç—å, —á—Ç–æ "–æ–¥–∏–Ω–∞–∫–æ–≤–æ–µ –≤—Ä–µ–º—è = –≤–µ—Ä—Ç–∏–∫–∞–ª—å–Ω–∞—è –ª–∏–Ω–∏—è".

–ü—É—Å—Ç—å $\tau_2 = 1 \Rightarrow \tanh 1 \approx 0.76 \Rightarrow t = 0.76 x$

````python
import numpy as np
import matplotlib.pyplot as plt

# –ü–∞—Ä–∞–º–µ—Ç—Ä—ã
c = 1
a = 1

# -----------------------------------------------------------------------------
# 1. –î–∏–∞–ø–∞–∑–æ–Ω—ã
t_vals = np.linspace(-2.5, 2.5, 500)
x_vals = np.linspace(0.1, 3.0, 500)

# -----------------------------------------------------------------------------
# 2. –°–µ—Ç–∫–∞ –†–∏–Ω–¥–ª–µ—Ä–∞

plt.figure(figsize=(10, 8))

# –ü—Ä–µ–¥–µ–ª—ã –≥—Ä–∞—Ñ–∏–∫–∞
plt.xlim(-1.6, 1.6)
plt.ylim(0.9, 2.3)

# -----------------------------------------------------------------------------
# 3. –õ–∏–Ω–∏–∏ Œæ = const (–º–∏—Ä–æ–≤—ã–µ –ª–∏–Ω–∏–∏)
xi_values = np.linspace(-0.5, 1.0, 7)  # –æ—Ç –¥–Ω–∞ –¥–æ –≤–µ—Ä—Ö–∞
for xi in xi_values:
    ell = 1 + xi
    x_world = np.sqrt(t_vals**2 + ell**2)
    valid = x_world >= 0.9  # –æ–≥—Ä–∞–Ω–∏—á–µ–Ω–∏–µ –ø–æ y
    plt.plot(t_vals[valid], x_world[valid], color='blue', lw=1.2, alpha=0.8)

# –ü–æ–¥–ø–∏—Å—å –æ–¥–Ω–æ–π –ª–∏–Ω–∏–∏ (–∏—Å–ø–æ–ª—å–∑—É–µ–º \mathrm –≤–º–µ—Å—Ç–æ \text)
plt.plot([], [], color='blue', lw=1.5, label=r'$\xi = \mathrm{const}$ (–º–∏—Ä–æ–≤—ã–µ –ª–∏–Ω–∏–∏)')

# -----------------------------------------------------------------------------
# 4. –õ–∏–Ω–∏–∏ œÑ = const (–ø–æ–≤–µ—Ä—Ö–Ω–æ—Å—Ç–∏ –æ–¥–Ω–æ–≤—Ä–µ–º–µ–Ω–Ω–æ—Å—Ç–∏)
tau_values = np.linspace(-1.0, 1.2, 7)
for tau in tau_values:
    tanh_tau = np.tanh(tau)
    x_line = np.linspace(0.1, 2.5, 200)
    t_line = tanh_tau * x_line
    plt.plot(t_line, x_line, color='red', linestyle='-', lw=1.0, alpha=0.7)

# –ê–Ω–∞–ª–æ–≥–∏—á–Ω–æ
plt.plot([], [], color='red', lw=1.5, label=r'$\tau = \mathrm{const}$ (–æ–¥–Ω–æ–≤—Ä–µ–º–µ–Ω–Ω–æ—Å—Ç—å)')

# -----------------------------------------------------------------------------
# 5. –û–±–ª–∞—Å—Ç—å –∏–Ω—Ç–µ–≥—Ä–∏—Ä–æ–≤–∞–Ω–∏—è –¥–µ–π—Å—Ç–≤–∏—è
xi_min = -0.3
xi_max = 0.7
tau_1 = 0.0
tau_2 = 1.0

# –í—Å–ø–æ–º–æ–≥–∞—Ç–µ–ª—å–Ω—ã–µ —Ñ—É–Ω–∫—Ü–∏–∏
def x_from_tau_xi(tau, xi):
    return (1 + xi) * np.cosh(tau)

def t_from_tau_xi(tau, xi):
    return (1 + xi) * np.sinh(tau)

# –í–¥–æ–ª—å œÑ = œÑ‚ÇÅ: –æ—Ç Œæ_min –¥–æ Œæ_max
xi_fill = np.linspace(xi_min, xi_max, 50)
t_tau1 = t_from_tau_xi(tau_1, xi_fill)
x_tau1 = x_from_tau_xi(tau_1, xi_fill)

# –í–¥–æ–ª—å Œæ = Œæ_max: –æ—Ç œÑ‚ÇÅ –¥–æ œÑ‚ÇÇ
tau_fill = np.linspace(tau_1, tau_2, 50)
t_xi_max = t_from_tau_xi(tau_fill, xi_max)
x_xi_max = x_from_tau_xi(tau_fill, xi_max)

# –í–¥–æ–ª—å œÑ = œÑ‚ÇÇ: –æ—Ç Œæ_max –¥–æ Œæ_min
xi_fill_rev = np.linspace(xi_max, xi_min, 50)
t_tau2 = t_from_tau_xi(tau_2, xi_fill_rev)
x_tau2 = x_from_tau_xi(tau_2, xi_fill_rev)

# –í–¥–æ–ª—å Œæ = Œæ_min: –æ—Ç œÑ‚ÇÇ –¥–æ œÑ‚ÇÅ
tau_fill_rev = np.linspace(tau_2, tau_1, 50)
t_xi_min = t_from_tau_xi(tau_fill_rev, xi_min)
x_xi_min = x_from_tau_xi(tau_fill_rev, xi_min)

# –°–æ–±–∏—Ä–∞–µ–º –ø–æ–ª–∏–≥–æ–Ω
t_poly = np.concatenate([t_tau1, t_xi_max, t_tau2, t_xi_min])
x_poly = np.concatenate([x_tau1, x_xi_max, x_tau2, x_xi_min])

plt.fill(t_poly, x_poly, color='gold', alpha=0.6, edgecolor='orange', lw=1.5, label=r'–û–±–ª–∞—Å—Ç—å –∏–Ω—Ç–µ–≥—Ä–∏—Ä–æ–≤–∞–Ω–∏—è $S$')

# -----------------------------------------------------------------------------
# 6. –°–≤–µ—Ç–æ–≤—ã–µ –ª—É—á–∏
x_light = np.linspace(0, 3, 100)
plt.plot( x_light,  x_light, 'k-', lw=1, alpha=0.4, label=r'–°–≤–µ—Ç–æ–≤–æ–π –ª—É—á ($x = t$)')
plt.plot(-x_light,  x_light, 'k-', lw=1, alpha=0.4)

# -----------------------------------------------------------------------------
# 7. –û—Ñ–æ—Ä–º–ª–µ–Ω–∏–µ
plt.xlabel(r"–ö–æ–æ—Ä–¥–∏–Ω–∞—Ç–Ω–æ–µ –≤—Ä–µ–º—è $t$ (–≤ –µ–¥–∏–Ω–∏—Ü–∞—Ö $c/a$)", fontsize=12)
plt.ylabel(r"–ü—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–æ $x$ (–≤ –µ–¥–∏–Ω–∏—Ü–∞—Ö $c^2/a$)", fontsize=12)
plt.title(r"–°–µ—Ç–∫–∞ –†–∏–Ω–¥–ª–µ—Ä–∞: $\tau, \xi$ –∏ –æ–±–ª–∞—Å—Ç—å –∏–Ω—Ç–µ–≥—Ä–∏—Ä–æ–≤–∞–Ω–∏—è –¥–µ–π—Å—Ç–≤–∏—è", fontsize=13)
plt.legend(loc='upper left', fontsize=10)
plt.grid(True, alpha=0.3)
plt.tight_layout()

# –ü–æ–∫–∞–∑–∞—Ç—å
plt.show()
````

### üîπ 1. –ß—Ç–æ —Ç–∞–∫–æ–µ "—Å–æ–±—ã—Ç–∏–µ"?

–ü—Ä–∏–º–µ—Ä—ã:
- –í—Å–ø—ã—à–∫–∞ —Å–≤–µ—Ç–∞ –≤ —Ç–æ—á–∫–µ $x = 2\,\text{–º}$, –≤ –º–æ–º–µ–Ω—Ç $t = 5\,\text{—Å}$,

### üîπ 2. –ß—Ç–æ —Ç–∞–∫–æ–µ $\tau$ –≤ —ç—Ç–æ–º –∫–æ–Ω—Ç–µ–∫—Å—Ç–µ?

$\tau$ ‚Äî —ç—Ç–æ **–∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–Ω–æ–µ –≤—Ä–µ–º—è –≤ —É—Å–∫–æ—Ä–µ–Ω–Ω–æ–π —Å–∏—Å—Ç–µ–º–µ –æ—Ç—Å—á—ë—Ç–∞**, –Ω–∞–ø—Ä–∏–º–µ—Ä, –≤ **–ª–∏—Ñ—Ç–µ, –¥–≤–∏–∂—É—â–µ–º—Å—è —Å —É—Å–∫–æ—Ä–µ–Ω–∏–µ–º**.

- –≠—Ç–æ **–Ω–µ —Å–æ–±—Å—Ç–≤–µ–Ω–Ω–æ–µ –≤—Ä–µ–º—è –æ–¥–Ω–æ–π —á–∞—Å—Ç–∏—Ü—ã**, –∞ ‚Äî **—Å–∏—Å—Ç–µ–º–∞ –æ—Ç—Å—á—ë—Ç–∞**, –ø—Ä–∏–≤—è–∑–∞–Ω–Ω–∞—è –∫–æ –≤—Å–µ–º—É —É—Å–∫–æ—Ä–µ–Ω–Ω–æ–º—É —Ç–µ–ª—É.

### üîπ 3. –ß—Ç–æ –∑–Ω–∞—á–∏—Ç "–ø–æ–≤–µ—Ä—Ö–Ω–æ—Å—Ç—å $\tau = \text{const}$"?

- –ï—Å–ª–∏ –≤—ã –∑–∞—Ñ–∏–∫—Å–∏—Ä—É–µ—Ç–µ $\tau = \tau_0$ (–Ω–∞–ø—Ä–∏–º–µ—Ä, $\tau_0 = 1$), —Ç–æ –≤—ã –ø–æ–ª—É—á–∏—Ç–µ **–º–Ω–æ–∂–µ—Å—Ç–≤–æ –≤—Å–µ—Ö —Å–æ–±—ã—Ç–∏–π**, –∫–æ—Ç–æ—Ä—ã–µ –∏–º–µ—é—Ç —ç—Ç–æ –∑–Ω–∞—á–µ–Ω–∏–µ $\tau$.
- –≠—Ç–æ –Ω–µ –æ–¥–Ω–∞ —Ç–æ—á–∫–∞, –∞ **—Ü–µ–ª–∞—è "–ø–æ–≤–µ—Ä—Ö–Ω–æ—Å—Ç—å"** (–≤ 4D ‚Äî 3D-–ø–æ–≤–µ—Ä—Ö–Ω–æ—Å—Ç—å, –≤ 2D-–¥–∏–∞–≥—Ä–∞–º–º–µ ‚Äî –ª–∏–Ω–∏—è).
- –≠—Ç–∞ –ø–æ–≤–µ—Ä—Ö–Ω–æ—Å—Ç—å –ø—Ä–æ—Ö–æ–¥–∏—Ç —á–µ—Ä–µ–∑ **—Ä–∞–∑–Ω—ã–µ —Ç–æ—á–∫–∏ –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–∞**, –Ω–æ **–≤ –æ–¥–∏–Ω –∏ —Ç–æ—Ç –∂–µ –º–æ–º–µ–Ω—Ç "–ø–æ —á–∞—Å–∞–º —É—Å–∫–æ—Ä–µ–Ω–Ω–æ–≥–æ –Ω–∞–±–ª—é–¥–∞—Ç–µ–ª—è"**.

### üîπ 4. –ß—Ç–æ –∑–Ω–∞—á–∏—Ç "—Å—á–∏—Ç–∞–µ—Ç –æ–¥–Ω–æ–≤—Ä–µ–º–µ–Ω–Ω—ã–º–∏"?

–í —Å–ø–µ—Ü–∏–∞–ª—å–Ω–æ–π —Ç–µ–æ—Ä–∏–∏ –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ—Å—Ç–∏ **–æ–¥–Ω–æ–≤—Ä–µ–º–µ–Ω–Ω–æ—Å—Ç—å ‚Äî –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–∞**:
- –¢–æ, —á—Ç–æ –æ–¥–Ω–æ–≤—Ä–µ–º–µ–Ω–Ω–æ –¥–ª—è –æ–¥–Ω–æ–≥–æ –Ω–∞–±–ª—é–¥–∞—Ç–µ–ª—è, **–Ω–µ –æ–¥–Ω–æ–≤—Ä–µ–º–µ–Ω–Ω–æ** –¥–ª—è –¥—Ä—É–≥–æ–≥–æ.

–î–ª—è **–∏–Ω–µ—Ä—Ü–∏–∞–ª—å–Ω–æ–≥–æ –Ω–∞–±–ª—é–¥–∞—Ç–µ–ª—è** (—Ä–∞–≤–Ω–æ–º–µ—Ä–Ω–æ –¥–≤–∏–∂—É—â–µ–≥–æ—Å—è) –ø–æ–≤–µ—Ä—Ö–Ω–æ—Å—Ç–∏ –æ–¥–Ω–æ–≤—Ä–µ–º–µ–Ω–Ω–æ—Å—Ç–∏ ‚Äî —ç—Ç–æ **–≥–æ—Ä–∏–∑–æ–Ω—Ç–∞–ª—å–Ω—ã–µ –ª–∏–Ω–∏–∏** –Ω–∞ –¥–∏–∞–≥—Ä–∞–º–º–µ –ú–∏–Ω–∫–æ–≤—Å–∫–æ–≥–æ: $t = \text{const}$.

–ù–æ –¥–ª—è **—É—Å–∫–æ—Ä–µ–Ω–Ω–æ–≥–æ –Ω–∞–±–ª—é–¥–∞—Ç–µ–ª—è**:
- –ï–≥–æ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ "–æ–¥–Ω–æ–≤—Ä–µ–º–µ–Ω–Ω–æ" **–º–µ–Ω—è–µ—Ç—Å—è —Å–æ –≤—Ä–µ–º–µ–Ω–µ–º**,
- –û–Ω –∏—Å–ø–æ–ª—å–∑—É–µ—Ç **—Å–≤–æ—é —Å–∏—Å—Ç–µ–º—É –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç** (–†–∏–Ω–¥–ª–µ—Ä–∞),
- –ò –≤ –Ω–µ–π **–æ–¥–Ω–æ–≤—Ä–µ–º–µ–Ω–Ω—ã–µ —Å–æ–±—ã—Ç–∏—è** ‚Äî —ç—Ç–æ —Ç–µ, —É –∫–æ—Ç–æ—Ä—ã—Ö **–æ–¥–∏–Ω–∞–∫–æ–≤–æ–µ $\tau$**.

### üîπ 5. –ü—Ä–æ—Å—Ç–æ–π –ø—Ä–∏–º–µ—Ä

–ü—Ä–µ–¥—Å—Ç–∞–≤—å—Ç–µ, —á—Ç–æ –≤—ã ‚Äî **–Ω–∞–±–ª—é–¥–∞—Ç–µ–ª—å –≤ —É—Å–∫–æ—Ä–µ–Ω–Ω–æ–º –∫–æ—Å–º–∏—á–µ—Å–∫–æ–º –∫–æ—Ä–∞–±–ª–µ** (–ª–∏—Ñ—Ç–µ), –∏ —É –≤–∞—Å –µ—Å—Ç—å **–º–Ω–æ–≥–æ —á–∞—Å–æ–≤**, —Ä–∞—Å—Å—Ç–∞–≤–ª–µ–Ω–Ω—ã—Ö –≤–¥–æ–ª—å –∫–æ—Ä–∞–±–ª—è: –≤ –Ω–æ—Å—É, –≤ —Ü–µ–Ω—Ç—Ä–µ, –≤ —Ö–≤–æ—Å—Ç–µ.

–í—ã —Ö–æ—Ç–∏—Ç–µ **—Å–∏–Ω—Ö—Ä–æ–Ω–∏–∑–∏—Ä–æ–≤–∞—Ç—å** –∏—Ö –ø–æ —Å–≤–æ–µ–π —Å–∏—Å—Ç–µ–º–µ.

- –í –º–æ–º–µ–Ω—Ç, –∫–æ—Ç–æ—Ä—ã–π –≤—ã –Ω–∞–∑—ã–≤–∞–µ—Ç–µ $\tau = 0$, –≤—ã –ø–æ—Å—ã–ª–∞–µ—Ç–µ **—Å–∏–≥–Ω–∞–ª –∏–∑ —Ü–µ–Ω—Ç—Ä–∞** –∏ –≥–æ–≤–æ—Ä–∏—Ç–µ:  
  ¬´–í—Å–µ —Å–æ–±—ã—Ç–∏—è, –∫–æ—Ç–æ—Ä—ã–µ –ø—Ä–æ–∏—Å—Ö–æ–¥—è—Ç **–≤ —ç—Ç–æ—Ç –º–æ–º–µ–Ω—Ç –ø–æ –º–æ–µ–π —à–∫–∞–ª–µ**, ‚Äî –æ–¥–Ω–æ–≤—Ä–µ–º–µ–Ω–Ω—ã¬ª.
- –ò–∑-–∑–∞ —É—Å–∫–æ—Ä–µ–Ω–∏—è –∏ —Ä–µ–ª—è—Ç–∏–≤–∏—Å—Ç—Å–∫–∏—Ö —ç—Ñ—Ñ–µ–∫—Ç–æ–≤:
  - –°–æ–±—ã—Ç–∏–µ –≤ **—Ö–≤–æ—Å—Ç–µ** –∫–æ—Ä–∞–±–ª—è –¥–æ–ª–∂–Ω–æ –ø—Ä–æ–∏–∑–æ–π—Ç–∏ **—Ä–∞–Ω—å—à–µ** –ø–æ –ª–∞–±–æ—Ä–∞—Ç–æ—Ä–Ω–æ–º—É –≤—Ä–µ–º–µ–Ω–∏ $t$,
  - –°–æ–±—ã—Ç–∏–µ –≤ **–Ω–æ—Å—É** ‚Äî **–ø–æ–∑–∂–µ**,
- –ù–æ –¥–ª—è –≤–∞—Å ‚Äî –æ–Ω–∏ **–æ–¥–Ω–æ–≤—Ä–µ–º–µ–Ω–Ω—ã**, –ø–æ—Ç–æ–º—É —á—Ç–æ $\tau$ –≤–µ–∑–¥–µ –æ–¥–∏–Ω–∞–∫–æ–≤–æ.

üìå –≠—Ç–∞ —Å–æ–≤–æ–∫—É–ø–Ω–æ—Å—Ç—å —Å–æ–±—ã—Ç–∏–π ‚Äî –∏ –µ—Å—Ç—å **–ø–æ–≤–µ—Ä—Ö–Ω–æ—Å—Ç—å $\tau = \text{const}$**.

### üîπ 6. –ö–∞–∫ —ç—Ç–æ –≤—ã–≥–ª—è–¥–∏—Ç –Ω–∞ –¥–∏–∞–≥—Ä–∞–º–º–µ?

–ù–∞ –¥–∏–∞–≥—Ä–∞–º–º–µ –ú–∏–Ω–∫–æ–≤—Å–∫–æ–≥–æ (–æ—Å—å $t$, –æ—Å—å $x$):
- –ü–æ–≤–µ—Ä—Ö–Ω–æ—Å—Ç—å $\tau = \text{const}$ ‚Äî —ç—Ç–æ **–ø—Ä—è–º–∞—è –ª–∏–Ω–∏—è, –ø—Ä–æ—Ö–æ–¥—è—â–∞—è —á–µ—Ä–µ–∑ –Ω–∞—á–∞–ª–æ**, —Å –Ω–∞–∫–ª–æ–Ω–æ–º $t = x \cdot \tanh \tau$.
- –û–Ω–∞ **–Ω–µ –≥–æ—Ä–∏–∑–æ–Ω—Ç–∞–ª—å–Ω–∞** ‚Äî –∑–Ω–∞—á–∏—Ç, —Å–æ–±—ã—Ç–∏—è –≤ —Ä–∞–∑–Ω—ã—Ö —Ç–æ—á–∫–∞—Ö –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–∞ –∏–º–µ—é—Ç **—Ä–∞–∑–Ω–æ–µ $t$**, –Ω–æ **–æ–¥–∏–Ω–∞–∫–æ–≤–æ–µ $\tau$**.

```python
# –î–∏–∞–ø–∞–∑–æ–Ω—ã

def x_world(tau, xi):
    return (1 + xi) * np.cosh(tau)

def t_world(tau, xi):
    return (1 + xi) * np.sinh(tau)

# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –≤—ã—á–∏—Å–ª–µ–Ω–∏—è –Ω–æ—Ä–º–∞–ª–∏
def normal_vector(tau, xi):
    dx_dtau = (1 + xi) * np.sinh(tau)
    dt_dtau = (1 + xi) * np.cosh(tau)
    return np.array([-dt_dtau, dx_dtau])

# –ì—Ä–∞—Ñ–∏–∫

# –ú–∏—Ä–æ–≤—ã–µ –ª–∏–Ω–∏–∏

    # –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –ø–æ–≤–µ—Ä—Ö–Ω–æ—Å—Ç–∏ –æ–¥–Ω–æ–≤—Ä–µ–º–µ–Ω–Ω–æ—Å—Ç–∏
    for i in range(len(points) - 1):
        p1 = points[i]
        p2 = points[i + 1]
        n1 = normals[i]
        n2 = normals[i + 1]
        
        # –õ–∏–Ω–µ–π–Ω–∞—è –∏–Ω—Ç–µ—Ä–ø–æ–ª—è—Ü–∏—è –º–µ–∂–¥—É —Ç–æ—á–∫–∞–º–∏
        t_interp = np.linspace(p1[0], p2[0], 100)
        x_interp = np.linspace(p1[1], p2[1], 100)
        
        # –ù–æ—Ä–º–∞–ª–∏–∑–∞—Ü–∏—è –Ω–æ—Ä–º–∞–ª–µ–π

        # –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –ø–æ–≤–µ—Ä—Ö–Ω–æ—Å—Ç–∏ –æ–¥–Ω–æ–≤—Ä–µ–º–µ–Ω–Ω–æ—Å—Ç–∏
        for t, x in zip(t_interp, x_interp):
            plt.plot([t, t + 0.1 * n1_norm[0]], [x, x + 0.1 * n1_norm[1]], 'r--', alpha=0.7)
            plt.plot([t, t + 0.1 * n2_norm[0]], [x, x + 0.1 * n2_norm[1]], 'r--', alpha=0.7)

# –û—Ñ–æ—Ä–º–ª–µ–Ω–∏–µ
plt.xlabel("–ö–æ–æ—Ä–¥–∏–Ω–∞—Ç–Ω–æ–µ –≤—Ä–µ–º—è $t$ (–≤ –µ–¥–∏–Ω–∏—Ü–∞—Ö $c/a$)", fontsize=12)

plt.title("–ü—Ä–∞–≤–∏–ª—å–Ω—ã–µ –ø–æ–≤–µ—Ä—Ö–Ω–æ—Å—Ç–∏ –æ–¥–Ω–æ–≤—Ä–µ–º–µ–Ω–Ω–æ—Å—Ç–∏ –¥–ª—è –≤–∞—Ä–∏–∞—Ü–∏–∏ –í", fontsize=14)
plt.grid(True, alpha=0.3)
plt.legend()
plt.tight_layout()

# –ü–æ–∫–∞–∑–∞—Ç—å
plt.show()
```

```python
    return np.array([-dt_dtau, dx_dtau])

plt.grid(True, alpha=0.3)
plt.legend()
plt.tight_layout()

# –ü–æ–∫–∞–∑–∞—Ç—å
```

## ‚úÖ –ú–∞—Ç–µ–º–∞—Ç–∏–∫–∞: –∫–∞–∫ –ø–æ—Å—Ç—Ä–æ–∏—Ç—å –æ—Ä—Ç–æ–≥–æ–Ω–∞–ª—å–Ω—ã–µ —Å–µ—á–µ–Ω–∏—è

$$

$$

–ö–∞—Å–∞—Ç–µ–ª—å–Ω—ã–π –≤–µ–∫—Ç–æ—Ä:
$$
\vec{v} = \left( \frac{dt}{d\tau}, \frac{dx}{d\tau} \right) = (1 + \xi) (\sinh \tau, \cosh \tau)
$$

–ù–æ—Ä–º–∞–ª—å–Ω—ã–π –≤–µ–∫—Ç–æ—Ä (–≤ 2D):
$$
\vec{n} = (-\cosh \tau, \sinh \tau)
\quad \text{(—Å —Ç–æ—á–Ω–æ—Å—Ç—å—é –¥–æ –∑–Ω–∞–∫–∞ –∏ –Ω–æ—Ä–º–∏—Ä–æ–≤–∫–∏)}

‚úÖ **–ü–æ–≤–µ—Ä—Ö–Ω–æ—Å—Ç—å –æ–¥–Ω–æ–≤—Ä–µ–º–µ–Ω–Ω–æ—Å—Ç–∏** –≤ —Ç–æ—á–∫–µ $(t, x)$ ‚Äî —ç—Ç–æ –ø—Ä—è–º–∞—è, –ø—Ä–æ—Ö–æ–¥—è—â–∞—è —á–µ—Ä–µ–∑ –Ω–µ—ë —Å –Ω–∞–ø—Ä–∞–≤–ª–µ–Ω–∏–µ–º $\vec{n}$.

–ù–æ: **–≤—Å–µ —Ç–æ—á–∫–∏ —Å –æ–¥–∏–Ω–∞–∫–æ–≤—ã–º $\tau$** –ª–µ–∂–∞—Ç –Ω–∞ –æ–¥–Ω–æ–π —Ç–∞–∫–æ–π –ø–æ–≤–µ—Ä—Ö–Ω–æ—Å—Ç–∏ ‚Äî —ç—Ç–æ –∏ –µ—Å—Ç—å $\tau = \text{const}$ –≤ **–≥–µ–æ–º–µ—Ç—Ä–∏—á–µ—Å–∫–æ–º —Å–º—ã—Å–ª–µ –§–µ—Ä–º–∏**.

## ‚úÖ –ò—Å–ø—Ä–∞–≤–ª–µ–Ω–Ω—ã–π –∏ —Ç–æ—á–Ω—ã–π Python-–∫–æ–¥

```python
import numpy as np
import matplotlib.pyplot as plt

# –ü–∞—Ä–∞–º–µ—Ç—Ä—ã
c = 1
a = 1  # —É—Å–∫–æ—Ä–µ–Ω–∏–µ (–≤ –µ–¥–∏–Ω–∏—Ü–∞—Ö c¬≤/–º–µ—Ç—Ä)

# –î–∏–∞–ø–∞–∑–æ–Ω—ã
tau_vals = np.linspace(-2.0, 2.0, 500)
xi_vals = np.linspace(-0.4, 0.6, 5)  # –º–∏—Ä–æ–≤—ã–µ –ª–∏–Ω–∏–∏

# –§—É–Ω–∫—Ü–∏–∏ –º–∏—Ä–æ–≤—ã—Ö –ª–∏–Ω–∏–π
def x_world(tau, xi):
    return (1 + xi) * np.cosh(tau)

def t_world(tau, xi):
    return (1 + xi) * np.sinh(tau)

# -----------------------------------------------------------------------------
# 1. –†–∏—Å—É–µ–º –¥–∏–∞–≥—Ä–∞–º–º—É
plt.figure(figsize=(10, 8))

# -----------------------------------------------------------------------------
# 2. –ú–∏—Ä–æ–≤—ã–µ –ª–∏–Ω–∏–∏ Œæ = const
for xi in xi_vals:
    x = x_world(tau_vals, xi)
    t = t_world(tau_vals, xi)
    valid = (x >= 0.8) & (x <= 2.5)
    plt.plot(t[valid], x[valid], color='blue', lw=1.8, label=f'$\\xi = {xi}$' if xi == xi_vals[0] else "")

# -----------------------------------------------------------------------------
# 3. –ü–æ–≤–µ—Ä—Ö–Ω–æ—Å—Ç–∏ –æ–¥–Ω–æ–≤—Ä–µ–º–µ–Ω–Ω–æ—Å—Ç–∏ œÑ = const (–æ—Ä—Ç–æ–≥–æ–Ω–∞–ª—å–Ω—ã–µ –º–∏—Ä–æ–≤—ã–º –ª–∏–Ω–∏—è–º)
tau_const = np.linspace(-0.8, 0.8, 5)  # –∑–Ω–∞—á–µ–Ω–∏—è œÑ –¥–ª—è —Å–µ—á–µ–Ω–∏–π

for tau in tau_const:
    # –°–æ–±–∏—Ä–∞–µ–º —Ç–æ—á–∫–∏ –Ω–∞ –≤—Å–µ—Ö –º–∏—Ä–æ–≤—ã—Ö –ª–∏–Ω–∏—è—Ö –ø—Ä–∏ —ç—Ç–æ–º œÑ
    points_t = []
    points_x = []
    normals = []

    for xi in xi_vals:
        t_point = t_world(tau, xi)
        x_point = x_world(tau, xi)
        # –ù–æ—Ä–º–∞–ª—å–Ω—ã–π –≤–µ–∫—Ç–æ—Ä: (-cosh œÑ, sinh œÑ) ‚Üí –Ω–∞–ø—Ä–∞–≤–ª–µ–Ω–∏–µ –ø–æ–≤–µ—Ä—Ö–Ω–æ—Å—Ç–∏
        n_t = -np.cosh(tau)
        n_x = np.sinh(tau)
        points_t.append(t_point)
        points_x.append(x_point)
        normals.append((n_t, n_x))

    # –ò–Ω—Ç–µ—Ä–ø–æ–ª–∏—Ä—É–µ–º –º–µ–∂–¥—É —Ç–æ—á–∫–∞–º–∏ ‚Äî —ç—Ç–æ –∏ –µ—Å—Ç—å –ø–æ–≤–µ—Ä—Ö–Ω–æ—Å—Ç—å œÑ = const
    t_surface = np.interp(np.linspace(0, 1, 100), np.linspace(0, 1, len(points_t)), points_t)
    x_surface = np.interp(np.linspace(0, 1, 100), np.linspace(0, 1, len(points_x)), points_x)

    # –†–∏—Å—É–µ–º –ø–æ–≤–µ—Ä—Ö–Ω–æ—Å—Ç—å
    plt.plot(t_surface, x_surface, color='red', linestyle='-', lw=2.5, alpha=0.9, label=f'$\\tau = {tau:.1f}$' if tau == tau_const[0] else "")

    # –û–ø—Ü–∏–æ–Ω–∞–ª—å–Ω–æ: –ø–æ–∫–∞–∑–∞—Ç—å –Ω–æ—Ä–º–∞–ª–∏
    # for i in range(0, len(points_t), 2):
    #     plt.arrow(points_t[i], points_x[i], 0.1 * normals[i][0], 0.1 * normals[i][1],
    #               head_width=0.05, color='red', alpha=0.6)

# -----------------------------------------------------------------------------
# 4. –û–±–ª–∞—Å—Ç—å –∏–Ω—Ç–µ–≥—Ä–∏—Ä–æ–≤–∞–Ω–∏—è –ê–í–ïF (–º–µ–∂–¥—É –¥–≤—É–º—è œÑ)
tau1, tau2 = -0.5, 0.5
points_t1 = [t_world(tau1, xi) for xi in xi_vals]
points_x1 = [x_world(tau1, xi) for xi in xi_vals]
points_t2 = [t_world(tau2, xi) for xi in xi_vals]
points_x2 = [x_world(tau2, xi) for xi in xi_vals]

# –°–æ–±–∏—Ä–∞–µ–º –ø–æ–ª–∏–≥–æ–Ω: –≤–¥–æ–ª—å œÑ1 ‚Üí –≤–¥–æ–ª—å Œæ_max ‚Üí –≤–¥–æ–ª—å œÑ2 (–æ–±—Ä–∞—Ç–Ω–æ) ‚Üí –≤–¥–æ–ª—å Œæ_min
t_poly = np.concatenate([
    np.interp(np.linspace(0, 1, 50), np.linspace(0, 1, len(points_t1)), points_t1),
    np.array([t_world(tau, xi_vals[-1]) for tau in np.linspace(tau1, tau2, 50)]),
    np.interp(np.linspace(0, 1, 50), np.linspace(0, 1, len(points_t2)), points_t2[::-1]),
    np.array([t_world(tau, xi_vals[0]) for tau in np.linspace(tau2, tau1, 50)])
])
x_poly = np.concatenate([
    np.interp(np.linspace(0, 1, 50), np.linspace(0, 1, len(points_x1)), points_x1),
    np.array([x_world(tau, xi_vals[-1]) for tau in np.linspace(tau1, tau2, 50)]),
    np.interp(np.linspace(0, 1, 50), np.linspace(0, 1, len(points_x2)), points_x2[::-1]),
    np.array([x_world(tau, xi_vals[0]) for tau in np.linspace(tau2, tau1, 50)])
])

plt.fill(t_poly, x_poly, color='gold', alpha=0.5, edgecolor='orange', lw=1.5, label='–û–±–ª–∞—Å—Ç—å –∏–Ω—Ç–µ–≥—Ä–∏—Ä–æ–≤–∞–Ω–∏—è $ABEF$')

# -----------------------------------------------------------------------------
# 5. –û—Ñ–æ—Ä–º–ª–µ–Ω–∏–µ
plt.xlabel("–ö–æ–æ—Ä–¥–∏–Ω–∞—Ç–Ω–æ–µ –≤—Ä–µ–º—è $t$ (–≤ –µ–¥–∏–Ω–∏—Ü–∞—Ö $c/a$)", fontsize=12)
plt.ylabel("–ü—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–æ $x$ (–≤ –µ–¥–∏–Ω–∏—Ü–∞—Ö $c^2/a$)", fontsize=12)
plt.title("–ü—Ä–∞–≤–∏–ª—å–Ω—ã–µ –ø–æ–≤–µ—Ä—Ö–Ω–æ—Å—Ç–∏ –æ–¥–Ω–æ–≤—Ä–µ–º–µ–Ω–Ω–æ—Å—Ç–∏ (–≤–∞—Ä–∏–∞—Ü–∏—è –í)\n–ø–æ –§–µ—Ä–º–∏ (1923)", fontsize=13)
plt.xlim(-1.2, 1.2)
plt.ylim(0.9, 2.1)
plt.grid(True, alpha=0.3)
plt.legend(loc='upper left', fontsize=10)
plt.tight_layout()

# –ü–æ–∫–∞–∑–∞—Ç—å
plt.show()
```

$$W=\sum_i\int de \int \varphi_i dx_i.$$

–í –õ–õ2 "–¢–µ–æ—Ä–∏—è –ø–æ–ª—è" –µ—Å—Ç—å –∞–Ω–∞–ª–æ–≥–∏—á–Ω–∞—è —Ñ–æ—Ä–º—É–ª–∞ (16,1), –æ—Ç–±—Ä–æ—Å–∏–≤ –æ—Ç –∫–æ—Ç–æ—Ä–æ–π –¥–µ–π—Å—Ç–≤–∏–µ –º–∞—Ç–µ—Ä–∏–∞–ª—å–Ω–æ–π –º–∞—Å—Å—ã –º–æ–∂–Ω–æ –∑–∞–ø–∏—Å–∞—Ç—å

$$S = \int\limits_{a}^{b} \left(-\frac{\int de}{c}A_i d x^i\right).\,\,\,  (16,1)$$


$$S = \int\limits_{a}^{b} \left(\frac{e}{c}\vec A d\vec r - e \varphi dt \right),$$

–≥–¥–µ $x^i = \left(ct, x, y, z\right)$

$$A^i=\left(\varphi, \vec A\right).\,\,\, (16,2)$$


$$\delta S = \delta \int\limits_{a}^{b} \left(-\frac{e}{c}A_i d x^i\right).\,\,\,  (23,1)$$

–ò–∑–º–µ–Ω–µ–Ω–∏–µ $S$ –ø—Ä–∏ –∑–∞–º–µ–Ω–µ $x^i$ –Ω–∞ $x^i + \delta x^i$ (–∏–ª–∏ –∏–∑–º–µ–Ω–µ–Ω–∏–µ $W$ –ø—Ä–∏ –∑–∞–º–µ–Ω–µ $x_i$ –Ω–∞ $x_i + \delta x_i$ –≤ –æ–±–æ–∑–Ω–∞—á–µ–Ω–∏—è—Ö –§–µ—Ä–º–∏)  –∏ –ø—Ä–∏ –∑–∞–º–µ–Ω–µ $A_i$ –Ω–∞ $A_i + \delta A_i$  ($\varphi_i$ –Ω–∞ $\varphi_i + \delta \varphi_i$ –≤ –æ–±–æ–∑–Ω–∞—á–µ–Ω–∏—è—Ö –§–µ—Ä–º–∏) –¥–∞—ë—Ç—Å—è —Ä–∞–∑–Ω–æ—Å—Ç—å—é

–≤ –æ–±–æ–∑–Ω–∞—á–µ–Ω–∏—è—Ö –õ–∞–Ω–¥–∞—É:

$$

$$

—Ä–∞–∑–ª–æ–∂–µ–Ω–∏–µ —ç—Ç–æ–π —Ä–∞–∑–Ω–æ—Å—Ç–∏ –ø–æ —Å—Ç–µ–ø–µ–Ω—è–º  $\delta A_i$ –∏ $\delta x^i$ –≤ –ø–æ–¥—ã–Ω—Ç–µ–≥—Ä–∞–ª—å–Ω–æ–º –≤—ã—Ä–∞–∂–µ–Ω–∏–∏ –Ω–∞—á–∏–Ω–∞–µ—Ç—Å—è —Å —á–ª–µ–Ω–æ–≤ –ø–µ—Ä–≤–æ–≥–æ –ø–æ—Ä—è–¥–∫–∞. –ù–µ–æ–±—Ö–æ–¥–∏–º—ã–º —É—Å–ª–æ–≤–∏–µ–º –º–∏–Ω–∏–º–∞–ª—å–Ω–æ—Å—Ç–∏ $W$ —è–≤–ª—è–µ—Ç—Å—è –æ–±—Ä–∞—â–µ–Ω–∏–µ –≤ –Ω—É–ª—å —Å–æ–≤–æ–∫—É–ø–Ω–æ—Å—Ç–∏ —ç—Ç–∏—Ö —á–ª–µ–Ω–æ–≤. –¢–∞–∫–æ–µ —Ä–∞–∑–ª–æ–∂–µ–Ω–∏–µ –Ω–∞–∑—ã–≤–∞—é—Ç –ø–µ—Ä–≤–æ–π –≤–∞—Ä–∏–∞—Ü–∏–µ–π –∏–Ω—Ç–µ–≥—Ä–∞–ª–∞. –ü—Ä–æ–∏–∑–≤–æ–¥—è –≤–∞—Ä—å–∏—Ä–æ–≤–∞–Ω–∏–µ –º–æ–∂–Ω–æ –∑–∞–ø–∏—Å–∞—Ç—å

  - –ó–∞—Ä—è–¥ $Q \gg q$, —Ç–∞–∫ —á—Ç–æ –º–æ–∂–Ω–æ —Å—á–∏—Ç–∞—Ç—å, —á—Ç–æ **–ø–æ–ª–µ –∏—Å—Ç–æ—á–Ω–∏–∫–∞ –¥–æ–º–∏–Ω–∏—Ä—É–µ—Ç**.

### –¶–µ–ª—å:

## üîπ –®–∞–≥ 1: –º–∏—Ä–æ–≤—ã–µ –ª–∏–Ω–∏–∏

## üîπ –®–∞–≥ 2: –∫–∞–∫–æ–≤–∞ –æ–±–ª–∞—Å—Ç—å –∏–Ω—Ç–µ–≥—Ä–∏—Ä–æ–≤–∞–Ω–∏—è –¥–µ–π—Å—Ç–≤–∏—è?

## üîπ –®–∞–≥ 3: –∫–∞–∫–∏–µ —Å–µ—á–µ–Ω–∏—è –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å?

### –ù–∞–±–ª—é–¥–∞—Ç–µ–ª—å **–ø–æ–∫–æ–∏—Ç—Å—è** ‚Üí –µ–≥–æ 4-—Å–∫–æ—Ä–æ—Å—Ç—å $u^\mu = (1, 0, 0, 0)$

## üîπ –®–∞–≥ 4: –ø–æ—á–µ–º—É —ç—Ç–æ –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ –ø–æ –§–µ—Ä–º–∏?

## ‚úÖ –ò—Ç–æ–≥: –ø—Ä–µ–¥–µ–ª—ã –∏–Ω—Ç–µ–≥—Ä–∏—Ä–æ–≤–∞–Ω–∏—è

- –≠—Ç–æ ‚Äî **–ø—Ä—è–º–æ—É–≥–æ–ª—å–Ω—ã–π —Ü–∏–ª–∏–Ω–¥—Ä** –≤ –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–µ-–≤—Ä–µ–º–µ–Ω–∏,

```python

import matplotlib.pyplot as plt

# –ü–∞—Ä–∞–º–µ—Ç—Ä—ã

x_obs_min, x_obs_max = 0.5, 1.5  # –ø—Ä–æ–±–Ω—ã–π –∑–∞—Ä—è–¥ (–ø–æ–∫–æ—è—â–∏–π—Å—è)
xi_source_min, xi_source_max = -0.3, 0.7  # –∏—Å—Ç–æ—á–Ω–∏–∫ (—É—Å–∫–æ—Ä—è—é—â–∏–π—Å—è)

tau_vals = np.linspace(-2, 2, 500)

# –§—É–Ω–∫—Ü–∏—è –º–∏—Ä–æ–≤–æ–π –ª–∏–Ω–∏–∏ –∏—Å—Ç–æ—á–Ω–∏–∫–∞
def x_world(tau, xi):
    return (1 + xi) * np.cosh(tau)

def t_world(tau, xi):
    return (1 + xi) * np.sinh(tau)

# -----------------------------------------------------------------------------
# –ì—Ä–∞—Ñ–∏–∫
plt.figure(figsize=(10, 8))

# –ú–∏—Ä–æ–≤—ã–µ –ª–∏–Ω–∏–∏ –∏—Å—Ç–æ—á–Ω–∏–∫–∞ (—É—Å–∫–æ—Ä—è—é—â–µ–≥–æ—Å—è)

    x = x_world(tau_vals, xi)
    t = t_world(tau_vals, xi)
    valid = (x > 0.1) & (x < 3.0)
    plt.plot(t[valid], x[valid], 'blue', lw=2, alpha=0.8, label='–ò—Å—Ç–æ—á–Ω–∏–∫ $\\xi = \\text{const}$' if xi == xi_source_min else "")

# –ú–∏—Ä–æ–≤—ã–µ –ª–∏–Ω–∏–∏ –Ω–∞–±–ª—é–¥–∞—Ç–µ–ª—è (–ø–æ–∫–æ—è—â–µ–≥–æ—Å—è)
x_obs_vals = np.linspace(x_obs_min, x_obs_max, 5)
for x_obs in x_obs_vals:
    t_line = np.linspace(t1, t2, 100)
    plt.plot(t_line, np.full_like(t_line, x_obs), 'green', lw=2.5, label='–ù–∞–±–ª—é–¥–∞—Ç–µ–ª—å (–ø–æ–∫–æ—è—â–∏–π—Å—è)' if x_obs == x_obs_vals[0] else "")

# –ü–æ–≤–µ—Ä—Ö–Ω–æ—Å—Ç–∏ –æ–¥–Ω–æ–≤—Ä–µ–º–µ–Ω–Ω–æ—Å—Ç–∏ (t = const)

    plt.axvline(t_c, color='gray', linestyle='--', alpha=0.6)

# –û–±–ª–∞—Å—Ç—å –∏–Ω—Ç–µ–≥—Ä–∏—Ä–æ–≤–∞–Ω–∏—è –¥–µ–π—Å—Ç–≤–∏—è
t_fill = np.linspace(t1, t2, 100)
x_bottom = np.full_like(t_fill, x_obs_min)

plt.fill_between(t_fill, x_bottom, x_top, color='gold', alpha=0.5, edgecolor='orange', lw=2, label='–û–±–ª–∞—Å—Ç—å –∏–Ω—Ç–µ–≥—Ä–∏—Ä–æ–≤–∞–Ω–∏—è $S$')

# –û—Ñ–æ—Ä–º–ª–µ–Ω–∏–µ
plt.xlabel("–í—Ä–µ–º—è $t$", fontsize=12)
plt.ylabel("–ü—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–æ $x$", fontsize=12)
plt.title("–ü—Ä–µ–¥–µ–ª—ã –∏–Ω—Ç–µ–≥—Ä–∏—Ä–æ–≤–∞–Ω–∏—è –¥–µ–π—Å—Ç–≤–∏—è\n(–ø–æ–∫–æ—è—â–∏–π—Å—è –Ω–∞–±–ª—é–¥–∞—Ç–µ–ª—å, —É—Å–∫–æ—Ä—è—é—â–∏–π—Å—è –∏—Å—Ç–æ—á–Ω–∏–∫)", fontsize=13)
plt.xlim(t1 - 0.5, t2 + 0.5)
plt.ylim(0.4, 2.0)
plt.legend(loc='upper left', fontsize=10)
plt.grid(True, alpha=0.3)
plt.tight_layout()

# –ü–æ–∫–∞–∑–∞—Ç—å
plt.show()
```

## üéØ –ß—Ç–æ –¥–∞–ª—å—à–µ?

```python
plt.ylim(t1 - 0.5, t2 + 0.5)
plt.legend(loc='upper left', fontsize=10)
```

## ‚úÖ Python-–∫–æ–¥: –≤–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è –æ–±–ª–∞—Å—Ç–∏ –∏–Ω—Ç–µ–≥—Ä–∏—Ä–æ–≤–∞–Ω–∏—è

```python
def x_world(tau, xi):
    return (1 + xi) * np.cosh(tau)

def t_world(tau, xi):
    return (1 + xi) * np.sinh(tau)

# –ú–∏—Ä–æ–≤—ã–µ –ª–∏–Ω–∏–∏ –∑–∞—Ä—è–¥–∞ 2 (–ø–æ–∫–æ—è—â–µ–≥–æ—Å—è)

x_intersect_min = max(x1_min, x2_min)
```

## ‚úÖ Python-–∫–æ–¥: –∏—Å–ø—Ä–∞–≤–ª–µ–Ω–Ω–∞—è –≤–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è

```python

import matplotlib.pyplot as plt

# –ü–∞—Ä–∞–º–µ—Ç—Ä—ã

x1_min, x1_max = 0.7, 1.7  # –∑–∞—Ä—è–¥ 1 (—É—Å–∫–æ—Ä—è—é—â–∏–π—Å—è)
x2_min, x2_max = 2.3, 3.3  # –∑–∞—Ä—è–¥ 2 (–ø–æ–∫–æ—è—â–∏–π—Å—è)

tau_vals = np.linspace(-2, 2, 500)

# –§—É–Ω–∫—Ü–∏—è –º–∏—Ä–æ–≤–æ–π –ª–∏–Ω–∏–∏ –∑–∞—Ä—è–¥–∞ 1 (—É—Å–∫–æ—Ä—è—é—â–µ–≥–æ—Å—è)
def x_world(tau, xi):
    return (1 + xi) * np.cosh(tau)

def t_world(tau, xi):
    return (1 + xi) * np.sinh(tau)

# -----------------------------------------------------------------------------
# –ì—Ä–∞—Ñ–∏–∫: –æ—Å—å x ‚Äî –≥–æ—Ä–∏–∑–æ–Ω—Ç–∞–ª—å–Ω–∞—è, t ‚Äî –≤–µ—Ä—Ç–∏–∫–∞–ª—å–Ω–∞—è

for xi in np.linspace(x1_min - 1, x1_max - 1, 5):
    x = x_world(tau_vals, xi)
    t = t_world(tau_vals, xi)
    valid = (x > 0.4) & (x < 2.0)
    plt.plot(x[valid], t[valid], 'blue', lw=2, alpha=0.8, label=r'$\xi = \mathrm{const}$ (–∑–∞—Ä—è–¥ 1)' if xi == x1_min - 1 else "")

# –ú–∏—Ä–æ–≤—ã–µ –ª–∏–Ω–∏–∏ –∑–∞—Ä—è–¥–∞ 2 (–ø–æ–∫–æ—è—â–µ–≥–æ—Å—è)

for x2 in x2_vals:
    t_line = np.linspace(t1, t2, 100)
    plt.plot(np.full_like(t_line, x2), t_line, 'green', lw=2.5, label='–ó–∞—Ä—è–¥ 2 (–ø–æ–∫–æ—è—â–∏–π—Å—è)' if x2 == x2_vals[0] else "")

# –û–±–ª–∞—Å—Ç—å –∏–Ω—Ç–µ–≥—Ä–∏—Ä–æ–≤–∞–Ω–∏—è –¥–ª—è –∑–∞—Ä—è–¥–∞ 1: —Å–µ—á–µ–Ω–∏—è œÑ = const
# –ù–∞—Ä–∏—Å—É–µ–º –Ω–µ—Å–∫–æ–ª—å–∫–æ —Å–µ—á–µ–Ω–∏–π œÑ = const
```

## üéØ –ß—Ç–æ –¥–∞–ª—å—à–µ?

–¢–µ–ø–µ—Ä—å –≤—ã –º–æ–∂–µ—Ç–µ:
- –†–∞—Å—Å–º–æ—Ç—Ä–µ—Ç—å, –∫–∞–∫ **–ø–æ–ª–µ –º–µ–∂–¥—É –∑–∞—Ä—è–¥–∞–º–∏** –≤–ª–∏—è–µ—Ç –Ω–∞ –¥–µ–π—Å—Ç–≤–∏–µ,
- –ò–ª–∏ –ø–µ—Ä–µ–π—Ç–∏ –∫ **–≤—ã—á–∏—Å–ª–µ–Ω–∏—é —Å–∏–ª—ã**, –¥–µ–π—Å—Ç–≤—É—é—â–µ–π –Ω–∞ –∫–∞–∂–¥—ã–π –∏–∑ –Ω–∏—Ö,
- –ò–ª–∏ –ø–æ—Å—Ç—Ä–æ–∏—Ç—å **–∫–∞–∂—É—â–µ–µ—Å—è –ø–æ–ª–µ** –≤ —Å–æ–ø—É—Ç—Å—Ç–≤—É—é—â–µ–π –°–û.

–ì–æ—Ç–æ–≤ —É–≥–ª—É–±–∏—Ç—å—Å—è. –•–æ—Ç–∏—Ç–µ ‚Äî –Ω–∞—á–Ω—ë–º —Å **–ø–æ–ª—è –º–µ–∂–¥—É –∑–∞—Ä—è–¥–∞–º–∏**?


````python
import numpy as np
import matplotlib.pyplot as plt

# –ü–∞—Ä–∞–º–µ—Ç—Ä—ã
t1, t2 = -1.0, 1.0
x1_min, x1_max = 0.7, 1.7  # –∑–∞—Ä—è–¥ 1 (—É—Å–∫–æ—Ä—è—é—â–∏–π—Å—è)
x2_min, x2_max = 2.3, 3.3  # –∑–∞—Ä—è–¥ 2 (–ø–æ–∫–æ—è—â–∏–π—Å—è)

tau_vals = np.linspace(-2, 2, 500)

# –§—É–Ω–∫—Ü–∏—è –º–∏—Ä–æ–≤–æ–π –ª–∏–Ω–∏–∏ –∑–∞—Ä—è–¥–∞ 1 (—É—Å–∫–æ—Ä—è—é—â–µ–≥–æ—Å—è)
def x_world(tau, xi):
    return (1 + xi) * np.cosh(tau)

def t_world(tau, xi):
    return (1 + xi) * np.sinh(tau)

# -----------------------------------------------------------------------------
# –ì—Ä–∞—Ñ–∏–∫: –æ—Å—å x ‚Äî –≥–æ—Ä–∏–∑–æ–Ω—Ç–∞–ª—å–Ω–∞—è, t ‚Äî –≤–µ—Ä—Ç–∏–∫–∞–ª—å–Ω–∞—è
plt.figure(figsize=(12, 8))

# –ú–∏—Ä–æ–≤—ã–µ –ª–∏–Ω–∏–∏ –∑–∞—Ä—è–¥–∞ 1 (—É—Å–∫–æ—Ä—è—é—â–µ–≥–æ—Å—è)
for xi in np.linspace(x1_min - 1, x1_max - 1, 5):
    x = x_world(tau_vals, xi)
    t = t_world(tau_vals, xi)
    valid = (x > 0.4) & (x < 2.0)
    plt.plot(x[valid], t[valid], 'blue', lw=2, alpha=0.8, label=r'$\xi = \mathrm{const}$ (–∑–∞—Ä—è–¥ 1)' if xi == x1_min - 1 else "")

# –ú–∏—Ä–æ–≤—ã–µ –ª–∏–Ω–∏–∏ –∑–∞—Ä—è–¥–∞ 2 (–ø–æ–∫–æ—è—â–µ–≥–æ—Å—è)
x2_vals = np.linspace(x2_min, x2_max, 5)
for x2 in x2_vals:
    t_line = np.linspace(t1, t2, 100)
    plt.plot(np.full_like(t_line, x2), t_line, 'green', lw=2.5, label='–ó–∞—Ä—è–¥ 2 (–ø–æ–∫–æ—è—â–∏–π—Å—è)' if x2 == x2_vals[0] else "")

# –û–±–ª–∞—Å—Ç—å –∏–Ω—Ç–µ–≥—Ä–∏—Ä–æ–≤–∞–Ω–∏—è –¥–ª—è –∑–∞—Ä—è–¥–∞ 1: —Å–µ—á–µ–Ω–∏—è œÑ = const
# –ù–∞—Ä–∏—Å—É–µ–º –Ω–µ—Å–∫–æ–ª—å–∫–æ —Å–µ—á–µ–Ω–∏–π œÑ = const
tau_const = np.linspace(-1.0, 1.0, 5)
for tau_c in tau_const:
    # –°—Ç—Ä–æ–∏–º —Å–µ—á–µ–Ω–∏–µ œÑ = const: t = (1+Œæ) sinh œÑ, x = (1+Œæ) cosh œÑ
    xi_vals = np.linspace(x1_min - 1, x1_max - 1, 100)
    t_s = (1 + xi_vals) * np.sinh(tau_c)
    x_s = (1 + xi_vals) * np.cosh(tau_c)
    plt.plot(x_s, t_s, 'red', linestyle='--', alpha=0.6, lw=1.5)

# –û–±–ª–∞—Å—Ç—å –∏–Ω—Ç–µ–≥—Ä–∏—Ä–æ–≤–∞–Ω–∏—è –¥–ª—è –∑–∞—Ä—è–¥–∞ 2: —Å–µ—á–µ–Ω–∏—è t = const
t_const = np.linspace(t1, t2, 5)
for t_c in t_const:
    plt.axhline(t_c, color='gray', linestyle='--', alpha=0.6)

# –û–±–ª–∞—Å—Ç—å –∏–Ω—Ç–µ–≥—Ä–∏—Ä–æ–≤–∞–Ω–∏—è G1 (–∑–∞—Ä—è–¥ 1): –ø—Ä—è–º–æ—É–≥–æ–ª—å–Ω–∏–∫ –≤ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–∞—Ö –†–∏–Ω–¥–ª–µ—Ä–∞
t_fill1 = np.linspace(t1, t2, 100)
x_bottom1 = np.full_like(t_fill1, x1_min)
x_top1 = np.full_like(t_fill1, x1_max)
plt.fill_betweenx(t_fill1, x_bottom1, x_top1, color='gold', alpha=0.5, edgecolor='orange', lw=2, label=r'–û–±–ª–∞—Å—Ç—å $G_1$')

# –û–±–ª–∞—Å—Ç—å –∏–Ω—Ç–µ–≥—Ä–∏—Ä–æ–≤–∞–Ω–∏—è G2 (–∑–∞—Ä—è–¥ 2): –ø—Ä—è–º–æ—É–≥–æ–ª—å–Ω–∏–∫ –≤ –∏–Ω–µ—Ä—Ü–∏–∞–ª—å–Ω–æ–π —Å–∏—Å—Ç–µ–º–µ
t_fill2 = np.linspace(t1, t2, 100)
x_bottom2 = np.full_like(t_fill2, x2_min)
x_top2 = np.full_like(t_fill2, x2_max)
plt.fill_betweenx(t_fill2, x_bottom2, x_top2, color='lightblue', alpha=0.5, edgecolor='blue', lw=2, label=r'–û–±–ª–∞—Å—Ç—å $G_2$')

# –û—Ñ–æ—Ä–º–ª–µ–Ω–∏–µ
plt.xlabel(r"–ü—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–æ $x$", fontsize=12)
plt.ylabel(r"–í—Ä–µ–º—è $t$", fontsize=12)
plt.title(r"–û–±–ª–∞—Å—Ç—å –∏–Ω—Ç–µ–≥—Ä–∏—Ä–æ–≤–∞–Ω–∏—è –¥–µ–π—Å—Ç–≤–∏—è: –¥–≤–∞ –Ω–µ–ø–µ—Ä–µ—Å–µ–∫–∞—é—â–∏—Ö—Å—è –∑–∞—Ä—è–¥–∞", fontsize=13)
plt.xlim(0.4, 4.0)
plt.ylim(t1 - 0.5, t2 + 0.5)
plt.legend(loc='upper left', fontsize=10)
plt.grid(True, alpha=0.3)
plt.tight_layout()

# –ü–æ–∫–∞–∑–∞—Ç—å
plt.show()
````

- –ê **—Å–µ—á–µ–Ω–∏–µ $\tau = \text{const}$** ‚Äî **–æ—Ä—Ç–æ–≥–æ–Ω–∞–ª—å–Ω–æ**, –Ω–æ **–Ω–µ –ø—Ä—è–º–æ—É–≥–æ–ª—å–Ω–æ–µ**.

üìå –≠—Ç–æ ‚Äî **–æ—à–∏–±–∫–∞**:  
> –û–±–ª–∞—Å—Ç—å –∏–Ω—Ç–µ–≥—Ä–∏—Ä–æ–≤–∞–Ω–∏—è –¥–ª—è —É—Å–∫–æ—Ä—è—é—â–µ–≥–æ—Å—è –∑–∞—Ä—è–¥–∞ **–Ω–µ –º–æ–∂–µ—Ç –±—ã—Ç—å –ø—Ä—è–º–æ—É–≥–æ–ª—å–Ω–∏–∫–æ–º** –≤ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–∞—Ö $(t,x)$.

## ‚úÖ –ü—Ä–∞–≤–∏–ª—å–Ω–∞—è –æ–±–ª–∞—Å—Ç—å –∏–Ω—Ç–µ–≥—Ä–∏—Ä–æ–≤–∞–Ω–∏—è –¥–ª—è —É—Å–∫–æ—Ä—è—é—â–µ–≥–æ—Å—è –∑–∞—Ä—è–¥–∞

### –ü–æ –§–µ—Ä–º–∏: –≤–∞—Ä–∏–∞—Ü–∏—è –í

  - –û–±–ª–∞—Å—Ç—å $G_1$ ‚Äî —ç—Ç–æ **4D-–æ–±—ä—ë–º –º–µ–∂–¥—É –¥–≤—É–º—è —Ç–∞–∫–∏–º–∏ —Å–µ—á–µ–Ω–∏—è–º–∏**.

### –ì–µ–æ–º–µ—Ç—Ä–∏—á–µ—Å–∫–∏:
- $G_1$ ‚Äî —ç—Ç–æ **—Ü–∏–ª–∏–Ω–¥—Ä**, –æ–±—Ä–∞–∑–æ–≤–∞–Ω–Ω—ã–π –≤—Å–µ–º–∏ –º–∏—Ä–æ–≤—ã–º–∏ –ª–∏–Ω–∏—è–º–∏ –∑–∞—Ä—è–¥–∞ 1 –º–µ–∂–¥—É $\tau_1$ –∏ $\tau_2$,

```python
t_fill2 = np.linspace(-1.5, 1.5, 100)
x_bottom2 = np.full_like(t_fill2, x2_min)
x_top2 = np.full_like(t_fill2, x2_max)

plt.title(r"–û–±–ª–∞—Å—Ç—å –∏–Ω—Ç–µ–≥—Ä–∏—Ä–æ–≤–∞–Ω–∏—è –¥–µ–π—Å—Ç–≤–∏—è: –¥–≤–∞ –Ω–µ–ø–µ—Ä–µ—Å–µ–∫–∞—é—â–∏—Ö—Å—è –∑–∞—Ä—è–¥–∞", fontsize=13)
```

## üîπ –®–∞–≥ 2: –∫—Ä–∞—Ç–Ω–æ—Å—Ç—å –∏–Ω—Ç–µ–≥—Ä–∏—Ä–æ–≤–∞–Ω–∏—è

> –£–≤–µ–ª–∏—á–∏–≤–∞–µ—Ç—Å—è –ª–∏ –∫—Ä–∞—Ç–Ω–æ—Å—Ç—å –∏–Ω—Ç–µ–≥—Ä–∏—Ä–æ–≤–∞–Ω–∏—è –ø—Ä–∏ –Ω–∞–ª–∏—á–∏–∏ –¥–≤—É—Ö –∑–∞—Ä—è–¥–æ–≤?

### –û—Ç–≤–µ—Ç:
‚úÖ **–ù–µ—Ç**, –∫—Ä–∞—Ç–Ω–æ—Å—Ç—å **–Ω–µ —É–≤–µ–ª–∏—á–∏–≤–∞–µ—Ç—Å—è**.

–ü–æ—Ç–æ–º—É —á—Ç–æ:
- –î–µ–π—Å—Ç–≤–∏–µ ‚Äî —ç—Ç–æ **—Ñ—É–Ω–∫—Ü–∏–æ–Ω–∞–ª –æ—Ç –ø–æ–ª–µ–π –∏ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç –≤—Å–µ—Ö –∑–∞—Ä—è–¥–æ–≤**,
- –ù–æ **–∏–Ω—Ç–µ–≥—Ä–∏—Ä–æ–≤–∞–Ω–∏–µ** ‚Äî –ø–æ **–ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤—É-–≤—Ä–µ–º–µ–Ω–∏**, –∞ –Ω–µ –ø–æ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–∞–º –∑–∞—Ä—è–¥–æ–≤,
- –ö–∞–∂–¥—ã–π –∑–∞—Ä—è–¥ –∏–º–µ–µ—Ç —Å–≤–æ—é **–º–∏—Ä–æ–≤—É—é —Ç—Ä—É–±–∫—É**, –∏ **–∏–Ω—Ç–µ–≥—Ä–∏—Ä–æ–≤–∞–Ω–∏–µ –ø–æ –Ω–µ–π** ‚Äî —ç—Ç–æ **—á–µ—Ç—ã—Ä—ë—Ö–º–µ—Ä–Ω—ã–π –∏–Ω—Ç–µ–≥—Ä–∞–ª**.

üìå –¢–æ –µ—Å—Ç—å:
- –î–ª—è –æ–¥–Ω–æ–≥–æ –∑–∞—Ä—è–¥–∞: $S_1 = \int_{G_1} j_1^\mu A_{\mu,2} d^4x$,
- –î–ª—è –≤—Ç–æ—Ä–æ–≥–æ: $S_2 = \int_{G_2} j_2^\mu A_{\mu,1} d^4x$,
- –û–±—â–µ–µ –¥–µ–π—Å—Ç–≤–∏–µ: $S = S_1 + S_2$,
- –ö–∞–∂–¥–æ–µ –∏–Ω—Ç–µ–≥—Ä–∞–ª—å–Ω–æ–µ –≤—ã—Ä–∞–∂–µ–Ω–∏–µ ‚Äî **—á–µ—Ç—ã—Ä—ë—Ö–º–µ—Ä–Ω—ã–π –∏–Ω—Ç–µ–≥—Ä–∞–ª**.

‚úÖ –ö—Ä–∞—Ç–Ω–æ—Å—Ç—å ‚Äî **–≤—Å–µ–≥–¥–∞ 4**, –ø–æ—Ç–æ–º—É —á—Ç–æ –º—ã –∏–Ω—Ç–µ–≥—Ä–∏—Ä—É–µ–º –ø–æ $d^4x$.

## üéØ –ß—Ç–æ –¥–∞–ª—å—à–µ?

–¢–µ–ø–µ—Ä—å –≤—ã –º–æ–∂–µ—Ç–µ:

- –ò–ª–∏ –ø–µ—Ä–µ–π—Ç–∏ –∫ **–≤—ã—á–∏—Å–ª–µ–Ω–∏—é —Å–∏–ª—ã**, –¥–µ–π—Å—Ç–≤—É—é—â–µ–π –Ω–∞ –∫–∞–∂–¥—ã–π –∏–∑ –Ω–∏—Ö,
- –ò–ª–∏ –ø–æ—Å—Ç—Ä–æ–∏—Ç—å **–∫–∞–∂—É—â–µ–µ—Å—è –ø–æ–ª–µ** –≤ —Å–æ–ø—É—Ç—Å—Ç–≤—É—é—â–µ–π –°–û.

–ì–æ—Ç–æ–≤ —É–≥–ª—É–±–∏—Ç—å—Å—è. –•–æ—Ç–∏—Ç–µ ‚Äî –Ω–∞—á–Ω—ë–º —Å **–ø–æ–ª—è –º–µ–∂–¥—É –∑–∞—Ä—è–¥–∞–º–∏**?


````python
import numpy as np
import matplotlib.pyplot as plt

# –ü–∞—Ä–∞–º–µ—Ç—Ä—ã
tau1, tau2 = -1.0, 1.0
xi_min, xi_max = 0.7 - 1, 1.7 - 1  # Œæ ‚àà [0.7, 1.7] ‚Üí Œæ ‚àà [-0.3, 0.7]

# –°–æ–∑–¥–∞–µ–º —Å–µ—Ç–∫—É œÑ –∏ Œæ
tau_vals = np.linspace(tau1, tau2, 50)
xi_vals = np.linspace(xi_min, xi_max, 50)

# –í—ã—á–∏—Å–ª—è–µ–º x –∏ t
X, T = np.meshgrid(xi_vals, tau_vals)
x = (1 + X) * np.cosh(T)
t = (1 + X) * np.sinh(T)

# -----------------------------------------------------------------------------
# –ì—Ä–∞—Ñ–∏–∫: –æ—Å—å x ‚Äî –≥–æ—Ä–∏–∑–æ–Ω—Ç–∞–ª—å–Ω–∞—è, t ‚Äî –≤–µ—Ä—Ç–∏–∫–∞–ª—å–Ω–∞—è
plt.figure(figsize=(10, 8))

# –ü–æ—Å—Ç—Ä–æ–∏—Ç—å –ø–æ–≤–µ—Ä—Ö–Ω–æ—Å—Ç—å G1 –∫–∞–∫ –∫–æ–Ω—Ç—É—Ä
for i in range(len(xi_vals)):
    plt.plot(x[i], t[i], 'blue', lw=1.5, alpha=0.8)

# –ü–æ—Å—Ç—Ä–æ–∏—Ç—å —Å–µ—á–µ–Ω–∏—è œÑ = const (–∫—Ä–∞—Å–Ω—ã–µ –ø—É–Ω–∫—Ç–∏—Ä–Ω—ã–µ)
tau_const = np.linspace(tau1, tau2, 5)
for tau_c in tau_const:
    xi_vals_s = np.linspace(xi_min, xi_max, 100)
    x_s = (1 + xi_vals_s) * np.cosh(tau_c)
    t_s = (1 + xi_vals_s) * np.sinh(tau_c)
    plt.plot(x_s, t_s, 'red', linestyle='--', alpha=0.6, lw=1.5)

# –ú–∏—Ä–æ–≤—ã–µ –ª–∏–Ω–∏–∏ –∑–∞—Ä—è–¥–∞ 2 (–ø–æ–∫–æ—è—â–∏–π—Å—è)
x2_min, x2_max = 2.3, 3.3
x2_vals = np.linspace(x2_min, x2_max, 5)
for x2 in x2_vals:
    t_line = np.linspace(-1.5, 1.5, 100)
    plt.plot(np.full_like(t_line, x2), t_line, 'green', lw=2.5, label='–ó–∞—Ä—è–¥ 2 (–ø–æ–∫–æ—è—â–∏–π—Å—è)' if x2 == x2_vals[0] else "")

# –û–±–ª–∞—Å—Ç—å –∏–Ω—Ç–µ–≥—Ä–∏—Ä–æ–≤–∞–Ω–∏—è G2 (–∑–∞—Ä—è–¥ 2): –ø—Ä—è–º–æ—É–≥–æ–ª—å–Ω–∏–∫ –≤ –∏–Ω–µ—Ä—Ü–∏–∞–ª—å–Ω–æ–π —Å–∏—Å—Ç–µ–º–µ
t_fill2 = np.linspace(-1.5, 1.5, 100)
x_bottom2 = np.full_like(t_fill2, x2_min)
x_top2 = np.full_like(t_fill2, x2_max)
plt.fill_betweenx(t_fill2, x_bottom2, x_top2, color='lightblue', alpha=0.5, edgecolor='blue', lw=2, label=r'–û–±–ª–∞—Å—Ç—å $G_2$')

# –û—Ñ–æ—Ä–º–ª–µ–Ω–∏–µ
plt.xlabel(r"–ü—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–æ $x$", fontsize=12)
plt.ylabel(r"–í—Ä–µ–º—è $t$", fontsize=12)
plt.title(r"–û–±–ª–∞—Å—Ç—å –∏–Ω—Ç–µ–≥—Ä–∏—Ä–æ–≤–∞–Ω–∏—è –¥–µ–π—Å—Ç–≤–∏—è: –¥–≤–∞ –Ω–µ–ø–µ—Ä–µ—Å–µ–∫–∞—é—â–∏—Ö—Å—è –∑–∞—Ä—è–¥–∞", fontsize=13)
plt.xlim(0.4, 4.0)
plt.ylim(-1.5, 1.5)
plt.legend(loc='upper left', fontsize=10)
plt.grid(True, alpha=0.3)
plt.tight_layout()

# –ü–æ–∫–∞–∑–∞—Ç—å
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt

# –ü–∞—Ä–∞–º–µ—Ç—Ä—ã
tau1, tau2 = -1.0, 1.0
xi_min, xi_max = 0.7 - 1, 1.7 - 1  # Œæ ‚àà [-0.3, 0.7]

# –°–æ–∑–¥–∞–µ–º —Å–µ—Ç–∫—É œÑ –∏ Œæ
tau_vals = np.linspace(tau1, tau2, 50)
xi_vals = np.linspace(xi_min, xi_max, 50)

# -----------------------------------------------------------------------------
# –ì—Ä–∞—Ñ–∏–∫: –æ—Å—å x ‚Äî –≥–æ—Ä–∏–∑–æ–Ω—Ç–∞–ª—å–Ω–∞—è, t ‚Äî –≤–µ—Ä—Ç–∏–∫–∞–ª—å–Ω–∞—è
plt.figure(figsize=(10, 8))

# –ú–∏—Ä–æ–≤—ã–µ –ª–∏–Ω–∏–∏ —É—Å–∫–æ—Ä—è—é—â–µ–≥–æ—Å—è –∑–∞—Ä—è–¥–∞ (Œæ = const)
for xi in xi_vals:
    x_line = []
    t_line = []
    for tau in tau_vals:
        x = (1 + xi) * np.cosh(tau)
        t = (1 + xi) * np.sinh(tau)
        x_line.append(x)
        t_line.append(t)
    plt.plot(x_line, t_line, 'blue', lw=1.5, alpha=0.8)

# –°–µ—á–µ–Ω–∏—è œÑ = const (–∫—Ä–∞—Å–Ω—ã–µ –ø—É–Ω–∫—Ç–∏—Ä–Ω—ã–µ)
tau_const = np.linspace(tau1, tau2, 5)
for tau_c in tau_const:
    xi_vals_s = np.linspace(xi_min, xi_max, 100)
    x_s = (1 + xi_vals_s) * np.cosh(tau_c)
    t_s = (1 + xi_vals_s) * np.sinh(tau_c)
    plt.plot(x_s, t_s, 'red', linestyle='--', alpha=0.6, lw=1.5)

# –û–±–ª–∞—Å—Ç—å –∏–Ω—Ç–µ–≥—Ä–∏—Ä–æ–≤–∞–Ω–∏—è G1: —Ü–∏–ª–∏–Ω–¥—Ä –º–µ–∂–¥—É œÑ1 –∏ œÑ2
# –î–ª—è –∫–∞–∂–¥–æ–≥–æ Œæ —Å—Ç—Ä–æ–∏–º –ª–∏–Ω–∏—é –º–µ–∂–¥—É œÑ1 –∏ œÑ2
for xi in xi_vals:
    x_line = []
    t_line = []
    for tau in [tau1, tau2]:
        x = (1 + xi) * np.cosh(tau)
        t = (1 + xi) * np.sinh(tau)
        x_line.append(x)
        t_line.append(t)
    plt.plot(x_line, t_line, 'blue', lw=1.5, alpha=0.8)

# –ó–∞—Ä—è–¥ 2 (–ø–æ–∫–æ—è—â–∏–π—Å—è)
x2_min, x2_max = 2.3, 3.3
x2_vals = np.linspace(x2_min, x2_max, 5)
for x2 in x2_vals:
    t_line = np.linspace(-1.5, 1.5, 100)
    plt.plot(np.full_like(t_line, x2), t_line, 'green', lw=2.5, label='–ó–∞—Ä—è–¥ 2 (–ø–æ–∫–æ—è—â–∏–π—Å—è)' if x2 == x2_vals[0] else "")

# –û–±–ª–∞—Å—Ç—å –∏–Ω—Ç–µ–≥—Ä–∏—Ä–æ–≤–∞–Ω–∏—è G2 (–∑–∞—Ä—è–¥ 2): –ø—Ä—è–º–æ—É–≥–æ–ª—å–Ω–∏–∫ –≤ –∏–Ω–µ—Ä—Ü–∏–∞–ª—å–Ω–æ–π —Å–∏—Å—Ç–µ–º–µ
t_fill2 = np.linspace(-1.5, 1.5, 100)
x_bottom2 = np.full_like(t_fill2, x2_min)
x_top2 = np.full_like(t_fill2, x2_max)
plt.fill_betweenx(t_fill2, x_bottom2, x_top2, color='lightblue', alpha=0.5, edgecolor='blue', lw=2, label=r'–û–±–ª–∞—Å—Ç—å $G_2$')

# –û—Ñ–æ—Ä–º–ª–µ–Ω–∏–µ
plt.xlabel(r"–ü—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–æ $x$", fontsize=12)
plt.ylabel(r"–í—Ä–µ–º—è $t$", fontsize=12)
plt.title(r"–û–±–ª–∞—Å—Ç—å –∏–Ω—Ç–µ–≥—Ä–∏—Ä–æ–≤–∞–Ω–∏—è –¥–µ–π—Å—Ç–≤–∏—è: –¥–≤–∞ –Ω–µ–ø–µ—Ä–µ—Å–µ–∫–∞—é—â–∏—Ö—Å—è –∑–∞—Ä—è–¥–∞", fontsize=13)
plt.xlim(0.4, 4.0)
plt.ylim(-1.5, 1.5)
plt.legend(loc='upper left', fontsize=10)
plt.grid(True, alpha=0.3)
plt.tight_layout()

# –ü–æ–∫–∞–∑–∞—Ç—å
plt.show()
````

–ù–æ—Ä–º–∞–ª—å–Ω—ã–π –≤–µ–∫—Ç–æ—Ä:
$$
\vec{n} = (-\cosh \tau, \sinh \tau)
\quad \text{(–≤ 2D)}

```python
import numpy as np
import matplotlib.pyplot as plt

# –ü–∞—Ä–∞–º–µ—Ç—Ä—ã
tau_vals = np.linspace(-2, 2, 500)
xi_vals = np.linspace(-0.3, 0.7, 5)  # –º–∏—Ä–æ–≤—ã–µ –ª–∏–Ω–∏–∏

# –§—É–Ω–∫—Ü–∏–∏ –º–∏—Ä–æ–≤—ã—Ö –ª–∏–Ω–∏–π
def x_world(tau, xi):
    return (1 + xi) * np.cosh(tau)

def t_world(tau, xi):
    return (1 + xi) * np.sinh(tau)

# -----------------------------------------------------------------------------
# –ì—Ä–∞—Ñ–∏–∫
plt.figure(figsize=(10, 8))

# –ú–∏—Ä–æ–≤—ã–µ –ª–∏–Ω–∏–∏ Œæ = const
for xi in xi_vals:
    x = x_world(tau_vals, xi)
    t = t_world(tau_vals, xi)
    valid = (x >= 0.8) & (x <= 2.5)
    plt.plot(t[valid], x[valid], color='blue', lw=1.8, label=f'$\\xi = {xi}$' if xi == xi_vals[0] else "")

# -----------------------------------------------------------------------------
# –ü–æ–≤–µ—Ä—Ö–Ω–æ—Å—Ç–∏ –æ–¥–Ω–æ–≤—Ä–µ–º–µ–Ω–Ω–æ—Å—Ç–∏ œÑ = const (–æ—Ä—Ç–æ–≥–æ–Ω–∞–ª—å–Ω—ã–µ –º–∏—Ä–æ–≤—ã–º –ª–∏–Ω–∏—è–º)
tau_const = np.linspace(-0.8, 0.8, 5)

for tau in tau_const:
    # –°–æ–±–∏—Ä–∞–µ–º —Ç–æ—á–∫–∏ –Ω–∞ –≤—Å–µ—Ö –º–∏—Ä–æ–≤—ã—Ö –ª–∏–Ω–∏—è—Ö –ø—Ä–∏ —ç—Ç–æ–º œÑ
    points_t = []
    points_x = []
    normals = []

    for xi in xi_vals:
        t_point = t_world(tau, xi)
        x_point = x_world(tau, xi)
        # –ù–æ—Ä–º–∞–ª—å–Ω—ã–π –≤–µ–∫—Ç–æ—Ä: (-cosh œÑ, sinh œÑ) ‚Üí –Ω–∞–ø—Ä–∞–≤–ª–µ–Ω–∏–µ –ø–æ–≤–µ—Ä—Ö–Ω–æ—Å—Ç–∏
        n_t = -np.cosh(tau)
        n_x = np.sinh(tau)
        points_t.append(t_point)
        points_x.append(x_point)
        normals.append((n_t, n_x))

    # –ò–Ω—Ç–µ—Ä–ø–æ–ª–∏—Ä—É–µ–º –º–µ–∂–¥—É —Ç–æ—á–∫–∞–º–∏ ‚Äî —ç—Ç–æ –∏ –µ—Å—Ç—å –ø–æ–≤–µ—Ä—Ö–Ω–æ—Å—Ç—å œÑ = const
    t_surface = np.interp(np.linspace(0, 1, 100), np.linspace(0, 1, len(points_t)), points_t)
    x_surface = np.interp(np.linspace(0, 1, 100), np.linspace(0, 1, len(points_x)), points_x)

    # –†–∏—Å—É–µ–º –ø–æ–≤–µ—Ä—Ö–Ω–æ—Å—Ç—å
    plt.plot(t_surface, x_surface, color='red', linestyle='-', lw=2.5, alpha=0.9, label=f'$\\tau = {tau:.1f}$' if tau == tau_const[0] else "")

plt.xlabel("–ö–æ–æ—Ä–¥–∏–Ω–∞—Ç–Ω–æ–µ –≤—Ä–µ–º—è $t$", fontsize=12)
plt.ylabel("–ü—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–æ $x$", fontsize=12)
plt.title("–ü—Ä–∞–≤–∏–ª—å–Ω—ã–µ –ø–æ–≤–µ—Ä—Ö–Ω–æ—Å—Ç–∏ –æ–¥–Ω–æ–≤—Ä–µ–º–µ–Ω–Ω–æ—Å—Ç–∏ (–≤–∞—Ä–∏–∞—Ü–∏—è –í)\n–ø–æ –§–µ—Ä–º–∏ (1923)", fontsize=13)
plt.xlim(-1.2, 1.2)
plt.ylim(0.9, 2.1)
plt.grid(True, alpha=0.3)
plt.legend(loc='upper left', fontsize=10)
plt.tight_layout()

# –ü–æ–∫–∞–∑–∞—Ç—å
plt.show()
```

- –û—Å—å $t$ –±—ã–ª–∞ –≥–æ—Ä–∏–∑–æ–Ω—Ç–∞–ª—å–Ω–æ–π ‚Äî —ç—Ç–æ **–Ω–µ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É–µ—Ç —Å—Ç–∞—Ç—å–µ –§–µ—Ä–º–∏**,

- –ò –≥–ª–∞–≤–Ω–æ–µ: **–Ω–æ—Ä–º–∞–ª—å –∫ –º–∏—Ä–æ–≤–æ–π –ª–∏–Ω–∏–∏** ‚Äî —ç—Ç–æ **–Ω–µ –≤–¥–æ–ª—å $\tau = \text{const}$**, –∞ **–ø–µ—Ä–ø–µ–Ω–¥–∏–∫—É–ª—è—Ä–Ω–æ –∫–∞—Å–∞—Ç–µ–ª—å–Ω–æ–π –∫ –≥–∏–ø–µ—Ä–±–æ–ª–µ**.

2. ‚úÖ –°–µ—á–µ–Ω–∏—è $\tau = \text{const}$ –∑–∞–º–µ–Ω–∏—Ç—å –Ω–∞ **—Å–µ—á–µ–Ω–∏—è, –æ—Ä—Ç–æ–≥–æ–Ω–∞–ª—å–Ω—ã–µ –º–∏—Ä–æ–≤—ã–º –ª–∏–Ω–∏—è–º** (–≤–∞—Ä–∏–∞—Ü–∏—è –í),
3. ‚úÖ –ü—Ä–∞–≤–∏–ª—å–Ω–æ –≤—ã—á–∏—Å–ª–∏—Ç—å **–Ω–æ—Ä–º–∞–ª—å –∫ –º–∏—Ä–æ–≤–æ–π –ª–∏–Ω–∏–∏** –≤ –∫–∞–∂–¥–æ–π —Ç–æ—á–∫–µ.

## üîç –ì–µ–æ–º–µ—Ç—Ä–∏—è: –∫–∞–∫ –Ω–∞–π—Ç–∏ –Ω–æ—Ä–º–∞–ª—å –∫ –º–∏—Ä–æ–≤–æ–π –ª–∏–Ω–∏–∏?

–ú–∏—Ä–æ–≤–∞—è –ª–∏–Ω–∏—è:
$$

$$

–ö–∞—Å–∞—Ç–µ–ª—å–Ω—ã–π –≤–µ–∫—Ç–æ—Ä:
$$
\vec{v} = \left( \frac{dx}{d\tau}, \frac{dt}{d\tau} \right) = (1 + \xi) (\sinh \tau, \cosh \tau)
$$

–ù–æ—Ä–º–∞–ª—å–Ω—ã–π –≤–µ–∫—Ç–æ—Ä (–≤ 2D):
$$
\vec{n} = (-\cosh \tau, \sinh \tau)
\quad \text{(–≤—Ä–∞—â–µ–Ω–∏–µ –Ω–∞ 90¬∞ –ø—Ä–æ—Ç–∏–≤ —á–∞—Å–æ–≤–æ–π —Å—Ç—Ä–µ–ª–∫–∏)}

```python
# –°–µ—á–µ–Ω–∏—è, –æ—Ä—Ç–æ–≥–æ–Ω–∞–ª—å–Ω—ã–µ –º–∏—Ä–æ–≤—ã–º –ª–∏–Ω–∏—è–º (–≤–∞—Ä–∏–∞—Ü–∏—è –í –ø–æ –§–µ—Ä–º–∏)

        t0 = (1 + xi) * np.sinh(tau_c)

        nt = np.sinh(tau_c)
        
        # –ú–∞—Å—à—Ç–∞–±–∏—Ä—É–µ–º –¥–ª—è –Ω–∞–≥–ª—è–¥–Ω–æ—Å—Ç–∏
        scale = 0.3
        x_end = x0 + scale * nx
        t_end = t0 + scale * nt
        
        # –†–∏—Å—É–µ–º –Ω–æ—Ä–º–∞–ª—å
        plt.arrow(x0, t0, scale * nx, scale * nt,
                  head_width=0.08, head_length=0.1, fc='red', ec='red', alpha=0.7)

    # –î–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω–æ: –∏–Ω—Ç–µ—Ä–ø–æ–ª–∏—Ä—É–µ–º –º–µ–∂–¥—É —Ç–æ—á–∫–∞–º–∏ —Å –æ–¥–∏–Ω–∞–∫–æ–≤—ã–º œÑ, —á—Ç–æ–±—ã –ø–æ–∫–∞–∑–∞—Ç—å "—Å–µ—á–µ–Ω–∏–µ"
    xi_interp = np.linspace(xi_min, xi_max, 100)

    t_section = (1 + xi_interp) * np.sinh(tau_c)
    plt.plot(x_section, t_section, 'red', linestyle='--', alpha=0.5, lw=1.5, label=f'$\\tau = {tau_c:.1f}$ (–Ω–∞–ø—Ä–∞–≤–ª–µ–Ω–∏–µ —Å–µ—á–µ–Ω–∏—è)' if tau_c == tau_const[0] else "")
```

- ‚úÖ –î–æ–±–∞–≤–ª–µ–Ω –ø–æ–∫–æ—è—â–∏–π—Å—è –∑–∞—Ä—è–¥ —Å–ø—Ä–∞–≤–∞,
- ‚úÖ –û–±–ª–∞—Å—Ç—å –∏–Ω—Ç–µ–≥—Ä–∏—Ä–æ–≤–∞–Ω–∏—è –¥–ª—è —É—Å–∫–æ—Ä—è—é—â–µ–≥–æ—Å—è –∑–∞—Ä—è–¥–∞ ‚Äî **—Ü–∏–ª–∏–Ω–¥—Ä –º–µ–∂–¥—É –¥–≤—É–º—è —Å–µ—á–µ–Ω–∏—è–º–∏ $\tau = \text{const}$**.

### ‚úÖ –ü–æ–ª–Ω—ã–π –∏ –∏—Å–ø—Ä–∞–≤–ª–µ–Ω–Ω—ã–π –∫–æ–¥

```python
import numpy as np
import matplotlib.pyplot as plt

# –ü–∞—Ä–∞–º–µ—Ç—Ä—ã
tau1, tau2 = -1.0, 1.0  # –ø—Ä–µ–¥–µ–ª—ã –ø–æ œÑ –¥–ª—è –æ–±–ª–∞—Å—Ç–∏ –∏–Ω—Ç–µ–≥—Ä–∏—Ä–æ–≤–∞–Ω–∏—è
xi_min, xi_max = -0.3, 0.7  # –ø–∞—Ä–∞–º–µ—Ç—Ä Œæ –¥–ª—è —É—Å–∫–æ—Ä—è—é—â–µ–≥–æ—Å—è –∑–∞—Ä—è–¥–∞ (x ‚àà [0.7, 1.7] –ø—Ä–∏ œÑ=0)

# –î–∏–∞–ø–∞–∑–æ–Ω œÑ –¥–ª—è –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è –º–∏—Ä–æ–≤—ã—Ö –ª–∏–Ω–∏–π
tau_vals = np.linspace(-2.0, 2.0, 500)

# –§—É–Ω–∫—Ü–∏–∏ –º–∏—Ä–æ–≤—ã—Ö –ª–∏–Ω–∏–π: x = (1+Œæ) cosh œÑ, t = (1+Œæ) sinh œÑ
def x_world(tau, xi):
    return (1 + xi) * np.cosh(tau)

def t_world(tau, xi):
    return (1 + xi) * np.sinh(tau)

# -----------------------------------------------------------------------------
# –ù–∞—á–∞–ª–æ –≥—Ä–∞—Ñ–∏–∫–∞
plt.figure(figsize=(12, 8))

# -----------------------------------------------------------------------------
# 1. –ú–∏—Ä–æ–≤—ã–µ –ª–∏–Ω–∏–∏ —É—Å–∫–æ—Ä—è—é—â–µ–≥–æ—Å—è –∑–∞—Ä—è–¥–∞ (Œæ = const)
for xi in np.linspace(xi_min, xi_max, 5):
    x_line = x_world(tau_vals, xi)
    t_line = t_world(tau_vals, xi)
    valid = (x_line >= 0.5) & (x_line <= 2.5)
    plt.plot(x_line[valid], t_line[valid], color='blue', lw=2.0, alpha=0.8, 
             label=f'$\\xi = {xi}$ (—É—Å–∫–æ—Ä—è—é—â–∏–π—Å—è –∑–∞—Ä—è–¥)' if xi == xi_min else "")

tau_const = np.linspace(tau1, tau2, 5)  # –∑–Ω–∞—á–µ–Ω–∏—è œÑ –¥–ª—è —Å–µ—á–µ–Ω–∏–π

for tau_c in tau_const:
    # –¢–æ—á–∫–∏ –Ω–∞ —Å–µ—á–µ–Ω–∏–∏: —Ä–∞–∑–Ω—ã–µ Œæ –ø—Ä–∏ —Ñ–∏–∫—Å–∏—Ä–æ–≤–∞–Ω–Ω–æ–º œÑ
    xi_interp = np.linspace(xi_min, xi_max, 100)
    x_section = (1 + xi_interp) * np.cosh(tau_c)
    t_section = (1 + xi_interp) * np.sinh(tau_c)
    plt.plot(x_section, t_section, 'red', linestyle='--', alpha=0.6, lw=1.5,
             label=f'$\\tau = {tau_c:.1f}$ (–Ω–∞–ø—Ä–∞–≤–ª–µ–Ω–∏–µ —Å–µ—á–µ–Ω–∏—è)' if tau_c == tau_const[0] else "")

        x0 = (1 + xi) * np.cosh(tau_c)
        t0 = (1 + xi) * np.sinh(tau_c)
        
        # –ù–æ—Ä–º–∞–ª—å–Ω—ã–π –≤–µ–∫—Ç–æ—Ä: (-cosh œÑ, sinh œÑ) ‚Äî –ø–µ—Ä–ø–µ–Ω–¥–∏–∫—É–ª—è—Ä–µ–Ω –∫–∞—Å–∞—Ç–µ–ª—å–Ω–æ–π
        nx = -np.cosh(tau_c)
        nt = np.sinh(tau_c)

        scale = 0.2
        plt.arrow(x0, t0, scale * nx, scale * nt,
                  head_width=0.08, head_length=0.1, fc='red', ec='red', alpha=0.7)

# -----------------------------------------------------------------------------
# 3. –û–±–ª–∞—Å—Ç—å –∏–Ω—Ç–µ–≥—Ä–∏—Ä–æ–≤–∞–Ω–∏—è G1: –º–µ–∂–¥—É œÑ1 –∏ œÑ2
# –ó–∞–∫—Ä–∞—à–∏–≤–∞–µ–º –æ–±–ª–∞—Å—Ç—å –º–µ–∂–¥—É œÑ = tau1 –∏ œÑ = tau2
xi_fill = np.linspace(xi_min, xi_max, 100)
x_tau1 = (1 + xi_fill) * np.cosh(tau1)
t_tau1 = (1 + xi_fill) * np.sinh(tau1)
x_tau2 = (1 + xi_fill) * np.cosh(tau2)
t_tau2 = (1 + xi_fill) * np.sinh(tau2)

# –û–±—Ä–∞—Ç–Ω—ã–π –ø–æ—Ä—è–¥–æ–∫ –¥–ª—è –∑–∞–º—ã–∫–∞–Ω–∏—è –ø–æ–ª–∏–≥–æ–Ω–∞
x_poly = np.concatenate([x_tau1, x_tau2[::-1]])
t_poly = np.concatenate([t_tau1, t_tau2[::-1]])

plt.fill(x_poly, t_poly, color='gold', alpha=0.4, edgecolor='orange', lw=2, label=r'–û–±–ª–∞—Å—Ç—å $G_1$ ($\tau_1 \leq \tau \leq \tau_2$)')

# -----------------------------------------------------------------------------
# 4. –ü–æ–∫–æ—è—â–∏–π—Å—è –∑–∞—Ä—è–¥ (—Å–ø—Ä–∞–≤–∞)
x2_min, x2_max = 2.5, 3.5

for x2 in np.linspace(x2_min, x2_max, 5):
    plt.plot(np.full_like(t_line, x2), t_line, 'green', lw=2.5, alpha=0.9,
             label='–ü–æ–∫–æ—è—â–∏–π—Å—è –∑–∞—Ä—è–¥' if x2 == x2_min else "")

# –û–±–ª–∞—Å—Ç—å G2
plt.fill_betweenx(t_line, x2_min, x2_max, color='lightblue', alpha=0.4,
                  edgecolor='blue', lw=2, label=r'–û–±–ª–∞—Å—Ç—å $G_2$')

# -----------------------------------------------------------------------------
# 5. –û—Ñ–æ—Ä–º–ª–µ–Ω–∏–µ
plt.xlabel(r"–ü—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–æ $x$", fontsize=12)
plt.ylabel(r"–í—Ä–µ–º—è $t$", fontsize=12)
plt.title(r"–í–∞—Ä–∏–∞—Ü–∏—è –í –ø–æ –§–µ—Ä–º–∏: –æ—Ä—Ç–æ–≥–æ–Ω–∞–ª—å–Ω—ã–µ —Å–µ—á–µ–Ω–∏—è –∏ –æ–±–ª–∞—Å—Ç—å –∏–Ω—Ç–µ–≥—Ä–∏—Ä–æ–≤–∞–Ω–∏—è", fontsize=13)
plt.xlim(0.4, 4.0)
plt.ylim(-2.0, 2.0)
plt.legend(loc='upper left', fontsize=10, framealpha=0.9)
plt.grid(True, alpha=0.3)
plt.tight_layout()

# –ü–æ–∫–∞–∑–∞—Ç—å –≥—Ä–∞—Ñ–∏–∫
plt.show()
```

## ‚úÖ 1. –û–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ 4-—Ç–æ–∫–∞ $j^\mu(x)$

$$

### –î–ª—è –ø—Ä–æ—Ç—è–∂—ë–Ω–Ω–æ–≥–æ –∑–∞—Ä—è–¥–∞

–ü—É—Å—Ç—å –∑–∞—Ä—è–¥ –∏–º–µ–µ—Ç **–æ–±—ä—ë–º $V$** –∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –≤ —Å–≤–æ–µ–π –°–û:

–¢–æ–≥–¥–∞ –≤ –ª–∞–±–æ—Ä–∞—Ç–æ—Ä–Ω–æ–π –°–û (–≥–¥–µ –æ–Ω –¥–≤–∏–∂–µ—Ç—Å—è —Å–æ —Å–∫–æ—Ä–æ—Å—Ç—å—é $\vec{v}(t)$):

üìå –≠—Ç–æ ‚Äî —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ.

## ‚úÖ 2. –û–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ 4-–ø–æ—Ç–µ–Ω—Ü–∏–∞–ª–∞ $A_\mu(x)$

–≥–¥–µ:

üìå –≠—Ç–æ ‚Äî **—Ç–æ—á–Ω–æ–µ —Ä–µ–ª—è—Ç–∏–≤–∏—Å—Ç—Å–∫–æ–µ –ø–æ–ª–µ**.

---

### –î–ª—è –ø—Ä–æ—Ç—è–∂—ë–Ω–Ω–æ–≥–æ –∑–∞—Ä—è–¥–∞

$$
A_\mu(x) = \frac{1}{c} \int G_{\text{ret}}(x - x') j_\mu(x') \, d^4x'

- $\vec{r}_1$ ‚Äî –æ—Ç —Ü–µ–Ω—Ç—Ä–∞ 1.

---

### üîπ –£—á—ë—Ç –õ–æ—Ä–µ–Ω—Ü–µ–≤–∞ —Å–æ–∫—Ä–∞—â–µ–Ω–∏—è

–ü—Ä–∏ $\vec{v}_1 \ne 0$, —Å—Ñ–µ—Ä–∞ 1 ‚Äî **—ç–ª–ª–∏–ø—Å–æ–∏–¥**:
- –ü–æ–ª—É–æ—Å—å –≤–¥–æ–ª—å $\vec{v}_1$: $R_1 / \gamma_1$,
- –ü–æ–ø–µ—Ä–µ—á–Ω—ã–µ: $R_1$,
- –û–±—ä—ë–º: $V_1 = \frac{4\pi}{3} R_1^2 (R_1 / \gamma_1) = \frac{4\pi R_1^3}{3\gamma_1}$,
- –ü–ª–æ—Ç–Ω–æ—Å—Ç—å: $\rho_1 = \frac{q_1}{V_1} = \frac{3q_1 \gamma_1}{4\pi R_1^3}$.

üìå –¢–æ –∂–µ –¥–ª—è —Å—Ñ–µ—Ä—ã 2 –ø—Ä–∏ –≤—ã—á–∏—Å–ª–µ–Ω–∏–∏ –ø–æ–ª—è.

---

### üîπ –í –ø—Ä–µ–¥–µ–ª–µ $R_1, R_2 \to 0$

$$
\vec{F}_{12} = \frac{q_1 q_2}{4\pi\varepsilon_0 D^2} \left(1 + \frac{\vec{\Gamma}_2 \cdot \vec{D}}{c^2}\right) \hat{D}
$$

$$
\boxed{
\mathcal{G} = 1 + \frac{(\vec{\Gamma}_1 + \vec{\Gamma}_2) \cdot \vec{D}}{2c^2}
}
$$

## üì¶ 2. –ß–∏—Å–ª–µ–Ω–Ω—ã–π –∫–æ–¥ –Ω–∞ Python

```python
import numpy as np
from scipy.integrate import tplquad
import matplotlib.pyplot as plt

# –ü–∞—Ä–∞–º–µ—Ç—Ä—ã
c = 3e8  # –º/—Å
eps0 = 8.854e-12  # –§/–º

# –ó–∞—Ä—è–¥—ã
q1 = 1e-9  # –ö–ª
q2 = 1e-9  # –ö–ª
R1 = 0.1  # –º
R2 = 0.1  # –º
D = 10.0  # –º

# –£—Å–∫–æ—Ä–µ–Ω–∏—è (–º/—Å¬≤)
Gamma1 = np.array([1e6, 0, 0])
Gamma2 = np.array([1e6, 0, 0])

# –°–∫–æ—Ä–æ—Å—Ç–∏ (–º/—Å) ‚Äî –¥–ª—è –õ–æ—Ä–µ–Ω—Ü–µ–≤–∞ —Å–æ–∫—Ä–∞—â–µ–Ω–∏—è
v1 = np.array([0.1*c, 0, 0])
v2 = np.array([0.1*c, 0, 0])
gamma1 = 1 / np.sqrt(1 - (v1[0]/c)**2)
gamma2 = 1 / np.sqrt(1 - (v2[0]/c)**2)

# –¶–µ–Ω—Ç—Ä—ã
center1 = np.array([0.0, 0.0, 0.0])
center2 = np.array([D, 0.0, 0.0])

# –ü–ª–æ—Ç–Ω–æ—Å—Ç–∏ (—Å —É—á—ë—Ç–æ–º —Å–æ–∫—Ä–∞—â–µ–Ω–∏—è)
rho1 = q1 * gamma1 / (4/3 * np.pi * R1**3)
rho2 = q2 * gamma2 / (4/3 * np.pi * R2**3)

# –ü–æ–ª–µ –õ–∏–µ–Ω–∞—Ä–∞‚Äì–í–∏—Ö–µ—Ä—Ç–∞ (—É–ø—Ä–æ—â—ë–Ω–Ω–æ–µ, –¥–ª—è –º–µ–¥–ª–µ–Ω–Ω–æ–≥–æ –¥–≤–∏–∂–µ–Ω–∏—è)
def E_LW(r1, r2, v2, Gamma2):
    R_vec = r1 - r2
    R = np.linalg.norm(R_vec)
    if R < 1e-10:
        return np.array([0.0, 0.0, 0.0])
    n = R_vec / R
    beta = v2 / c
    beta_dot = Gamma2 / c
    n_dot_beta = np.dot(n, beta)
    
    # –≠–ª–µ–∫—Ç—Ä–∏—á–µ—Å–∫–æ–µ –ø–æ–ª–µ (—Ç–æ–ª—å–∫–æ –∫—É–ª–æ–Ω–æ–≤—Å–∫–∞—è —á–∞—Å—Ç—å + —É—Å–∫–æ—Ä–µ–Ω–∏–µ)
    term1 = (1 - np.dot(beta, beta)) * (n - beta) / (R**2 * (1 - n_dot_beta)**3)
    term2 = np.cross(n, np.cross(n - beta, beta_dot)) / (c * R * (1 - n_dot_beta)**3)
    E = (q2 / (4 * np.pi * eps0)) * (term1 + term2)
    return E

# –ò–Ω—Ç–µ–≥—Ä–∏—Ä–æ–≤–∞–Ω–∏–µ –ø–æ —Å—Ñ–µ—Ä–µ 1
def integrand(x1, y1, z1):
    r1 = np.array([x1, y1, z1])
    if np.linalg.norm(r1 - center1) > R1:
        return np.array([0.0, 0.0, 0.0])
    
    # –ü–æ–ª–µ –æ—Ç —Å—Ñ–µ—Ä—ã 2 (—É—Å—Ä–µ–¥–Ω—ë–Ω–Ω–æ–µ –∏–ª–∏ –≤ —Ü–µ–Ω—Ç—Ä–µ ‚Äî –¥–ª—è –ø—Ä–æ—Å—Ç–æ—Ç—ã)
    r2 = center2  # –ø—Ä–∏–±–ª–∏–∂–µ–Ω–∏–µ: –ø–æ–ª–µ –æ—Ç —Ü–µ–Ω—Ç—Ä–∞
    E = E_LW(r1, r2, v2, Gamma2)
    
    # –ü–æ–ø—Ä–∞–≤–∫–∞ –§–µ—Ä–º–∏
    r1_rel = r1 - center1
    fermit_weight = 1 + np.dot(Gamma1, r1_rel) / c**2
    
    dF = rho1 * E * fermit_weight
    return dF

# –í—ã—á–∏—Å–ª–µ–Ω–∏–µ —Å–∏–ª—ã
def F12_integrand(z, y, x):
    return integrand(x, y, z)

# –ì—Ä–∞–Ω–∏—Ü—ã –∏–Ω—Ç–µ–≥—Ä–∏—Ä–æ–≤–∞–Ω–∏—è ‚Äî —Å—Ñ–µ—Ä–∞
def y_limit(x):
    return [-np.sqrt(R1**2 - x**2), np.sqrt(R1**2 - x**2)]

def z_limit(x, y):
    return [-np.sqrt(R1**2 - x**2 - y**2), np.sqrt(R1**2 - x**2 - y**2)]

result, _ = tplquad(F12_integrand,
                    -R1, R1,
                    lambda x: y_limit(x)[0], lambda x: y_limit(x)[1],
                    lambda x, y: z_limit(x, y)[0], lambda x, y: z_limit(x, y)[1])

F12 = result

print(f"–°–∏–ª–∞ F12: {F12} –ù")
print(f"–ú–æ–¥—É–ª—å: {np.linalg.norm(F12):.3e} –ù")
```

–ø—Ä–∏ –≤—ã—á–∏—Å–ª–µ–Ω–∏–∏ —Å–∏–ª—ã –Ω–∞ $i$-–π –∑–∞—Ä—è–¥,

## üîπ –®–∞–≥ 1: –æ–±—â–∏–π –≤–∏–¥ –ø–æ–ø—Ä–∞–≤–∫–∏ –¥–ª—è $N$ –∑–∞—Ä—è–¥–æ–≤

–î–ª—è **–∫–∞–∂–¥–æ–≥–æ –∑–∞—Ä—è–¥–∞ $i$** –≤ –º–æ–º–µ–Ω—Ç –≤—Ä–µ–º–µ–Ω–∏ $t$:
$$
\boxed{
\mathcal{G}_i(t) = 1 + \frac{1}{2c^2} \sum_{\substack{j=1 \\ j \ne i}}^N \left[ \vec{\Gamma}_i(t) + \vec{\Gamma}_j(t_{\text{–∑–∞–ø},ij}) \right] \cdot \vec{r}_{ij}(t)
}
$$

–≥–¥–µ:
- $\vec{\Gamma}_i(t)$ ‚Äî —É—Å–∫–æ—Ä–µ–Ω–∏–µ $i$-–≥–æ –∑–∞—Ä—è–¥–∞ (–Ω–∞–±–ª—é–¥–∞—Ç–µ–ª—è) –≤ –º–æ–º–µ–Ω—Ç $t$,
- $\vec{\Gamma}_j(t_{\text{–∑–∞–ø},ij})$ ‚Äî —É—Å–∫–æ—Ä–µ–Ω–∏–µ $j$-–≥–æ –∑–∞—Ä—è–¥–∞ (–∏—Å—Ç–æ—á–Ω–∏–∫–∞) –≤ **–∑–∞–ø–∞–∑–¥—ã–≤–∞—é—â–∏–π –º–æ–º–µ–Ω—Ç** $t_{\text{–∑–∞–ø},ij}$,
- $\vec{r}_{ij}(t) = \vec{r}_i(t) - \vec{r}_j(t_{\text{–∑–∞–ø},ij})$ ‚Äî –∑–∞–ø–∞–∑–¥—ã–≤–∞—é—â–∏–π –≤–µ–∫—Ç–æ—Ä,
- $t_{\text{–∑–∞–ø},ij}$ –æ–ø—Ä–µ–¥–µ–ª—è–µ—Ç—Å—è –∏–∑ —É—Ä–∞–≤–Ω–µ–Ω–∏—è:
  $$
  |\vec{r}_i(t) - \vec{r}_j(t_{\text{–∑–∞–ø},ij})| = c(t - t_{\text{–∑–∞–ø},ij})
  $$

---

## üîπ –®–∞–≥ 2: –∞–ª–≥–æ—Ä–∏—Ç–º –≤—ã—á–∏—Å–ª–µ–Ω–∏—è $\mathcal{G}_i(t)$

### üîÑ –î–ª—è –∫–∞–∂–¥–æ–≥–æ –≤—Ä–µ–º–µ–Ω–Ω–æ–≥–æ —à–∞–≥–∞ $t$:

#### 1. –î–ª—è –∫–∞–∂–¥–æ–≥–æ –∑–∞—Ä—è–¥–∞ $i$:
   - –ü–æ–ª—É—á–∏—Ç—å $\vec{r}_i(t)$, $\vec{v}_i(t)$, $\vec{\Gamma}_i(t)$.

#### 2. –î–ª—è –∫–∞–∂–¥–æ–π –ø–∞—Ä—ã $(i, j)$, $j \ne i$:
   - **–†–µ—à–∏—Ç—å —É—Ä–∞–≤–Ω–µ–Ω–∏–µ –∑–∞–ø–∞–∑–¥—ã–≤–∞–Ω–∏—è**:
     $$
     |\vec{r}_i(t) - \vec{r}_j(t_{\text{–∑–∞–ø}})| = c(t - t_{\text{–∑–∞–ø}})
     $$
     (—á–∏—Å–ª–µ–Ω–Ω–æ, –Ω–∞–ø—Ä–∏–º–µ—Ä, –º–µ—Ç–æ–¥–æ–º –ù—å—é—Ç–æ–Ω–∞ –∏–ª–∏ –¥–∏—Ö–æ—Ç–æ–º–∏–∏).
   - –ù–∞–π—Ç–∏ $t_{\text{–∑–∞–ø},ij}$, $\vec{r}_j(t_{\text{–∑–∞–ø},ij})$, $\vec{\Gamma}_j(t_{\text{–∑–∞–ø},ij})$.
   - –í—ã—á–∏—Å–ª–∏—Ç—å $\vec{r}_{ij}(t) = \vec{r}_i(t) - \vec{r}_j(t_{\text{–∑–∞–ø},ij})$.

#### 3. –°—É–º–º–∏—Ä–æ–≤–∞—Ç—å –≤–∫–ª–∞–¥—ã:
   $$
   \mathcal{G}_i(t) = 1 + \frac{1}{2c^2} \sum_{j \ne i} \left[ \vec{\Gamma}_i(t) + \vec{\Gamma}_j(t_{\text{–∑–∞–ø},ij}) \right] \cdot \vec{r}_{ij}(t)
   $$

---

## üîπ –®–∞–≥ 3: –∏–Ω—Ç–µ–≥—Ä–∏—Ä–æ–≤–∞–Ω–∏–µ —Å –ø–æ–ø—Ä–∞–≤–∫–æ–π

–ü—Ä–∏ –≤—ã—á–∏—Å–ª–µ–Ω–∏–∏ **–∏–º–ø—É–ª—å—Å–∞** –∏–ª–∏ **–¥–µ–π—Å—Ç–≤–∏—è** –¥–ª—è $i$-–≥–æ –∑–∞—Ä—è–¥–∞:
$$
\vec{p}_i = \int \vec{F}_i(t) \, \mathcal{G}_i(t) \, dt
$$
–≥–¥–µ:
- $\vec{F}_i(t) = \sum_{j \ne i} q_i \left[ \vec{E}_{\text{LV},j}(\vec{r}_i, t) + \vec{v}_i(t) \times \vec{B}_{\text{LV},j}(t) \right]$,
- $\vec{E}_{\text{LV},j}$, $\vec{B}_{\text{LV},j}$ ‚Äî –ø–æ–ª–µ –õ–∏–µ–Ω–∞—Ä–∞‚Äì–í–∏—Ö–µ—Ä—Ç–∞ –æ—Ç $j$-–≥–æ –∑–∞—Ä—è–¥–∞.

üìå **–ö–ª—é—á–µ–≤–æ–π –º–æ–º–µ–Ω—Ç**:  
> –ü–æ–ø—Ä–∞–≤–∫–∞ $\mathcal{G}_i(t)$ **–Ω–µ —É–º–Ω–æ–∂–∞–µ—Ç—Å—è –Ω–∞ –ø–æ–ª–µ**,  
> –∞ **–≤—Ö–æ–¥–∏—Ç –≤ –º–µ—Ä—É –∏–Ω—Ç–µ–≥—Ä–∏—Ä–æ–≤–∞–Ω–∏—è –ø–æ –≤—Ä–µ–º–µ–Ω–∏**.

## üîπ –®–∞–≥ 4: —á–∏—Å–ª–µ–Ω–Ω–∞—è —Ä–µ–∞–ª–∏–∑–∞—Ü–∏—è (–ø—Å–µ–≤–¥–æ–∫–æ–¥)

```python
def compute_fermi_correction(t, positions, accelerations, c=3e8):
    """
    –í—ã—á–∏—Å–ª—è–µ—Ç –ø–æ–ø—Ä–∞–≤–∫—É –§–µ—Ä–º–∏ –¥–ª—è –≤—Å–µ—Ö –∑–∞—Ä—è–¥–æ–≤ –≤ –º–æ–º–µ–Ω—Ç t.
    
    –ê—Ä–≥—É–º–µ–Ω—Ç—ã:
    positions[i] = [r_i(t), r_i(t-dt), ...] ‚Äî —Ç—Ä–∞–µ–∫—Ç–æ—Ä–∏—è i-–≥–æ –∑–∞—Ä—è–¥–∞
    accelerations[i] = [Gamma_i(t), Gamma_i(t-dt), ...]
    
    –í–æ–∑–≤—Ä–∞—â–∞–µ—Ç:
    G[i] ‚Äî –ø–æ–ø—Ä–∞–≤–∫–∞ –¥–ª—è i-–≥–æ –∑–∞—Ä—è–¥–∞

    G = np.ones(N)

        r_i = positions[i][-1]  # —Ç–µ–∫—É—â–µ–µ –ø–æ–ª–æ–∂–µ–Ω–∏–µ i-–≥–æ –∑–∞—Ä—è–¥–∞

            if i == j:
                continue
                
            # –†–µ—à–∞–µ–º —É—Ä–∞–≤–Ω–µ–Ω–∏–µ –∑–∞–ø–∞–∑–¥—ã–≤–∞–Ω–∏—è: |r_i(t) - r_j(t_zap)| = c(t - t_zap)
            t_zap = solve_retarded_time(t, r_i, positions[j], c)
            
            # –ü–æ–ª—É—á–∞–µ–º –ø–æ–ª–æ–∂–µ–Ω–∏–µ –∏ —É—Å–∫–æ—Ä–µ–Ω–∏–µ j-–≥–æ –∑–∞—Ä—è–¥–∞ –≤ t_zap
            r_j_zap = interpolate_position(t_zap, positions[j])

            r_ij = r_i - r_j_zap
            
            # –í–∫–ª–∞–¥ –ø–∞—Ä—ã (i, j) –≤ –ø–æ–ø—Ä–∞–≤–∫—É
            contribution = (Gamma_i + Gamma_j_zap) @ r_ij / (2 * c**2)
            G[i] += contribution
            
    return G

# –ü—Ä–∏–º–µ—Ä –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è –≤ —Ü–∏–∫–ª–µ –ø–æ –≤—Ä–µ–º–µ–Ω–∏
for t in time_steps:
    # –û–±–Ω–æ–≤–ª—è–µ–º —Ç—Ä–∞–µ–∫—Ç–æ—Ä–∏–∏ –∏ —É—Å–∫–æ—Ä–µ–Ω–∏—è (–Ω–∞–ø—Ä–∏–º–µ—Ä, —á–µ—Ä–µ–∑ —É—Ä–∞–≤–Ω–µ–Ω–∏—è –¥–≤–∏–∂–µ–Ω–∏—è)
    update_positions_and_accelerations(t)
    
    # –í—ã—á–∏—Å–ª—è–µ–º –ø–æ–ø—Ä–∞–≤–∫—É –§–µ—Ä–º–∏
    G = compute_fermi_correction(t, positions, accelerations)
    
    # –í—ã—á–∏—Å–ª—è–µ–º —Å–∏–ª—É —Å —É—á—ë—Ç–æ–º –ø–æ–ø—Ä–∞–≤–∫–∏
    for i in range(N):
        F_i = compute_lorentz_force(i, t)  # —Å–∏–ª–∞ –±–µ–∑ –ø–æ–ø—Ä–∞–≤–∫–∏

```

```python
import numpy as np
from scipy.optimize import root_scalar

def solve_retarded_time(t_current, r_i, trajectory_j, c=3e8, max_iter=50, tol=1e-12):
    """
    –†–µ—à–∞–µ—Ç —É—Ä–∞–≤–Ω–µ–Ω–∏–µ |r_i(t) - r_j(t_zap)| = c(t - t_zap)
    
    –ê—Ä–≥—É–º–µ–Ω—Ç—ã:
    t_current -- —Ç–µ–∫—É—â–µ–µ –≤—Ä–µ–º—è
    r_i -- –ø–æ–ª–æ–∂–µ–Ω–∏–µ –Ω–∞–±–ª—é–¥–∞—Ç–µ–ª—è –≤ t_current
    trajectory_j -- —Å–ø–∏—Å–æ–∫ [ (t, r_j(t)) ] –¥–ª—è –∏—Å—Ç–æ—á–Ω–∏–∫–∞ j
    c -- —Å–∫–æ—Ä–æ—Å—Ç—å —Å–≤–µ—Ç–∞
    
    –í–æ–∑–≤—Ä–∞—â–∞–µ—Ç:
    t_zap -- –∑–∞–ø–∞–∑–¥—ã–≤–∞—é—â–∏–π –º–æ–º–µ–Ω—Ç –≤—Ä–µ–º–µ–Ω–∏
    """
    # –ò–Ω—Ç–µ—Ä–ø–æ–ª—è—Ü–∏—è —Ç—Ä–∞–µ–∫—Ç–æ—Ä–∏–∏ –∏—Å—Ç–æ—á–Ω–∏–∫–∞
    times = np.array([t for t, _ in trajectory_j])
    positions = np.array([r for _, r in trajectory_j])
    
    def equation(t_zap):
        # –ò–Ω—Ç–µ—Ä–ø–æ–ª–∏—Ä—É–µ–º –ø–æ–ª–æ–∂–µ–Ω–∏–µ –∏—Å—Ç–æ—á–Ω–∏–∫–∞ –≤ t_zap
        idx = np.searchsorted(times, t_zap)
        if idx == 0 or idx == len(times):
            return np.inf  # –≤–Ω–µ –¥–∏–∞–ø–∞–∑–æ–Ω–∞
        
        # –õ–∏–Ω–µ–π–Ω–∞—è –∏–Ω—Ç–µ—Ä–ø–æ–ª—è—Ü–∏—è
        t_low, t_high = times[idx-1], times[idx]
        r_low, r_high = positions[idx-1], positions[idx]
        r_j = r_low + (r_high - r_low) * (t_zap - t_low) / (t_high - t_low)
        
        R = np.linalg.norm(r_i - r_j)
        return R - c * (t_current - t_zap)
    
    # –ù–∞—á–∞–ª—å–Ω–æ–µ –ø—Ä–∏–±–ª–∏–∂–µ–Ω–∏–µ
    t_guess = t_current - np.linalg.norm(r_i - positions[-1]) / c
    
    # –†–µ—à–∞–µ–º —É—Ä–∞–≤–Ω–µ–Ω–∏–µ
    sol = root_scalar(equation, bracket=[max(t_guess-0.1, times[0]), t_current], 
                      method='brentq', xtol=tol, maxiter=max_iter)
    return sol.root

def compute_fermi_correction(t, trajectories, accelerations, c=3e8):
    """
    –í—ã—á–∏—Å–ª—è–µ—Ç –ø–æ–ø—Ä–∞–≤–∫—É –§–µ—Ä–º–∏ –¥–ª—è –≤—Å–µ—Ö –∑–∞—Ä—è–¥–æ–≤ –≤ –º–æ–º–µ–Ω—Ç t
    
    –ê—Ä–≥—É–º–µ–Ω—Ç—ã:
    trajectories[i] = [(t0, r_i(t0)), (t1, r_i(t1)), ...] -- —Ç—Ä–∞–µ–∫—Ç–æ—Ä–∏—è i-–≥–æ –∑–∞—Ä—è–¥–∞
    accelerations[i] = [(t0, Gamma_i(t0)), ...] -- —É—Å–∫–æ—Ä–µ–Ω–∏—è
    
    –í–æ–∑–≤—Ä–∞—â–∞–µ—Ç:
    G[i] -- –ø–æ–ø—Ä–∞–≤–∫–∞ –¥–ª—è i-–≥–æ –∑–∞—Ä—è–¥–∞
    """
    N = len(trajectories)
    G = np.ones(N)
    
    for i in range(N):
        # –¢–µ–∫—É—â–µ–µ –ø–æ–ª–æ–∂–µ–Ω–∏–µ –∏ —É—Å–∫–æ—Ä–µ–Ω–∏–µ –Ω–∞–±–ª—é–¥–∞—Ç–µ–ª—è
        r_i = trajectories[i][-1][1]
        Gamma_i = accelerations[i][-1][1]
        
        for j in range(N):
            if i == j:
                continue
                
            # –†–µ—à–∞–µ–º —É—Ä–∞–≤–Ω–µ–Ω–∏–µ –∑–∞–ø–∞–∑–¥—ã–≤–∞–Ω–∏—è
            try:
                t_zap = solve_retarded_time(t, r_i, trajectories[j], c)
            except:
                continue
                
            # –ü–æ–ª—É—á–∞–µ–º –ø–æ–ª–æ–∂–µ–Ω–∏–µ –∏ —É—Å–∫–æ—Ä–µ–Ω–∏–µ –∏—Å—Ç–æ—á–Ω–∏–∫–∞ –≤ t_zap
            idx = np.searchsorted([t for t, _ in trajectories[j]], t_zap)
            r_j = np.interp(t_zap, 
                           [t for t, _ in trajectories[j][idx-1:idx+1]], 
                           [np.linalg.norm(r) for _, r in trajectories[j][idx-1:idx+1]])
            Gamma_j = np.interp(t_zap, 
                               [t for t, _ in accelerations[j][idx-1:idx+1]], 
                               [np.linalg.norm(g) for _, g in accelerations[j][idx-1:idx+1]])
            
            # –ó–∞–ø–∞–∑–¥—ã–≤–∞—é—â–∏–π –≤–µ–∫—Ç–æ—Ä
            r_ij = r_i - r_j
            
            # –í–∫–ª–∞–¥ –ø–∞—Ä—ã (i, j) –≤ –ø–æ–ø—Ä–∞–≤–∫—É
            contribution = (Gamma_i + Gamma_j) @ r_ij / (2 * c**2)
            G[i] += contribution
            
    return G

# –ü—Ä–∏–º–µ—Ä –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è –≤ —Ü–∏–∫–ª–µ –ø–æ –≤—Ä–µ–º–µ–Ω–∏

for t in time_steps:
    # 1. –û–±–Ω–æ–≤–ª—è–µ–º —Ç—Ä–∞–µ–∫—Ç–æ—Ä–∏–∏ –∏ —É—Å–∫–æ—Ä–µ–Ω–∏—è

    # 3. –ò–Ω—Ç–µ–≥—Ä–∏—Ä—É–µ–º —Å–∏–ª—É —Å –ø–æ–ø—Ä–∞–≤–∫–æ–π

        F_i = compute_lorentz_force(i, t)  # –≤–∞—à–∞ —Ñ—É–Ω–∫—Ü–∏—è
```

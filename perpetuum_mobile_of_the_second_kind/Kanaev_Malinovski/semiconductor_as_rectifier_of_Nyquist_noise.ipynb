# полупроводник как выпрямитель шума Найквиста

Кстати вчера во время лекции по физике твердого тела, пишу чтобы не забыть возникла идея, которую я сформулировал в виде вопроса преподавателю. Лекция была о теории полупроводников Блоха, об энергии Ферми о ширине запрещённой зоны, о зависимости сопротивления от этой ширины. А от величины сопротивления, как Вы, надеюсь, помните, зависит энергия шума Найквиста

Так вот в чем идея. Имеем монокристалл полупроводника с отсуствием зеркальной симметрии. Выберем ось вдоль которой имеется зеркальная асимметрия за икс

прикладываем к кристалу полупроводника внешнее электрическое поле вдоль оси икс. Внешнее электрическое поле должно повлиять на вид решения уравнений Шредингера для электронов в кристалле. Во первых через добавку к гамильтониану каждого электрона (что как бы тривиально) а во вторых через смещение узлов кристаллической решётки благодаря эффекту изменения формы кристалла из-за его поляризации. Таким образом влияние внешнего электрического поля на вид волновых функций электронов, вытекающий из решений уравнений Шредингера приведет к изменению величины энергетического уровня каждого электрона и следовательно через это изменение повлияет на ширину запрещённой зоны

А поскольку кристалл имеет ось зеркальной асимметрии то при изменении направления внешнего поля вдоль оси икс ответное изменение конфигурации узлов кристаллической решётки будет другим (асимметричным) и поэтому решение уравнений Шредингера изменится также по другому (асимметрично) и поэтому смещение уровней энергии электронов в стационарных состояниях тоже произойдетнесколько по другому (асимметрично)!

Отсюда вывод - я ожидаю, что для монокристалла полупроводника с зеркальной асимметрией изменение направления внешнего поля вдоль оси асимметрии будет по разному влиять на ширину запрещённой зоны и как следствие на проводимость или сопротивление кристалла

что в принципе и наблюдал Малиновский в своих экспериментах на сегнетоэлектрике

А если учесть зависимость величины шума Найквиста от сопротивления то можно ожидать что полупроводник с зеркальной асимметрией будет работать как выпрямитель шума Найквиста
